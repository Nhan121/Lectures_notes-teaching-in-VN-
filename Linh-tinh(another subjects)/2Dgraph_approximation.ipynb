**`Step function`** Let $c_1, \ldots, c_n$ be the constants and $A_1, \ldots, A_n$ be the disjoint sequence sets ($A_i \cap A_j = \emptyset, \quad \forall i \neq j$), the `step function` on $A_1, \ldots, A_n$ is defined as

$$ SF(x) = \sum_{k = 1}^n c_k \mathbb{I}_{A_k}(x) $$

```python
## Define the constants
const = [1, 3, 5, 4, 7, 2, 6, 9]

## define the partitions
partt = np.linspace(0, 8, len(const) + 1 )

## define the range of values
u = np.linspace(0, 8, 401)

sum = 0
## Simple step function
for k in range(len(const)):
    sum = sum + (const[k] * (u >= partt[k]) * (u < partt[k+1]))
    
## display
plt.figure(figsize = (9, 4))
plt.grid(axis='x', color='0.95')
plt.plot(u, sum)
```

`Example 2.` Use the Riemann sum to approximate the function by the step_function; by using the partitions 

$$ \left\lbrace x_1, x_2, \ldots, x_k \right\rbrace $$ 

on the interval $(a, b)$; hence we consider

$$ f(x) \approx \sum_{k = 1}^{n - 1} \frac{f(x_k) + f(x_{k+1})}{2} \mathbb{I}_{(x_k, x_{k+1})}(x) $$


```python
## Define the function
def func(t):
    return abs(t**3 - 4*t)

plt.figure(figsize = (10, 6))

x = np.linspace(-3, 3, 1001)

nb_parts = [8, 20, 100]
linestyle = ['--', '-.', '-']

for idx in range(len(nb_parts)):       
    
    z = np.linspace(-3, 3, nb_parts[idx] + 1)
    w = func(z)

    sum = 0
    
    ## Simple step function
    
    for k in range(len(z) - 1):
        ck = (w[k] + w[k + 1])/2
        sum = sum + (ck * (x >= z[k]) * (x < z[k+1]))

    ## display
    plt.plot(x, sum, label = 'number of partitions = '+str(nb_parts[idx]), ls = linestyle[idx], linewidth = 2.0)
    plt.legend()

plt.plot(x, func(x), 'darkmagenta', linewidth = 3.0, label = "real_function")
plt.show()
```

**approximate the heart_shape with the available function: `plt.step`**

```python
plt.figure(figsize = (7, 6))
t = np.linspace(-2,2, 250); 
def heart_plot(t):
    f1 = np.sqrt( 1 - (np.abs(t) - 1)**2 ) 
    f2 = -3*np.sqrt(1-(np.abs(t)/2)**0.5) 
    plt.plot(t, f1, 'darkblue')
    plt.plot(t, f2, 'darkblue') 
heart_plot(t)

t = np.linspace(-2,2, 25)   
def heart_step(t):
    f1 = np.sqrt( 1 - (np.abs(t) - 1)**2 ) 
    f2 = -3*np.sqrt(1-(np.abs(t)/2)**0.5) 
    plt.step(t, f1, 'r-', where='mid')
    plt.step(t, f2, 'r-', where='mid') 
heart_step(t)
```

**Example 4. plot multiple hearts and their step_function**

```python
plt.figure(figsize = (8, 8))

def hearts(n): ## n is the numbers of heart
    for k in np.arange(1,n+1,1):
        t = np.linspace(-2*k,2*k, 251)
        f1 = np.sqrt( k**2 - (np.abs(t) - k)**2 )
        f2 = -(2*k+1)*np.sqrt(k**0.5 -(np.abs(t)/2)**0.5)
        plt.plot(t, f1, 'b' ,t, f2, 'b')
        
        t = np.linspace(-2*k,2*k, 11)
        f1 = np.sqrt( k**2 - (np.abs(t) - k)**2 )
        f2 = -(2*k+1)*np.sqrt(k**0.5 -(np.abs(t)/2)**0.5)        
        plt.step(t, f1, 'r', where = 'mid')
        plt.step(t, f2, 'r', where = 'mid')
        
hearts(5)
```
