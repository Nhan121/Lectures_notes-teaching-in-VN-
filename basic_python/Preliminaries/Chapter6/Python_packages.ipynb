## 1. Python packages introduction.
When a project gets larger, we naturally split the code into several [modules](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/basic_python/Preliminaries/Chapter6/Python_Modules.ipynb), each one being adpated for a specific purpose (`getting the data`, `computing statistics`, `producing outputs`, `...` ). A `package` is a collection of such `modules` simply based on `directory hierarchy`.

Let start with a basic `package` called `my_package1_basic_graph`.

So, create a directory named [my_package1_basic_graph](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/tree/master/basic_python/Preliminaries/Chapter6/my%20packages) an 2 `module` [basic_graph.py](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/basic_python/Preliminaries/Chapter6/my%20packages/my_package1_basic_graph/basic_graph.py) and [ellipse_shape](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/basic_python/Preliminaries/Chapter6/my%20packages/my_package1_basic_graph/ellipse_shape.py) inside it. 

Moreover, create an empty file [__init__.py](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/basic_python/Preliminaries/Chapter6/my%20packages/my_package1_basic_graph/__init__.py) in `my_package1_basic_graph`. Your directory should be as follows:

    .../ my_package1_basic_graph/
            __init__.py
            basic_graph.py
            ellipse_shape.py

```python
## hidden link!
```

- The special file `__init__.py` is required to treat the directory as a package. This file **can be** left **`empty`** but it can contain code to `initialize the package`. We are not going into details abo0ut this file (see the documentation) and we simply use it to load the `module` `basic_graph` on initialization. 

- So, putting `from .. import basic_graph` in `__init__.py` is sufficient for now. 

- The statement `from . ...` stands for importing directly from current directory!

> **Import a `package` like a `module`**

```python
import my_package1_basic_graph
```

> **`Functions` are in `module's namespace`, nested in `package's namespace` (automated by `__init__.py`)**

```python
my_package1_basic_graph.basic_graph.draw_pine()
```

> **Import just a `module` (with a saving `alias`)**

```python
from my_package1_basic_graph import basic_graph as bg
bg.draw_splited_sqr(20, 'o')
```

> **Importing in `global scope` from a `package`**

```python
from my_package1_basic_graph.basic_graph import draw_rect_boundary

draw_rect_boundary(12, 7, '*', '.')
```

Now, import our 2 modules `basic_graph` and `ellipse_shape` from the `package : my_package1_basic_graph`

```python
import my_package1_basic_graph
from my_package1_basic_graph import basic_graph as bg
from my_package1_basic_graph import ellipse_shape as es

bg.draw_sq_triag(5)
```

With only one module, a package is useless but it becomes more interesting when you collect various modules and subpackages. Consider the follwing file hierarchy:


    .../your_package_name/
        __init__.py
        tools/
            __init__.py
            first_tool_module.py
            an_other_module.py
        processing/
            __init__.py
            method_1.py
            method_2.py
        some_stuff.py
        other_stuff.py
        ...
Filling the `modules` and `__init__.py` files, you will learn how to properly organize a **`Python`** project for your future developments.
