```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
help(cv2.inpaint)
```

```python
img = cv2.imread(r"C:\Users\Admin\Desktop\Data\computer_vision_data\girl_face_1_noise.jpg")
img[:,:,2] = img[:,:,0] 
mask = cv2.imread(r"C:\Users\Admin\Desktop\Data\computer_vision_data\mask_1_noise.jpg", 0)

print('shape of_img_noise = %s \t of mask_noise = %s' %(img.shape, mask.shape))

dst_TELEA_45 = cv2.inpaint(img,mask, 45, cv2.INPAINT_TELEA)
dst_NS_45 = cv2.inpaint(img,mask, 45, cv2.INPAINT_NS)

dst_TELEA_75 = cv2.inpaint(img,mask, 75, cv2.INPAINT_TELEA)
dst_NS_75 = cv2.inpaint(img,mask, 75, cv2.INPAINT_NS)
```

```python
plt.figure(figsize = (20, 16))
plt.subplot(2,3,1), plt.imshow(img), plt.title('noise_image')
plt.subplot(2,3,4), plt.imshow(mask, 'gray'), plt.title('mask_noise')
plt.subplot(2,3,3), plt.imshow(dst_TELEA_45), plt.title('cv2.inpaint using TELEA & radius = 45')
plt.subplot(2,3,2), plt.imshow(dst_NS_45), plt.title('cv2.inpaint using NS & radius = 45')
plt.subplot(2,3,5), plt.imshow(dst_TELEA_75), plt.title('cv2.inpaint using TELEA & radius = 75')
plt.subplot(2,3,6), plt.imshow(dst_NS_75), plt.title('cv2.inpaint using NS & radius = 75')
plt.tight_layout()
plt.show()
```
