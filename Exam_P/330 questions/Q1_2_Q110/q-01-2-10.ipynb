#### Question 01.

A survey of a group’s viewing habits over the last year revealed the following information:
- (i) 28% watched gymnastics
- (ii) 29% watched baseball
- (iii) 19% watched soccer
- (iv) 14% watched gymnastics and baseball
- (v) 12% watched baseball and soccer
- (vi) 10% watched gymnastics and soccer
- (vii) 8% watched all three sports.

Calculate the percentage of the group that watched none of the three sports during the last year.

----------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt sở thích của nhóm người tham gia khảo sát lần lượt là $G, H, S$ ứng với các bộ môn họ theo dõi (`gymnastic`, `baseball` & `soccer`).
- Theo 3 giả thiết đầu tiên, xác suất của mỗi nhóm đơn là
$$ P(G) = 0.28, \quad P(B) = 0.29, \quad P(S) = 0.19 $$
- 3 giả thiết kế tiếp `(iv)`, `(v)` và `(vi)` cho biết thông tin về những người theo dõi cả 2 môn thể thao,
$$ P(G \cap B) = 0.14, \quad P(B \cap S) = 0.12, \quad P(G \cap S) = 0.10$$
- Giả thiết cuối cùng cho biết thông tin số người theo dõi cả 3 môn
$$ P(G \cap B \cap S) = 0.08. $$
- Yêu cầu BT là tìm XS của nhóm đ** quan tâm bộ môn nào, tức là

$$ P((G^c \cap B^c \cap S^c)) = P\left( \left( G \cup B \cup S \right)^c \right) = 1 - P\left( G \cup B \cup S \right) $$
- Mà ta có
$$ \begin{array}{ccl} P\left( G \cup B \cup S \right)  &=& \left( P(B) + P(G) + P(S) \right) - \left( P(G\cap B) + P(G\cap S) + P(B\cap S) \right) + P(G \cap B \cap S) \\ &=& 0.29 + 0.28 + 0.19 - ( 0.14 + 0.10 + 0.12 ) + 0.08 \\ &=& 0.48 \end{array} $$
do đó,
$$ P((G^c \cap B^c \cap S^c)) = 1 - P\left( G \cup B \cup S \right) = 0.52. $$

#### Question 02.

The probability that a visit to a `primary care physician’s (PCP)` office results in neither `lab work` nor referral to a `specialist` is `35%`. Of those coming to a PCP’s office, `30% are referred to specialists` and `40% require lab work`.

Calculate the probability that a visit to a PCP’s office results in both `lab work` and referral to a `specialist`.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $R$, $L$ lần lượt là các biến cố thể hiện mục đích chuyến thăm của `PCP` đến là hướng vào `referral to specialist` hay `lab-work (phòng lab)`.
- Từ các giả thiết về số liệu đề bài, ta được
$$ P(R) = 0.3, \quad P(L) = 0.4, \quad \text{ và } \; P(R^c \cap L^c) = 0.35 $$
- BT yêu cầu tính XS để chuyến thăm đó có liên quan đến cả 2, tức là $P(R \cap L)$. Dễ thấy rằng vì

$$  P(R \cup L) = P(R) + P(L) - P(R \cap L) $$

nên

$$ \begin{array}{cclr} P(R \cap L) &=& P(R) + P(L) - P(R \cup L) \\ &=& P(R) + P(L) - \left( 1 - P\left( \left( R \cup L \right)^c \right) \right) & \text{ (bởi } \left. \left( R \cup L \right) = \Omega - \left( R \cup L \right)^c \right) \\ &=& P(R) + P(L) - 1 + P(R^c \cap L^c) \\ &=& 0.30 + 0.40 - 1 + 0.35 \\ &=& 0.05 \end{array} $$

#### Question 03.

You are given $P[ A \cup B ] = 0.7$ and $P[ A \cup B^c ] = 0.9$. Calculate $P[A]$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Cả 2 dữ kiện $P[ A \cup B ] = 0.7$ và $P[ A \cup B^c ] = 0.9$ đều có điểm giống nhau là phép hội với biến cố $A$. Do đó, để tìm $P(A)$ thì ta phải khai triển và thu gọn các dữ kiện này thông qua $B$ và phần bù của nó $B^c$,
- Ta có
$$ \left\lbrace \begin{array}{ccl} P[A \cup B] &=& P(A) + P(B) - P(A \cap B) \\ P[A \cup B^c] &=& P(A) + P(B^c) - P(A \cap B^c) \end{array} \right. $$
cộng 2 vế của hệ trên, ta được
$$ P[A \cup B] + P[A \cup B^c] = 2 P(A) + \left( P(B) + P(B^c) \right) - \left( P(A \cap B) + P(A \cap B^c) \right) $$
- Thay các số trong giả thiết vào và nhớ rằng $$P(B) + P(B^c) = 1$$ cũng như $$(A \cap B) \cap (A \cap B^c) = \emptyset,$$ nên tổng xác suất 2 biến cố rời nhau sẽ là xác suất của hội 2 biến cố đấy, $$ P(A \cap B) + P(A \cap B^c) = P \left( \left( A \cap B \right) \cup \left( A \cap B^c \right) \right). $$

Ta được

$$ \begin{array}{cccl} & 0.7 + 0.9 &=& 2P(A) + 1 - \left( P(A \cap B) + P(A \cap B^c) \right) \\ \Leftrightarrow & 1.6 &=& 1 + 2P(A) - P\left( \left( A \cap B \right) \cup \left( A \cap B^c \right) \right) \\ \Leftrightarrow & 0.6 &=& 2P(A) - P(A \cap (B \cup B^c)) \\ \Leftrightarrow & 0.6 &=& 2P(A) - P(A) \end{array} $$

Vậy $P(A) = 0.6$

#### Question 04.

An urn contains 10 balls: 4 red and 6 blue. A second urn contains 16 red balls and an unknown number of blue balls. A single ball is drawn from each urn. The probability that both balls are the same color is 0.44.

Calculate the number of blue balls in the second urn. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $R_1, R_2$ lần lượt là sự kiện lấy được bóng đỏ trong bình thứ nhất và bình thứ hai; tương tự, $U_1, U_2$ là cho bóng xanh trong 2 bình đó.
- Theo giả thiết về bình thứ nhất, có 4 đỏ, 6 xanh nên $$P(R_1) = \frac{4}{10} = 0.4 \text{ và } P(B_1) = 0.6 $$
- Đặt số bóng xanh trong bình thứ hai là $x$, và do bình thứ hai có 16 banh đỏ và đ biết số bóng xanh, do đó
$$ P(R_2) = \dfrac{16}{16 + x} \text{ và } P(B_2) = \dfrac{x}{16 + x} $$
- Giả thiết cuối cùng cho biết xác suất lây được bóng cùng màu trong mỗi bình (sau 1 lượt duy nhất) là 0.44, tức là có 2 trường hợp: cùng xanh (tức $B_1 \cap B_2$) hoặc cùng đỏ (là $R_1 \cap R_2$), do đó

$$ P \left[ \left( R_1 \cap R_2 \right) \cup \left( B_1 \cap B_2 \right) \right] = 0.44 $$

- Hơn nữa, sự kiện 2 bình mỗi bình rút được 1 bóng được cùng màu xanh $(B_1 \cap B_2)$ là độc lập với sự kiện cùng màu đỏ $(R_1 \cap R_2)$, do đó

$$ P \left( R_1 \cap R_2 \right) + P \left( B_1 \cap B_2 \right) = 0.44 $$

- Tiếp theo, khả năng lấy mẫu ở mỗi bình là độc lập với nhau, tức là $P \left( R_1 \cap R_2 \right) = P\left( R_1 \right) P\left( R_2 \right)$ và
$P \left( B_1 \cap B_2 \right) = P\left( B_1 \right) P\left( B_2 \right)$, do đó,

$$ \begin{array}{clcr} & P \left[ \left( R_1 \cap R_2 \right) \cup \left( B_1 \cap B_2 \right) \right] &=& 0.44 \\  \Leftrightarrow & P(R_1) P(R_2) + P(B_1) P(B_2) &=& 0.44 \\ \Leftrightarrow & \dfrac{4}{10} \cdot \dfrac{16}{x + 16} + \dfrac{6}{10} \cdot \dfrac{x}{x + 16} &=& 0.44 \\ \Leftrightarrow & \qquad \dfrac{3x + 32}{x + 16} &=& 2.2  \end{array} $$

Vậy $x = 4.$

#### Question 05.

An auto insurance company has 10,000 policyholders. Each policyholder is classified as
- (i) `Age`: young or old;
- (ii) `Gender`: male or female; and
- (iii) `Marital status`: married or single.

Of these policyholders, 3000 are young, 4600 are male, and 7000 are married. The policyholders can also be classified as 1320 young males, 3010 married males, and 1400 young married persons. Finally, 600 of the policyholders are young married males.

Calculate the number of the company’s policyholders who are young, female, and single. 

--------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A, G, M$ lần lượt tương ứng với 3 `nhóm tính chất` được khảo sát là `Age, Gender` và `marital status`, do đó về số phần tử trong nó là như nhau, tức là $$|A| = |M| = |G| = 10000;$$ trong đó, ta ký hiệu $|X|$ là số phần tử trong tập $X$
- Mỗi nhóm này đều có 2 mức và ta sẽ quy ước 

$\qquad \diamond$ $A_1:$ `young` và $A_2:$ `old`

$\qquad \diamond$ $G_1:$ `female` và $G_2:$ `male`

$\qquad \diamond$ $M_1:$ `single` và $M_2:$ `married`

- Theo khảo sát (về số người), ta có 

$\qquad \diamond |A_1| = 3000, |G_2| = 4600, |M_2| = 7000$ 

$\qquad \diamond |A_1 \cap G_2| = 1320, |G_2 \cap M_2| = 3010, |A_1 \cap M_2| = 1400$ và 

$$ | A_1 \cap G_1 \cap M_2 | = 600 $$

- BT yêu cầu tính số người thỏa $A_1 \cap G_1 \cap M_1$, lưu ý rằng

$$ (A =) A_1 \cup A_2 = (G =) G_1 \cup G_2 = (M =) M_1 \cup M_2 = \Omega $$

ta có

$$ \begin{array}{ccl} | A_1 \cap G_1 \cap M_1 | &=& \left\vert \left( A_1 \cap G_1 \cap M \right) - \left( A_1 \cap G_1 \cap M_2 \right) \right\vert \\ &=& |A_1 \cap G_1| - |A_1 \cap G_1 \cap M_2| \\ &=& \left( |A_1| - |A_1 \cap G_2| \right) - \left( |A_1 \cap M_2| - |A_1 \cap M_2 \cap G_2| \right) \\ &=& (3000 - 1320) - (1400 - 500) = 880 \end{array} $$

#### Question 06.

- A public health researcher examines the medical records of a group of `937` men who died in 1999 and discovers that `210` of the men died from causes related to heart disease.

- Moreover, `312` of the `937` men had at least one parent who suffered from heart disease, and, of these `312` men, `102` died from causes related to heart disease.

Calculate the probability that a man randomly selected from this group died of causes
related to heart disease, given that neither of his parents suffered from heart disease.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $H$ là sự kiện mà người được khảo sát chết vì bệnh tim.
- $F$ là sự kiện có ít nhất cha hoặc mẹ của người đó từng mắc bệnh tim.
- Theo giả thiết, ta có
$$ |\Omega| = 937, \quad |H| = 210, \quad |F| = 312, \quad \text{ và } \quad |H \cap F| = 102, $$
- Yêu cầu BT là tính XSCĐK: `"chọn ngẫu nhiên một người trong nhóm chết vì bệnh tim mạch`, `biết rằng cha mẹ của người này chưa từng bị bệnh tim"`, tức là

$$ P \left[ \left. H_{} \right| F^c \right] = \dfrac{P(H \cap F^c)}{P(F^c)} = \dfrac{ \dfrac{|H| - |H \cap F|}{|\Omega|} }{ \dfrac{ |\Omega| - |F| }{|\Omega |} } = \dfrac{|H| - |H \cap F|}{ |\Omega| - |F| } = \dfrac{210 - 102}{937 - 312} = 0.173 $$

#### Question 07.

- An insurance company estimates that `40%` of policyholders who have `only` an `auto policy` will renew next year and `60%` of policyholders who have only a `homeowners` policy will renew next year. 

- The company estimates that `80%` of policyholders who have `both an auto policy and a homeowners` policy will renew at least one of those policies next year.
 
- Company records show that `65%` of policyholders have an `auto policy`, `50% of policyholders` have a homeowners policy, and `15%` of policyholders have `both an auto policy and a homeowners policy`.

Using the company’s estimates, calculate the percentage of policyholders that will renew at least one policy next year.

-------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A$ và $H$ lần lượt là các sự kiện về hợp đồng BH đó là `auto` hay `homeowner`
- Từ các giả thiết trên, ta được bảng phân phối xác suất và giá trị như sau

| policies $(S)$ | `auto & not-homeowner` | `not-auto & homeowner` | `auto & homeowner` |
|-|:-:|:-:|:-:|
|probabilities | `P(A and H^c)` | `P(A^c and H)` | `P(A and H)` |
| corresponding values/proportions (theo 2 GT đầu tiên) | 0.4 | 0.6 | 0.8 |

mà theo GT thứ 3, ta được

$$ \begin{array}{ccl} P(A \cap H) &=& 0.15 \\ P(A \cap H^c) &=& P(A) - P(A \cap H) = 0.65 - 0.15 = 0.5 \\ P(A^c \cap H) &=& P(H) - P(A \cap H) = 0.5 - 0.15 = 0.35 \end{array} $$

Từ đó ta được

$$ \mathbb{E}S = 0.4\cdot P(A \cap H^c) + 0.6 \cdot P(A^c \cap H) + 0.8 \cdot P(A \cap H) = 0.53 $$

#### Question 08.

- Among a large group of patients recovering from shoulder injuries, it is found that 22% visit both a physical therapist and a chiropractor, whereas 12% visit neither of these. 

- The probability that a patient visits a chiropractor exceeds by 0.14 the probability that a patient visits a physical therapist.

Calculate the probability that a randomly chosen member of this group visits a physical therapist.

-------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ và $C$ lần lượt là sự kiện nếu bệnh nhân đó phải thực hiện điều trị `Vật lý trị liệu (physical "therapist")` hoặc `chấn thương chỉnh hình ("chiropractor")`.
- Theo giả thiết đầu tiên của đề bài, ta có
$$ P(C \cap T) = 0.22 \quad \text{ trong khi } \quad P(T^c \cap C^c) = 0.12 $$
Hơn nữa, giả thiết sau cùng khẳng định xác suất số người thực hiện điều trị chỉnh hình vượt 0.14 so với nhóm thực hiện VLTL, tức là
$$ P(C) = P(T) + 0.14 $$
- BT yêu cầu tính $P(T)$, và để làm điều đó, chú ý biến đổi sau:

$$ 1 - P(T^c \cap C^c) = P(T \cup C) = P(T) + P(C) - P(T \cap C) $$
tương đương với
$$ 1 - 0.12 = P(T) + (P(T) + 0.14) - 0.22 \Leftrightarrow 0.88 = 2P(T) - 0.08 \Leftrightarrow P(T) = 0.48. $$

#### Question 09.

An insurance company examines its pool of auto insurance customers and gathers the following information:
- (i) All customers insure at least one car.
- (ii) 70% of the customers insure more than one car.
- (iii) 20% of the customers insure a sports car.
- (iv) Of those customers who insure more than one car, 15% insure a sports car.

Calculate the probability that a randomly selected customer insures exactly one car and that car is not a sports car.

-------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $M$ và $S$ lần lượt là các sự kiện "khách hàng có nhiều hơn 1 chiếc xe được BH" và "khách hàng mua BH cho một chiếc xe thể thao".
- Các giả thiết (ii), (iii) và (iv) lần lượt cho biết

$$ P(M) = 0.7, \quad P(S) = 0.2 ,\quad P(S | M) = 0.15 $$
- BT yêu cầu tính XS để KH không mua loại xe nào trong 2 loại xe đã kể đến, tức là 

$$P(M^c \cap S^c)$$

Theo luật DeMorgan’s, ta có $(M^c \cap S^c) = (M \cup S)^c$ và do đó

$$ \begin{array}{ccl} P(M^c \cap S^c) &=& 1 - P(M \cup S) \\ &=& 1 - \left[ P(M) + P(S) - P(M \cap S) \right] \\ &=& 1 - \left[ P(M) + P(S) - P(S | M) P(M) \right] = 0.205. \end{array} $$
