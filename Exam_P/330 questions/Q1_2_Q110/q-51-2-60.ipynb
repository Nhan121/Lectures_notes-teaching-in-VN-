#### Question 51.

- A manufacturer’s annual losses follow a distribution with density function 
$$ f(x) = \left\lbrace \begin{array}{ccl} \dfrac{2.5 (0.6)^{2.5}}{x^{3.5}} & \text{ if } & x > 0.6 \\ 0, & & \text{otherwise} \end{array} \right. $$
- To cover its losses, the manufacturer purchases an insurance policy with an annual deductible of 2.

Calculate the mean of the manufacturer’s annual losses not paid by the insurance policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện mức tổn thất hàng năm mà bảo hiểm chi trả, $X$ có hàm MĐXS $f$ như đề.
- Đề bài yêu cầu tính kỳ vọng chi phí mà chính sách BH phải chi trả với khấu hao hàng năm bằng 2, tức là $T = \min(X, 2)$ do đó

$$ \begin{array}{ccl} \mathbb{E} T &=& \displaystyle \int_{\mathbb{R}} \min(t, 2) f(t) dt \\ &=& \displaystyle \int_{0.6}^2 t \cdot \dfrac{2.5 (0.6)^{2.5}}{t^{3.5}} dt + \int_{2}^{\infty} 2 \cdot \dfrac{2.5 (0.6)^{2.5}}{t^{3.5}} dt \approx 0.9343 \end{array} $$

#### Question 52.
- An insurance company sells a one-year automobile policy with a `deductible` of 2. 

- The probability that the insured will incur a loss is 0.05. 

- If there is a loss, the probability of a loss of amount N is K/N, for N = 1, . . . , 5 and K a constant. These are the only possible loss amounts and no more than one loss can occur.

Calculate the expected payment for this policy. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là BNN thể hiện số tổn thất hàng năm của BH tai nạn xe cộ, theo giả thiết $N$ chỉ nhận $5$ giá trị với phân phối xác suất như sau.

| N | 1 | 2 | 3 | 4 | 5 | 
|-|:-:|:-:|:-:|:-:|:-:|
|`Prob(N=n)`| `K` | `K/2` | `K/3` | `K/4` | `K/5` |

Vì tổng các xác suất ứng với tất cả các giá trị của môt BNN rời rạc phải bằng 1, nên

$$ \sum_{n=1}^5 \dfrac{K}{n} = 1 \Leftrightarrow K \sum_{n=1}^5 \dfrac{1}{n} = 1 \Leftrightarrow K = \dfrac{60}{137} $$

- Giả thiết còn lại trong bài cho biết xác suất khi có tai nạn xảy ra là $0.05$.

- Cuối cùng, để thực hiện yêu cầu BT, tức là tìm kỳ vọng số tiền phải trả hàng năm khi có tai nạn xảy ra; sau khi khấu hao với $2$, tức là

$$ \mathbb{E}\max\left(0,  N - 2 \right) = \sum_{n \geq 3} (n - 2) \left( 0.05 \cdot P \left[ N = n \right] \right) = 0.05 \cdot K \cdot \left[ \sum_{n=3}^5 \left( \dfrac{n-2}{n}\right) \right] = 0.0314. $$

#### Question 53.

An insurance policy reimburses a loss up to a benefit limit of 10. The policyholder’s loss, Y, follows a distribution with density function: 

$$ f(y) = \left\lbrace \begin{array}{ccl} 2y^{-3} & \text{ if } & y > 1 \\ 0, & & \text{otherwise} \end{array} \right. $$

Calculate the expected value of the benefit paid under the insurance policy. 

-----------------------------

**Giải đáp / SOLUTIONs.**

Đây tương tự như BT 51, cũng là câu hỏi lý thuyết về tìm kỳ vọng của $T = \min(Y, 10)$ với $Y$ là BNN được mô tả trong đề bài, khi đó:

$$ \begin{array}{ccl} \mathbb{E} T &=& \displaystyle \int_{\mathbb{R}} \min(t, 10) f(t) dt \\ &=& \displaystyle \int_{1}^{10} t \cdot \dfrac{2}{t^{3}} dt + \int_{10}^{\infty} 10 \cdot \dfrac{2}{t^{3}} dt = 1.9 \end{array} $$

#### Question 54.

- An auto insurance company insures an automobile worth 15,000 for one year under a policy with a 1,000 deductible. 
- During the policy year there is a 0.04 chance of partial damage to the car and a 0.02 chance of a total loss of the car. 
- If there is partial damage to the car, the amount $X$ of damage (in thousands) follows a distribution with density function

$$ f(x) = \left\lbrace \begin{array}{ccl} 0.5003 e^{-x/2} & \text{ if } & x \in (0, 15) \\ 0, & & \text{otherwise} \end{array} \right. $$

Calculate the expected claim payment. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện mức độ chi trả cho các thiệt hại liên quan đến riêng các tai nạn từ xe cộ (`automobile`) và theo giả thiết đầu tiên lẫn cuối cùng, $Y$ có 3 trường hợp đáng quan tâm như sau

| Cases | `partial damage to the car` | ` total loss of car ` | `no related car (otherwise)` |
|-|:-:|-|:-:|
|Purchased_fee: $\mathbb{E} Y$| $\mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right] $ | $15 - 1$ = $14$ | $0$ |

trong đó $X$ là BNN thể hiện số tiền thiệt hại liên quan đến xe hơi được tính theo đơn vị hàng nghìn.

- Đặt $F$ là mức phí mà BH sẽ chi trả cho các tai nạn liên quan đến xe cộ trong năm, và kết hợp với giả thiết thứ hai theo đó nó có phân phối xác suất như sau

| Cases | `partial damage to the car` | ` total loss of car ` | `no related car (otherwise)` |
|-|:-:|:-:|:-:|
| Prob (`change of case`) | $0.04$ | $0.02$ | $1 - 0.04 - 0.02 = 0.94$ |
| Purchased_fee (in thousand) | $\mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right]$ | $15 - 1$ = $14$ | $0$ |

- Do đó, tổng phí chi trả sẽ được tính bởi

$$ \begin{array}{ccl} \mathbb{E} F &=& 0.04 \cdot \mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right] + 0.02 \cdot 14 + 0.94 \cdot 0 \\ &=& \displaystyle 0.04 \cdot \left( \int_{1}^{15} (x-1) f(x) dx \right) + 0.28 + 0 \\ &=& 0.04\cdot \displaystyle 0.5003 \cdot \left( \int_{1}^{15} (x-1) \cdot e^{-x/2} dx \right) + 0.28 \\ &=& 0.020012 \cdot \left( 2.408 \right) + 0.28 \\ &=& 0.328 \end{array} $$

Sau đó chuyển đơn vị hàng nghìn này về hàng đơn vị và ta được mức phí chi trả phải tính là $328$ (đồng)

#### Question 55.

An insurance company’s monthly claims are modeled by a continuous, positive random variable $X$, whose probability density function is proportional to $(1 + x)^{- 4} \mathbf{1}_{ \lbrace x \in (0, \infty) \rbrace }.$

Calculate the company’s expected monthly claims.

------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Do BNN $X$ có hàm mật độ $f$ tỷ lệ với $(1 + x)^{- 4} \mathbf{1}_{ \lbrace x \in (0, \infty) \rbrace }$ nên tồn tại một hằng số $C$ sao cho $f(x) = C(1 + x)^{- 4} \mathbf{1}_{ \lbrace x \in (0, \infty) \rbrace }$ thỏa

$$ \left \lbrace \begin{array}{cclr} f(x) & \geq & 0, & \forall x \\ \displaystyle \int_{\mathbb{R}} f(x) dx &=& 1 \end{array} \right.$$ 

tương đương với

$$ \Leftrightarrow \left \lbrace \begin{array}{cclr} C (1 + x)^{- 4} \mathbf{1}_{ \lbrace x \in (0, \infty) \rbrace } & \geq & 0, & \forall x \\ \displaystyle C \int_{0}^{\infty} (1 + x)^{- 4} dx &=& 1 \end{array} \right. \Leftrightarrow C = 3$$

- Đề bài yêu cầu tính kỳ vọng của nó, tức là

$$ \mathbb{E}X = \displaystyle \int_{\mathbb{R}} x f(x) dx = \int_0^{\infty} x \dfrac{3}{(1 + x)^4} dx = \int_1^{\infty} \dfrac{3(u-1)}{u^4} du $$

dấu bằng cuối cùng có được bởi cách đổi biến $u = x + 1$, sau đó bằng tích phân từng phần, ta được

$$ \mathbb{E}X = \left. \left( \dfrac{-3}{u^2} - \dfrac{1}{u^3} \right) \right\vert_1^{\infty} = \lim_{r \to \infty} \left. \left( \dfrac{-3}{u^2} - \dfrac{1}{u^3} \right) \right\vert_1^r = \dfrac{1}{2} $$

#### Question 56.

- An insurance policy is written to cover a loss, $X$, where $X$ has a uniform distribution on $[0, 1000]$. 
- The policy has a `deductible`, $d$, and the expected payment under the policy is `25%` of what it would be with `no deductible`.

Calculate $d$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:

- Như đã biết, một BNN có phân phối đều $\mathcal{U}([a, b])$ sẽ có hàm MDXS $f$ xác định bởi

$$ f(x) = \dfrac{1}{b - a} \mathbf{1}_{\lbrace x \in [a, b] \rbrace} $$

- Theo giả thiết đầu tiên, $X$ là BNN có phân phối đều $\mathcal{U}([0, 1000])$ nên có kỳ vọng được tính bởi 

$$ \mathbb{E}X = \int_{\mathbb{R}} x \frac{1}{1000} \mathbf{1}_{\lbrace x \in [0, 1000] \rbrace} dx = \dfrac{1}{1000} \int_0^{1000} x dx = 500 $$

- Mức khấu hao của chính sách này $d$ với $d > 0$ được kỳ vọng bởi $25\%$ giá trị kỳ vọng $\mathbb{E}(X)$ khi không có khấu hao ($d = 0$), do đó ta cần tìm $d$ thỏa phương trình

$$ \mathbb{E}(X - d) = 0.25 \cdot \mathbb{E}X = 125. $$

như vậy, ta cần tìm $d$ thỏa phương trình

$$ \int_{0}^{1000} \frac{(x-d)}{1000} dx = 125 \Leftrightarrow \left. \left( \frac{(x - d)^2}{2} \right) \right\vert_0^{1000} = 125000 \Leftrightarrow d = 500. $$

#### Question 57.

An actuary determines that the claim size for a certain class of accidents is a random variable, $X$, with `moment generating function`

$$ M_X(t) = \dfrac{1}{(1 - 2500 t)^4}. $$

Calculate the standard deviation of the claim size for this class of accidents.

-------------------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Nhắc lại rằng nếu BNN $X$ có hàm sinh moment $M_X(t)$ thì

$$ \mathbb{E}X = M'_X(0) \quad \text{ và } \quad \mathbb{E}X^2 = M_X^{''}(0) $$

- Do đó, giờ ta sẽ tính các đạo hàm cấp 1 và cấp 2 của hàm sinh moment $M_X(t)$, ta có

$$ M'_X(t) = \dfrac{d}{dt} \left( \frac{1}{(1 - 2500 t)^4} \right) = \dfrac{-4\cdot \left[ -2500 \cdot (t - 2500)^3 \right] }{\left( (1 - 2500 t)^4 \right)^2} = \frac{10000}{(1 - 2500 t)^5} $$

suy ra $\mathbb{E}X = M'_X(0) = 10000.$ Tương tự,

$$ M^{''}_X(t) = \dfrac{d^2}{dt^2} \left( \frac{1}{(1 - 2500 t)^4} \right) = \dfrac{d}{dt} \left( \frac{10000}{(1 - 2500 t)^5} \right) = \dfrac{125000000}{(1 - 2500t)^6} $$

suy ra $\mathbb{E} X^2 = M^{''}_X(0) = 125000000.$

- Như vậy, độ lệch chuẩn $sd_X$ sẽ được tính bởi

$$ sd_X = \sqrt{\text{Var} X} = \sqrt{ \left[ \mathbb{E}X^2 \right] - \left[ \mathbb{E}X \right]^2 } = \sqrt{125000000 - 10000^2} = 5000. $$

#### Question 58.

A company insures homes in three cities, $J, K$, and $L$. Since sufficient distance separates the cities, it is reasonable to assume that the losses occurring in these cities are mutually independent.

The moment generating functions for the loss distributions of the cities are: 

$$ M_J(t) = (1-2t)^{-3}, \quad M_K(t) = (1 - 2t)^{-2.5}, \quad M_L(t) = (1 - 2t)^{-4.5} $$

Let $X$ represent the combined losses from the three cities.

Calculate $\mathbb{E}(X^3)$ .


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài cho các BNN $J, K, L$ lần lượt thể hiện số tiền mà BH phải chi trả cho 3 thành phố $J, K, L$.
- Câu hỏi của bài là khi $X$ là tính tổn thất ở cả 3 thành phố trên, tức là

$$ X = J + K + L $$

- Theo ý cuối cùng trong giả thiết đầu tiên thì 3 BNN này là độc lập với nhau, do đó hàm sinh moment của $X$ sẽ được tính bởi tích 3 hàm sinh moment của $J, K, L$, bởi

$$ M_X(t) = \mathbb{E} e^{Xt} = \mathbb{E} e^{(J + K + L)t} = \mathbb{E} \left[ e^{Jt} \cdot e^{Kt} e^{Lt} \right] = \mathbb{E} e^{Jt} \cdot \mathbb{E} e^{Jt} \cdot \mathbb{E} e^{Lt} = M_J(t) \cdot M_K(t) \cdot M_L(t)$$

trong đó, dấu bằng đầu tiên là định nghĩa của hàm sinh moment bằng kỳ vọng và dấu bằng kế cuối là do 3 BNN này độc lập với nhau.

- Như vậy, thay các giá trị hàm sinh moment của $J,K,L$ vào và ta được 

$$M_X(t) = (1 - 2t)^{-3 + (-2.5) + (-4.5)} = (1-2t)^{-10}.$$ 

- Cuối cùng, để tính $\mathbb{E}(X^3)$ thì ta lại tính đạo hàm cấp 3 của $M_X(t)$ tại $t = 0$, ta có

$$ M^{(3)}_X(t) = \dfrac{d^3}{dt^3} M_X(t) = \dfrac{d^3}{dt^3} \left( \dfrac{1}{(1 - 2t)^{10}} \right) = \dfrac{2^3\cdot 10 \cdot 11 \cdot 12}{(1 - 2t)^{10 + 3}} $$

do đó

$$ \mathbb{E}X^3 = M^{(3)}_X(0) = 10560.$$

#### Question 59.

An insurer's annual weather-related loss, X, is a random variable with density function 

$$ f(x) = \left\lbrace \begin{array}{lcr} \dfrac{2.5 (200)^{2.5}}{x^{3.5}} & \text{ if } & x > 200 \\ 0, & & \text{otherwise} \end{array} \right.$$

Calculate the difference between the 30th and 70th percentiles of $X$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Trước tiên, ta ký hiệu $\alpha_p = q_p(X)$ là mức phân vị thứ $p$ (với $p \in [0, 100]$) của một BNN $X$ nào đó.
- Khi đó, theo định nghĩa của phân vị, thì

$$ \dfrac{p}{100} = P(X < \alpha_p) $$

- Vì $X$ có hàm MDXS $f$ như đề nên ta có phương trình liên hệ phân vị $\alpha_p$ với mức $p$ như sau

$$ \dfrac{p}{100} = \int_{-\infty}^{\alpha_p} f(x) dx = \int_{200}^{\alpha_p} \dfrac{2.5 (200)^{2.5}}{x^{3.5}} dx = 1 - \dfrac{(200)^{2.5}}{\alpha_p^{2.5}} $$

- Đề bài yêu cầu tính hiệu số giữa 2 mức $p=70$ và $p=30$, do đó ta sẽ lần lượt tìm $\alpha_{30}$ và $\alpha_{70}$. Tổng quát, ta có

$$ \frac{p}{100} = 1 - \left( \dfrac{200}{\alpha_p} \right)^{2.5} \Leftrightarrow \left( \dfrac{200}{\alpha_p} \right) = \left( 1 - \frac{p}{100} \right)^{1/2.5} \Leftrightarrow \alpha_{p} = \dfrac{200}{\left( 1 - \frac{p}{100} \right)^{0.4}} $$

Suy ra

$$ \alpha_{70} - \alpha_{30} = \dfrac{200}{\left( 1 - 0.7 \right)^{0.4}} - \dfrac{200}{\left( 1 - 0.3 \right)^{0.4}} \approx 93.06 $$

#### Question 60.

- A recent study indicates that the annual cost of maintaining and repairing a car in a town in `Ontario` `averages 200` with a `variance of 260`.
- A tax of 20% is introduced on all items associated with the maintenance and repair of cars (i.e., everything is made 20% more expensive).

Calculate the variance of the annual cost of maintaining and repairing a car after the tax is introduced. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện phí bảo trì và sửa chửa xe hơi tạo `Ontario`, theo giả thiết đầu tiên thì $\mathbb{E}X = 200$ và Var$X = 260$.
- Đặt $Y$ là BNN thể hiện giá tiền sau thuế của $X$, và theo giả thiết thứ hai thì

$$ Y = (1 + 0.2) X = 1.2 \cdot X $$

- Yêu cầu đề bài là tìm Var$Y$, do đó

$$ \text{Var}Y = \text{Var}(1.2 \cdot X) = 1.2^2 \cdot \text{Var}X = 1.44 \cdot 260 = 374.4.$$
