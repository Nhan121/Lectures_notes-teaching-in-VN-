#### Question 51.

- A manufacturer’s annual losses follow a distribution with density function 
$$ f(x) = \left\lbrace \begin{array}{ccl} \dfrac{2.5 (0.6)^{2.5}}{x^{3.5}} & \text{ if } & x > 0.6 \\ 0, & & \text{otherwise} \end{array} \right. $$
- To cover its losses, the manufacturer purchases an insurance policy with an annual deductible of 2.

Calculate the mean of the manufacturer’s annual losses not paid by the insurance policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện mức tổn thất hàng năm mà bảo hiểm chi trả, $X$ có hàm MĐXS $f$ như đề.
- Đề bài yêu cầu tính kỳ vọng chi phí mà chính sách BH phải chi trả với khấu hao hàng năm bằng 2, tức là $T = \min(X, 2)$ do đó

$$ \begin{array}{ccl} \mathbb{E} T &=& \displaystyle \int_{\mathbb{R}} \min(t, 2) f(t) dt \\ &=& \displaystyle \int_{0.6}^2 t \cdot \dfrac{2.5 (0.6)^{2.5}}{t^{3.5}} dt + \int_{2}^{\infty} 2 \cdot \dfrac{2.5 (0.6)^{2.5}}{t^{3.5}} dt \approx 0.9343 \end{array} $$

#### Question 52.
- An insurance company sells a one-year automobile policy with a `deductible` of 2. 

- The probability that the insured will incur a loss is 0.05. 

- If there is a loss, the probability of a loss of amount N is K/N, for N = 1, . . . , 5 and K a constant. These are the only possible loss amounts and no more than one loss can occur.

Calculate the expected payment for this policy. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là BNN thể hiện số tổn thất hàng năm của BH tai nạn xe cộ, theo giả thiết $N$ chỉ nhận $5$ giá trị với phân phối xác suất như sau.

| N | 1 | 2 | 3 | 4 | 5 | 
|-|:-:|:-:|:-:|:-:|:-:|
|`Prob(N=n)`| `K` | `K/2` | `K/3` | `K/4` | `K/5` |

Vì tổng các xác suất ứng với tất cả các giá trị của môt BNN rời rạc phải bằng 1, nên

$$ \sum_{n=1}^5 \dfrac{K}{n} = 1 \Leftrightarrow K \sum_{n=1}^5 \dfrac{1}{n} = 1 \Leftrightarrow K = \dfrac{60}{137} $$

- Giả thiết còn lại trong bài cho biết xác suất khi có tai nạn xảy ra là $0.05$.

- Cuối cùng, để thực hiện yêu cầu BT, tức là tìm kỳ vọng số tiền phải trả hàng năm khi có tai nạn xảy ra; sau khi khấu hao với $2$, tức là

$$ \mathbb{E}\max\left(0,  N - 2 \right) = \sum_{n \geq 3} (n - 2) \left( 0.05 \cdot P \left[ N = n \right] \right) = 0.05 \cdot K \cdot \left[ \sum_{n=3}^5 \left( \dfrac{n-2}{n}\right) \right] = 0.0314. $$

#### Question 53.

An insurance policy reimburses a loss up to a benefit limit of 10. The policyholder’s loss, Y, follows a distribution with density function: 

$$ f(y) = \left\lbrace \begin{array}{ccl} 2y^{-3} & \text{ if } & y > 1 \\ 0, & & \text{otherwise} \end{array} \right. $$

Calculate the expected value of the benefit paid under the insurance policy. 

-----------------------------

**Giải đáp / SOLUTIONs.**

Đây tương tự như BT 51, cũng là câu hỏi lý thuyết về tìm kỳ vọng của $T = \min(Y, 10)$ với $Y$ là BNN được mô tả trong đề bài, khi đó:

$$ \begin{array}{ccl} \mathbb{E} T &=& \displaystyle \int_{\mathbb{R}} \min(t, 10) f(t) dt \\ &=& \displaystyle \int_{1}^{10} t \cdot \dfrac{2}{t^{3}} dt + \int_{10}^{\infty} 10 \cdot \dfrac{2}{t^{3}} dt = 1.9 \end{array} $$

#### Question 54.

- An auto insurance company insures an automobile worth 15,000 for one year under a policy with a 1,000 deductible. 
- During the policy year there is a 0.04 chance of partial damage to the car and a 0.02 chance of a total loss of the car. 
- If there is partial damage to the car, the amount $X$ of damage (in thousands) follows a distribution with density function

$$ f(x) = \left\lbrace \begin{array}{ccl} 0.5003 e^{-x/2} & \text{ if } & x \in (0, 15) \\ 0, & & \text{otherwise} \end{array} \right. $$

Calculate the expected claim payment. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện mức độ chi trả cho các thiệt hại liên quan đến riêng các tai nạn từ xe cộ (`automobile`) và theo giả thiết đầu tiên lẫn cuối cùng, $Y$ có 3 trường hợp đáng quan tâm như sau

| Cases | `partial damage to the car` | ` total loss of car ` | `no related car (otherwise)` |
|-|:-:|-|:-:|
|Purchased_fee: $\mathbb{E} Y$| $\mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right] $ | $15 - 1$ = $14$ | $0$ |

trong đó $X$ là BNN thể hiện số tiền thiệt hại liên quan đến xe hơi được tính theo đơn vị hàng nghìn.

- Đặt $F$ là mức phí mà BH sẽ chi trả cho các tai nạn liên quan đến xe cộ trong năm, và kết hợp với giả thiết thứ hai theo đó nó có phân phối xác suất như sau

| Cases | `partial damage to the car` | ` total loss of car ` | `no related car (otherwise)` |
|-|:-:|:-:|:-:|
| Prob (`change of case`) | $0.04$ | $0.02$ | $1 - 0.04 - 0.02 = 0.94$ |
| Purchased_fee (in thousand) | $\mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right]$ | $15 - 1$ = $14$ | $0$ |

- Do đó, tổng phí chi trả sẽ được tính bởi

$$ \begin{array}{ccl} \mathbb{E} F &=& 0.04 \cdot \mathbb{E} \left[ (X - 1)\textbf{1}_{X \geq 1} \right] + 0.02 \cdot 14 + 0.94 \cdot 0 \\ &=& \displaystyle 0.04 \cdot \left( \int_{1}^{15} (x-1) f(x) dx \right) + 0.28 + 0 \\ &=& 0.04\cdot \displaystyle 0.5003 \cdot \left( \int_{1}^{15} (x-1) \cdot e^{-x/2} dx \right) + 0.28 \\ &=& 0.020012 \cdot \left( 2.408 \right) + 0.28 \\ &=& 0.328 \end{array} $$

Sau đó chuyển đơn vị hàng nghìn này về hàng đơn vị và ta được mức phí chi trả phải tính là $328$ (đồng)
