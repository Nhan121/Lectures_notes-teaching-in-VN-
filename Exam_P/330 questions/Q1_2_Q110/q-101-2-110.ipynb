#### Question 101, 
The profit for a new product is given by $Z = 3X – Y - 5$. $X$ and $Y$ are independent random variables with $\text{Var}X = 1$ and $\text{Var}Y = 2$. Calculate $\text{Var}(Z).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta có $$ \text{Var}Z = \text{Var}(3X-Y-5)=\text{Var}(3X-Y) $$
- Vì $X, Y$ độc lập nên $\text{Var}(aX + bY) = a^2\text{Var}X + b^2\text{Var}Y$; suy ra

$$ \text{Var}Z = (3)^2 \cdot \text{Var}(X) + (-1)^2 \cdot \text{Var}Y = 11.$$

#### Question 102.
A company has two electric generators. The time until failure for each generator follows an exponential distribution with mean 10. The company will begin using the second generator immediately after the first one fails.

Calculate the variance of the total time that the generators produce electricity

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là BNN thể hiện thời gian của các máy phát điện thứ nhất và thứ hai. Theo giả thiết thì cả $X, Y$ đều độc lập với cùng phân phối mũ $\mathcal{E}(10)$; do đó
$$ \text{Var}X = \text{Var}Y = 10^2 = 100 $$
- BT yêu cầu tính $\text{Var}(X+Y)$, ta có

$$ \text{Var}(X+Y) = \text{Var}X + \text{Var}Y = 200. $$

#### Question 103.
In a small metropolitan area, annual losses due to storm, fire, and theft are assumed to be mutually independent, exponentially distributed random variables with respective means 1.0, 1.5, and 2.4.

Calculate the probability that the maximum of these losses exceeds 3. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $S, F, T$ lần lượt là các BNN thể hiện số cơn bão (`storm`), vụ cháy (`fire`) và trộm cắp (`theft`).
- Giả thiết cho biết chúng lần lượt tuân theo phân phối mũ với các trung bình $1, 1.5$ và $2.4$; cụ thể

$$ S \sim \mathcal{E}(1), F \sim \mathcal{E}(1.5), T \sim \mathcal{E}(2.4)$$
và lưu ý rằng
$$ P[\mathcal{E}(\lambda) \leq t] = 1 - e^{-t /\lambda}, \quad \forall t \geq 0$$
- Đặt $$ Y = \max(S, F, T) $$ là số tai nạn lớn nhất đến từ 3 vấn đề :`"storm", "fire", "theft"` và yêu cầu BT là tính $P[Y > 3]$; ta có

$$ \begin{array}{ccl} P[Y > 3] &=& 1 - P[Y \leq 3] \\ &=& 1 - P[\max(S, F, T) \leq 3] \\ &=& 1 - P[S \leq 3]P[F \leq 3]P[T \leq 3]\\ &=& 1 - \left( 1 - e^{-3/1} \right)\left( 1 - e^{-3/(1.5)} \right)\left( 1 - e^{-3/(2.4)} \right) \\ &\approx & 0.414. \end{array} $$

#### Question 104.
A joint density function is given by 

$$ f(x,y) = \left\lbrace \begin{array}{cl} kx, & x\in (0,1), y \in (0, 1) \\ 0, & \text{otherwise} \end{array} \right. $$
where $k$ is a constant.

Calculate $\text{Cov}(X,Y).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Trước tiên, ta phải xác định $k > 0$ để $f$ là hàm MDXS, tức là 

$$ \iint f(x,y)dxdy = 1 \Leftrightarrow \int_0^1 \int_0^1 kx dx dy = 1 \Leftrightarrow k = 2. $$
- Để tính hiệp phương sai $\text{Cov}(X,Y)$, lưu ý rằng

$$\text{Cov}(X,Y) = \mathbb{E}(XY) - (\mathbb{E}X) (\mathbb{E}Y).$$
ta có

$$ \begin{array}{cll} \mathbb{E}X &=& \displaystyle \iint x f(x, y)dx dy = \int_0^1 \int_0^1 x(2x)dxdy = \dfrac{2}{3} \\ \mathbb{E}Y &=& \displaystyle \iint y f(x, y)dx dy = \int_0^1 \int_0^1 y(2x)dxdy = \dfrac{1}{2} \\ \mathbb{E}(XY) &=& \displaystyle \iint xy f(x, y)dx dy = \int_0^1 \int_0^1 xy(2x)dxdy = \dfrac{1}{3} \end{array} $$

Do đó $\text{Cov}(X,Y) = 0.$

#### Question 105.
Let $X$ and $Y$ be continuous random variables with joint density function
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{8}{3}xy, & x\in [0, 1], x \leq y \leq 2x \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the covariance of $X$ and $Y$. 

**Giải đáp / SOLUTIONs.**

Tương tự câu 104, nhớ rằng $$\text{Cov}(X,Y) = \mathbb{E}(XY) - (\mathbb{E}X) (\mathbb{E}Y).$$
ta có

$$ \begin{array}{cll} \mathbb{E}X &=& \displaystyle \iint x f(x, y)dx dy = \int_0^1 \int_0^1 x \left( \dfrac{8xy}{3} \right) dxdy = \dfrac{4}{5} \\ \mathbb{E}Y &=& \displaystyle \iint y f(x, y)dx dy = \int_0^1 \int_0^1 y\left( \dfrac{8xy}{3} \right)dxdy = \dfrac{56}{45} \\ \mathbb{E}(XY) &=& \displaystyle \iint xy f(x, y)dx dy = \int_0^1 \int_0^1 xy\left( \dfrac{8xy}{3} \right)dxdy = \dfrac{28}{27} \end{array} $$
Do đó $\text{Cov}(X,Y) = \dfrac{1}{25}.$

#### Question 106.
- Let $X$ and $Y$ denote the values of two stocks at the end of a five-year period. 
- $X$ is uniformly distributed on the interval $(0, 12)$. Given $X = x, Y$ is uniformly distributed on the interval $(0, x)$.

Calculate $\text{Cov}(X, Y)$ according to this model. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ các giả thiết, ta có 

$\qquad \diamond f(x) = \dfrac{1}{12} \textbf{1}_{\lbrace x \in (0, 12) \rbrace}$ 

$\qquad \diamond f(y|x) = \dfrac{1}{x} \textbf{1}_{\lbrace y \in (0, x) \rbrace}$ 

từ đó, hàm MDXS đồng thời của cặp $(X, Y)$ sẽ được xác định bởi
$$ f(x, y) = f(x) f(y|x) = \dfrac{1}{12x} \textbf{1}_{\lbrace x \in (0, 12) \rbrace} \textbf{1}_{\lbrace y \in (0, x)\rbrace} $$
- Trở về cách giải như 2 bài trước(105 và 104), ta có

$$ \begin{array}{cll} \mathbb{E}X &=& \displaystyle \iint x f(x, y)dx dy = \int_0^{12} \int_0^x x \left( \dfrac{1}{12x} \right) dxdy = 6 \\ \mathbb{E}Y &=& \displaystyle \iint y f(x, y)dx dy = \int_0^{12} \int_0^x y\left( \dfrac{1}{12x} \right)dxdy = 3 \\ \mathbb{E}(XY) &=& \displaystyle \iint xy f(x, y)dx dy = \int_0^{12} \int_0^x xy\left( \dfrac{1}{12x} \right)dxdy = 24 \end{array} $$
Do đó $\text{Cov}(X,Y) = 6.$

#### Question 107.
Let $X$ denote the size of a surgical claim and let $Y$ denote the size of the associated hospital claim. An actuary is using a model in which 

$$ \mathbb{E}X = 5, \mathbb{E}X^2 = 27.4, \mathbb{E}Y=7, \mathbb{E}Y^2 = 51.4, \text{ Var}(X + Y)= 8. $$

Let $C_1 = X + Y$ denote the size of the combined claims before the application of a 20% `surcharge` on the hospital portion of the claim, and let $C_2$ denote the size of the combined claims after the application of that surcharge.

Calculate $\text{Cov}(C_1, C_2)$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ các giả thiết đã cho, ta được

$$ \begin{array}{rcl} \text{Var}X &=& \mathbb{E}X^2 - [\mathbb{E}X]^2 = 2.4 \\ \text{Var}Y &=& \mathbb{E}Y^2 - [\mathbb{E}Y]^2 = 2.4 \\ \text{Cov}(X, Y) &=& \dfrac{1}{2} \left( \text{Var}(X+Y) - \text{Var}X - \text{Var}Y \right) = 1.6 \\ C_2 &=& X + Y + 0.2 Y = X + 1.2Y \end{array} $$
- Do đó, 

$$ \begin{array}{cll} \text{Cov}(C_1, C_2) &=& \text{Cov}(X+Y, X + 1.2Y) \\ &=& \text{Cov}(X,X) + \text{Cov}(X, 1.2Y) + \text{Cov}(Y, X) + \text{Cov}(Y, 1.2Y) \\ &=& \text{Var}X + 1.2 \text{Cov}(X, Y) + \text{Cov}(X, Y) + 1.2\text{Var}Y \\ &=& 8.8\end{array} $$

#### Question 108.
A device containing two key components fails when, and only when, both components fail. The lifetimes, $T_1$ and $T_2$ of these components are independent with common density function 
$$ f(t) = \left\lbrace \begin{array}{cl} e^{-t}, & 0 < t \\ 0, & \text{otherwise} \end{array} \right. $$
The cost, $X$, of operating the device until failure is $2T_1 + T_2$. Let $g$ be the density function
for $X$.

Determine $g(x)$, for $x > 0$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ giả thiết $T_1, T2$ độc lập cùng phân phối với hàm MDXS $f$, ta được hàm MDXS đồng thời của cặp $(T_1, T_2)$ là

$$ h(t_1, t_2) = f(t_1) f(t_2) = e^{-(t_1 + t_2)} \mathbf{1}_{\lbrace t_1 \geq 0, t_2 \geq 0 \rbrace} $$
- Giả thiết cuối cùng cho biết, $X$ là giới hạn thời gian hoạt động của thiết bị không vượt quá $2T_1 + T_2$, tức là 

$$ G(x) = P[X < x] = P[2T_1 + T_2 < x] $$
là hàm phân phối tích lũy của $X$, và ta được

$$ \begin{array}{ccl} G(x) &=& P\left[ T_1 \geq 0, T_2 \geq 0, 2T_1 + T_2 < x \right] \\ &=& \displaystyle \int_0^{x/2} \int_0^{x - 2t_1} e^{-t_1} e^{-t_2} dt_2 dt_1 \\ &=& 1 + e^{-x} - 2e^{-x/2} \end{array} $$
- Hàm MDXS $g$ của $X$ sẽ được suy ra từ đạo hàm của $G$, và do đó

$$ g(x) = \dfrac{dG}{dx} = e^{-x/2} - e^{-x}. $$

#### Question 109.
- A company offers earthquake insurance. Annual premiums are modeled by an `exponential random variable` with mean 2. Annual claims are modeled by an `exponential random variable` with mean 1. 
- Premiums and claims are independent. Let $X$ denote the ratio of claims to premiums, and let $f$ be the density function of $X$. 

Determine $f(x)$, where it is positive.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $U, V$ lần lượt tương ứng với các gói `annual claim` và `annual premiums` và theo giả thiết đầu tiên thì
$$ U \sim \mathcal{E}(1) \text{ và } V \sim \mathcal{E}(2). $$
- Hơn nữa, do $U, V$ độc lập nên hàm MDXS đồng thời của chúng là
$$ g(u, v) = \left( e^{-u} \mathbf{1}_{\lbrace u \geq 0 \rbrace} \right) \cdot \left( e^{-v/2} \mathbf{1}_{\lbrace v \geq 0 \rbrace} \right) = e^{-u - v/2} \mathbf{1}_{\lbrace u, v \geq 0 \rbrace} $$
- Bài toán yêu cầu tìm hàm MDXS $f$ của $X$ với $X$ là tỷ số giữa `annual claims` và `annual premiums`, tức là $X=U/V$. Trước tiên, ta cần xác định hàm phân phối tích lũy $F$ của $X$,

$$ \begin{array}{ccl} F(x) = P[X < x] &=& P[(U / V) < x] \\ &=& P[U < Vx] \\ &=& \displaystyle \int_0^{\infty} \int_0^{vx} g(u, v) du dv \\ &=& \dfrac{2x}{2x + 1} \end{array} $$
suy ra

$$ f(x) = \dfrac{dF}{dx} = \dfrac{2}{(2x+1)^2} $$

#### Question 100.
Let X and Y be continuous random variables with joint density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} 24xy, & x \in (0, 1), 0 < y < 1- x \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate $P\left[ Y < X \left\vert X = \dfrac{1}{3} \right. \right].$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta có
$$P\left[ Y < X \left\vert X = \dfrac{1}{3} \right. \right] = P\left[ Y < 1/3 \left\vert X = \dfrac{1}{3} \right. \right] = \int_0^{1/3} f(y | x = 1/3) dy.$$
vì vậy trước tiên ta phải tính được $f(y | x = 1/3).$
- Trước tiên, nhắc lại rằng
$$ f(x,y) = \left\lbrace \begin{array}{cl} 24xy, & x \in (0, 1), 0 < y < 1- x \\ 0, & \text{otherwise} \end{array} \right. $$
nên $\forall x_0 \in (0, 1)$ thì

$$ f(y | x = x_0) = \dfrac{f(x_0, y)}{f_X(x_0)}, \text{ khi } 0<y<1 -x_0, $$
trong đó

$$ f_X(x) = \int_0^{1-x} f(x, y) dy $$
là hàm mật độ lề của $X$.
- Thay $x_0 = 1/3$, ta có
$$ \begin{array}{rcll} f(1/3, y) &=& 24 \cdot (1/3) y = 8y & \text{ khi } 0 < y < 2/3  \\ f_X(1/3) &=& \displaystyle \int_0^{2/3} 24\cdot (1/3)y dy = \dfrac{16}{9}  \end{array} $$
do đó
$$ f(y | x = 1/3) = \dfrac{8y}{16/9} = \dfrac{9y}{2}, \quad \text{khi } y \in \left(0, \dfrac{2}{3} \right) $$
và
$$P\left[ Y < X \left\vert X = \dfrac{1}{3} \right. \right] = \int_0^{1/3} \dfrac{9y}{2}dy = \dfrac{1}{4}. $$
