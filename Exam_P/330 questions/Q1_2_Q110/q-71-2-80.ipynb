#### Question 71, 
The time, $T$ that a manufacturing system is out of operation has `cumulative distribution function` 

$$ F(t) = \left\lbrace \begin{array}{lr} 1 - \left( \dfrac{2}{t} \right)^2, & t > 2 \\ 0, & \text{ otherwise } \end{array} \right. $$

The resulting cost to the company is $Y = T^2$. Let $g$ be the density function for $Y$.

Determine $g(y)$, for $y > 4$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính $g(y)$: hàm MDXS của $Y = T^2$ thông qua hàm phân phối tích lũy $F(t)$ từ $T$; trước tiên ta cần xác định hàm phân phối tích lũy $G(y)$ của $Y$ rồi suy ra $g(y) = \dfrac{d}{dy} G(y)$
- Theo định nghĩa, với mọi $y \geq 0,$ ta có
$$ G(y) = P[Y < y] = P [T^2 < y] = P\left[ -\sqrt{y} < T < \sqrt{y} \right] = F\left( \sqrt{y} \right) - F\left( -\sqrt{y} \right)$$
- Vì đề bài yêu cầu tìm $g(y)$ với $y > 4$ nên $t = \sqrt{y} > 2$ và $-\sqrt{y} < - 2$, do đó

$$ F\left( -\sqrt{y} \right) = 0 $$
và
$$ F\left( \sqrt{y} \right) = 1 - \left( \frac{2}{\sqrt{y}} \right)^2 = 1 - \dfrac{4}{y}$$
Suy ra
$$ G(y) = F\left( \sqrt{y} \right) - F\left( -\sqrt{y} \right) = \left\lbrace \begin{array}{cl} 1 - \dfrac{4}{y}, & y > 4 \\ 0, & y \leq 4 \end{array} \right.$$
- Lấy đạo hàm theo $y$, ta được

$$ g(y) = \dfrac{d}{dy} G(y) = \left\lbrace \begin{array}{cl} \dfrac{4}{y^2}, & y > 4 \\ 0, & y \leq 4 \end{array} \right. $$

do đó $g(y) = \dfrac{4}{y^2}, \quad \forall y > 4.$

#### Question 72.
- An investment account earns an annual interest rate $R$ that follows a uniform distribution on the interval `(0.04, 0.08)`. The value of a 10,000 initial investment in this account after one year is given by $V = 10000 e^{R}$
- Let $F$ be the cumulative distribution function of $V$.

Determine $F(v)$ for values of $v$ that satisfy $0 < F(v) < 1.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài yêu cầu tính $F(v)$ là hàm phân phối tích lũy của $V$ với $V = 10000 e^R$, tức là

$$ F(v) = P [V < v] = P \left[ e^R < \frac{v}{10000} \right], \quad \forall v > 0$$

do đó,

$$ F(v) = P\left[ 0 \leq R \leq \ln(v) - \ln(10000) \right], \quad \forall v > 0. $$
- Giả thiết về BNN $R$ cho biết nó là phân phối đều với hàm MDXS là $$f(t) = \dfrac{1}{0.04} \cdot \textbf{1}_{\lbrace x\in (0.04, 0.08) \rbrace} $$ và do đó,

$$ \begin{array}{rcl}  F(v) &=& P\left[ 0 \leq R \leq \ln(v) - \ln(10000) \right] \\ &=& \displaystyle \int_{0.04}^{\ln(v) - \ln(10000)} \dfrac{1}{0.04} \cdot \textbf{1}_{\lbrace x\in (0.04, 0.08) \rbrace} dx \\ &=& 25 \left[ \dfrac{\ln(v)}{10000} - 0.04 \right] dx \end{array} $$

#### Question 73.
- An actuary models the lifetime of a device using the random variable $Y = 10X^{0.8}$, where $X$ is an exponential random variable with mean 1.
- Let $f(y)$ be the density function for $Y$.

Determine $f (y),$ for $y > 0$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Lập luận tương tự câu 71, ta có hàm phân phối tích lũy của $Y$ xác định bởi

$$ F(y) = P[Y < y] = P [10 X^{0.8} < y] = P[X < (0.1y)^{1.25}] = G[(0.1y)^{1.25}], \quad \forall y > 0$$

với $G(x)$ là hàm phân phối tích lũy của $X$.
- Theo giả thiết đầu tiên thì $X$ có phân phối mũ với trung bình $1$ nên 

$$ G(x) = 1 - e^{-x}, \quad x \geq 0$$
do đó
$$ F(y) = G[(0.1y)^{1.25}] = 1 - e^{-(0.1y)^{1.25}} $$

- Để có được hàm MDXS $f(y)$, ta sẽ lấy đạo hàm của $F(y)$ theo $y$, 

$$ \begin{array}{ccl} f(y) = \dfrac{d}{dy} F(y) &=& \dfrac{d}{dy} \left[ 1 - e^{-(0.1y)^{1.25}} \right] \\ &=& -[-(0.1)\cdot(1.25)] (0.1y)^{0.25} \cdot e^{(0.1y)^{1.25}}\\ &=& 0.125 \cdot (0.1y)^{0.25} \cdot e^{(0.1y)^{1.25}}  \end{array} $$

#### Question 74.
- Let $T$ denote the time in minutes for a customer service representative to respond to 10 telephone inquiries. 
- $T$ is uniformly distributed on the interval with endpoints 8 minutes and 12 minutes.
- Let $R$ denote the average rate, in customers per minute, at which the representative responds to inquiries, and let $f(r)$ be the density function for $R$. 

Determine $f(r)$ for $r \in \left( \frac{10}{12}, \frac{10}{8} \right)$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ giả thiết đầu tiên, ta được $R = \dfrac{10}{T}$
- Giả thiết thứ 2 cho ta biết $T$ là BNN có phân phối đều với hàm MDXS

$$ g(t) = (0.25) \cdot \textbf{1}_{\lbrace x \in [8, 12] \rbrace} $$

- Để tìm hàm MDXS $f(r)$ của $R$ khi $r \in \left( \frac{10}{12}, \frac{10}{8} \right)$, ta lại phải tính thông qua hàm phân phối tích lũy của nó

$$ \forall r \in \left( \frac{10}{12}, \frac{10}{8} \right), \quad F(r) = P(R < r) = P[T > 10 / r] = \int_{10/r}^{+\infty} f(t) dt = \int_{10 / r}^{12} (0.25) dt= \dfrac{12 - (10 / r)}{4} $$
suy ra

$$ f(r) = \dfrac{d}{dr} F(r) = \dfrac{5}{2r^2}, \quad \forall r \in \left( \frac{10}{12}, \frac{10}{8} \right) $$

#### Question 75.
- The monthly profit of `Company I` can be modeled by a continuous random variable with density function $f$. `Company II` has a monthly profit that is twice that of `Company I`.
- Let g be the density function for the distribution of the monthly profit of `Company II`.

Determine $g(y)$ where it is not zero. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là BNN thể hiện lợi nhuận hàng tháng ở công ty `I` và `II` và theo giả thiết đầu tiên thì $Y = 2X$
- Đề bài yêu cầu xác định hàm MDXS của $Y$ theo MDXS của $X$; do đó ta lại đặt tiếp $g(y), G(y)$ và $f(x), F(x)$ lần lượt là các hàm `MDXS`, `phân phối tích lũy` của 2 thằng này.
- Để tìm liên hệ hàm $g$ theo $f$ thì trước tiên ta phải tìm liên hệ từ 2 hàm phân phối tích lũy $G$ theo $F$, ta có

$$ G(y) = P[Y < y] = P[X < y/2] = F(y/2) $$
do đó,
$$ g(y) = \dfrac{d}{dy} [G(y)] = \dfrac{d}{dy} [F(y / 2)] = \dfrac{1}{2} f\left( \dfrac{y}{2} \right). $$

#### Question 76.
Claim amounts for wind damage to insured homes are mutually independent random variables with `common density function`

$$ f(x) = \left\lbrace \begin{array}{rl} 3x^{-4}, & x > 1 \\ 0, & \text{ otherwise} \end{array} \right. $$

where $x$ is the amount of a claim in thousands.
 
Suppose 3 such claims will be made.

Calculate the expected value of the largest of the three claims. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_1, X_2, X_3$ lần lượt là 3 yêu cầu về số tiền bồi thường đầu tiên mà BH phải trả.
- $Y$ là số tiền lớn nhất trong 3 số tiền trên tức là $Y = \max(X_1, X_2, X_3)$ và theo yêu cầu BT ta phải tính $\mathbb{E}Y$.
- Vì $Y = \max(X_1, X_2, X_3)$ và $X_1, X_2, X_3$ là i.i.d với hàm MDXS $f(x)$ nên

$$ P[Y < y] = P[\max(X_1, X_2, X_3) < y] = P[X_1 < y]P[X_2 < y]P[X_3 < y] $$
và với mọi $y > 1$, ta có

$$ P[X_1 < y]=P[X_2 < y]=P[X_3 < y] = \int_{-\infty}^y f(x) dx = \int_1^y 3 x^{-4}dx = 1 - \dfrac{1}{y^3}. $$
Do đó
$$ F(y) = P[Y < y] = \left( 1 - \dfrac{1}{y^3} \right)^3 $$
suy ra
$$ f(y) = \dfrac{d}{dy} [F(y)] = 3 \left( \dfrac{3}{y^4} \right) \left( 1 - \dfrac{1}{y^3} \right)^2 = \dfrac{9}{y^4} \left( 1 - \dfrac{1}{y^3} \right)^2 $$
- Cuối cùng
$$ \mathbb{E}Y = \int y f(y) dy = \int_1^{\infty} y \cdot \dfrac{9}{y^4} \left( 1 - \dfrac{1}{y^3} \right)^2 dy = 2.025 $$


#### Question 77.
A device runs until either of two components fails, at which point the device stops running. The `joint density function` of the `lifetimes` of the two components, both measured in hours, is 
$$ f(x, y) = \left\lbrace \begin{array}{rl} \dfrac{x+y}{8}, & 0 < x, y < 2 \\ 0, & \text{ otherwise} \end{array} \right. $$

Calculate the probability that the device fails during its first hour of operation.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Z = (X, Y)$ là BNN thể hiện tuổi thọ của 2 thành phần trong thiết bị.
- BT yêu cầu tính XS để thiết bị đó hỏng ngay trong giờ hoạt động đầu tiên, tức là $P(Z \not \in [1,2] \times [1,2])$
- Vì $Z$ có hàm MDXS $f$ nên

$$ P(Z \not \in [1,2] \times [1,2]) = \int_1^2 \int_1^2 f(x,y) dx dy = \int_1^2 \int_1^2 \dfrac{x+y}{8}dx dy = \dfrac{3}{8} = 0.375$$

do đó

$$ P(Z \in [1,2] \times [1,2]) = 1 - P(Z \not \in [1,2] \times [1,2]) = 0.625. $$

#### Question 79.
A device contains two components. The device fails if either component fails. The `joint density function` of the lifetimes of the components, measured in hours, is $f(s,t)$, where $0 < s < 1$ and $0 < t < 1$.

Determine which formular represents the probability that the device fails during the **first half hour of operation.**

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ là 2 thành phần trong thiết bị trên và $Z = (X, Y)$ là BNN thể hiện tuổi thọ của cả 2 thành phần.
- Theo giả thiết, $Z$ có hàm MDXS là $f(s, t)$ với $s, t \in (0, 1)$ và thiết bị chỉ bị hỏng nếu $X$ hoặc $Y$ bị hỏng. Đề bài yêu cầu tính XS để $Z$ hỏng trong nửa giờ đầu tiền, tức là hoặc $X$ hoặc $Y$ hỏng trong nửa giờ đầu tiên, tức là 

$$ Z \in A = \lbrace (x,y): x\in(0, 0.5), y \geq 0.5 \text{ hoặc } x \in (0, 1), y \in (0, 0.5) \rbrace $$
- Do đó, 
$$ P(Z \in A) = \int_0^{0.5} \int_{0.5}^1 f(s, t) ds dt + \int_0^1 \int_0^{0.5} f(s, t) ds dt $$

#### Question 80.
A charity receives `2025` contributions. Contributions are assumed to be mutually independent and identically distributed with `mean 3125` and `standard deviation 250`.

Calculate the approximate `90th percentile` for the distribution of the total contributions received.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_1, ..., X_n$ là BNN thể hiện số tiền quyên góp của tổ chức từ thiện với $n = 2025.$
- Theo giả thiết thì $X_1, \ldots, X_{2025}$ là i.i.d với trung bình 3125 và độ lệch chuẩn 250, do đó theo Định lý giới hạn trung tâm thì

$$ S = \sum_{k=1}^{2025} X_k \sim \mathcal{N} (\mu, \sigma^2 )$$
trong đó
$$ \mu = \mathbb{E} S = \sum_{k=1}^{2025} \mathbb{E} X_k = 2025\cdot 3125 = 6328125. $$
và
$$ \sigma = \sqrt{\text{Var}S} = \sqrt{\sum_{k=1}^{2025} \text{Var} X_k} = \sqrt{2025 \cdot 250^2} = 11250. $$
- Đê tìm phân vị mức 90% cho $S$, tức là tìm $x_{\alpha}$ sao cho
$$ P(S < x_{\alpha}) = 0.9 $$
Vì $S$ là BNN có phân phối chuẩn, do đó
$$ x_{\alpha} = \mathbb{E}S + \Phi^{-1}(0.9) \cdot \sqrt{\text{Var}S} $$
trong đó 
$$ \Phi^{-1}(s) = P[\mathcal{N}(0, 1) < s] $$
- Lúc này tra bảng và thay các giá trị $\mu, \sigma$, ta được

$$ x_{\alpha} = 6328125 + (1.282) \cdot 11250 = 6342548. $$
