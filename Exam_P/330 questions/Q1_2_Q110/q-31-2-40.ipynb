#### Question 31, 
A company establishes a fund of **120** from which it wants to pay an amount, `C`, to any of its **20** employees who achieve a **`high performance level`** during the `coming year`. 
- (i) Each employee has a `2%` chance of achieving a `high performance level` during the coming year. 
- (iii) The events of different employees achieving a **`high performance level`** during the coming year are mutually independent.

Calculate the maximum value of C for which the probability is less than 1% that the fund will be inadequate to cover all payments for high performance.

----------------------------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là BNN (`rvs`) thể hiện năng suất của nhân viên thứ $i$ trong công ty đó có làm việc hiệu quả hay không, tức
$$ X_i = \left \lbrace \begin{array}{ccl} 1 & \text{ nếu nhân viên đó đạt hiệu quả cao trong công việc} \\ 0 & \text{dek (otherwise)} \end{array} \right. $$
- Như vậy $X_i$ là các BNN tuân theo pp Bernoulli, và hơn nữa; theo (i), ta có
$$ P(X_i = 1) = 2\% = 0.02, \qquad \forall i$$
- Khi đó $\displaystyle X = \sum_{i=1}^{n} X_i$ là BNN thể hiện số nhân viên đạt năng suất công việc hiệu quả cao và $X \sim \mathcal{B}(n, 0.02)$ do các BNN này là độc lập (theo giả thiết (ii)) với $n = 20$.
- Vì nguồn quỹ của công ty là 120 nên mức thưởng `C` mà họ trích ra phải không được vượt quá 120, tức là $C \leq 120$.
- Yêu cầu bài toán là **tính lượng tiền thưởng `C` dành cho các nhân viên đạt hiệu quả cao sao cho xác suất{mà lượng tiền này không đủ chi trả} sẽ không vượt quá 0.01**, tức là tìm số nguyên dương nhỏ nhất `x` (số nhân viên đạt hiệu quả cao) sao cho
$$ P(X > x) < 0.01 $$
tương đương với
$$ P(X \leq x) > 0.99 $$
sau đó, ta sẽ xác định $C$ bằng cách chia đều số tiền `120` cho `x` tìm được từ phương trình trên.
- Do $X \sim B(20, 0.02)$ nên bài toán trở thành tìm $x$ sao cho
$$ P(X \leq x) = \sum_{k=0}^n C_{20}^k 0.02^k (1 - 0.02)^{20 - k} \geq 0.99$$
Dễ thấy rằng các BNN như `nhị thức, Poisson,.. có xu hướng giảm về xác suất khi tăng giá trị` nên ta sẽ tập trung quan sảt vài giá trị đầu khi tính tổng tích lũy của chúng tại các bất phương trình có dạng trên. Ta có 
$$P(X = 0) = 0.668, \, P(X = 1) = 0.272\text{ và }P(X = 2) = 0.053,$$
như vậy, 

$\qquad$ Nếu $x = 1$ thì tổng 2 xác suất đầu tiên chưa vượt 0.99. 

$\qquad$ Nếu $x = 2$, tổng 3 xác suất đầu tiên đạt mức $0.993$ và ta sẽ chọn nó (số nhỏ nhất thỏa BPT)

- Cuối cùng, khi đã tìm được `x = 2` thì dễ dàng suy ra được $C = \dfrac{120}{x} = 60$

#### Question 32, 
A large pool of adults earning their first driver’s license includes **50% low-risk** drivers, **30% moderate-risk** drivers, and **20% high-risk** drivers. 
- Because these drivers have no prior driving record, an insurance company considers each driver to be **randomly selected** from the pool.
- This month, the insurance company writes **four new policies** for adults earning their **first driver’s license**.

Calculate the probability that these four will contain at least two more high-risk drivers than low-risk drivers. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $L, M, H$ lần lượt là các sự kiện mà khi chọn ngẫu nhiên một người bất kỳ thì người tài xế đó có rủi ro gây tai nạn `thấp (Low)`, `trung bình (moderate)` và `cao (high)`. Khi đó, trong câu đầu tiên của đề bài ta được
$$ P(L) = 0.5, \quad P(M) = 0.3 \; \text{ và } \; P(H) = 0.2. $$
- Tiếp theo, do việc lựa chọn quan sát này là ngẫu nhiên nên mỗi quan trắc $X_i$ có thể đưọc xem như độc lập đôi một với nhau.
- Giả thiết sau cùng cho biết công ty bảo hiểm chỉ phát hành `4 hợp đồng bảo hiểm (HĐBH) mới` đối với nhóm các tài xế nhận giấy phép lái xe lần đầu, trong tháng này. Đặt `x, y, z` lần lượt là số sự kiện liên quan đến `L, M, H` khi quan sát ngẫu nhiên một tài xế trong tháng này, do đó
$$ x + y + z = 4; \quad \forall x,y,z \in \lbrace 0,1,2,3,4 \rbrace $$
- Đặt $f(x, y, z) = P(L = x, M = y, H = z)$ là xác suất đồng thời mà các sự kiện `L, M, H` lần lượt xảy ra `x, y, z` (lần), khi đó

$$ f(x,y,z) = P(L=x,M=y,H=z) = \left( \dfrac{(x+y+z)!}{x! y! z!} P^x(L) \right) P^y(M) P^z(H), $$
$\qquad$ Yêu cầu bài toán là tính **xác suất mà trong 4 `HĐBH` đó thì nhóm `rủi ro cao (H)` nhiều hơn ít nhất 2 `HĐBH` so với nhóm `rủi ro thấp`**, tức là $z \geq x+2$, mà do $x,y,z$ là các số nguyên dương bé hơn $4$ nên ta chỉ có 4 trường hợp sau đây

$$ \left\lbrace \left. x,y,z \in \overline{0,\ldots 4} \right \vert z \geq x+2 \right\rbrace = \left\lbrace (x=0, y=0, z=4); (x=0, y=1, z=3), (x=1, y=0, z=3), (x=0, y=2, z=2) \right\rbrace $$
Như vây,
$$ \begin{array}{ccl} f(z \geq x+2) &=& f(0,0,4) + f(0,1,3) + f(1,0,3) + f(0,2,2) \\ &=& (P(H=4))^4 + \dfrac{4!}{1! 3!} P(M=1) (P(H=3))^3 + \dfrac{4!}{1! 3!} P(L=1) (P(H=3))^3 + \dfrac{4!}{2! 2!} (P(M=2))^2 (P(H=2))^2 \\ &=& (0.2)^4 + 4*(0.3)*(0.2)^3 + 4*(0.3)*(0.2)^3 + 6*(0.3)^2*(0.2)^2 \\ &=& 0.0488 \end{array} $$

#### Question 33.

The loss due to a fire in a commercial building is modeled by a random variable $X$ with a density function

$$ f(x) = \left \lbrace \begin{array}{ccl} 0.0005 (20 - x) & \text{ nếu } & 0 < x < 20 \\ 0 & \text{dek (otherwise)} \end{array} \right. $$

Given that a fire loss exceeds 8, calculate the probability that it exceeds 16. 

------------------------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Yêu cầu BT là tính $P(X > 16 | X > 8)$, theo công thức xác suất có điều kiện $P(A | B) = \dfrac{P(A \cap B)}{P(B)}$, thì

$$ P(X > 16 | X > 8) = \dfrac{P\left( (X > 16) \cap (X > 8) \right)}{P(X > 8)} = \dfrac{P\left( X > 16\right)}{P(X > 8)} $$

bởi $$\lbrace X > 16 \rbrace = \lbrace \omega : X(\omega) > 16 \rbrace \subset \lbrace \omega : X(\omega) > 8 \rbrace = \lbrace X > 8 \rbrace $$
- Như vậy, để đơn giản hóa tính toán, ta chỉ cần tính công thức tổng quát cho $P(X > x)$ với $x$ bất kỳ. Mà theo định nghĩa về hàm MĐXS của $X$ thì nó sẽ triệt tiêu (bằng 0) bên ngoài tập compact [0, 20], và do đó $\forall x \in (0, 20]$, ta có

$$ P(X > x) = 1 - P(X \geq x) = 1 - \int_{-\infty}^x f(t) dt = 1 - \int_0^x f(t) dt = 0.005 \left( 200 - 20x + \dfrac{x^2}{2} \right) $$

Suy ra,

$$ P(X > 16 | X > 8) = \dfrac{0.005 \left( 200 - 20*16 + \dfrac{16^2}{2} \right) }{0.005 \left( 200 - 20*8 + \dfrac{8^2}{2} \right)} = \dfrac{1}{9}. $$

#### Question 34.

The `lifetime` of a machine part has a continuous distribution on the interval (0, 40) with probability density function $f(x)$, where $f(x)$ is proportional to $(10 + x)^{− 2}$ on the interval.

Calculate the probability that the lifetime of the machine part is less than 6.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện `tuổi thọ của máy` (`lifetime`) với hàm mật độ XS là $f(x)$. Theo giả thiết, hàm MĐXS $f(x)$ liên tục trên khoảng $(0, 40)$ và có tỷ lệ với $(10 + x)^{-2}$ trên đoạn này, do đó tồn tại một hằng số dương $C$ nào đó sao cho
$$ f(x) = \left\lbrace \begin{array}{cl} C(10 + x)^{-2} & \text{ khi } x \in (0, 40) \\ 0, & \text{otherwise} \end{array} \right. $$
- Điều kiện đủ để $f$ là hàm MĐXS thì $f$ phải thỏa $\displaystyle \int_{\mathbb{R}} f = 1$, tức là
$$ C \int_0^{40} (10 + x)^{-2} dx = 1 \Leftrightarrow \left. \left( \dfrac{-C}{10 + x} \right) \right\vert_0^{40} = 1 \Leftrightarrow C = 12.5$$
- Bấy giờ, ta sẽ xử lý yêu cầu bài toán, tức là tính XS để `tuổi thọ của máy` bé hơn 6, tức là
$$ P(X < 6) = \int_0^6 f(x) dx = \left. \left( \dfrac{-12.5}{10 + x} \right) \right\vert_0^{6} = 0.47. $$

#### Question 36.
A `group insurance policy` covers the medical claims of the employees of a small company. The value, V, of the claims made in one year is described by $$ V = 100000 Y, $$
where $Y$ is a random variable with density function $$ f(y) = \left\lbrace \begin{array}{ccl} k(1-y)^4, & 0 < y < 1 \\ 0, & \text{ otherwise} \end{array} \right. $$
where $k$ is a constant.

Calculate the conditional probability that $V$ exceeds $40,000$, given that $V$ exceeds $10,000$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Tương tự câu 35, đầu tiên cần xác định hằng số $k$ sao cho $f$ là hàm MĐXS, tức là tìm $k$ thỏa
$$ \left\lbrace \begin{array}{ccl} f(y) & \geq & 0 \\ \displaystyle \int f &=& 1 \end{array} \right. \Leftrightarrow \left\lbrace \begin{array}{ccl} k & \geq & 0 \\ \displaystyle k \int_0^1 (1 - y)^4 dy &=& 1 \end{array} \right. \Leftrightarrow k = 5. $$
- Yêu cầu BT là tính XS có điều kiện,
$$ P(V > 40000 | V > 10000) = \dfrac{P\left( (V > 40000) \cap (V > 10000) \right)}{P(V > 10000)} = \dfrac{P\left( V > 40000\right)}{P(V > 10000)} $$
- Theo giả thiết với $V = 100000 Y$ và $y \in (0, 1)$ thì 
$$ P(V > 40000) = P(1000000Y > 40000) = P(Y > 0.4) = 1 - P(Y \leq 0.4) = 1 - \int_0^{0.4} 5(1-y)^4 dy = 0.078$$
Tương tự, với mọi $y \in (0, 1)$ ta có
$$ P(V > 10000) = P(Y > 0.1) = 1 - \int_0^{0.1} 5(1-y)^4 dy = 0.59$$
do đó,
$$ P(V > 40000 | V > 10000) = \dfrac{0.078}{0.59} = 0.132 $$

#### Question 37.

- The `lifetime` of a `printer costing` 200 is **`exponentially`** distributed with `mean 2 years`. 
- The manufacturer agrees to pay a `full refund` to a buyer if the printer fails **during** the `first year` following its purchase, a `one-half` refund if it fails during the `second year`, and `no refund` for failure **after** the `second year`.

Calculate the expected total amount of refunds from the sale of `100 printers`

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ là BNN thể hiện `tuổi thọ (lifetime)` của `máy in (printer)`. 
- Theo giả thiết, $T$ có phân phối mũ với trung bình $\lambda = 2$ năm, tức hàm phân phối tích lũy của nó có dạng $F_T(t) = 1 - e^{- t / \lambda}$, do đó

$\qquad \diamond$ XS để máy in hỏng ngay trong năm đầu tiên là $P(T \leq 1) = F_T(1) = 1 - e^{-0.5} \approx 0.3935$.

$\qquad \diamond$ XS để máy in hỏng trong năm thứ hai là $P(1 < T \leq 2) = (1 - e^{-1}) - (1 - e^{-0.5}) \approx 0.2386$.

$\qquad \diamond$ XS để máy in hỏng ngay sau năm thứ hai trở đi là $P(T > 2) = 1 - F_T(2) = e^{-2} \approx 0.3679$.

- Các thông tin khác:

$\qquad \diamond$ Chi phí bồi thường cho mỗi máy in là 200.

$\qquad \diamond$ Chính sách hoàn tiền: toàn bộ (200) nếu máy hỏng ngay trong năm đầu, một nửa (100) nếu máy hỏng trong năm thứ hai và không bồi thường nếu máy in hỏng sau 2 năm, tức là

| | hỏng trong năm đầu $(T \leq 1)$ | $T \in (1,2]$ | $T > 2$ |
|-|:-:|:-:|:-:|
| $Z:$ Chi phí hoàn tiền/ mỗi sản phẩm| 200 | 100 | 0 |
| Xác suất hỏng | $P(T \leq$ 1) =0.3935 | $P(T\in$ (1, 2]) = 0.2386 | $P(T$ > 2) =0.3679 | 

- Theo đó, số tiền phải trả sẽ được tính bởi công thức kỳ vọng của $Z$

$$ \mathbb{E}Z = \sum_{z \in Z(T)} z P(Z(T) = z) = 200*P(T \leq 1) + 100*P(T \in (1, 2]) + 0*P(T > 2) \approx 10.256 $$

#### Question 38.
An insurance company insures a large number of homes. The insured value, $X$, of a randomly selected home is assumed to follow a distribution with density function 
$$ f(x) = \left\lbrace \begin{array}{ccl} 3x^{-4}, & x > 1 \\ 0, & \text{ otherwise} \end{array} \right.  $$
Given that a randomly selected home is insured for at least 1.5, calculate the probability that it is insured for less than 2. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài đã cho $X$ là BNN có hàm MĐXS $f(x)$ như trên và yêu cầu tính xác suất có điều kiện
$$ P(X < 2 | X \geq 1.5) = \dfrac{P(1.5 \leq X < 2)}{P(X \geq 1.5)} = \dfrac{P(X < 2) - P(X < 1.5)}{1 - P(X < 1.5)} = \dfrac{F(2) - F(1.5)}{1 - F(1.5)} $$
trong đó,
$$ F(x) = P(X < x) = \int_{-\infty}^x f(t) dt $$
- Theo định nghĩa của $f(x)$, ta tìm được
$$ F(1.5) = \int_1^{1.5} 3 t^{-4} dt = 1 - (1.5)^{-3} = \dfrac{19}{27} $$
và tương tự
$$ F(2) = 1 - (2)^{-3} = \dfrac{7}{8} $$
do đó $$ P(X < 2 | X \geq 1.5) = \dfrac{7/8 - 19/27}{1 - (19/27)} = \dfrac{37}{64} = 0.578. $$

#### Question 39

A company prices its hurricane insurance using the following assumptions:
- (i) In any calendar year, there can be at most one hurricane.
- (ii) In any calendar year, the probability of a hurricane is 0.05.
- (iii) The numbers of hurricanes in different calendar years are mutually independent.

Using the company’s assumptions, calculate the probability that there are fewer than 3 hurricanes in a 20-year period. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $$ X_i = \left \lbrace \begin{array}{ccl} 1 & \text{ nếu lần quan sát trong năm thứ i có cơn bão} \\ 0 & \text{dell bão (otherwise)} \end{array} \right. $$
là BNN quan sát có cơn bão thứ $i$ xảy ra và $$X = \displaystyle \sum_{i=1}^n X_i$$ với $n$ là BNN thể hiện số cơn bão trong $n$ năm.
- Theo (i), mỗi năm có nhiều nhất 1 cơn bão $X_i$ chỉ nhận đúng 2 giá trị {0, 1} như ta đã thiết lập.
- Theo (ii), xác suất xảy ra mỗi cơn bão là 0.05; có nghĩa là
$$ P(X_i = 1) = 0.05 \text{ và } P(X_i = 0) = 0.95 $$
- Theo (iii), mỗi lần quan sát này là độc lập với nhau nên $X$ là tổng độc lập của các BNN $(X_i)_i$ có cùng phân phối Bernoulli $B(1, p)$ nên $X \sim B(n, p)$ với $p = 0.05$ (phân phối nhị thức)
- Yêu cầu đề bài là tính xác suất xảy ra ít hơn 3 cơn bão trong 20 năm, tức là 
$$ P(X < 3), \quad \text{với } X = \sum_{i=1}^{20} X_i (\text{ tức là } n = 20) $$
- Vì $X$ có phân phối nhị thức $B(20, 0.05)$ nên
$$ P(X < 3) = \sum_{k=0}^2 P(X = k) = \sum_{k=0}^2 C_{20}^k (0.05)^k (1 - 0.05)^{20 - k} = 0.9245. $$

#### Question 40. 

An insurance policy pays for a random loss $X$ subject to a deductible of $C$, where $0 < C < 1$. The loss amount is modeled as a continuous random variable with density function
$$ f(x) = \left\lbrace \begin{array}{ccl} 2x, & x \in (0, 1) \\ 0, & \text{ otherwise} \end{array} \right. $$
Given a random loss $X$, the probability that the insurance payment is less than 0.5 is equal to 0.64.

Calculate $C$.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là chi phí phải trả theo chính sách $X$ và khấu hao $C$, thì
$$ Y = \left\lbrace \begin{array}{ccl} X - C, & X \in (C, 1) \\ 0, & X \in (0, C] \end{array} \right. $$
- Đề yêu cầu tìm $C$ sao cho
$$ P(Y < 0.5) = 0.64 $$
có nghĩa là
$$ P(0 < X-C < 0.5) < 0.64 $$
- Mà
$$ P(0 < X-C < 0.5) = P(X \in (0, 0.5+C) ) = \int_0^{0.5 + C} 2x dx = (0.5 + C)^2 $$
Lúc này 
$$ (0.5 + C)^2 = 0.64 \Leftrightarrow \left[ \begin{array}{cclc} C &=& 0.3 \in (0, 1) & \textbf{ [ nhan ]} \\ C &=& -1.3 < 0 & \textbf{ [ loai ]} \end{array} \right.$$
Do đó, $C = 0.3$
