#### Question 81, 
Claims filed under auto insurance policies follow a `normal distribution` with `mean 19,400` and `standard deviation 5,000`.

Calculate the `probability` that the `average of 25` randomly selected claims `exceeds 20,000`.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là BNN thể hiện số tiền bồi thường thứ $i$ mà BH phải chi trả; ta có $X_i \overset{\text{i.i.d}}{\sim} \mathcal{N}(\mu, \sigma^2)$ và theo giả thiết đầu tiên thì $\mu = 19400$ và $\sigma = 5000$.
- Đề yêu cầu tính XS để trung bình 25 gói bồi thường vượt quá 20000, tức là

$$ P(\bar{X} > 20000), \quad \text{ với } \bar{X} = \frac{1}{25} \sum_{k=1}^{25} X_k $$
- Ta có $X_i \overset{\text{i.i.d}}{\sim} \mathcal{N}(\mu, \sigma^2)$ nên
$$ \bar{X} \sim \mathcal{N}\left( \mu, n^{-1} \sigma^{2} \right) = \mathcal{N}\left( 19400, \dfrac{5000^2}{25} \right) $$
Suy ra

$$ P(\bar{X} > 20000) = P \left( \mathcal{N}(0, 1) > \dfrac{20000 - 19400}{\sqrt{\dfrac{5000^2}{25}}} \right)= P\left[ \mathcal{N}(0,1) > 0.6 \right] = 1 - \Phi(0.6) \approx 0.2743 $$

#### Question 82.
An insurance company issues 1250 vision care insurance policies. The number of claims filed by a policyholder under a vision care insurance policy during one year is a `Poisson random variable` with `mean 2`. Assume the numbers of claims filed by different policyholders are mutually independent.

Calculate the approximate `probability` that there is a total of **between 2450 and 2600 claims** during a `one-year period`?

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là BNN thể hiện số yêu cầu bồi thường của người chủ hợp đồng thứ $i$; theo giả thiết thì $X_i \overset{\text{i.i.d}}{\sim} P(2)$; khi đó

$$ \mathbb{E}X_i = 2 \text{ và Var} X_i = 2, \quad \forall i $$
- Ngoài ra, giả thiết còn cho biết công ty bảo hiểm phát hành 1250 chính sách bảo hiểm chăm sóc thị lực, nên $n = 1250$ rồi kết hợp với giả thiết trước đó, với $S = \displaystyle \sum_{k=1}^{1250} X_k$ là tổng số yêu cầu bồi thường trong các chính sách đã phát hành và ta được

$$ \mathbb{E}S = 1250\cdot 2 = 2500 \text{ và Var}S = 1250\cdot 2 = 2500 $$

- Yêu cầu BT là tính XS để có tổng số 2450 đến 2600 yêu cầu bồi thường trong suốt 1 năm, tức là

$$ P\left( S \in [ 2450, 2600 ] \right) $$

và áp dụng nguyên lý xấp xỉ Poisson bởi phân phối chuẩn [khi $n = 1250$ đủ lớn] và $S \sim \mathcal{N}(\mu, \sigma^2)$ với $\mu = 2500$ và $\sigma^2 = 2500$, ta được

$$ \begin{array}{ccl} P\left( S \in [ 2450, 2600 ] \right) &=& P \left( \dfrac{2400 - \mu}{\sigma} \leq \mathcal{N}(0, 1) \leq \dfrac{2600 - \mu}{\sigma} \right) \\ &=& \Phi \left( \dfrac{2600 - 2500}{\sqrt{1250}} \right) - \Phi \left( \dfrac{2450 - 2500}{\sqrt{1250}} \right) \\ &=& \Phi(2) - \left( 1 - \Phi(1) \right) \\ &=& 0.9772 – (1 – 0.8413) = 0.8185. \end{array} $$

#### Question 83.
A company manufactures a brand of light bulb with a lifetime in months that is `normally distributed` with `mean 3` and `variance 1`. A consumer buys a number of these bulbs with the intention of replacing them successively as they burn out. The light bulbs have mutually independent lifetimes.

Calculate the `smallest number of bulbs` to be purchased so that the succession of light bulbs produces light for `at least 40 months` **with probability `at least 0.9772`**?

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_k$ là BNN thể hiện tuổi thọ của bóng đèn (tính bằng tháng), theo giả thiết đầu tiên thì $X_k \overset{\text{i.i.d}}{\sim} \mathcal{N}(\mu, \sigma^2)$ với $\mu = 3$ và $\sigma^2 = 1$.
- Đặt $n$ là số bóng đèn ít nhất cần mua để XS mà tổng số bóng có thể thắp sáng được ít nhất $40$ tháng với XS tối thiểu $0.9772$, tức là $n$ là nghiệm của phương trình
$$ P \left[ \sum_{k=1}^n X_k > 40 \right] \leq 0.9772 $$
tức là
$$ P \left[ \sum_{k=1}^n X_k \leq 40 \right] \geq 1 - 0.9772 = 0.0228 $$
- Theo giả thiết đầu tiên; ta có
$$ \sum_{k=1}^n X_k \sim \mathcal{N} \left( n\mu, n \sigma^2 \right) = \mathcal{N}(3n, n) $$
suy ra
$$ 0.0228 = P \left[ \sum_{k=1}^n X_k \leq 40 \right] = P \left[ \mathcal{N}(0,1) \leq \dfrac{40 -3n}{\sqrt{n}} \right] $$
mà $$\Phi^{-1}(0.0228) = -1.999077$$ nên
$$ -1.999077 = \dfrac{40 -3n}{\sqrt{n}} \text{ với } n \in \mathbb{N} $$
và ta được $n = 16.$

#### Question 84.
- Let $X$ and $Y$ be the number of hours that a randomly selected person watches movies and sporting events, respectively, during a three-month period. The following information is known about $X$ and $Y$:
$$ \mathbb{E}(X) = 50, \quad \mathbb{E}(Y) = 20, \quad \text{Var}(X) = 50, \text{Var}(Y) = 30, \text{Cov}(X,Y) = 10.$$
- The totals of hours that different individuals watch movies and sporting events during the three months are mutually independent.
- One hundred people are randomly selected and observed for these three months. Let $T$ be the total number of hours that these one hundred people watch movies or sporting events during this three-month period.

Approximate the value of $P[T < 7100]$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Z = X+Y$ là tổng số giờ từ của việc cá nhân đó xem phim và xem thể thao; khi đó
$$ \begin{array}{rccl} & \mathbb{E}Z &=& \mathbb{E}(X) + \mathbb{E}Y = 70 \\ \text{ and} & \\ & \text{Var}Z &=& \text{Var}X + \text{Var}Y + 2\text{Cov}(X, Y) = 100 \end{array} $$
- Đặt $Z_k$ là BNN thể hiện tổng số giờ mà người được khảo sát thứ $k$ đã xem phim hoặc thể thao trong suốt 3 tháng và $Z_k \overset{\text{i.i.d}}{\sim} Z$
- Bài toán cho biết đã khảo sát 100 người trong 3 tháng qua, ta đặt $T = \sum_{k=0}^{100} Z_k$ là tổng thời gian mà những người được khảo sát này đã xem trong thời gian đó. BT yêu cầu ta tính $P[T < 7100]$.
- Từ 2 giả thiết đầu tiên và áp dụng Định lý giới hạn trung tâm; ta được 

$$T \sim \mathcal{N} \left( 100\cdot \mathbb{E}Z, 100\cdot \text{Var}Z \right) = \mathcal{N}(7000, 10000)$$
do đó

$$P[T < 7100] = P\left[ \mathcal{N}(0, 1) < \dfrac{7100 - 7000}{\sqrt{10000}} \right] = P\left[ \mathcal{N}(0, 1) < 1 \right] = \Phi(1) \approx 0.8413. $$

#### Question 85.
The total claim amount for a health insurance policy follows a distribution with `density function` 

$$ f(x) = \dfrac{1}{1000} e^{-(x/1000)}, \quad x > 0 $$

The `premium` for the policy is set at the expected total claim amount plus 100.

If 100 policies are sold, calculate the approximate probability that the insurance company will have claims exceeding the `premiums collected`.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện tổng số tiền mà BH phải chi trả sau khi cộng thêm $100$ vào phí BH cơ bản $X$; tức là $Y = X +100$. Ta có

$$ \begin{array}{lll} \mathbb{E}X &=& \displaystyle \int x f(x) dx = \int_0^{\infty} \dfrac{x}{1000}e^{-(x/1000)} dx = 1000 \\ \mathbb{E}Y &=& \mathbb{E}X + 100 = 1100 \\ \text{Var}X &=& \mathbb{E}X^2 - [\mathbb{E}X ]^2 = \displaystyle \int_0^{\infty} \dfrac{x^2}{1000}e^{-(x/1000)} dx - 1000^2 = 1000000 \end{array} $$

do đó 
$$ \mathbb{E} X = 1000 \text{ và Var}X = 1000000 $$
- Đặt $X_k$ là BNN thể hiện tổng số tiền mà BH phải chi cho KH thứ $k$; khi đó $X_k \overset{\text{i.i.d}}{\sim} X$ và theo Định lý giới hạn trung tâm,
$$ \sum_{k=1}^{100} X_k \sim \mathcal{N}(100 \mathbb{E}X, 100 \text{Var}X) $$
- Ta cần tính

$$ P\left[ \sum_{k=1}^{100} X_k > 100 \mathbb{E}Y \right] =  P\left[ \mathcal{N}(0,1) > \dfrac{100 \mathbb{E}Y - 100 \mathbb{E}X}{\sqrt{100 \text{Var}X}} \right] =  P\left[ \mathcal{N}(0,1) > 1 \right] = 1 - \Phi(1) \approx 0.1587. $$

#### Question 86.
A city has just added 100 new female recruits to its police force. The city will provide a pension to each new hire who remains with the force until retirement. In addition, if the new hire is married at the time of her retirement, a second pension will be provided for
her husband. A consulting actuary makes the following assumptions:
- (i) Each new recruit has a 0.4 probability of remaining with the police force until retirement.
- (ii) Given that a new recruit reaches retirement with the police force, the probability that she is not married at the time of retirement is 0.25.
- (iii) The events of different new hires reaching retirement and the events of different new hires being married at retirement are all mutually independent events.

Calculate the probability that the city will provide at most 90 pensions to the 100 new
hires and their husbands.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_k$ là BNN thể hiện số khoản lương hưu của nhân viên thứ $k$, theo giả thiết thì $X_k$ độc lập và có phân phối xác suất như sau (dựa vào (i) và (ii))

| number of pension $(X_k = x)$ | probabilities $P[X_k = x]$ | Reasons / explainations |
|:-:|:-|:-|
| 0 | 1 - 0.4 | Theo (i), nhân viên sẽ không có lương hưu nếu nghỉ trước tuổi về hưu (XS nghỉ việc khi đến tuổi hưu là 0.4)
| 1 | 0.4*0.25 | Theo (ii), nếu nhân viên đến tuổi hưu và nghỉ, XS người đó còn độc thân là 0.25 nên XS người nhân viên đó chỉ nhận được 1 khoản lương (cho riêng họ) là 0.25*0.4.
| 2 | 0.4*(1-0.25)| Còn lại, nếu đã kết hôn thì họ sẽ nhận được tổng cộng 2 khoản lương hưu (cho họ và cho chồng) |

do đó,

$$ \begin{array}{rcll} & \mathbb{E}X_k &=& 0\cdot P[X_k = 0] + 1\cdot P[X_k = 1] + 2\cdot P[X_k = 2] = 0.7 \\ \text{ và } \\ &\mathbb{E}X_k^2 &=&  0^2\cdot P[X_k = 0] + 1^2\cdot P[X_k = 1] + 2^2\cdot P[X_k = 2] = 1.3 \\ \text{suy ra} \\ & \text{Var}X_k &=& \mathbb{E}X_k^2 - [\mathbb{E}X_k ]^2 = 0.81  \end{array} $$
- Lúc này, $X_k \overset{\text{i.i.d}}{\sim} X_1$ với $\mathbb{E}X_1 = 1.3$ và Var$X_1 = 0.81.$ Áp dụng Định lý giới hạn trung tâm, ta có
$$ S = \sum_{k=1}^{100} X_k \sim \mathcal{N}(100\cdot \mathbb{E}X_1, 100\cdot \text{Var}X_1) $$
- BT yêu cầu trong 100 nhân viên này thì có quá lắm 90 khoản lương sẽ đươc phát cho họ và chồng của họ, tức là $P[S \leq 90]$, do đó

$$ P[S \leq 90] = P\left[ \mathcal{N}(0, 1) \leq \dfrac{90 - 100\mathbb{E}X_1}{\sqrt{100 \text{Var}X_1}} \right] = P[\mathcal{N}(0,1) < 2.22] = \Phi(2.22) \approx 0.98  $$

#### Question 87.
- In an analysis of healthcare data, ages have been rounded to the nearest multiple of 5 years. 
- The difference between the true age and the rounded age is assumed to be `uniformly distributed` on the interval from - 2.5 years to 2.5 years. 
- The healthcare data are based on a random sample of 48 people.

Calculate the approximate probability that the mean of the rounded ages is within 0.25 years of the mean of the true ages. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_k$ là BNN thể hiện hiệu số giữa tuổi thật (`true age`) và tuổi làm tròn (`rounded age`).
- Giả thiết thứ 2 cho ta biết $X_k \overset{\text{i.i.d}}{\sim} \mathcal{U}([-2.5, 2.5])$, tức là

$$ \mathbb{E}X_k = \dfrac{2.5 + (-2.5)}{2} = 0, \text{ và  Var} X_k = \dfrac{[2.5 - (-2.5)]^2}{12} = \dfrac{25}{12} $$
- Đặt $$ \bar{X} = n^{-1} \sum_{k=1}^n X_k $$
là trung bình về hiệu số giữa tuổi thật với tuổi được làm tròn trên mẫu $n$ người được khảo sát. 
- Theo Định lý giới hạn trung tâm, ta có 
$$\bar{X} \sim \mathcal{N}\left( \mathbb{E}X_1, n^{-1}\text{Var}X_1 \right) = \mathcal{N}\left( 0, \dfrac{25}{576} \right) $$
- Giả thiết còn lại cho biết $n = 48$ và yêu cầu BT là tính XS để trung bình mẫu giữa tuổi được làm tròn trừ đi tuổi thật là 0.25, tức là 

$$P\left[ \vert \bar{X} \vert < 0.25 \right],$$ ta có

$$ \begin{array}{ccl} P[-0.25 < \bar{X} < 0.25] &=& P\left[\dfrac{-0.25}{\sqrt{\dfrac{25}{576}}} < \mathcal{N}(0, 1) < \dfrac{0.25}{\sqrt{\dfrac{25}{576}}} \right] \\ &=& \Phi(1.2) - \Phi(-1.2) \\ &=& \Phi(1.2) - (1 - \Phi(1.2)) \\ &=& 0.8849 - (1- 0.8849) = 0.7698 \end{array} $$
 

#### Question 88.
The waiting time for the first claim from a good driver and the waiting time for the first claim from a bad driver are independent and follow exponential distributions with means 6 years and 3 years, respectively.

Calculate the probability that the first claim from a good driver will be filed within 3 years and the first claim from a bad driver will be filed within 2 years

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là sự kiện về năm mà ta nhận được đơn khiếu nại đến các tài xế lái xe tốt (trong vòng 3 năm) và lái xe tệ (trong vòng 2 năm). BT yêu cầu ta tính
$$ P[X \leq 3, Y \leq 2] $$
- Theo giả thiết, $X$ là BNN có phân phối mũ với trung bình bằng 6 nên $P[X \leq x] = 1 - e^{-x/6}$
- Tương tự, $Y$ là BNN có phân phối mũ với trung bình bằng 3 nên $P[Y \leq y] = 1 - e^{-y/3}$
- Vì khả năng lái xe của 2 thằng $X$ và $Y$ là độc lập với nhau, do đó
$$ \begin{array}{ccl} P[X \leq 3, Y \leq 2] &=& P[X \leq 3] \cdot P[X \leq 2] \\ &=& \left( 1 - e^{-3/6} \right)\cdot \left( 1 - e^{-2/3} \right)\\ &=& 1 - e^{-1/2} - e^{-2/3} + e^{-7/6} \end{array} $$

#### Question 89.
The future lifetimes (in months) of two components of a machine have the following `joint density function`: 

$$ f(x,y) = \left\lbrace \begin{array}{rl} \dfrac{6 \cdot (50 - x - y)}{125000}, & 0 < x < 50 - y < 50 \\ 0, & \text{otherwise} \end{array} \right. $$

Determine which of the following represents the probability that both components are still functioning 20 months from now.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài yêu cầu tính XS để cả 2 thành phần $X, Y$ hoạt động từ tháng thứ 20 trở đi, tức là 
$$ P[X > 20, Y > 20] $$
- Lưu ý rằng $(X,Y)$ có hàm MDXS đồng thời là $f(x,y)$ và hàm này bị triệt tiêu bên ngoài miền tam giác
$$ \lbrace (s, t): 0 < s < 50 - t < 50 \rbrace $$
- Vì vậy đặt $A$ là giao của 2 miền,

$$ \begin{array}{ccl} A &=& \lbrace (s, t): 0 < s < 50 - t < 50 \rbrace \cap \lbrace (s, t): s > 20, t > 20 \rbrace \\ &=& \lbrace (s, t): 20 < s < 30; 20 < t < 50 -s \rbrace \end{array} $$

sẽ là miền giá trị mà $f(x,y)$ không bị triệt tiêu khi có ràng buộc; do đó

$$ P[X > 20, Y > 20] = \iint_A f(s, t) ds dt = \dfrac{6}{125000} \int_0^{20} \int_{20}^{50-s}(50 - s - t) ds dt $$

#### Question 90.
- An insurance company sells two types of auto insurance policies: `Basic` and `Deluxe`. 
- The time until the next `Basic Policy` claim is an `exponential random variable` with `mean two days`. 
- The time until the next `Deluxe Policy` claim is an `independent exponential random variable` with `mean three days`.

Calculate the probability that the next claim will be a Deluxe Policy claim. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là các BNN tương ứng với thời gian của các yêu cầu tiếp theo trong gói BH là `Basic` và `Deluxe`
- Theo đó, $X$ có phân phối mũ với trung bình 2 ngày và độc lập với $Y$ có phân phối mũ với trung bình 3 ngày.
- Như vậy, hàm MDXS đồng thời của $(X, Y)$ xác định bởi

$$ f(x, y) = \left( \dfrac{1}{2} e^{-x/2} \right) \cdot \left( \dfrac{1}{3} e^{-x/3} \right) \mathbf{1}_{ \lbrace x \geq 0, y \geq 0 \rbrace } $$
- Đề bài yêu cầu tính XS để yêu cầu kế tiếp sẽ là `Deluxe`, tức $P[X > Y]$, do đó

$$ \begin{array}{ccl} P[X > Y] &=& \displaystyle \int_{0}^{\infty} \int_0^x f(x, y) dy dx \\ &=& \displaystyle \dfrac{1}{6}  \int_{0}^{\infty}  e^{-x/2} \left( \int_0^x e^{-y/2} dy \right) dx \\ &=& \displaystyle \dfrac{1}{2}  \int_{0}^{\infty}  e^{-x/2} \left( 1 - e^{-x/3} \right) dx \\ &=& \displaystyle \lim_{r \to \infty} \left. \left( \dfrac{3}{5}e^{-5x/6} - e^{-x/2} \right) \right\vert_0^r \\ &=& \dfrac{2}{5}. \end{array} $$
