#### Question 21, 
Upon arrival at a hospital’s emergency room, patients are categorized according to their condition as critical, serious, or stable. In the past year:
- (i) 10% of the emergency room patients were critical;
- (ii) 30% of the emergency room patients were serious;
- (iii) the rest of the emergency room patients were stable;
- (iv) 40% of the critical patients died;
- (v) 10% of the serious patients died; and
- (vi) 1% of the stable patients died.

Given that a patient survived, calculate the probability that the patient was categorized as serious upon arrival. 

-----------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C$ (`critical`), $S$ (`serious`), $T$ (`stable`) là các biến cố thể hiện tình trạng của một bệnh nhân (lần lượt nghĩa `"nguy kịch", "nghiêm trọng", "còn tỉnh táo"`) và $D$ là biến cố nếu bệnh nhân đã chết.
- Khi đó, 
$$ \begin{array}{llll} P(C) &=& 0.1 & \text{ theo (i)} \\ P(S) &=& 0.3 & \text{ theo (ii)} \\ P(T) &=& 0.6 & \text{ theo (iii)} \\  P[D^c | C] &=& 1 - P[D | C] = 0.6 & \text{ vì theo (iv):} P[D | C] = 0.4 \\ P[D^c | S] &=& 0.9 & \text{ vì theo (v):} P[D | S] = 0.1 \\ P[D^c | T] &=& 0.99 & \text{ vì theo (vi):} P[D | T] = 0.01 \end{array} $$

- Yêu cầu BT là tính $P[S | D^c]$ và lại theo công thức Bayes, ta có

$$ P[S | D^c] = \dfrac{P[D^c | S] P(S)}{P[D^c | S] P(S) + P[D^c | C] P(C) + P[D^c | T] P(T)} \approx 0.0141. $$

#### Question 22.

- A health study tracked a group of persons for five years. At the beginning of the study, 20% were classified as heavy smokers, 30% as light smokers, and 50% as nonsmokers.
- Results of the study showed that light smokers were twice as likely as nonsmokers to die during the five-year study, but only half as likely as heavy smokers.
- A randomly selected participant from the study died during the five-year period.

Calculate the probability that the participant was a heavy smoker.

----------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $H, L, N$ lần lượt là các biến cố thể hiện mức độ hút thuốc của một người được khảo sát là `"nhiều (heavy)", "ít (light smoker)", "không hút (none)"` và $D$ là biến cố nếu bệnh nhân đó chết trong vòng 5 năm (tính từ thời điểm khảo sát).
- Theo các giả thiết của đề bài, ta có
$$ \begin{array}{lll} P(H) = 0.2, & P(L) = 0.3, & P(N) = 0.5 \end{array} $$
và
$$ \begin{array}{ll} P[D | L] = 2P[D | N], & P[D | L] = 0.5 P[D | H] \end{array} $$
- Yêu cầu BT là tính $P[H | D]$ và

$$ \begin{array}{cll} P[H | D] &=& \dfrac{P[D | H] P(H)}{P[D | H] P(H) + P[D | L] P(L) + P[D | N] P(N)} \\ &=& \dfrac{ (2\cdot P[D | L]) 0.2}{ (0.5 \cdot P[D | L] )\cdot 0.2 + P[D | L] \cdot 0.3 + (2 \cdot P[D | N]) \cdot 0.2} \\ &=& \dfrac{2 \cdot 0.2}{0.5 \cdot 0.2 + 0.3 + 2 \cdot 0.2} \\ &=& 0.42 \end{array} $$

#### Question 23.

An actuary studied the likelihood that different types of drivers would be involved in at least one collision during any one-year period. The results of the study are

| Type of driver | Percentage of all drivers | Probability of at least one collision |
|-|:-:|:-:|
| Teen | 8% | 0.15 |
| Young adult| 16% | 0.08 |
| Midlife| 45% | 0.04|
| Senior | 31% | 0.05 |
| Total | 100% |

$$ $$
Given that a driver has been involved in at least one collision in the past year, calculate the probability that the driver is a young adult driver. 

-----------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C$ là biến cố nếu KH có mua BH tai nạn và $T, Y, M, S$ lần lượt là các biến cố về nhóm tuổi của tài xế: `"Teen", "Young adult", "mid-life", "senior"`.
- Theo các giả thiết trong bảng trên, thì

$$ \begin{array}{rllll} & P(T) = 0.08 & P(Y) = 0.16 & P(M) = 0.45 & P(S) = 0.31 \\ \text{ và} \\ & P[C | T] = 0.15 & P[C | Y] = 0.08 & P[C | M] = 0.04 & P[C | S] = 0.05  \end{array} $$

- BT yêu cầu tính $P[Y | C]$ và lại theo công thức Bayes, ta có

$$ P[Y | C] = \dfrac{P[C | Y] P(Y)}{P[C | Y] P(Y) + P[C | T] P(T) + P[C | M] P(M) + P[C | S] P(S)} \approx 0.22. $$

#### Question 24.
The number of injury claims per month is modeled by a random variable $N$ with 

$$ P[N = n] = \dfrac{1}{(n+1)(n+2)}, \qquad \forall n = 0,1,2, \ldots $$

Calculate the probability of at least one claim during a particular month, given that there have been at most four claims during that month.

---------------

**Giải đáp / SOLUTIONs.**

BT yêu cầu tính XSCDK 

$$ P[N \geq 1 | N \leq 4] = \dfrac{P[(N \geq 1) \cap (N \leq 4)]}{P[N \leq 4]} = \dfrac{\sum_{k=1}^4 P[N = k]}{\sum_{k=0}^4 P[N = k]} = \dfrac{ \frac{20}{60}}{ \frac{50}{ 60}} = \dfrac{2}{5} $$

#### Question 25.
- A blood test indicates the presence of a particular disease 95% of the time when the disease is actually present. 
- The same test indicates the presence of the disease 0.5% of the time when the disease is not actually present. 
- One percent of the population actually has the disease.

Calculate the probability that a person actually has the disease given that the test indicates the presence of the disease.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $D, Y$ lần lượt là các biến cố về số người mắc bệnh hiện tại trong quần thể (xã hội) và số người có kết quả kiểm định là dương tính.
- Lần lượt 3 giả thiết trong bài cho biết $P[Y | D] = 0.95; $ $P[Y | D^c] = 0.005$ và $P[D] = 0.01$.
- Yêu cầu BT là tìm $P[D | Y]$ và lại dùng Bayes, ta được

$$ P[D | Y] = \dfrac{P[Y|D] P(D)}{P[Y|D] P(D) + P[Y|D^c] P(D^c)} = \dfrac{(0.95)(0.01)}{(0.95)(0.01) + (0.005)(1 - 0.01)} = 0.657. $$

#### Question 26.

The probability that a randomly chosen male has a blood circulation problem is 0.25.
Males who have a blood circulation problem are twice as likely to be smokers as those
who do not have a blood circulation problem.

Calculate the probability that a male has a blood circulation problem, given that he is a
smoker.

-----------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $B, S$ lần lượt là các biến cố mà bệnh nhân đó có vấn đề về `tuần hoàn máu (blood circulation problem)` hay `hút thuốc`.
- Theo giả thiết, ta có $P[B] = 0.25$ và $P[S | B] = 2 P[S | B^c]$
- BT yêu cầu tính $P[B | S]$ và ta lại áp dụng công thức Bayes,

$$ P[B | S] = \dfrac{P[S | B] P(B)}{P[S | B] P(B) + P[S | B^c] P(B^c)} = \dfrac{2 P[S | B^c] P(B)}{2P[S | B^c] P(B) + P[S | B^c] P(B^c)} = \frac{2}{5} $$

#### Question 27.
A study of automobile accidents produced the following data: 

| Model year | Proportion of all vehicles | Probability of involvement in an accident |
|-|:-:|:-:|
|2014 | 0.16 | 0.05|
|2013| 0.18 | 0.02 |
|2012| 0.20 | 0.03 |
|Other| 0.46 | 0.04 |

$$ $$
An automobile from one of the model years 2014, 2013, and 2012 was involved in an
accident.
Calculate the probability that the model year of this automobile is 2014.

-----------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y_1, Y_2, Y_3$ lần lượt là biến cố về tai nạn trong các năm `2014, 2013, 2012` và $Z = Y_1 \cup Y_2 \cup Y_3$ là số tai nạn xảy ra một trong 3 năm đó.
- Đề yêu cầu tính XS mà một phương tiện gây ra tai nạn trong năm 2014 biết rằng năm nào trong 3 năm nó cũng gây tai nạn, tức $P[Y_1 | Z]$ và áp dụng công thức Bayes, ta có

$$ \begin{array}{ccl} P[Y_1 | Z] &=& \dfrac{P[Z | Y_1] P(Y_1)}{P[Z | Y_1] P(Y_1) + P[Z | Y_2] P(Y_2) + P[Z | Y_3] P(Y_3)} \\ &=& \dfrac{(0.05)(0.16)}{(0.05)(0.16) + (0.02)(0.18) + (0.03)(0.20)} = 0.45 \end{array} $$

#### Question 28.

- (i) A hospital receives 1/5 of its flu vaccine shipments from Company X and the remainder of its shipments from other companies. Each shipment contains a very large number of vaccine vials.

- (ii) For Company X’s shipments, 10% of the vials are ineffective. For every other company, 2% of the vials are ineffective. The hospital tests 30 randomly selected vials from a shipment and finds that one vial is ineffective.

Calculate the probability that this shipment came from Company X. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, I$ lần lượt là các biến cố "chuyến (hàng) vacxin đến từ cty $X$" và "có một vacxin không có hiệu quả (ineffective)".
- Theo các giả thiết, ta có $P[X] = 1/5$ nên $P[X^c] = 0.8$
- Giả thiết (ii) cho biết

$\qquad \diamond (I | X)$ sẽ tuân theo phân phối nhị thức $B(n, 0.1)$; tức là số vacxin không hiệu quả từ cty $X$.

$\qquad \diamond (I | X^c)$ sẽ tuân theo phân phối nhị thức $B(n, 0.02)$; tức là số vacxin không hiệu quả đé* đến từ cty $X$.

$\qquad \diamond n = 30$ và do đó

$$ \begin{array}{cl} & P[I | X] = P[B(30, 0.10) = 1] = C_{30}^1 (0.1) (1 - 0.1)^{29} = 0.141 \\ \text{  và } \\ & P[I | X^c] = P[B(30, 0.02) = 1] = C_{30}^1 (0.02) (1 - 0.02)^{29} = 0.334  \end{array} $$
- Đề bài yêu cầu tính $P[X | I ]$ và do đó,

$$ P[X | I] = \dfrac{P[I | X] P(X)}{P[I | X] P(X) + P[I | X^c] P(X^c)} \approx 0.096 $$

#### Question 29.
The number of days that elapse between the beginning of a calendar year and the moment a high-risk driver is involved in an accident is exponentially distributed. 

An insurance company expects that 30% of high-risk drivers will be involved in an accident during the first 50 days of a calendar year.

Calculate the portion of high-risk drivers are expected to be involved in an accident during the first 80 days of a calendar year.

----------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số ngày tính đến thời điểm xảy ra tai nạn, khi đó $X \sim$ `Exp`$(\lambda)$ với $\lambda > 0$, hay $X$ có hàm MDXS $f$ xác định bởi

$$ f(x) = \lambda e^{-\lambda x} \mathbf{1}_{\lbrace x \geq 0 \rbrace} $$
- Yêu cầu BT là tìm XS tai nạn trong 80 ngày đầu, hay $P(X \leq 80)$ thông qua hàm MDXS của nó,

$$ P(X \leq 80) = \int_0^{80} \lambda e^{-\lambda x} dx $$
và do đó phải biết thằng $\lambda$ trước!!
- GT (ii) cho biết 30% số vụ tai nạn xảy ra trong 50 ngày đầu tiên, nghĩa là $P(X \leq 50) = 0.3$, giải PT này, ta có

$$ 0.3 = \int_0^{50} \lambda e^{-\lambda x} dx = 1 - e^{-50 \lambda} \Leftrightarrow \lambda = \dfrac{-\ln (0.7)}{50} $$
thay nó ngược lên, ta được
$$ P(X \leq 80) = 1 - e^{-80 \lambda} \approx 0.435. $$

#### Question 30.
- An actuary has discovered that policyholders are three times as likely to file two claims as to file four claims.
- The number of claims filed has a Poisson distribution.
 
Calculate the variance of the number of claims filed. 

----------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số các yêu cầu của KH, và theo giả thiết "XS nhận được 2 yêu cầu cao gấp 3 lần XS khi nhận 4 yêu cầu" nên ta có

$$ P(X = 2) = 3 P(X = 4) $$

- Giả thiết (ii) còn lại cho biết $X$ có phân phối Poisson với trung bình $\lambda > 0$ chưa biết; khi đó phương trình trên tương đương với

$$ \dfrac{e^{-\lambda} \lambda^2}{2!} = 3 \cdot \dfrac{e^{-\lambda} \lambda^4}{4!} \Leftrightarrow \lambda^2 = 4 $$

Kết hợp với lưu ý $\lambda > 0$ ta được $\lambda = 2$, do đó phương sai của một BNN có phân phối Poisson(2) sẽ là $2$.
