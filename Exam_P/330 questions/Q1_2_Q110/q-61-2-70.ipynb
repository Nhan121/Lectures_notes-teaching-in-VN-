#### Question 62.
A random variable X has the cumulative distribution function 
$$ F(x) = \left\lbrace \begin{array}{ll} 0, & x < 1 \\ \dfrac{x^2 - 2x + 2}{2}, & 1 \leq x < 2 \\ 1, & x \geq 2  \end{array} \right. $$

Calculate the variance of X.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta thấy rằng hàm phân phối tích lũy $F(x)$ như trên không liên tục tại $x = 1$ do $F(1) = 0.5$ nhưng $F(1^-) = 0$, ta có công thức tổng quát cho moment bậc $k$ như sau

$$ \mathbb{E}X^k = 1^k \cdot F(1) + \int_1^2 x^k d F(x) $$

và ta tính được $dF(x) = (x - 1) \textbf{1}_{\lbrace x \in (1,2) \rbrace }$.
- Để tính phương sai cho $X$ như câu hỏi BT, chú ý rằng

$$ \text{Var}X = \mathbb{E}X^2 - [\mathbb{E}X ]^2$$

và theo công thức tính moment trên đó, thì

$$ \begin{array}{rcl} \mathbb{E}X &=& \displaystyle 1\cdot (0.5) + \int_1^2 x\cdot (x - 1)dx = 0.5 + \left. \left( \dfrac{x^3}{3} - \dfrac{x^2}{2} \right)\right\vert_1^2 = \dfrac{4}{3} \\ \mathbb{E}X^2  &=& \displaystyle 1^2\cdot (0.5) + \int_1^2 x^2\cdot (x - 1)dx = 0.5 + \left. \left( \dfrac{x^4}{4} - \dfrac{x^3}{3} \right)\right\vert_1^2 = \dfrac{23}{12} \\ \text{Var}X &=& \mathbb{E}X^2 - [\mathbb{E}X ]^2 = \dfrac{23}{12} - \left( \dfrac{4}{3} \right)^2 = \dfrac{5}{36}. \end{array} $$

#### Question 64.
A probability distribution of the claim sizes for an auto insurance policy is given in the table below:

| Claim Size $y_i$ | Probability $P(Y$ = $y_i)$ |
|:-:|:-:|
|20| 0.15 |
|30| 0.10 |
|40| 0.05 |
|50| 0.20 |
|60| 0.10 |
|70| 0.10 |
|80| 0.30 |

Calculate the percentage of claims that are within one standard deviation of the mean claim size. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện số yêu cầu (`claim size`) và theo bảng phân phối xác suất như trên, thì $Y$ nhận 7 giá trị rời rạc $y_1 = 20, y_2 = 30, \ldots, y_7 = 80$ và ta có

$$ \begin{array}{rcl} \mathbb{E}Y &=& \displaystyle \sum_{k=1}^7 y_k \cdot P(Y = y_k) = 20\cdot (0.15) + 30 \cdot (0.10) + \ldots + 80 \cdot (0.30) = 55 \\ \mathbb{E}Y^2 &=& \displaystyle \sum_{k=1}^7 y_k^2 \cdot P(Y = y_k) = 20^2\cdot (0.15) + 30^2 \cdot (0.10) + \ldots + 80^2 \cdot (0.30) = 3500 \\ \text{Var} Y &=& \mathbb{E}Y^2 - [ \mathbb{E} Y ]^2 = 475 \end{array} $$

- Ta cần tính $P\left( \left\vert \frac{Y - \mathbb{E}Y}{1\cdot \sqrt{\text{Var} Y}} \right\vert \leq 1 \right)$ tức là

$$ \begin{array}{rcl} P\left( \left\vert \frac{Y - 55}{\sqrt{ 475 }} \right\vert \leq 1 \right) &=& P\left( |Y - 55| \leq \sqrt{475} \right) \\ &=& P \left( 33.21 \leq Y \leq 76.29 \right) \\ &=& P(Y = 40) + P(Y = 50) + P(Y = 60) + P(Y = 70) \\ &=& 0.05 + 0.20 + 0.10 + 0.10 \\ &=& 0.45 \end{array} $$

#### Question 65.
- The owner of an automobile insures it against damage by purchasing an insurance policy with a `deductible of 250`. 
- In the event that the automobile is damaged, repair costs can be modeled by a `uniform random variable` on the interval `(0, 1500)`.

Calculate the `standard deviation` of the insurance payment in the event that the automobile is damaged. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện chi phí sửa chữa, và theo giả thiết thứ hai, nó tuân theo phân phối đều $\mathcal{U}_{(0, 1500)}$ với hàm MDXS $$f(x) = \dfrac{1}{1500} \mathbf{1}_{\lbrace x \in (0, 1500) \rbrace}$$
- $Y$ là BNN thể hiện khoản tiền mà BH phải chi trả, và theo giả thiết đầu tiên, BH sẽ được thực hiện với khấu hao `250` nên

$$ Y = \max(0, X - 250) $$

- Yêu cầu BT là tìm độ lệch chuẩn của $Y$ tức là $\sqrt{\text{Var} Y}$, ta có

$$ \begin{array}{rcl} \mathbb{E}Y &=& \displaystyle \int [ \max(0, x - 250) ] \cdot f(x) dx = \int_{250}^{1500} (x-250) \dfrac{1}{1500} dx = 521 \\ \mathbb{E}Y^2 &=& \displaystyle \int [ \max(0, x - 250) ]^2 \cdot f(x) dx = \int_{250}^{1500} (x-250)^2 \dfrac{1}{1500} dx = 434028 \\  \text{Var} Y &=& \mathbb{E}Y^2 - [ \mathbb{E} Y ]^2 = 162587\end{array} $$

do đó $\sqrt{\text{Var} Y} \approx 403$

#### Question 67.

Calculate the `standard deviation` of the amount the insurance company will have to pay.

|-|-|-|-|
| Xác suất $P(N = n)$ | $e^{-0.6}$ $\cdot \dfrac{(0.6)}{1!}$ | $e^{-0.6}$ $\cdot \dfrac{(0.6)^2}{2!}$ | $\sum_{n=3}^{\infty} e^{-0.6} \cdot \dfrac{(0.6)^n}{n!}$ |
| Số tiền sẽ trả $Y =$ $Y(N = n)$ | `1000*1 = 1000` | `1000*2 = 2000` | `0` (do hoãn : postpone)|

$$ $$

- Yêu cầu BT là tìm độ lệch chuẩn của chi phí mà BH phải chi trả, tức $\sqrt{\text{Var} Y}$, ta có

#### Question 68.
An insurance policy reimburses dental expense, $X$, up to a maximum benefit of 250. The probability density function for $X$ is:
$$ f(x) = \left\lbrace \begin{array}{ll} ce^{-0.004 x}, & x \geq 0 \\ 0, & \text{otherwise} \end{array} \right. $$
where $c$ is a constant.

Calculate the median benefit for this policy.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đầu tiên, ta cần tìm $c$ để $f$ là hàm MDXS, tức là $$\left\lbrace \begin{array}{rcl} \displaystyle \int f(x) dx &=& 1 \\ f(x) &\geq & 0, \quad \forall x \end{array} \right. ,$$ hê trên tương đương với 
$$\left\lbrace \begin{array}{rcl} \displaystyle c \int_0^{\infty} e^{-0.004 x} dx &=& 1 \\ c &> & 0, \quad \forall x \end{array} \right. \Rightarrow c = 0.004$$
- Để xác định trung vị (median) tức là tìm giá trị $x_{\alpha}$ sao cho 

$$ P\left( X < x_{\alpha} \right) = 0.5 $$

- Với $c$ vừa tìm được, ta dễ dàng tính được hàm phân phối tích lũy

$$ F(x) = P(X < x) = \int_{-\infty}^x f(t) dt = 1 - e^{-0.004 x}$$

do đó, $x_{\alpha}$ chính là nghiệm của phương trình $1 - e^{-0.004 x} = 0.05$ suy ra 

$$ x_{\alpha} = -250 \ln(0.5) \approx 173.29 $$

#### Question 69.
The time to failure of a component in an electronic device has an `exponential distribution` with a `median of four hours`.

Calculate the `probability` that the component will work without failing for `at least five hours`. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ là BNN thể hiện thời gian sử dụng của linh kiện trong thiết bị điện và giả thiết đề bài cho biết $T$ có phân phối mũ với `trung vị (median)` là 4 giờ, tức là $T$ có hàm phân phối tích lũy là
$$ F(t) = 1 - e^{-\lambda t} $$ 
và
$$ P(T < 4) = 0.5 $$
- Từ đó, ta tìm được $\lambda$ thỏa phương trình $1 - e^{-4\lambda} = 0.5$; do đó $$\lambda = \dfrac{-\ln(0.5)}{4}$$
- Yêu cầu BT là tính XS để linh kiện này hỏng sau ít nhất 5 giờ sử dụng, nghĩa là
$$ P(T \geq 5) = 1 - P(T < 5) = 1 - F(5) = 1 - \left[ 1 - \exp\left(\frac{-5\ln(0.5)}{4} \right) \right] \approx 0.42 $$

#### Question 70.
- An insurance company sells an auto insurance policy that covers losses incurred by a policyholder, subject to a `deductible of 100`.
- Losses incurred follow an `exponential distribution` with mean 300.

Calculate the `95th percentile` of losses that exceed the deductible

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện các khoản bị thiệt hại và giả thiết sau cùng cho biết $X$ có phân phối mũ với trung bình $300$.
- Đề bài yêu cầu tìm mức phân vị $x_{\alpha}$ của phân phối có điều kiện sao cho nó vượt qua khoản khấu hao 100 mà cty đã đưa ra, tức là tìm $x_{\alpha}$ thỏa

$$ P \left( \left. X \leq x_{\alpha} \right\vert X > 100 \right) = 0.95 $$

- Áp dụng định nghĩa xs có điều kiện và giả thiết phân phối mũ với hàm phân phối tích lũy $$F(x) = P(X < x) = 1 - e^{-\lambda x} \quad \text{ voi } \lambda = 300,$$ ta được

$$ \dfrac{P(100 < X \leq x_{\alpha})}{P(X > 100)} = 0.95 \Leftrightarrow \dfrac{F(x_{\alpha}) - F(100)}{1 - F(100)} = 0.95 \Leftrightarrow x_{\alpha} = 999. $$
