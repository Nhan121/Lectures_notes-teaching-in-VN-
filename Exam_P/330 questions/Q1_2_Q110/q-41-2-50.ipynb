#### Question 41.
- A study is being conducted in which the `health` of `two independent groups` of 10 `policyholders` is being monitored over a `one-year period` of time. 
- `Individual participants` in the study `drop out` before the end of the study with `probability` 0.2 (`independently of the other participants`).

Calculate the `probability` that at least nine participants complete the study in one of the two groups, but not in `both groups`? 

--------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là BNN thể hiện người được khảo sát thứ $i$ có `"bỏ học (drop out of studying)"` hay không, tức là
$$ X_i = \left \lbrace \begin{array}{ccl} 1 & \text{ nếu SV đó bỏ học} \\ 0 & \text{dek (otherwise)} \end{array} \right. $$
và
$$ X = \sum_{i=1}^{n} X_i $$
là tổng số người bỏ học trong $n$ người tham gia khảo sát. Theo giả thiết, ta có
$$ P(X_i = 1) = 0.2 $$
và như vậy $X \sim B(n, p)$ với $p = 0.2$
- Vì yêu cầu BT là tính XS để có ít nhất 9 trong 10 người được khảo sát đã hoàn thành chương trình học (tối đa 1 người bỏ học) đến từ một trong 2 nhóm $A$ và $B$ **(chứ không phải xuất hiện trong cả 2 nhóm đồng thời)**, tức là có 2 trường hợp: nhóm $A$ hoặc nhóm $B$ (các nhóm đơn) có ít nhất 9 người hoàn thành (khóa học) và nhóm còn lại có nhiều nhất 1 người không hoàn thành, hay

$$ P_A(X \geq 8)P_B(X \leq 1) + P_A(X \leq 1)P_B(X \geq 8) $$
- Mà
$$ P_A(X \leq 1) = P_B(X \leq 1) = P(X \leq 1) = \sum_{k=0}^2 C_{10}^{k} 0.2^k (1 - 0.2)^{10 - k} = 0.376 $$
và
$$ P_A(X \geq 8) = P_B(X \geq 8) = 1 - P(X \leq 1) = 0.624. $$
- Do đó, xác suất cần tìm bằng
$$ 2 \cdot P_A(X \geq 8) \cdot P_B(X \leq 1) = 2\cdot0,376\cdot0.624 = 0.469. $$

#### Question 42.
- For `Company A` there is a `60%` chance that `no claim` is made during the coming year. If `one or more claims` are made, the `total claim amount` is `normally distributed` with `mean 10,000` and `standard deviation 2,000`.
- For `Company B` there is a `70%` chance that `no claim` is made during the coming year. If `one or more claims` are made, the `total claim amount` is `normally distributed` with `mean 9,000` and `standard deviation 2,000`.
- The total claim amounts of the two companies are independent.

Calculate the probability that, in the coming year, `Company B’s total claim amount` will **exceed** `Company A’s total claim amount`. 

--------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A, B$ lần lượt là các BNN thể hiện số hợp đồng trong năm tới của các công ty $A$ và $B$.
- Theo giả thiết về Comp.$A$, ta có
$$ P(A = x) = \left\lbrace \begin{array}{lcr} 0.6, &\text{ khi }& x = 0 \\ 0.4*P(\mathcal{N}(10000, 2000)^2 \leq x), &\text{ khi }& x \geq 1 \end{array} \right.$$
tương tự với Comp.$B$,
$$ P(B = x) = \left\lbrace \begin{array}{lcr} 0.7, &\text{ khi }& x = 0 \\ 0.3*P(\mathcal{N}(9000, 2000)^2 \leq x), &\text{ khi }& x \geq 1 \end{array} \right.$$
- Đề bài yêu cầu, tính xác suất mà tổng hợp đồng của Comp.$B$ vượt Comp.$A$, tức là $P(B - A \geq 1)$
- Để đơn giản hóa, ta sẽ xét 2 trường hợp của biến cố $\lbrace B - A \geq 1 \rbrace$, đó là

$\qquad \diamond$ Khi $A=0$ và $B - A \geq 1$, lúc này khá đơn giản vì khi đó

$$ P(A = 0, B - A \geq 1) = P(A = 0)\cdot P(B \geq 1) = P(A=0)\cdot (1 - P(B= 0)) = 0.6 \cdot (1 - 0.7) = 0.18 $$

$\qquad \diamond$ Khi cả 2 công ty đều nhận hợp đồng và $B - A \geq 1$, thì

$$ \begin{array}{ccl} P(A \geq 1, B \geq 1, B-A \geq 1) &=& P(A \geq 1) \cdot P(B \geq 1)\cdot P\left( \left( B - A \geq 1 \right) | B, A \geq 1 \right) \\ &=& (1 - P(A=0))\cdot(1 - P(B=0)) \cdot P\left( \left( B - A \geq 1 \right) | B, A \geq 1 \right) \\ &=& 0.12 \cdot P\left( \left( B - A \geq 1 \right) | B, A \geq 1 \right) \end{array} $$

$\qquad $ Theo như giả định phân phối khi $x \geq 1$ thì cả $A, B$ đều tuân theo phân phối chuẩn, tức là 

$$ (A | A \geq 1) \sim \mathcal{N}(\mu_A, \sigma_A^2) \text{ và } (B| B \geq 1) \sim \mathcal{N}(\mu_B, \sigma_B^2) $$

$\qquad $ trong đó $\mu_A = 10000, \, \mu_B = 9000$ và $\sigma_A = \sigma_B = 2000,$ suy ra

$$ (B - A | B, A \geq 1) \sim \mathcal{N} \left( \mu_B - \mu_A, \sigma_A^2 + \sigma_B^2 \right) $$
$\qquad $ vì vậy, 

$$ Z = \dfrac{(B - A | B, A \geq 1) - (\mu_B - \mu_A)}{\sqrt{\sigma_A^2 + \sigma_B^2}} \sim \mathcal{N}(0, 1) $$

$\qquad $ và do đó,

$$ P\left( \left( B - A \geq 1 \right) | B, A \geq 1 \right) = P\left( Z \geq \dfrac{1 - (9000 - 10000)}{\sqrt{2000^2 + 2000^2}} \right) \approx 1 - \Phi(0.354) \approx 0.362 $$

$\qquad $ trong đó $\Phi$ là hàm phân phối tích lũy của BNN $\mathcal{N}(0, 1)$.
- Cuối cùng, ta được

$$ P(B - A \geq 1) = 0.18 + 0.12 \cdot 0.362 = 0.223 $$

#### Question 43.

A company takes out an insurance policy to cover accidents that occur at its manufacturing plant. 
- The probability that one or more accidents will occur during any given month is 0.60. 
- The numbers of accidents that occur in different months are mutually independent.

Calculate the probability that there will be at least four months in which no accidents occur before the fourth month in which at least one accident occurs

--------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là quan sát về việc tháng thứ $i$ có tai nạn tại nhà máy sản suất hay không, tức là
$$ X_i = \left \lbrace \begin{array}{ccl} 1 & \text{ nếu có tai nạn (ít nhất 1 tai nạn)} \\ 0 & \text{dek (otherwise)} \end{array} \right. $$
và
$$ X = \sum_{i=1}^n X_i $$
thể hiện số tai nạn trong $n$ tháng
- Giả thiết đầu tiên cho ta biết $P(X_i = 1) = 0.6$, và do đó $X_i \sim B(1, p)$ với $p = 0.6$
- Giả thiết thứ hai kết hợp với hệ quả từ giả thiết đầu; cho biết $X \sim B(n, p)$.
- BT yêu cầu tính XS để `có ít nhất bốn tháng không xảy ra tai nạn xảy ra trước tháng thứ tư`, `trong đó ít nhất một vụ tai nạn xảy ra`. Điều này đồng nghĩa với việc ta phải thực hiện quan sát trong 7 tháng đầu tiên mới có được sự kiện trên, như vậy tổng số tháng cần quan sát là $n = 7$ và điều kiện là có ít nhất 4 tháng không xảy ra tai nạn và **ta cần tìm**

$$ P(X \geq 4) = \sum_{k=4}^7 C_{7}^k \cdot 0.4^k \cdot (1 - 0.4)^{7 - k} = 0.2897.$$

#### Question 44.

An insurance policy pays 100 per day for up to three days of hospitalization and 50 per day for each day of hospitalization thereafter.

The number of days of hospitalization, $X$, is a discrete random variable with probability function

$$ P(X = k) = \left \lbrace \begin{array}{ccl} \dfrac{6-k}{15} & \text{ if} & k = 1,2,3,4,5 \\ 0, & &\text{ otherwise}& \end{array} \right. $$

Determine the expected payment for hospitalization under this policy. 

--------------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số ngày tại bệnh viện mà bảo hiểm sẽ chi trả và $Z = Z(X)$ là số tiền mà bảo hiểm sẽ chi trả
- Từ giả thiết, bảo hiểm sẽ chi trả $100$ cho 3 ngày đầu tiên và 50 cho các ngày kế tiếp, và do đó ta được bảng quan sát sau

| k | $1$ | $2$ | $3$ | $4$ | $5$ | `(`$> 5$`)` |
|-|:-:|:-:|:-:|:-:|:-:|:-:|
| P(X = k) | 5/15 | 4/15 | 3/15 | 2/15 | 1/15 | 0 |
| Z(X = k) | `100` | `2*100` | `3*100` | `3*100` + `(4-3)*50` | `3*100 + (5 - 3)*50` | `3*100 + (k-3)*50` |

do đó,

$$ \begin{array}{ccl} \mathbb{E}Z &=& \displaystyle \sum_{k \geq 1} Z(X = k) \cdot P(X = k) \\ &=& \left( 100\cdot \dfrac{5}{15} \right) + \left( 200\cdot \dfrac{4}{15} \right) + \left( 300\cdot \dfrac{3}{15} \right) + \left( 350\cdot \dfrac{2}{15} \right) + \left( 400\cdot \dfrac{1}{15} \right) + 0 \\ &=& 220 \end{array} $$

#### Question 45.

Let X be a continuous random variable with density function $$ f(x) = \dfrac{|x|}{10} \cdot \mathbf{1}_{x \in [-2, 4]} = \left \lbrace \begin{array}{ccl} \dfrac{|x|}{10} & \text{ if} & -2 \leq x \leq 4 \\ 0, & &\text{ otherwise}& \end{array} \right.  $$

Calculate the expected value of X. 

----------------------------

**Giải đáp / SOLUTIONs.**

Dễ dàng kiểm chứng $f(x)$ là hàm MĐXS, nên kỳ vọng của $X$ sẽ được tính bởi

$$ \mathbb{E}X = \int_{\mathbb{R}} x f(x) dx = \int_{-2}^4 \dfrac{|x|}{10} dx = \dfrac{28}{15},$$

#### Question 46.

A device that continuously measures and records seismic activity is placed in a remote region. 
- The time, $T$, to failure of this device is exponentially distributed with mean 3 years. 
- Since the device will not be monitored during its first two years of service, the time to discovery of its failure is $X = \max(T, 2)$.

Calculate $\mathbb{E}(X)$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Giả thiết đầu tiên cho biết $T$ có phân phối mũ với trung bình 3 năm nên $T$ có hàm MĐXS là $f(x) = \dfrac{1}{3} e^{-x/3} \textbf{1}_{ \lbrace x \geq 0 \rbrace }$
- Từ giả thiết thứ hai, suy ra
$$ \begin{array}{ccl} \mathbb{E}X &=& \mathbb{E} \max(T, 2) \\ &=& \displaystyle \int_{\mathbb{R}} \max(x, 2) f(x)dx \\ &=& \displaystyle \int_0^{2} 2 f(x) dx + \int_2^{\infty} x f(x) dx \\ &=& \displaystyle \dfrac{2}{3} \int_0^2 e^{-x/3} dx + \dfrac{1}{3} \int_2^{\infty} x e^{-x/3} dx = 2 + 3e^{-2/3} \end{array} $$

#### Question 47.

A piece of equipment is being insured against early failure. 
- The time from purchase until failure of the equipment is exponentially distributed with mean 10 years.
- The insurance will pay an amount x if the equipment fails during the first year, and it will pay 0.5x if failure occurs during the second or third year. If failure occurs after the first three years, no payment will be made.

Calculate x such that the expected payment made under this insurance is 1000. 

-----------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ là BNN thể hiện thời gian sử dụng ,cho đến khi xuất hiện hư hỏng thiết bị (tính bằng năm); và theo giả thiết thứ nhất $T$ có phân phối mũ với trung bình 10 năm, nên $T$ có hàm MĐXS $f(t) = \dfrac{1}{10} e^{-t/10} \mathbf{1}_{ \lbrace x \geq 0 \rbrace }$
- Đặt $F$ `(fee)` là phí chi trả của phía bảo hiểm và theo giả thiết còn lại thì

$\qquad \diamond$ BH chi trả $x$ (USD / GBP) nếu thiết bị hỏng ngay trong năm đầu

$\qquad \diamond$ trả $0.5 x$ nếu hỏng trong năm thứ hai hoặc thứ ba.

$\qquad \diamond$ $0$ nếu thiết bị hỏng sau 3 năm.

Do đó, số tiền mà BH phải trả được kỳ vọng ở mức

$$ \begin{array}{ccl} \mathbb{E} (F) &=& x\cdot P(T \leq 1) + 0.5x \cdot P( 1 < T \leq 3) + 0 \cdot P( T > 3) \\ &=& \displaystyle x \int_0^1 f(t) dt + 0.5x \int_1^3 f(t) dt \\ &=& 0.1772 \cdot x\end{array} $$

- Đề bài yêu cầu tìm $x$ để mức tiền chi trả là $1000$, nghĩa là

$$ \mathbb{E} F = 1000 \Leftrightarrow 0.1772 x = 1000 \Leftrightarrow x = 5644$$

Vậy $x = 5644$

#### Question 48.

An insurance policy on an electrical device pays a benefit of 4000 if the device fails during the first year. 
- The amount of the benefit decreases by 1000 each successive year until it reaches 0. 
- If the device has not failed by the beginning of any given year, the probability of failure during that year is 0.4.

Calculate the expected benefit under this policy. 

--------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $F$ là số tiền mà chính sách này sẽ chi trả và $E = k$ là biến cố thiết bị điện xuất hiện tình trạng hư hỏng trong năm thứ $k$ với $k \geq 1$
- Theo giả thiết đầu tiên, nếu thiết bị điện có hư/hỏng vào năm đầu BH sẽ chi trả 4000
- Và giả thiết thứ hai; mỗi năm sau đó sẽ giảm đi 1000 cho đến khi dừng ở 0 đồng.
- Ở giả thiết còn lại, các thiết bị điện hỏng trong một năm với xác suất 0.4, tức là không hỏng với XS 0.6. Điều này cho biết xác suất thiết bị điện hỏng tại năm thứ $k$ (nghĩa là $k-1$ năm trước đó đéo hỏng) sẽ được tính bởi công thức
$$ P(E = k) = 0.4 \cdot (1 - 0.4)^{k-1}, \qquad \forall k = 1,2,\ldots $$

Từ đây, ta nhận được bảng phân phối xác suất như sau

| k (năm xuất hiện tình trạng hư hỏng) | $1$ | $2$ | $3$ | $4$ | $>4$ |
|-|:-:|:-:|:-:|:-:|:-:|
| Tiền phải trả $F(k)$ nếu hỏng ở năm thứ $k$ | 4000 | 3000| 2000| 1000 | 0 |
|Xác suất hỏng `P(E = k)`| `0.4` | `0.4*0.6` | `0.4*(0.6)^2` | `0.4*(0.6)^3` | `0.4*(0.6)^(k-1)`

do đó, số tiền phải trả của chính sách này là

$$ \mathbb{E}(F) = 4000\cdot 0.4 + 3000\cdot 0.4 \cdot (0.6)^1 + 2000\cdot 0.4 \cdot (0.6)^2 + 1000\cdot 0.4 \cdot (0.6)^3 = 2694. $$

#### Question 50.

A company buys a policy to insure its revenue in the event of major snowstorms that shut down business. 
- The policy pays nothing for the first such snowstorm of the year and 10,000 for each one thereafter, until the end of the year. 
- The number of major snowstorms per year that shut down business is assumed to have a `Poisson distribution` with mean 1.5.

Calculate the expected amount paid to the company under this policy during a one-year period,

------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ là BNN thể hiện số cơn bão lớn hàng năm có thể khiến các doanh nghiệp tạm dừng sản xuất. Theo giả thiết thứ 2 thì $T$ tuân theo phân phối Poisson với trung bình $1.5$
- Đặt $F(T)$ là chi phí mà BH phải trả theo số cơn bão $T$, và giả thiết đầu tiên cho biết, mỗi năm hãng sẽ không chi tiền trong cơn bão đầu tiên và sẽ chi 10000 cho mỗi cơn bão sau đó, điều này có nghĩa nếu có $k \geq 1$ cơn bão trong năm thì 
$$ F(T=k) = 10000 (k-1) ; \qquad k \geq 1. $$
- Như vậy, số tiền mà BH sẽ chi trả được tính bởi công thức

$$ \begin{array}{ccl} \mathbb{E}(F(T)) &=& \displaystyle \sum_{k=1}^{\infty} F(k) \cdot P(T = k) \\ &=& \displaystyle \sum_{k=1}^{\infty} 10000(k-1) \cdot \left( \dfrac{(1.5)^k e^{-1.5}}{k!} \right) \\ &=& 10000 \displaystyle \left[ \left( \sum_{k=0}^{\infty} (k-1) \cdot \left( \dfrac{(1.5)^k e^{-1.5}}{k!} \right) \right) - (0 - 1) \cdot \left( \dfrac{(1.5)^0 e^{-1.5}}{0!} \right) \right] \\ &=& 10000 \left[ \; \mathbb{E}(T - 1) \; + \; e^{-3/2} \right] \\ &=& 10000 \left( 0.5 + e^{-3/2} \right) \\ &=& 7321 \end{array} $$

trong đó, 2 dấu bằng kế cuối có được bởi $T$ là BNN có phân phối `Poisson(1.5)` nên $\mathbb{E}(T) = 1.5$ và lưu ý rằng

$$ \mathbb{E}(T - 1) = \sum_{k=0}^{\infty} (k-1) P(T = k) $$

Như vậy, số tiền phải chi trả là $7321.$
