#### Question 41.
- A study is being conducted in which the `health` of `two independent groups` of 10 `policyholders` is being monitored over a `one-year period` of time. 
- `Individual participants` in the study `drop out` before the end of the study with `probability` 0.2 (`independently of the other participants`).

Calculate the `probability` that at least nine participants complete the study in one of the two groups, but not in `both groups`? 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là BNN thể hiện người được khảo sát thứ $i$ có `"bỏ học (drop out of studying)"` hay không, tức là
$$ X_i = \left \lbrace \begin{array}{ccl} 1 & \text{ nếu SV đó bỏ học} \\ 0 & \text{dek (otherwise)} \end{array} \right. $$
và
$$ X = \sum_{i=1}^{n} X_i $$
là tổng số người bỏ học trong $n$ người tham gia khảo sát. Theo giả thiết, ta có
$$ P(X_i = 1) = 0.2 $$
và như vậy $X \sim B(n, p)$ với $p = 0.2$
- Vì yêu cầu BT là tính XS để có ít nhất 9 trong 10 người được khảo sát đã hoàn thành chương trình học (tối đa 1 người bỏ học) đến từ một trong 2 nhóm $A$ và $B$ **(chứ không phải xuất hiện trong cả 2 nhóm đồng thời)**, tức là có 2 trường hợp: nhóm $A$ hoặc nhóm $B$ (các nhóm đơn) có ít nhất 9 người hoàn thành (khóa học) và nhóm còn lại có nhiều nhất 1 người không hoàn thành, hay

$$ P_A(X \geq 8)P_B(X \leq 1) + P_A(X \leq 1)P_B(X \geq 8) $$
- Mà
$$ P_A(X \leq 1) = P_B(X \leq 1) = P(X \leq 1) = \sum_{k=0}^2 C_{10}^{k} 0.2^k (1 - 0.2)^{10 - k} = 0.376 $$
và
$$ P_A(X \geq 8) = P_B(X \geq 8) = 1 - P(X \leq 1) = 0.624. $$
- Do đó, xác suất cần tìm bằng
$$ 2 \cdot P_A(X \geq 8) \cdot P_B(X \leq 1) = 2\cdot0,376\cdot0.624 = 0.469. $$

#### Question 42.
- For `Company A` there is a `60%` chance that `no claim` is made during the coming year. If `one or more claims` are made, the `total claim amount` is `normally distributed` with `mean 10,000` and `standard deviation 2,000`.
- For `Company B` there is a `70%` chance that `no claim` is made during the coming year. If `one or more claims` are made, the `total claim amount` is `normally distributed` with `mean 9,000` and `standard deviation 2,000`.
- The total claim amounts of the two companies are independent.

Calculate the probability that, in the coming year, `Company B’s total claim amount` will **exceed** `Company A’s total claim amount`. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A, B$ lần lượt là các BNN thể hiện số hợp đồng trong năm tới của các công ty $A$ và $B$.
- Theo giả thiết về Comp.$A$, ta có
$$ P(A = x) = \left\lbrace \begin{array}{lcr} 0.6, &\text{ khi }& x = 0 \\ 0.4*P(\mathcal{N}(10000, 2000)^2 \leq x), &\text{ khi }& x \geq 1 \end{array} \right.$$
tương tự với Comp.$B$,
$$ P(B = x) = \left\lbrace \begin{array}{lcr} 0.7, &\text{ khi }& x = 0 \\ 0.3*P(\mathcal{N}(9000, 2000)^2 \leq x), &\text{ khi }& x \geq 1 \end{array} \right.$$
- Đề bài yêu cầu, tính xác suất mà tổng hợp đồng của Comp.$B$ vượt Comp.$A$, tức là $P(B - A \geq 1)$
- Để đơn giản hóa, ta sẽ xét 2 trường hợp của biến cố $\lbrace B - A \geq 1 \rbrace$, đó là

$\qquad \diamond$ Khi $A=0$ và $B - A \geq 1$, lúc này khá đơn giản vì khi đó

$$ P(A = 0, B - A = 1) = P(A = 0)\cdot P(B \geq 1) = P(A=0)\cdot (1 - P(B= 0)) = 0.6 \cdot (1 - 0.7) = 0.18 $$

$\qquad \diamond$ Khi cả 2 công ty đều nhận hợp đồng và $B - A \geq 1$, thì

$$ \begin{array}{ccl} P(A \geq 1, B \geq 1, B-A \geq 1) &=& P(A \geq 1) \cdot P(B \geq 1)\cdot P(B - A \geq 1) \\ &=& (1 - P(A=0))\cdot(1 - P(B=0)) \cdot P(B - A \geq 1) \\ &=& 0.12 \cdot P(B - A \geq 1) \end{array} $$

$\qquad $ Theo như giả định phân phối khi $x \geq 1$ thì cả $A, B$ đều tuân theo phân phối chuẩn, tức là 

$$ (A \geq 1) \sim \mathcal{N}(\mu_A, \sigma_A^2) \text{ và } (B \geq 1) \sim \mathcal{N}(\mu_B, \sigma_B^2) $$

$\qquad $ trong đó $\mu_A = 10000, \, \mu_B = 9000$ và $\sigma_A = \sigma_B = 2000,$ suy ra

$$ B - A \sim \mathcal{N} \left( \mu_B - \mu_A, \sigma_A^2 + \sigma_B^2 \right) $$
$\qquad $ vì vậy, 

$$ Z = \dfrac{B-A - (\mu_B - \mu_A)}{\sqrt{\sigma_A^2 + \sigma_B^2}} \sim \mathcal{N}(0, 1) $$

$\qquad $ và do đó,

$$ P(B - A \geq 1) = P\left( Z \geq \dfrac{1 - (9000 - 10000)}{\sqrt{2000^2 + 2000^2}} \right) \approx 1 - \Phi(0.354) \approx 0.362 $$

$\qquad $ trong đó $\Phi$ là hàm phân phối tích lũy của BNN $\mathcal{N}(0, 1)$.
- Cuối cùng, ta được

$$ P(B - A \geq 1) = 0.18 + 0.12 \cdot 0.362 = 0.223 $$
