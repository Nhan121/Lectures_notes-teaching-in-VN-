#### Question 11, 

An actuary studying the insurance preferences of automobile owners makes the following conclusions:
- (i) An automobile owner is twice as likely to purchase collision coverage as disability coverage.
- (ii) The event that an automobile owner purchases collision coverage is independent of the event that he or she purchases disability coverage.
- (iii) The probability that an automobile owner purchases both collision and disability coverages is 0.15.

Calculate the probability that an automobile owner purchases neither collision nor disability coverage.

--------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C$ và $D$ lần lượt là biến cố khách hàng mua `BH tai nạn` và `BH khuyết tật`.
- Khi đó:

$\qquad \diamond$ Theo (i), ta có $P(C) = 2P(D)$

$\qquad \diamond$ Theo (iii) thì $P(C\cap D) = 0.15$

$\qquad \diamond$ Theo (ii), thì $C, D$ độc lập nên $P(C \cap D) = P(C) \cdot P(D)$ và kết hợp với (i) ta được 

$$P(C \cap D) = 2\left[ P(D) \right]^2,$$ 

$\qquad \quad$ kết hợp với (iii) thì 

$$0.15 = 2 [ P(D) ]^2$$ 

$\qquad \quad$ do đó $P(D) = \sqrt{0.075}$ và $P(C) = 2\sqrt{0.075}$

- Đề bài yêu cầu tính XS để khách hàng không mua loại BH nào trong 2 loại kể trên, tức là $P(C^c \cap D^c)$. Lúc này, vì (ii) nên $C, D$ độc lập cũng kéo theo phần bù của nó (tức $C^c, D^c$) độc lập, vì vậy

$$ P(C^c \cap D^c) = P(C^c) P(D^c) = \left( 1 - P(C) \right) \left( 1 -P(D) \right) \approx 0.33 $$

#### Question 13.

- (i) An actuary is studying the prevalence of three health risk factors, denoted by $A$, $B$, and $C$, within a population of women. 
- (ii) For each of the three factors, the probability is 0.1 that a woman in the population has only this risk factor (and no others). For any two of the three factors, the probability is 0.12 that she has exactly these two risk factors (but not the other). 
- (iii) The probability that a woman has all three risk factors, given that she has A and B, is 1/3.

Calculate the probability that a woman has none of the three risk factors, given that she does not have risk factor A. 

------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ (iii), ta được $$P[ (A \cap B \cap C) | (A \cap B) ] = 1/3$$
- Từ (ii), ta được
$$ P(A \cap B \cap C^c) = P(A \cap B^c \cap C) = P(A^c \cap B \cap C) = 0.12 $$
- Từ (i),
$$ P(A \cap B^c \cap C^c) = P(A^c \cap B \cap C^c) = P(A^c \cap B^c \cap C) = 0.1 $$
- Yêu cầu BT là tính XSCĐK
$$ P\left( \left. (A^c \cap B^c \cap C^c) \right\vert A^c \right) = \dfrac{P[\left( A^c \cap B^c \cap C^c \right) \cap A^c]}{P[A^c]} = \dfrac{P[\left( A \cup B \cup C \right)^c]}{P[A^c]} = \dfrac{1 - P[\left( A \cup B \cup C \right)]}{1 - P[A]}$$
Mà ta có

$\qquad \qquad A \cup B \cup C =$ $\lbrace \exists$ chỉ duy nhất 1 triệu chứng $\rbrace \cup $ $\lbrace \exists$ đúng 2 triệu chứng $\rbrace \cup $ {dính cả 3}

nên

$\qquad \qquad P[A \cup B \cup C] =$ $P [\lbrace \exists$ chỉ duy nhất 1 triệu chứng $\rbrace ] + $ $P[ \lbrace \exists$ đúng 2 triệu chứng $\rbrace ] +$ $P[A \cap B \cap C]$

$\qquad \diamond $  $P [\lbrace \exists$ chỉ duy nhất 1 triệu chứng $\rbrace ] = P(A \cap B^c \cap C^c) + P(A^c \cap B \cap C^c) + P(A^c \cap B^c \cap C) = 0.3 $

$\qquad \diamond $  $P [\lbrace \exists$ có đúng 2 triệu chứng $\rbrace ] = P(A \cap B \cap C^c) + P(A \cap B \cap C^c) + P(A^c \cap B \cap C) = 0.36$

Như vậy $P(A \cup B \cup C) = 0.66 + P(A \cap B \cap C)$

- Hơn nữa,

$$ \begin{array}{ccl} P(A) &=& P(A \cap B^c) + P(A \cap B ) \\ &=& \left( P[ (A \cap B^c) \cap C^c ] + P[ (A \cap B^c) \cap C] \right) + \left( P[(A \cap B) \cap C^c] + P[ (A \cap B) \cap C] \right) \\ &=& 0.34 + P(A \cap B \cap C) \end{array} $$

- Do đó $$ P\left( \left. (A^c \cap B^c \cap C^c) \right\vert A^c \right) = \dfrac{1 - (0.66 + P(A \cap B \cap C))}{1 - \left( 0.34 + P(A \cap B \cap C) \right)} = \dfrac{0.34 + P(A \cap B \cap C)}{0.66 - P(A \cap B \cap C)} $$

- Vì vậy ta sẽ tìm $P(A \cap B \cap C)$, đặt $x = P(A \cap B \cap C)$ và theo (iii), ta có

$$ \dfrac{1}{3} = \dfrac{P [ (A \cap B \cap C) \cap (A \cap B) ]}{P(A \cap B)} = \dfrac{P(A \cap B \cap C)}{P(A \cap B)}  $$

Lại có $P(A \cap B) = P(A \cap B \cap C) + P(A \cap B \cap C^c) = x + 0.12$; do đó 

$$\dfrac{1}{3} = \dfrac{x}{x+0.12} \Leftrightarrow x = 0.06 $$

cuối cùng, ta được

$$ P\left( \left. (A^c \cap B^c \cap C^c) \right\vert A^c \right) = \dfrac{0.34 - 0.06}{0.66 - 0.06} \approx 0.467. $$

#### Question 14.

In modeling the number of claims filed by an individual under an automobile policy during a three-year period, an actuary makes the simplifying assumption that for all integers $n \geq 0$, $p(n+1) = 0.2 p(n)$ where $p(n)$ represents the probability that the policyholder files $n$ claims during the period.

Under this assumption, calculate the probability that a policyholder files more than one claim during the period. 

---------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số yêu cầu được nhận trong khoảng thời gian đó.
- Yêu cầu BT là tính XS để nhận được nhiều hơn 1 yêu cầu, tức $P(X > 1)$.
- Theo giả thiết thì với $n \geq 0$ thì

$$ P(X = n) = p(n) = 5p(n+1) $$

Đặt $a_n = p(n)$ thì $(a_n)_n$ trở thành cấp số nhân với công bội $0.2$ và ta thu được hệ sau 

$$ \left\lbrace \begin{array}{ccl} a_{n+1} &=& 0.2 a_n \\ \displaystyle \sum_{n=0}^{\infty} a_n &=& 1 \end{array} \right. \Leftrightarrow a_0 = \dfrac{4}{5} $$

Như vậy

$$ P(X > 1) = 1 - P(X = 0) - P(X = 1) = 1 - a_0 - 0.2 a_0 = \dfrac{4}{25}. $$

#### Question 15.

- An insurer offers a health plan to the employees of a large company. As part of this plan, the individual employees may choose exactly two of the supplementary coverages $A, B,$ and $C,$ or they may choose no supplementary coverage. 
- The proportions of the company’s employees that choose coverages $A, B,$ and $C$ are $1/4, 1/3$, and $5/12$ respectively.

Calculate the probability that a randomly chosen employee will choose no supplementary coverage.

---------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A, B, C$ là biến cố chọn các hạng mục BH $A, B, C$ và theo giả thiết.
- Đặt $x$ là xác suất chọn $A, B$ nhưng không chọn $C$, tức $x = P(A \cap B \cap C^c)$

 $\quad y$ là xác suất chọn $A, C$ nhưng không chọn $B$, tức $y = P(A \cap B^c \cap C)$

 $\quad z$ là xác suất chọn $B, C$ nhưng không chọn $A$, tức $z = P(A^c \cap B \cap C)$
 
 $\quad w$ là xác suất không chọn cả $A, B, C$, tức $w = P(A^c \cap B^c \cap C^c)$
 
- Yêu cầu BT là tính XS để nhân viên không chọn hạng mục BH nào, tức là $w$.

- Theo giả thiết (câu thứ 2) thì các nhân viên chỉ chọn đúng 2 hạng mục hoặc không chọn hạng mục nào nên các trường hợp chỉ có đúng 1 hạng mục hay cả 3 hạng mục sẽ không có và xác suất các trường hợp không đươc liệt kê là $0$, do đó
$$ x + y + z + w  = 1 $$

- Giả thiết cuối cùng trong BT giúp ta thiết lập được hệ sau $$ \left\lbrace \begin{array}{ccl} x+y &=& 1/4 \\ x+z &=& 1/3 \\ y+z &=& 5/12 \end{array} \right. $$

- Giải hệ trên và ta được $w = 0.5$.

#### Question 16. 

An insurance company determines that $N$, the number of claims received in a week, is a random variable with

$$ P[N = n] = \dfrac{1}{2^{n+1}}, \quad n \geq 0. $$

The company also determines that the number of claims received in a given week is independent of the number of claims received in any other week.

Calculate the probability that exactly seven claims will be received during a given two-week period.

--------------------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N_1, N_2$ lần lượt là các BNN thể hiện số yêu cầu được nhận tương ứng với tuần 1 và tuần 2.
- Giả thiết cuối cùng trong đề cho biết $N_1$ và $N_2$ độc lập với nhau.
- Yêu cầu BT là tính $P(N_1 + N_2 = 7)$, do đó

$$ \begin{array}{ccl} P(N_1 + N_2 = 7) &=& \displaystyle \sum_{k=0}^7 P(N_1 = k, N_2 = 7 - k) \\ &=& \displaystyle \sum_{k=0}^7 P(N_1 = k) P(N_2 = 7 - k) \\ &=& \displaystyle \sum_{k=0}^7 2^{-(k+1)} \cdot 2^{-[(7 - k) + 1]} = \dfrac{8}{2^9} = \dfrac{1}{64}. \\ \end{array} $$

#### Question 17.

- An insurance company pays hospital claims. The number of claims that include emergency room or operating room charges is 85% of the total number of claims. 
- The number of claims that do not include emergency room charges is 25% of the total number of claims. 
- The occurrence of emergency room charges is independent of the occurrence of operating room charges on hospital claims.

Calculate the probability that a claim submitted to the insurance company includes operating room charges.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $O, E$ lần lượt là các biến cố mà chi phí BH phải trả cho `"phòng giải phẫu (operating room)"` hoặc `"phòng cấp cứu (emergency room)"`; khi đó

$$ P(O \cup E) = 0.85 \quad \text{ và } \quad  P(E^c) = 0.25 $$

- Giả thiết cuối cùng cho biết $O, E$ độc lập với nhau, do đó

$$ P(O \cap E) = P(O) P(E) $$
hơn nữa, nhớ rằng
$$ \begin{array}{cccl} & P(O \cup E) &=& P(O) + P(E) - P(O \cap E) \\ \Rightarrow  & P(O \cup E) &=& P(O) + P(E) - P(O)P(E) \\ \Rightarrow  & P(O \cup E) &=& P(O) + [1 - P(E^c)] - P(O)[ 1 - P(E^c) ] \end{array} $$

- Yêu cầu BT là tính $P(O)$, thay các giá trị đã biết, ta tìm được $P(O) = 0.4$

#### Question 18.

- Two instruments are used to measure the height, $h$, of a tower. The error made by the less accurate instrument is normally distributed with mean 0 and standard deviation $0.0056h$. 
- The error made by the more accurate instrument is normally distributed with mean $0$ and standard deviation $0.0044h$.
- The errors from the two instruments are independent of each other.

Calculate the probability that the average value of the two measurements is within $0.005h$ of the height of the tower. 

------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là các BNN thể hiện sai số của các thiết bị đo đạc chiều cao $h$ tại tòa tháp nào đó.
- Theo 2 giả thiết đầu tiên thì cả $X, Y$ đều có phân phối chuẩn với $X \sim \mathcal{N}(\mu_X = 0, \sigma_X = 0.0056h)$ và $Y \sim \mathcal{N}(\mu_Y = 0, \sigma_Y = 0.0044h)$ và theo giả thiết sau cùng thì $X, Y$ độc lập với nhau, do đó

$$ X + Y \sim \mathcal{N} \left( \mu_X + \mu_Y, \sigma = \sqrt{\sigma_X^2 + \sigma_Y^2} \right) $$

- Yêu cầu BT là tính XS để trung bình 2 thiết bị này có sai số không vượt quá $0.005h$, tức là $P\left( \left\vert \dfrac{X+Y}{2} \right\vert \leq 0.005h \right)$, mà ta có

$$ \dfrac{X+Y}{2} \sim \mathcal{N} \left( 0, 0.00356h \right) $$

do đó $Z = \dfrac{X+Y}{0.00356h} \sim \mathcal{N}(0, 1)$ và

$$ \begin{array}{ccl} P\left( \left\vert \dfrac{X+Y}{2} \right\vert \leq 0.005h \right) &=& P\left( - \dfrac{0.005h}{0.00356h} \leq Z \leq \dfrac{0.005h}{0.00356h} \right) \\ &=& P(-1.4 \leq Z \leq 1.4) \\ &=& \Phi(1.4) - \Phi(-1.4) \\ &=& 2\Phi(1.4) - 1 \approx 0.84  \end{array} $$

trong đó $\Phi(z) = P(\mathcal{N}(0, 1) < z).$

#### Question 19.

An auto insurance company insures drivers of all ages. An actuary compiled the following statistics on the company’s insured drivers:

| Age of driver | Probability of Accident | Portion of Company’s Insured Drivers |
|:-:|:-:|:-:|
| 16-20 | 0.06 | 0.08 |
| 21-30 | 0.03 | 0.15 |
| 31-65 | 0.02 | 0.49 |
| 66-99 | 0.04 | 0.28 |

A randomly selected driver that the company insures has an accident.

Calculate the probability that the driver was age `16-20`. 

------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A$ `(accident)` là biến cố nếu có tai nạn xảy ra.
- Đặt $G_1, G_2, G_3, G_4$ lần lượt là là biến cố liên quan đến nhóm tuổi của tài xế `"16-20"`; `"21-30"`; `"31-65"` và `"66-99"`
- Đề bài yêu cầu tính $P( \left. G_1 \right\vert A)$, ta sẽ dùng công thức Bayes để tìm nó, nhưng trước tiên cần kiểm tra các biến cố về độ tuổi có đầy đủ hay chưa?
- Theo bảng phân phối trên thì
$$ P(G_1) = 0.08; P(G_2) = 0.15; P(G_3) = 0.49 \text{ và } P(G_4) = 0.28 $$
thỏa
$$ \sum_{k=1}^4 P(G_k) = 1 $$
nên họ các biến cố $\lbrace G_k : k=1,2,3,4 \rbrace$ tạo thành 1 họ đầy đủ các biến cố về độ tuổi. Khi đó, áp dụng công thức Bayes, thì

$$ \begin{array}{ccl} P( \left. G_1 \right\vert A) &=& \dfrac{P( \left. A \right\vert G_1) P(G_1)}{\sum_{k=1}^4 P( \left. A \right\vert G_k) P(G_k)} \\ &=& \dfrac{(0.06)(0.08)}{(0.06)(0.08) + (0.03)(0.15) + (0.02)(0.49) + (0.04)(0.28)} \\ &=& 0.1584. \end{array} $$

#### Question 20.

- An insurance company issues life insurance policies in three separate categories: standard, preferred, and ultra-preferred. 
- Of the company’s policyholders, 50% are standard, 40% are preferred, and 10% are ultra-preferred. Each standard policyholder has probability 0.010 of dying in the next year, each preferred policyholder has probability 0.005 of dying in the next year, and each ultra-preferred policyholder has probability 0.001 of dying in the next year.
- A policyholder dies in the next year.

Calculate the probability that the deceased policyholder was `ultra-preferred`. 

-------------

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $R, S, U$ lần lượt là các biến cố thể hiện chính sách BH của cty thuộc loại `"prefered", "standard"` hay `"ultra-prefered"` 
- Đặt $D$ là biến cố mà KH mua BH đã chết.
- Theo giả thiết đề bài, ta có

$$ \left\lbrace \begin{array}{clll} & P(S) = 0.6, & P(R) = 0.4, & P(U) = 0.1 \\ \text{và}\\ &  P[D | S] = 0.01, & P[D | R] = 0.005 & P[D | U] = 0.001 \end{array} \right.$$

- BT yêu cầu tính $P[U | D]$ và ta lại sử dụng công thức Bayes lần nữa (vì $U, R, S$ đã tạo thành họ đầy đủ các biến cố về `hạng mục BH`); do đó

$$ P[U | D] = \dfrac{P[D | U] P(U)}{P[D | U] P(U) + P[D | R] P(R) + P[D | S] P(S)} \approx 0.0141. $$
