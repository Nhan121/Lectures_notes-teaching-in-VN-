#### Question 91, 
An insurance company insures a large number of drivers. Let $X$ be the random variable representing the company’s losses under collision insurance, and let $Y$ represent the company’s losses under liability insurance. $X$ and $Y$ have `joint density function`

$$ f(x, y) = \left\lbrace \begin{array}{cl} \dfrac{2x + 2 - y}{4}, & x \in (0, 1), y \in (0, 2) \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the probability that the total company loss is at least 1. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu ta tính XS để tổng tổn thất tối thiểu là 1, tức là $P[X + Y \geq 1]$
- Giả thiết đề bài đã cho $(X, Y)$ có hàm MDXS $f(x, y)$, do đó

$$ \begin{array}{ccl} P[X + Y \geq 1] &=& \displaystyle \iint_{x + y \geq 1} f(x,y) dx dy \\ &=& \displaystyle \int_0^1 \int_{\lbrace y \in (0, 2) : x+y \geq 1 \rbrace} f(x, y) dy dx \\ &=& \displaystyle \int_0^1 \left( \int_{1-x}^2 \dfrac{2x+2-y}{4}dy \right) dx \\ &=& \displaystyle \int_0^1 \left. \left( \dfrac{2xy + 2y - 0.5y^2}{4} \right) \right\vert_{y=1-x}^{y=2} dx \\ &=& \displaystyle \int_0^1 \dfrac{2.5x^2 + 3x + 0.5}{4} dx \\ & \approx & 0.708 \end{array} $$

#### Question 92.
- Two `insurers` provide bids on an `insurance policy` to a large company. The bids must be between `2000` and `2200`. 
- The company decides to accept the `lower bid` if the two bids differ by 20 or more. Otherwise, the company will consider the two bids further.
- Assume that the two bids are independent and are `both uniformly distributed` on the interval from 2000 to 2200.

Calculate the probability that the company considers the two bids further. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là các BNN thể hiện số tiền của các gói thầu của 2 công ty $X, Y$.
- Theo giả thiết, $X, Y$ là có cùng phân phối với phân phối đều $\mathcal{U}([2000, 2200])$
- Mặt khác, công ty chỉ chấp nhận giá thầu thấp hơn nếu hai giá thầu chênh lệch nhau từ 20 trở lên; nếu không họ sẽ xét thêm 2 gói thầu nữa. Do đó khi BT yêu cầu tính XS để cty xem xét thêm 2 gói thầu nữa có nghĩa là $P[|X - Y| \leq 20]$
- Ta có $(X, Y)$ có hàm MDXS đồng thời là
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{200^2}, & \text{ if } x,y \in [2000, 2200], \\ 0, &\text{otherwise} \end{array} \right.$$
suy ra
$$ \begin{array}{ccc} P\left( |X - Y| > 20 \right) &=& P\left( X,Y \in [2000, 2200]: X-Y > 20 \right) + P\left( X,Y \in [2000, 2200]: Y-X > 20 \right) \\ &=& \displaystyle \int_{2020}^{2200}\int_{2000}^{x-20} \dfrac{1}{40000}dydx + \int_{2020}^{2200}\int_{2000}^{y-20} \dfrac{1}{40000}dxdy \\ &=& \displaystyle \int_{2020}^{2200} \left( \dfrac{x - 2020}{40000} \right) dx + \int_{2020}^{2200} \left( \dfrac{y - 2020}{40000} \right) dy\\ &=& \dfrac{81}{200} + \dfrac{81}{200} \\ &=& \dfrac{81}{100} \end{array} $$
cuối cùng ta được
$$ P\left( |X - Y| \leq 20 \right) = 1 - P\left( |X - Y| > 20 \right) = \dfrac{19}{100} $$

#### Question 93.
- A family buys two policies from the same insurance company. Losses under the two policies are independent and have `continuous uniform distributions` on the interval from 0 to 10. 
- One policy has a deductible of 1 and the other has a `deductible of 2`. The family experiences exactly one loss under each policy.

Calculate the probability that the total benefit paid to the family does not exceed 5.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là khoản tổn thất trên hợp đồng với khoản khấu hao tương ứng là $1$ và $2$, theo giả thiết đầu tiên thì $X, Y$ trước khi khấu hao sẽ có cùng phân phối đều $\mathcal{U}([0, 10])$, do đó $(X, Y)$ có hàm MDXS là
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{100}, & \text{ if } x,y \in [0, 10], \\ 0, &\text{otherwise} \end{array} \right.$$
- Yêu cầu BT cần tính XS để tổng tiền mỗi hộ gia đình sẽ nhận không vượt quá 5 (sau khi khấu hao), tức là 

$$ P[X \geq 1, (X+1) + (Y-2) < 8 ] + P[X < 1, Y < 7] $$
trong đó 

$\qquad \diamond P[X < 1, Y < 7]$ là XS mà không phải chi trả khoản nào trong hợp đồng, do $X \in [0, 1)$.

$\qquad \diamond P[X \geq 1, (X+1) + (Y-2) < 8]$ là XS mà khi $X$ có bồi thường thì $X, Y$ phải thỏa sau khi khấu hao thì tổng tiền không vượt quá 5.
- Ta có
$$ P[X \geq 1, (X+1) + (Y-2) < 8 ] = P[X \geq 1, X+Y < 8] = \int_{1}^{10}\int_{0}^{8-x} \dfrac{1}{100}dydx = \dfrac{9}{40}$$
và
$$ P[X < 1, Y<7] = \int_0^1 \int_0^7 \dfrac{1}{100}dydx = \dfrac{7}{100} $$
cuối cùng, xác suất cần tìm là $\dfrac{9}{40} + \dfrac{7}{100} = 0.295.$

#### Question 94.
- Let $T_1$ be the time between a `car accident` and reporting a claim to the insurance company. 
- Let $T_2$ be the time between the report of the `claim` and `payment` of the `claim`.
- The `joint density function` of $T_1$ and $T_2$, is $f(t_1 , t_2 )$ , is `constant` over the region $\lbrace 0 < t_1, t_2 < 6 : 0 < t_1 + t_2 < 10 \rbrace$ , and zero otherwise.

Calculate $\mathbb{E}(T_1 + T_2)$, the expected time between a car accident and payment of the claim.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Do hàm MDXS của $(T_1, T_2)$ là một hằng số trên miền $\lbrace 0 < t_1, t_2 < 6 : 0 < t_1 + t_2 < 10 \rbrace$ và bằng 0 ở phần còn lại, nên tồn tại $C > 0$ sao cho $f$ xác định bởi

$$ f(t_1, t_2) = \left\lbrace \begin{array}{ll} C, & \text{ if } t_1, t_2 \in (0, 6) \text{ sao cho } t_1 + t_2 \in (0, 10) \\ 0, & \text{otherwise} \end{array}  \right. $$
thỏa điều kiện
$$ \displaystyle \iint_{\mathbb{R}^2} f(t_1, t_2)dt_1 dt_2 = 1 \Leftrightarrow \iint_A C dt_2 dt_1 = 1 $$
trong đó,
$$ \begin{array}{lcl} A &=& \lbrace t_1, t_2 \in (0, 6) \text{ sao cho } t_1 + t_2 \in (0, 10) \rbrace \\ &=& \lbrace (t_1, t_2): t_1 \in (0, 4), t_2 \in (0, 6) \rbrace \cup \lbrace (t_1, t_2): t_1 \in (4, 6), t_2 \in (0, 10 - t_1) \rbrace \end{array} $$
- Ta có
$$ \begin{array}{rcrl} & \displaystyle \iint_{\mathbb{R}^2} f(t_1, t_2)dt_1 dt_2 &=& 1 \\ \Leftrightarrow &  \displaystyle \int_0^4 \int_0^6 C dt_2 dt_1 + \int_4^6 \int_0^{10 - t_1} C dt_2 dt_1 &=& 1 \\ \Leftrightarrow & \displaystyle 24C + \int_4^6 C(10 - t_1) dt_1 &=& 1 \\ \Leftrightarrow & C & = & \dfrac{1}{34} \end{array} $$
- Từ đó, suy ra
$$ \begin{array}{crl} \mathbb{E}(T_1 + T_2) &=&  \displaystyle \iint_A (t_1 + t_2) f(t_1, t_2) dt_2 dt_1 \\ &=& \displaystyle \int_0^4 \int_0^6 \dfrac{t_1 + t_2}{34} dt_2 dt_1 + \int_4^6 \int_0^{10 - t_1} \dfrac{t_1 + t_2}{34} dt_2 dt_1 \\ &=& \displaystyle \int_0^4 \left. \left( \dfrac{t_1 t_2 + 0.5t_2^2}{34} \right) \right\vert_{t_2=0}^6 dt_1 + \int_4^6 \left. \left( \dfrac{t_1 t_2 + 0.5t_2^2}{34} \right) \right\vert_0^{t_2 = 10 - t_1} dt_1  \\ &=& \dfrac{120}{34} + \dfrac{224}{102} \\ &=& \dfrac{292}{51} \approx 5.72 \end{array} $$

#### Question 95.
- $X$ and $Y$ are independent random variables with common `moment generating function` $M(t) = \exp( t^2 / 2)$.
- Let $W = X + Y$ and $Z = Y - X$.

Determine the `joint moment generating function` $M(t_1, t_2)$, of $W$ and $Z$.

**Giải đáp / SOLUTIONs.**

Ta có
$$ \begin{array}{crll} M(t_1, t_2) &=& \mathbb{E}e^{(t_1W + t_2 Z)} & \text{(định nghĩa hàm sinh moment)} \\ &=& \mathbb{E}e^{t_1(X+Y) + t_2(Y - X)} \\ &=& \mathbb{E}[ e^{X(t_1 + t_2)}\cdot e^{Y(t_1 - t_2)} ] \\ &=& \mathbb{E}[ e^{X(t_1 + t_2)}] \cdot \mathbb{E}[ e^{Y(t_1 - t_2)} ] & \text{(do } X, Y \text{ độc lập.)} \\ &=& M(t_1 - t_2) \cdot M(t_1 + t_2) & \text{( do } X, Y \text{ cùng phân phối với hàm sinh moment là } M(t)) \\ &=& \exp\left( \frac{(t_1 + t_2)^2}{2} \right) \cdot \exp\left( \frac{(t_1 - t_2)^2}{2} \right) & \text{( do định nghĩa:} \left. M(t) = \exp\left( \frac{t^2}{2} \right) \right) \\ &=& e^{t_1^2 + t_2^2} \end{array} $$

#### Question 96.
- A tour operator has a bus that can accommodate 20 tourists. The operator knows that tourists may not show up, so he sells 21 tickets. The `probability` that an individual tourist will not show up is 0.02, independent of all other tourists.
- Each ticket costs 50, and is `non-refundable` if a tourist fails to show up. 
- If a tourist shows up and a seat is not available, the tour operator has to pay 100 (ticket cost + 50 penalty) to the tourist.

Calculate the expected revenue of the tour operator. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$  là BNN thể hiện người khách du lịch thứ $i$ có xuất hiện trong chuyến đi hay không sau khi đã mua vé, theo đó $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, \theta)$ với $\theta = 1 - 0.02$ (vì XS người đó không có mặt là 0.02)
- BT cho biết cty đã bán 21 vé, nhưng xe chỉ có thể chở được 20 người; và nếu 1 người có mặt khi xe không còn chỗ thì đoàn phải bồi thường với số tiền là 100. 
- Với 21 vé đã bán, thì
$$ S = \sum_{i=1}^{21}X_i \sim \mathcal{B}(n=21, \theta = 0.98) $$
là tổng số người sẽ xuất hiện trong chuyến xe lần này.
- Như vậy, lợi nhuận của đoàn xe sẽ được tính bởi tổng số tiền bán vé trừ đi trường hợp nếu xe có đủ 21 khách (và đoàn phải bồi thường 100), tức là

$$ 21 \cdot 50 - 100 \cdot P[S = 21] = 1050 - 100 \cdot (0.98)^{21} \approx 985. $$

-----------

**Mở rộng.** Tổng quát cho trường hợp nếu có trên 21 vé được bán, khi đó, chi phí tối đa phải bồi thường sẽ được tính bởi công thức
$$ \sum_{k=21}^n (k - 20)\cdot 100 P[S = k] $$
trong đó $n$ là số người có mặt trên xe và $S \sim \mathcal{B}(n, \theta)$, do đó khoản lợi nhuận sẽ là
$$ 50n - 100 \cdot \left( \sum_{k=21}^n (k - 20) P[S = k] \right)$$

#### Question 97.
- Let $T_1$ and $T_2$ represent the lifetimes in hours of two linked components in an electronic device. 
- The `joint density function` for $T_1$ and $T_2$ is `uniform` over the region defined by $0 \leq t_1 \leq t_2 \leq L$ where $L$ is a positive constant.

Determine the expected value of the sum of the squares of $T_1$ and $T_2$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo giả thiết, $T_1, T_2$ là các BNN có phân phối đều và có hàm MDXS $f$ xác định bởi
$$ f(t_1, t_2) = \left\lbrace \begin{array}{cl} C, & 0 \leq t_1 \leq t_2 \leq L \\ 0, & \text{otherwise} \end{array} \right. $$
tương tự, ta sẽ tìm $C$ như sau
$$ \begin{array}{rcrl} & \displaystyle \iint_{\mathbb{R}^2} f(t_1, t_2)dt_1 dt_2 &=& 1 \\ \Leftrightarrow &  \displaystyle \int_0^L \int_0^{t_2} C dt_1 dt_2 &=& 1 \\ \Leftrightarrow & \displaystyle C \int_0^L t_2 dt_2 &=& 1 \\ \Leftrightarrow & C & = & \dfrac{2}{L^2} \end{array} $$
- Đề bài yêu cầu ta tính kỳ vọng của tổng các bình phương $T_1$ và $T_2$, tức là

$$ \begin{array}{crl} \mathbb{E}(T_1^2 + T_2^2) &=& \displaystyle \int_0^L \int_0^{t_2} (t_1^2 + t_2^2) f(t_1, t_2) dt_1 dt_2 \\ &=& \dfrac{2}{L^2} \displaystyle \int_0^L \int_0^{t_2} (t_1^2 + t_2^2) dt_1 dt_2 \\ &=& \dfrac{2}{L^2} \displaystyle \int_0^L \left.\left( \dfrac{t_1^3}{3} + t_1t_2^2 \right)\right\vert_0^{t_1 = t_2} dt_2 \\&=& \displaystyle \int_0^L \dfrac{8 t_2^3}{3L^2} dt_2 \\ &=& \dfrac{2L^2}{3} \end{array} $$

#### Question 98.
Let $X_1, X_2, X_3$ be a random sample from a discrete distribution with probability function 

$$ p(x) = \left\lbrace \begin{array}{rl} 1/3, & x = 0 \\ 2/3, & x=1 \\ 0, & \text{otherwise} \end{array} \right. $$

Determine the moment generating function, $M(t)$, of $Y = X_1 X_2 X_3$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta thấy rằng $X_1, X_2, X_3$ với hàm xác suất như vậy nó sẽ là các BNN rời rạc chỉ nhận 2 giá trị $\lbrace 0, 1 \rbrace$, do đó $Y = X_1 X_2 X_3$ cũng chỉ nhận 2 giá trị $\lbrace 0, 1 \rbrace$ với xác suất

$\qquad \diamond P(Y = 1) = P[X_1 = 1, X_2 = 1, X_3 = 1] = P[X_1 = 1]P[X_2 = 1]P[X_3 = 1] = \left( \dfrac{2}{3} \right)^3 = \dfrac{8}{27}$ 

$\qquad \diamond P(Y = 0) = 1 - P[Y = 1] = \dfrac{19}{27}$ 
- Theo đó, hàm sinh moment của BNN rời rạc $Y$, sẽ được xác định bởi

$$ \begin{array}{ccl} M(t) &=& \displaystyle \sum_{y \in \Omega_Y} (e^{ty}) P[Y = y] \\ &=& \left( e^{0 \cdot t} \right) P[Y = 0] + \left( e^{1 \cdot t} \right) \cdot P[Y = 1] \\&=& \dfrac{8}{27} + \dfrac{19}{27}e^t \end{array} $$

#### Question 99.
An insurance policy pays a total medical benefit consisting of two parts for each claim.
Let $X$ represent the part of the benefit that is paid to the surgeon, and let $Y$ represent the
part that is paid to the hospital. The variance of $X$ is 5000, the variance of $Y$ is 10,000,
and the variance of the total benefit, $X + Y$, is 17,000.

Due to increasing medical costs, the company that issues the policy decides to increase $X$
by a flat amount of 100 per claim and to increase $Y$ by `10%` per claim.

Calculate the variance of the total benefit after these revisions have been made.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo trình tự giả thiết, ta có 
$$ \text{Var}X = 5000, \text{Var}Y = 10000, \text{Var}(X + Y) = 17000 $$
suy ra
$$ \text{Cov}(X, Y) = \dfrac{1}{2} \left( \text{Var}(X+Y) - \text{Var}X - \text{Var}Y \right) = 1000. $$
- Giả thiết cuối nói rằng: vì chi phí y tế ngày càng tăng, công ty ban hành chính sách quyết định tăng $X$ số tiền cố định là 100 cho mỗi yêu cầu và tăng $Y$ thêm 10% cho mỗi yêu cầu. Đặt $Z$ là tổng lợi ích sau khi các sửa đổi này được thực hiện các sửa đổi như trên, khi đó

$$ Z = 100 + X + (Y + 0.1 Y) = 100 + X + 1.1Y $$
- BT yêu cầu ta tìm phương sai của $Z$, ta có

$$ \begin{array}{cll} \text{Var}Z &=& \text{Var} \left( 100 + X + 1.1Y \right) \\ &=& \text{Var}(X+100) + \text{Var}(1.1Y) + 2\text{Cov}(X + 100, 1.1Y) \\ &=& \text{Var}X + (1.1^2)\text{Var}Y + 2\text{Cov}(X, Y) \\ &=& 19300 \end{array} $$

với các lưu ý
$$ \begin{array}{rcll} \text{Var}(X + c) &=& \text{Var}X, & \forall c \in \mathbb{R} \\ \text{Var}(cX) &=& c^2 \cdot \text{Var}X, & \forall c \in \mathbb{R} \\ \text{Cov}(X + c, Y) &=& \text{Cov}(X, Y), & \forall c \in \mathbb{R} \end{array} $$
Như vậy $\text{Var}Z = 19300.$

#### Question 100.
A car dealership sells 0, 1, or 2 luxury cars on any day. When selling a car, the dealer
also tries to persuade the customer to buy an extended warranty for the car. Let $X$ denote
the number of luxury cars sold in a given day, and let $Y$ denote the number of extended
warranties sold. 

$$ \begin{array}{ccl} P[X = 0, Y = 0] &=& 1/6 \\ P[X = 1, Y = 0] &=& 1/12 \\ P[X = 1, Y = 1] &=& 1/6 \\ P[X = 2, Y = 0] &=& 1/12 \\ P[X = 2, Y = 1] &=& 1/3 \\ P[X = 2, Y = 2] &=& 1/6 \end{array} $$

Calculate the variance of $X$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ bảng phân phối xác suất đồng thơi như trên, ta có thể tìm được phân phối lề của $X$ bằng cách

$$ \begin{array}{ccl} P[X=0] &=& P[X=0, Y= 0] = \dfrac{1}{6} \\ P[X=1] &=& P[X=1,Y=0] + P[X=1, Y=1] = \dfrac{3}{12} \\ P[X = 2] &=& P[X=2, Y=0]+P[X=2, Y=1]+P[X=2, Y=2] = \dfrac{7}{12} \end{array} $$
- Đề bài yêu cầu tính phương sai của $X$; ta có

$$ \begin{array}{ccl} \mathbb{E}X &=& 0\cdot P[X=0] + 1\cdot P[X=1] + 2\cdot P[X =2] = \dfrac{17}{12} \\ \mathbb{E}X^2 &=& (0^2)\cdot P[X=0] + (1^2)\cdot P[X =1] + (2^2)\cdot P[X=2] = \dfrac{31}{12} \end{array} $$
suy ra
$$ \text{Var}X = \mathbb{E}X^2 - [\mathbb{E} X]^2 = 0.58. $$
