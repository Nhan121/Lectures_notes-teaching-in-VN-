#### Question 181, 
A company issues auto insurance policies. There are 900 insured individuals. 
- `(i)` Fifty-four percent of them are male. 
- `(ii)` If a female is randomly selected from the 900, the probability she is over 25 years old is 0.43. 
- `(iii)` There are 395 total insured individuals over 25 years old.

A person under 25 years old is randomly selected. Calculate the probability that the person selected is male.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $M$ là các sự kiện về giới tính của người được khảo sát là nam và $A$ là sự kiện về tuổi của họ.
- Theo `(i)`; có đến `54%` là nam trong tổng số 900 nhân viên nên $P[M] = 0.54$ dẫn đến XS nhận được người nữ bằng 

$$ P[M^c] = 1 - P[M] = 0.46$$
- Theo `(ii)`, XS mà nhân viên nữ trên 25 tuổi bằng 0.43 trong tổng số 900 nhân viên, tức là

$$ P[A > 25 | M^c] = 0.43 $$
vì vậy 
$$ P[(A > 25) \cap M^c] = P[(A > 25) | M^c] \cdot P[M^c] = 0.1978 $$
- Theo `(iii)` thì số người trên 25 tuổi có đến 395 (trong tổng số 900 người được khảo sát), do đó 
$$ P[A > 25] = \dfrac{395}{900} \approx 0.4389 $$
- BT yêu cầu tính XS khi chọn ngẫu nhiên một người để được giới tính nam biết rằng người đó dưới 25 tuổi, tức là $P[M | A \leq 25]$. Áp dụng công thức XS có điều kiện, ta có

$$ \begin{array}{ccl} P[M | A \leq 25] &=& \dfrac{P[(A \leq 25) \cap M]}{P[A \leq 25] } \\ &=& \dfrac{P[A \leq 25] - P[(A > 25) \cap M^c]}{P[A \leq 25]} \\ &=& \dfrac{ (1 - P[A > 25]) - \left( P[M^c] - P[(A > 25) \cap M] \right)}{1 - P[A > 25] } \\ &=& \dfrac{(1 - 0.4389) - (0.46 - 0.1978)}{1 - 0.4389} \\ &=& 0.5327  \\ \end{array} $$


#### Question 182.
An insurance company insures red and green cars. An actuary compiles the following data:

|Color of car | Red | Green |
|:-|-|-|
| Number insured | 300 | 700 |
| **Prob an accident occurs** | 0.1 | 0.05 |
| **Prob that the claim exceeds the deductible, given an accident occurs from this group** | 0.9 | 0.8 |

$$ \qquad $$

The actuary randomly picks a claim from all claims that exceed the deductible.

Calculate the probability that the claim is on a red car

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $R, G$ lần lượt là các sự kiện mà chiếc xe được BH có màu đỏ (`red`) hay xanh lục (`green`).
- Đăt $E$ là sự kiện mà yêu cầu đã vượt quá khoản khấu hao (`claim exceeds deductible`) và $A$ là sự kiện nếu có tai nạn xảy ra.
- Khi đó dựa vào bảng thông tin trong bài, ta có các dữ kiện sau:

$$ \begin{array}{ccc} P[R] = 0.3, & P[E | A \cap R] = 0.10, & P[R | A] = 0.9, \\ P[G] = 0.7, & P[ E| A \cap G] = 0.05,  & P[G | A] = 0.8 \end{array} $$
- BT yêu cầu tính XS để chọn được xe đỏ biết rằng xe này đã vượt mức khấu hao, tức $P[R|E]$; ta có

$$ \begin{array}{ccl} P[R|E] &=& \dfrac{P[E|R] \cdot P[R]}{P[E|R] \cdot P[R] + P[E|G] \cdot P[G]} \\ &=& \dfrac{P[E|R] \cdot (P[E | (R\cap A)] \cdot P [R | A])}{P[E|R] \cdot (P[E | (R\cap A)] \cdot P [R | A]) + P[E|G] \cdot (P[E | (G\cap A)] \cdot P [G | A])} \\ &=& \dfrac{(0.3)(0.1)(0.9)}{(0.3)(0.1)(0.9) + (0.7)(0.05)(0.8)} \\ &=& 0.491 \\ \end{array} $$

#### Question 183.
$X$ is a random variable with probability density function
$$ f(x) = \left\lbrace \begin{array}{cl} e^{-2x}, & x \geq 0 \\ 2e^{4x}, & x < 0 \end{array} \right. $$

Let $T = X^2$; determine the probability density function for $T$ for positive values of $t$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Trước tiên, cần xác định hàm phân phối tích lũy $G_T(t)$ cho $T = X^2$ với $t > 0$ và hàm MDXS của $X$ là $f$ như trong bài, ta có

$$ \begin{array}{cclcl} G_T(t) &=& P[T < t] &=& P\left[ - \sqrt{t} < X < \sqrt{t} \right] \\ &&&=& \displaystyle \int_{- \sqrt{t}}^{\sqrt{t}} f(x) dx \\ &&&=& \displaystyle \int_{- \sqrt{t}}^0 2e^{4x}dx + \int_0^{\sqrt{t}} e^{-2x}dx \\ &&&=& 1 - \left( \dfrac{e^{-2\sqrt{t}} + e^{-4\sqrt{t}} }{2} \right) \end{array} $$
- Sau đó, suy ra hàm MDXS bằng cách lấy đạo hàm của $G_T(t)$, 

$$ g_T(t) = \dfrac{dG_T(t)}{dt} = \dfrac{e^{-2\sqrt{t}} + 2e^{-4\sqrt{t}} }{2\sqrt{t}} $$

#### Question 184.
George and Paul play a betting game. 
- `(i)` Each chooses an integer from 1 to 20 (inclusive) at random. 
- `(ii)` If the two numbers differ by more than 3, George wins the bet. 
- `(iii)` Otherwise, Paul wins the bet.

Calculate the probability that Paul wins the bet.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $x, y$ là các số mà `Paul` và `Geogre` đã chọn. Theo đó, 

$\qquad \diamond$ Không gian mẫu: $\Omega = \lbrace (x,y): x,y = 1,2,\ldots,20 \rbrace$ nên $|\Omega| = 20^2 = 400$.

$\qquad \diamond$ Tập giá trị để Geogre thắng là khi $| x - y| > 3$ với $x, y = 1,2,\ldots,20$ và để Paul thắng thì

$$ \begin{array}{cclrl} \text{Paul win} &=& \lbrace (x,y): |x-y| \leq 3, \quad \forall x,y = 1,2,\ldots,20 \rbrace \\ &=& \displaystyle \bigcup_{x=1}^{20} \left\lbrace y = 1,2,\ldots,20: |y - x| \leq 3 \right\rbrace &=& \displaystyle \bigcup_{x=1}^{20} A_x\end{array} $$
$\qquad$ trong đó
$$ A_x = \left\lbrace y = 1,2,\ldots,20: |y - x| \leq 3 \right\rbrace, \quad \forall x = 1,2,\ldots,20 $$

$\qquad \quad$ Cụ thể
$$ \begin{array}{cclclrl} A_1 &=& \lbrace y = 1,2,\ldots,20: |y - 1| \leq 3 \rbrace &=& \lbrace 1,2,3,4 \rbrace & \Rightarrow & |A_1| = 4 \\ A_2 &=& \lbrace y = 1,2,\ldots,20: |y - 2| \leq 3 \rbrace &=& \lbrace 1,2,3,4,5 \rbrace & \Rightarrow & |A_2| = 5 \\ A_3 &=& \lbrace y = 1,2,\ldots,20: |y - 3| \leq 3 \rbrace &=& \lbrace 1,2,3,4,5,6 \rbrace & \Rightarrow & |A_3| = 6 \\ A_4 &=& \lbrace y = 1,2,\ldots,20: |y - 4| \leq 3 \rbrace &=& \lbrace 1,2,3,4,5,6,7 \rbrace & \Rightarrow & |A_4| = 7 \\ A_5 &=& \lbrace y = 1,2,\ldots,20: |y - 5| \leq 3 \rbrace &=& \lbrace 2,3,4,5,6,7,8 \rbrace & \Rightarrow & |A_5| = 7 \\ \ldots \\ A_{17} &=& \lbrace y = 1,2,\ldots,20: |y - 17| \leq 3 \rbrace &=& \lbrace 14,15,16,17,18,19,20 \rbrace & \Rightarrow & |A_{17}| = 7 \\ A_{18} &=& \lbrace y = 1,2,\ldots,20: |y - 18| \leq 3 \rbrace &=& \lbrace 15,16,17,18,19,20 \rbrace & \Rightarrow & |A_{18}| = 6 \\ A_{18} &=& \lbrace y = 1,2,\ldots,20: |y - 19| \leq 3 \rbrace &=& \lbrace 16,17,18,19,20 \rbrace & \Rightarrow & |A_{19}| = 5 \\ A_{18} &=& \lbrace y = 1,2,\ldots,20: |y - 20| \leq 3 \rbrace &=& \lbrace 17,18,19,20 \rbrace & \Rightarrow & |A_{20}| = 4\end{array} $$

Do đó,

$$ P[\text{Paul win}] = \dfrac{\displaystyle \sum_{x=1}^{20} |A_x|}{|\Omega|} = \dfrac{4+5+6+ 14\cdot (7) + 6+5+4}{400} \approx 0.32.  $$

#### Question 185.
A student takes an examination consisting of $20$ true-false questions. 
- The student knows the answer to $N$ of the questions, which are answered correctly, and guesses the answers
to the rest. 
- The conditional probability that the student knows the answer to a question,
given that the student answered it correctly, is $0.824$.

Calculate $N.$ 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C, K$ lần lượt là sự kiện mà SV được gọi trả lời đúng câu hỏi (`correct answers`) và đã biết trước đáp án (`know the A of Q`), theo đó

$$ P[C|K^c]=0.5, \quad P[C | K] = 1, \quad P[K|C] = 0.824, \quad \text{ và } \quad P[K] = \dfrac{N}{20} $$
- Để tìm $N$ thì ta phải cân bằng hệ thức từ các dữ kiện trên qua công thức Bayes:

$$ \begin{array}{rccl} & P[K|C] &=& \dfrac{P[C|K] \cdot P[K]}{P[C|K]\cdot P[K] + P[C|K^c]\cdot P[K^c]} \\ \Rightarrow & 0.824 &=& \dfrac{\frac{N}{20}}{\frac{N}{20} + 0.5 \left( 1 - \frac{N}{20} \right)} \\ \Rightarrow & 0.824 &=& \dfrac{N}{N+0.5(20 - N)} \\ \Rightarrow & N &=& 14  \end{array} $$

#### Question 186.
The minimum force required to break a particular type of cable is **normally distributed** with `mean 12,432` and `standard deviation 25`. 

A random sample of 400 cables of this type is selected.

Calculate the probability that at least 349 of the selected cables will not break under a force of 12,400. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN về sức chịu của sợi cáp, theo đó $Y \sim \mathcal{N}(\mu, \sigma^2)$ với $\mu = 12432$ và $\sigma = 25$. Do đó XS để sợi cáp không bị đứt bởi tác động của một ngoại lực F là
$$ P[Y > F] = P\left[ \mathcal{N}(0,1) > \dfrac{F - \mu}{\sigma} \right] = 1 - \Phi \left( \dfrac{F - 12432}{25} \right) $$
với $\Phi$ là hàm phân phối tích lũy của $\mathcal{N}(0,1)$.
- Đặt $X_i$ là khả năng chống chịu của sợi cáp thứ $i^{th}$ với xác suất $P[Y > F]$
- Vì BT yêu cầu tính XS ít nhất 349 sợi (trên tổng số 400) sẽ không đứt với một lực 12400, nên $F = 12400$ và do đó, ta cần tính

$$ \displaystyle P \left( \sum_{i=1}^{400} X_i \geq 349 \right), \quad \forall X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}\left( 1, P[Y > 12400] \right) $$ 
Ta có

$\qquad \diamond$  Thay $F = 12400$, thì $$ P[Y > 12400] = 1 - \Phi \left( \dfrac{12400 - 12432}{25} \right) \approx 0.9 $$
$\qquad \diamond$ Do đó $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.9)$ nên $\displaystyle X = \sum_{i=1}^{400} X_i \sim \mathcal{B}(400, 0.9)$ 

$\qquad \diamond$ Áp dụng công thức `hiệu chỉnh liên tục [continuity correction]`:
$$ P[\mathcal{B}(n,p) > N] = P\left[ \mathcal{N}(0,1) > \dfrac{N - 0.5 - np}{\sqrt{np(1-p)}} \right], \qquad \forall N = 0,1,\ldots,n, \quad \forall p \in (0,1) $$
$\qquad \diamond$ Cuối cùng; thay $N = 349, n = 400, p = 0.9$; ta được

$$ \displaystyle P \left( \sum_{i=1}^{400} X_i \geq 349 \right) = P\left[ \mathcal{N}(0,1) > \dfrac{349 - 0.5 - 360}{\sqrt{360(0.9)(1-0.9)}} \right] = 1 - \Phi(-1.9167) \approx 0.97. $$

#### Question 187.
- The number of policies that an agent sells has a **Poisson distribution** with `modes` at `2` and `3`.
- $K$ is the smallest number such that the probability of selling more than $K$ policies is less than `25%`.

Calculate $K.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $\lambda$ là tham số cần tìm từ phân phối Poisson. Khi đó
$$ P[\text{Pois}(\lambda) = k] = e^{-\lambda} \dfrac{\lambda^k}{k!}, \qquad \forall k = 0,1,2,\ldots $$
- Đề bài cho biết phân phối này đạt modes tại $2$ và $3$; nói cách khác: XS tại 2 và 3 bằng nhau và có trọng số cao nhất trong tất cả các số khác, tức là
$$ \displaystyle P[\text{Pois}(\lambda) = 2] = P[\text{Pois}(\lambda) = 3] = \max_{k = 0,1,2,\ldots } P[\text{Pois}(\lambda) = k] $$
- BT yêu cầu tìm số nguyên $K$ nhỏ nhất sao cho 
$$ P[\text{Pois}(\lambda) > K] > 0.25 $$
nên trước hết ta phải xác định được tham số $\lambda$. Ta có

$$ \begin{array}{rccl} & P[\text{Pois}(\lambda) = 2] &=& P[\text{Pois}(\lambda) = 3] \\ \Rightarrow & e^{-\lambda} \dfrac{\lambda^2}{2!} &=& e^{-\lambda} \dfrac{\lambda^3}{3!} \\ \Rightarrow & \lambda &=& 3 \end{array} $$
$\qquad$ Bấy giờ, dựa vào bảng phân phối xác suất tích lũy của phân phối `Poisson(3)`; ta thấy rằng

$$ P[\text{Pois}(3) < 3] = 0.64723  \quad \text{ va } \quad P[\text{Pois}(3) \leq 4 ] = 0.81526 $$
$\qquad$ do đó $K=4$ thì XS 
$$ P[\text{Pois}(3) > 4 ] < 0.25 \quad \text{ trong khi } \quad P[\text{Pois}(3) > 3] > 0.25 $$

Vậy $K = 4$ là số nguyên nhỏ nhất cần tìm.

#### Question 188.
Two fair dice are tossed. One die is red and one die is green.

Calculate the probability that the sum of the numbers on the two dice is an odd number given that the number that shows on the red die is larger than the number that shows on the green die.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $x, y$ lần lượt là số chấm trên viên xúc sắc đỏ và viên xanh. Theo đó không gian mẫu các phép thử

$$ \Omega = \lbrace (x,y): x,y=1,2,\ldots,6 \rbrace \Rightarrow | \Omega | = 36 $$
- BT yêu cầu tính XS để tổng số chấm $x+y$ là số lẻ biết rằng số chấm viên đỏ lớn hơn viên xanh $x > y$. do đó

$\qquad \diamond$ Đặt $A$ là không gian sự kiện tổng số chấm là số lẻ, tức là

$$ \begin{array}{ccl} A &=& \lbrace (x,y): x+y \not \vdots 2, \quad \forall x,y=1,2,\ldots,6 \rbrace \\ &=& \lbrace x \vdots 2, y \not \vdots 2 \rbrace \cup \lbrace x \not \vdots 2, y \vdots 2 \rbrace \end{array} $$
$\qquad $ do đó $$P[A] = \dfrac{|A|}{|\Omega|} = \dfrac{3\cdot 3 + 3\cdot 3}{36} = \dfrac{1}{2}$$

$\qquad \diamond$ Đặt $B$ là không gian sự kiện $x > y$, thì

$$ \begin{array}{ccl} B &=& \lbrace (x,y): x > y, \quad \forall x,y=1,2,\ldots,6 \rbrace \\ &=& \displaystyle \bigcup_{x=2}^6 \lbrace (x, y) : y = 1,\ldots,x-1 \rbrace  \end{array} $$
$\qquad $ do đó $$P[B] = \dfrac{|B|}{|\Omega|} = \dfrac{1+2+3+4+5}{36} = \dfrac{5}{12}$$

$\qquad \diamond$ Như vậy $A \cap B$ tức là

$$ \begin{array}{ccl} A \cap B &=& \lbrace (x,y): x>y \text{ and } x+y \not \vdots 2, \quad \forall x,y=1,2,\ldots,6 \rbrace \\ &=& \lbrace (2, 1); (3, 2); (4, 3), (4,1); (5,4), (5,2); (6, 5), (6,3), (6,1) \rbrace \end{array} $$
nên $$ P[A \cap B] = \dfrac{|A \cap B|}{|\Omega|} = \dfrac{9}{36} = \dfrac{1}{4} $$

Do đó, XS cần tìm là $$ P[A|B] = \dfrac{P[A \cap B]}{P[B]} = \dfrac{3}{5}. $$

**Comments. Bài này có thể xử lý nhanh hơn nếu ta chỉ quan tâm tần suất thay vì xác suất.** Thật vậy, 

$\qquad \diamond$ Sau khi xác định được tập giá trị của $B$ và $A \cap B$ thì ta đã biết $B$ chứa 15 phần tử còn $A \cap B$ chứa 9 phần tử. 

$\qquad \diamond$ Lúc này có thể thực hiện trực tiếp $\dfrac{P[A \cap B]}{P[B]} = \dfrac{|A\cap B|}{|B|} = \dfrac{3}{5}$

#### Question 189.
- `(i)` In 1982 Abby’s mother scored at the 93rd percentile in the math SAT exam. 
- `(ii)` In 1982 the mean score was 503 and the variance of the scores was 9604.
- `(iii)` In 2008 Abby took the math SAT and got the same numerical score as her mother had received 26 years before. 
- `(iv)` In 2008 the mean score was 521 and the variance of the scores was 10,201.

Math **`SAT`** scores are **normally distributed** and stated in multiples of ten.

Calculate the percentile for Abby’s score.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là BNN thể hiện điểm số mà một thí sinh đạt được trong kỳ thi **SAT** vào các năm 1982 và 2008; theo `(ii)` và `(iv)`, ta có

$$ X\sim \mathcal{N}(503, 9604), \quad Y \sim \mathcal{N}(521, 10201) $$
- Theo `(iii)` thì điểm số của Abby (thi vào 2008) và mẹ cô ấy (thi vào 1982) là bằng nhau nên ta sẽ đặt $x_0$ là điểm của cả 2 cùng đạt được.
- Theo `(i)`, ta được 
$$ P[X < x_0] = 0.93 $$
tương đương với

$$ \begin{array}{crcl} & P \left[ \mathcal{N}(0,1) < \dfrac{x_0 - \mathbb{E}X}{\sqrt{\text{Var}X}} \right] &=& 0.93 \\ \Rightarrow & \dfrac{x_0 - 503}{\sqrt{9604}}  &=& \Phi^{-1} \left( 0.93 \right) \\ \Rightarrow & x_0 &=& 650  \end{array} $$

- Vì BT yêu cầu tìm phân vị của điểm mà Abby đạt theo năm 2008, tức là $P[Y < x_0]$, ta có

$$ P[Y < x_0] =  P \left[ \mathcal{N}(0,1) < \dfrac{x_0 - \mathbb{E}Y}{\sqrt{\text{Var}Y}} \right] =  P \left[ \mathcal{N}(0,1) < \dfrac{950 - 521}{\sqrt{10201}} \right] = \Phi(1.277) = 0.90. $$
do đó điểm của Abby là 650 trong năm 2008 tương ứng mức phân vị `90 th`

#### Question 190.
- A certain brand of refrigerator has a useful life that is normally distributed with mean 10 years and standard deviation 3 years. 
- The useful lives of these refrigerators are independent.

Calculate the probability that the total useful life of two randomly selected refrigerators will exceed 1.9 times the useful life of a third randomly selected refrigerator.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_1, X_2, X_3$ tương ứng với thời hạn sử dụng của 3 cái tủ lạnh được chọn từ cùng một hãng sản xuất, khi đó BT yêu cầu tính 

$$ P[X_1 + X_2 > 1.9 X_3] \quad \text{ hay } \quad P[X_1 + X_2 - 1.9 X_3 > 0] $$
- Đặt $W = X_1 + X_2 - 1.9 X_3$.
- Theo giả thiết thì $X_1, X_2, X_3 \overset{\text{i.i.d}}{\sim} \mathcal{N}(\mu, \sigma^2)$ với $\mu = 10, \sigma = 3$, do đó

$$ W \sim \mathcal{N}\left( \mu + \mu - 1.9 \mu, \sigma^2 + \sigma^2 + (-1.9)^2 \sigma^2 \right) = \mathcal{N}(1, 50.49) $$
- Do đó

$$ P[X_1 + X_2 > 1.9 X_3] = P[W > 0] = P\left[ \mathcal{N}(0,1) > \dfrac{0 - 1}{\sqrt{50.49}} \right] = 1 - \Phi(-0.141) = \Phi(0.141) = 0.556. $$
