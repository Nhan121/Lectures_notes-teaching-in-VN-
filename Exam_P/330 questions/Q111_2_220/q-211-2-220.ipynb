#### Question 211,
In a casino game, a gambler selects four different numbers from the first twelve positive integers. 
- The casino then randomly draws **nine numbers without replacement** from the **first twelve positive integers**. 
- The gambler **wins the jackpot** if the **casino draws all four of the gambler’s selected numbers**.

Calculate the probability that the gambler wins the jackpot

**Giải đáp / SOLUTIONs.**

Đặt $\Omega = \lbrace 1,2,\ldots,12 \rbrace$ là KG mẫu các kết quả (outcome) có thể xuất hiện khi chơi `jackpot` ở sòng bạc.
Theo đó,
- `Sòng bạc (casino)` sẽ rút ngẫu nhiên không hoàn lại 9 trong 12 số ở $\Omega$; tức là ta có tất thảy $C_{12}^9$ trường hợp có thể xảy ra
- Một `con bạc (gambler)` sẽ phải rút ngẫu nhiên 4 số khác nhau từ $\Omega$; và con bạc chỉ giành giải độc đắc nếu 4 con số này trùng với 4 trong 9 con số mà sòng bạc đã rút; do đó ta có tất cả

$$ \text{[select 4 in 4 numbers which be correct]}.\text{[(9 - 4) in (12 - 4) numbers which is incorrect]} = C_{4}^4 C_{8}^5 $$
- Như vậy, XS để trúng giải là
$$ \dfrac{C_{4}^4 C_{8}^5}{C_{12}^9} \approx 0.2545. $$

#### Question 212.
The number of days an employee is sick each month is modeled by a Poisson distribution
with mean 1. The numbers of sick days in different months are mutually independent.

Calculate the probability that an employee is sick more than two days in a three-month
period. 

**Giải đáp / SOLUTIONs.**

Đặt $N_1, N_2, N_3$ lần lượt là số ngày bị ốm của nhân viên trong 3 tháng. Theo giả thiết thì

$$N_1, N_2, N_3 \overset{\text{i.i.d}}{\sim} \text{Pois}(\lambda = 1)$$

do đó số ngày nghỉ của một nhân viên trong 3 tháng chính là

$$ N = N_1 + N_2 + N_3 \sim \text{Pois}(\lambda + \lambda + \lambda = 3) $$

BT yêu cầu tính XS để nhân viên nghỉ nhiều hơn 2 ngày trong suốt 3 tháng; tức là

$$ P[N > 2] = 1 - P\left[ \text{Pois}(3) \leq 2 \right] = 1 - \displaystyle \sum_{k=0}^2 e^{-3} \dfrac{3^k}{k!} = 0.577. $$

#### Question 213.
- The number of traffic accidents per week at intersection Q has a Poisson distribution with
mean 3. 
- The number of traffic accidents per week at intersection R has a Poisson distribution with mean 1.5.

Let A be the probability that the number of accidents at intersection Q exceeds its mean.
Let B be the corresponding probability for intersection R. 

Calculate B – A. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Q, R$ lần lượt là các BNN thể hiện số vụ tai nạn trong tuần tại các giao lộ $Q$ và $R$, theo đó 

$$ Q \sim \text{Pois}(3), \qquad R \sim \text{Pois}(1.5) $$
- BT yêu cầu tính $B-A$ với
$$ \displaystyle A = P[Q > \mathbb{E}Q] = P\left[ Q > 3 \right] = 1 - \sum_{k=0}^3 P[Q = k] = 1 - \sum_{k=0}^3 e^{-3} \dfrac{3^k}{k!} = 0.3528 $$
và
$$ \displaystyle B = P[R > \mathbb{E}R] = P\left[ R > 1.5 \right] = 1 - \sum_{k=0}^1 P[R = k] = 1 - \sum_{k=0}^1 e^{-1.5} \dfrac{3^k}{k!} =0.4422 $$
- Suy ra,
$$ B - A = 0.0894 $$

#### Question 214.
Losses due to accidents at an amusement park are exponentially distributed. 
- An insurance company offers the park owner two different policies, with different premiums,
to insure against losses due to accidents at the park.
- Policy A has a deductible of 1.44. For a random loss, the probability is 0.640 that under
this policy, the insurer will pay some money to the park owner. 
- Policy B has a deductible of $d$. For a random loss, the probability is 0.512 that under this policy, the insurer will
pay some money to the park owner.

Calculate $d$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số tiền mà BH sẽ trả cho công viên bồi thường tổn thất do tai nạn; theo đó $X \sim \text{E}(\lambda)$ với $\lambda > 0$ chưa biết; do đó

$$ P[X > x] = e^{-x/\lambda}, \qquad \forall x \geq 0 $$
- Đối với chính sách $A$; với khấu hao $1.44$ thì XS mà BH sẽ bồi thường là 0.64; tức là

$$ 0.64 = P[X > 1.44] \Rightarrow e^{-1.44/\lambda} = 0.64 \Rightarrow \lambda =\dfrac{-1.44}{\ln (0.64)} \approx 3.2266. $$

- Đối với chính sách $B$; với khấu hao $d$ thì XS mà BH sẽ bồi thường là 0.512; tức là

$$ 0.512 = P[X > d] \Rightarrow e^{-d/\lambda} = 0.512 \Rightarrow d = -\lambda \ln (0.512) \approx 2.159. $$

Vậy $d \approx 2.159.$

#### Question 215.
The distribution of the size of claims paid under an insurance policy has probability
density function 

$$ f(x) = \left\lbrace \begin{array}{cl} cx^a, & 0 < x < 5, \\ 0, & \text{otherwise} \end{array} \right. $$
where $a >0$ and $c > 0.$

For a randomly selected claim, the probability that the size of the claim is less than 3.75
is 0.4871.

Calculate the probability that the size of a randomly selected claim is greater than 4.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số các yêu cầu trả phí BH với hàm MDXS $f(x)$ như bài toán đã cho.
- Trước tiên, ta cần xác định các hằng số $c,a$ để $f$ là hàm MDXS, tức là

$$ f(x) \geq 0, \quad \forall x, \qquad \text{and} \qquad \displaystyle \int f(x)dx = 1 $$
tương đương với
$$ cx^a \geq 0, \quad \forall x, \qquad \text{and} \qquad \displaystyle c \int_0^5 x^a dx = \left. \left( \dfrac{c x^{a+1}}{a+1} \right) \right\vert_0^{x=5} = 1 $$
vậy
$$ c > 0, a > 0, \qquad \text{and} \qquad \displaystyle c = \dfrac{a+1}{5^{a+1}} $$
- Hơn nữa, BT còn cho biết XS mà số yêu cầu < 3.75 bằng 0.4871, nghĩa là

$$ \begin{array}{cclrl} \displaystyle P[X < 3.75] = 0.4871 & \Leftrightarrow & \displaystyle c\int_0^{3.75} x^a dx &=& 0.4871 \\ & \Leftrightarrow & \dfrac{a+1}{5^{a+1}} \left. \left( \dfrac{x^{a+1}}{a+1} \right) \right\vert_0^{x=3.75} &=& 0.4871 \\ & \Leftrightarrow & \left( \dfrac{3.75}{5} \right)^{a+1} &=& 0.4871 \end{array} $$
lấy `logarith` ở 2 vế, ta được
$$ \ln(0.4871) = (a+1) \ln \left( \dfrac{3.75}{5} \right) = (a+1) \ln(0.75) $$
suy ra
$$ a = \left( \dfrac{\ln (0.4781)}{\ln (0.75)} \right) - 1 = 1.5$$

- Cuối cùng, BT yêu cầu ta tính XS mà số yêu cầu vượt quá $4$, tức là

$$ P[X > 4] = c \int_4^5 x^a dx = \dfrac{a+1}{5^{a+1}} \left. \left( \dfrac{x^{a+1}}{a+1} \right) \right\vert_4^{x=5} = 1 - \left( \dfrac{4}{5} \right)^{1.5 + 1} = 0.42757. $$

#### Question 216.
Company $XYZ$ provides a warranty on a product that it produces. Each year, the number
of warranty claims follows a **Poisson distribution with mean $c$**. The probability that no
warranty claims are received in any given year is $0.60.$

Company $XYZ$ purchases an insurance policy that will reduce its overall warranty claim
payment costs. The insurance policy will pay nothing for the first warranty claim
received and $5000$ for each claim thereafter until the end of the year.

Calculate the expected amount of annual insurance policy payments to Company $XYZ$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là số yêu cầu bảo hành (`warranty`) đã được nhận và theo giả thiết đầu tiên thì $N \sim \text{Pois}(c)$ do đó

$$ P[N = k] = e^{-c} \dfrac{c^k}{k!}, \qquad \forall k = 0,1,2, \ldots. $$
Cũng trong giả thiết đầu tiên này, XS không nhận được yêu cầu bảo hành bằng 0.6; tức là

$$ P[N = 0] = 0.6  \Leftrightarrow e^{-c} = 0.6 \Leftrightarrow c = -\ln 0.6 \approx 0.5108. $$
- Giả thiết còn lại cho biết cty sẽ không trả tiền khi $N \leq 1$ và $5000$ cho mỗi yêu cầu bảo hành sau đó cho đến cuối năm; do đó số tiền kỳ vọng mà cty phải trả hằng năm sẽ được tính bởi

$$ 5000 \displaystyle \sum_{n=2}^{\infty} (n-1) P[N = n] $$
- Mà ta có
$$ \begin{array}{ccl} \displaystyle \sum_{n=2}^{\infty} (n-1) P[N = n] &=& \displaystyle \left( \sum_{n=2}^{\infty} n P[N = n] \right) - \left( \sum_{n=2}^{\infty} P[N = n] \right) \\ &=& \displaystyle \left( \sum_{n=1}^{\infty} n P[N = n] \right) - \left( \sum_{n=1}^{\infty} P[N = n] \right) \\ &=& \displaystyle \left( \sum_{n=0}^{\infty} n P[N = n] \right) - \left( \sum_{n=1}^{\infty} P[N = n] \right) \\ &=& \left( \mathbb{E} N \right) - \left( 1 - P[N=0] \right) \\ &=& 0.5108 - 0.4 \\ &=& 0.1108 \end{array} $$
do đó số tiền phải trả bằng

$$ 5000 \displaystyle \sum_{n=2}^{\infty} (n-1) P[N = n] = 554. $$
