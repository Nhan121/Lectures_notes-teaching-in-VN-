#### Question 151.
- From 27 pieces of luggage, an airline luggage handler damages a random sample of four.
- The probability that exactly one of the damaged pieces of luggage is insured is twice the probability that none of the damaged pieces are insured.

Calculate the probability that exactly two of the four damaged pieces are insured.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $r$ là số `khoang kiện của một hành lý` được đảm bảo (`pices of luggage which be insured`).
- Giả thiết đầu tiên cho biết có tổng cộng 27 `kiện (pieces)` nên $r \in [0, 27]$. Đồng thời cũng cho biết chỉ có $4$ kiện được làm hỏng (`airline luggage handler damages`)
- Giả thiết còn lại `(ii)`: XS mà có đúng `một kiện hỏng (damaged pieces of luggage)` hành lý được đảm bảo sẽ **gấp đôi** XS mà `không có kiện hỏng` hành lý nào được đảm bảo. 
- BT yêu cầu tính XS mà có đúng 2 trong số 4 kiện bị hỏng được đảm bảo. Ta có

$\qquad \diamond$ Số cách chọn 4 kiện bất kỳ trong tổng thể 27 là $C_{27}^4$

$\qquad \diamond$ Chọn ngẫu nhiên 4 `kiện`; nếu có $k$ `kiện` hỏng (mà chúng được đảm bảo) trong $r$ `kiện` **được đảm bảo**; và tức là có $4-k$ kiện không hỏng trong số $27 - r$ kiện **không được đảm bảo**; sẽ có $C_r^k C_{27 - r}^{4-k}$ trường hợp

$\qquad \diamond$ Do đó, khi chọn ngẫu nhiên 4 kiện hàng, XS để nhận được $k$ kiện hỏng (mà nó đã **được đảm bảo**) là: $\dfrac{C_r^k C_{27 - r}^{4-k}}{C_{27}^4}$

$\qquad \diamond$ Như vậy, `(ii)` có thể viết lại thành:

$$ 2 = \dfrac{\left. \left( \dfrac{C_r^k C_{27 - r}^{4-k}}{C_{27}^4} \right) \right\vert_{k=1}}{\left. \left( \dfrac{C_r^k C_{27 - r}^{4-k}}{C_{27}^4} \right) \right\vert_{k=0}} \Leftrightarrow \dfrac{r C_{27-1}^3}{C_{27-r}^4} = 2 \Leftrightarrow \dfrac{4r}{24 - r} = 2 \Leftrightarrow r = 8 $$

Thay $r=8$, khi đó XS để khi chọn ngẫu nhiên 4 kiện hàng thì có đúng 2 kiện bị hỏng trong 4 kiện được đảm bảo sẽ là

$$ \left. \left( \dfrac{C_r^k C_{27 - r}^{4-k}}{C_{27}^4} \right) \right\vert_{k=2}^{r=8} = \dfrac{C_8^2 C_{19}^2}{C_{27}^4} = \dfrac{266}{975} \approx 0.27 $$

#### Question 152.
- `(i)` Automobile policies are separated into two groups: `low-risk` and `high-risk`. 
- `(ii)` **`Actuary Rahul`** examines low-risk policies, continuing until a policy with a claim is found and then stopping. 
- `(iii)` **`Actuary Toby`** follows the same procedure with high-risk policies. 
- `(iv)` Each `low-risk` policy has a 10% probability of having a claim. 
- `(v)` Each high-risk policy has a `20%` probability of having a claim. 
- `(vi)` The claim statuses of polices are mutually independent.

Calculate the probability that `Actuary Rahul` examines fewer policies than `Actuary Toby`.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $R, T$ lần lượt là các phương thức định phí BH `Rahaul` và `Toby`. Theo đó,

$\qquad \diamond$ Theo `(ii)`, $R$ sẽ kiểm định các gói BH  cho đến khi gặp được đơn là `rủi ro thấp (low-risk)`. Do đó $R$ có phân phối hình học và kết hợp với `(iv)`, $R \sim \text{Geom}(0.1)$, tức là $$P[R = k] = (0.1)(0.9)^{k-1}; \quad \forall k \geq 1$$
$\qquad \diamond$ Theo `(iii)`, $T$ sẽ kiểm định tất cả các gói BH nào là `rủi ro cao (low-risk)`. Do đó XS để $T$ gặp được ít nhất $k$ đơn hàng là `rủi ro cao`, tức là $$P[T > k] = (0.8)^{k}; \quad \forall k \geq 1$$
- Đề bài yêu cầu tính XS mà `Rahaul` kiểm định được ít hơn `Toby`; tức là
$$ P[R < T, T \geq 1] = \displaystyle \sum_{n=1}^{\infty} P[R = n, T > n] $$
Vì `(vi)` cho biết $R, T$ được thực hiện độc lập với nhau, do đó
$$ \begin{array}{ccl} \displaystyle \sum_{n=1}^{\infty} P[R = n, T > n] &=& \displaystyle \sum_{n=1}^{\infty} P[R = n] P[T > n] \\ &=& \displaystyle \sum_{n=1}^{\infty} (0.1)(0.9)^{n-1} \cdot (0.8)^n \\ &=& (0.1)(0.9)^{-1}\displaystyle \sum_{n=1}^{\infty} (0.72)^n \\ &=& \dfrac{0.1}{0.9} \left( \displaystyle \sum_{n=1}^{\infty} (0.72)^n \right) \\ &=& \dfrac{1}{9} \left( \dfrac{0.72}{1 - 0.72} \right) = 0.2857 \end{array} $$

trong đó, dấu bằng ở dòng cuối trong biểu thức trên có được từ quy tắc tổng của cấp số cộng với công bội không âm nhỏ hơn 1 (hay quy tắc chuỗi hình học)
$$ \displaystyle \sum_{n=1}^{\infty} \theta^n = \dfrac{\theta}{1 - \theta}, \qquad \forall \theta \in (0, 1). $$

#### Question 153.
Let $X$ represent the number of customers arriving during the morning hours and let $Y$ represent the number of customers arriving during the afternoon hours at a diner. You are given:
- `i)` $X$ and $Y$ are `Poisson` distributed.
- `ii)` The first moment of $X$ is less than the first moment of $Y$ by $8$.
- `iii)` The second moment of $X$ is `60%` of the second moment of $Y$.

Calculate the variance of $Y.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $\theta, \lambda$ lần lượt là các tham số trong phân phối Poisson $X, Y$ (theo `(i)`); tức là

$$ X \sim \text{Pois}(\theta), \quad Y \sim \text{Pois}(\lambda) $$
- Theo `(ii)` thì moment bậc 1 của $X$ (tức $\mathbb{E}X$) bé hơn moment bậc một của $Y$ và

$$ \mathbb{E}Y - \mathbb{E}X = 8 \Leftrightarrow \lambda - \theta = 8 \Leftrightarrow \theta = \lambda - 8. $$
- Theo `(iii)`, moment bậc 2 của $X$ bằng `60%` moment bậc 2 của $Y$; tức là

$$ \begin{array}{rccl} & \mathbb{E}X^2 = (0.6) \mathbb{E}Y^2 \\ \Rightarrow & \text{Var}X + [\mathbb{E}X]^2 &=& 0.6 \left( \text{Var}Y + [\mathbb{E}Y]^2 \right) \\ \Rightarrow & \theta + \theta^2 &=& 0.6 \left( \lambda + \lambda^2 \right) \end{array}$$

Thay $\theta = \lambda - 8$ vào đẳng thức cuối cùng, ta được

$$ \begin{array}{crcl} & \lambda - 8 + [\lambda - 8]^2 &=& 0.6 \left( \lambda + \lambda^2 \right) \\ \Rightarrow & 0.4\lambda^2 - 15.6\lambda + 56 &=& 0 \\ \Rightarrow & \lambda &=& \lbrace 4, 35 \rbrace \end{array} $$

**Biện luận.** Vì các tham số $\theta, \lambda$ là trung bình của phân phối Poisson nên $\theta, \lambda > 0$; do đó

$\qquad \diamond$ Nếu $\lambda = 4$ thì $\theta = \lambda - 8 = -4 < 0 \Rightarrow$ **loại**.

$\qquad \diamond$ Nếu $\lambda = 35$ thì $\theta = \lambda - 8 = 27 > 0 \Rightarrow$ **nhận**.

- Như vậy; phương sai $\text{Var}Y$ của $Y$ bằng $\lambda = 35$.

#### Question 154.
In a certain game of chance, a square board with area 1 is colored with sectors of either `red` or `blue`. 
- `(i)` A player, who cannot see the board, must specify a point on the board by giving an x-coordinate and a y-coordinate. 
- `(ii)` The player **wins** the game if the specified point is in a **`blue sector`**. 
- `(iii)` The game can be arranged with any number of red sectors, and the *`red sectors`* are designed so that

$R_i = \left( \dfrac{9}{20} \right)^i$, where $R_i$ is the area of the $i^{th}$ red sector.

Calculate the minimum number of red sectors that makes the chance of a player winning less than `20%.`

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là số các ô màu đỏ trong hình vuông (chỉ có 2 màu xanh và đỏ).
- Giả sử $w$ là XS để một người chơi giành chiến thắng (nếu ghi điểm ở một ô màu xanh bất kỳ, theo `(ii)`). Khi đó, $w = \lbrace$ XS mà người chơi trượt tất cả các ô màu đỏ $\rbrace$. Ta có:

$\qquad \diamond$ Theo `(iii)` thì XS rơi vào ô màu đỏ ở lần thứ $i^{th}$ là $R_i = \left( \dfrac{9}{20} \right)^i$

$\qquad \diamond$ Do đó, XS để rớt khỏi ô màu đỏ sau $N$ lần là $1 - \displaystyle \sum_{i=1}^N R_i$

$\qquad \diamond$ Áp dụng công thức chuỗi hình học, ta có
$$ w = 1 - \displaystyle \sum_{i=1}^N R_i = 1 - \dfrac{9}{20} \sum_{i=0}^{N-1} \left( \dfrac{9}{20} \right)^i = 1 - \dfrac{9}{20}  \left[ \dfrac{1 - \left( \dfrac{9}{20} \right)^N}{1 - \dfrac{9}{20}} \right] = \dfrac{2}{11} + \dfrac{9}{11} \left( \dfrac{9}{20} \right)^N $$
- Vì BT yêu cầu tính số ô màu đỏ để người chơi có ít hơn `20%` cơ hội chiến thắng, tức là $w < 20,$ tương đương với

$$ \dfrac{2}{11} + \dfrac{9}{11} \left( \dfrac{9}{20} \right)^N < 0.2 \Leftrightarrow N > \dfrac{\ln(45)}{\ln(20/9)} \approx 4.76 $$

Vì vậy số nguyên nhỏ nhất lớn hơn $4.76$ là $N = 5$ và đây là số ô màu đỏ cần thiết để XS chiến thắng thấp hơn $20\%$

The probability of y losses in year 2 given x losses in year 1 is given by the table:

Calculate the probability of exactly 2 losses in 2 years.

$$ P[X+Y = 2] = P[X=0, Y=2] + P[X=1, Y=1] + P[X=2, Y=0] $$

$$ \begin{array}{crl} P[X=0, Y=2] &=& P[Y=2 | X=0] P[X = 0] \\ &=& (0.05)(0.5)^{0+1} \\ &=& 0.025 \\ P[X=1, Y=1] &=& P[Y=1 | X=1] P[X = 1] \\ &=& (0.3)(0.5)^{1+1}\\ &=& 0.075 \\ P[X=2, Y=0] &=& P[Y=0 | X=2] P[X = 2] \\ &=& (0.25)(0.5)^{2+1}\\ &=& 0.03125 \end{array} $$
vì vậy
$$ P[X+Y = 2] = P[X=0, Y=2] + P[X=1, Y=1] + P[X=2, Y=0] = 0.13125. $$

#### Question 157.
Let $X$ be a continuous random variable with density function $$ f(x) = \left\lbrace \begin{array}{cl} \dfrac{p-1}{x^p}, & x > 1 \\ 0, & \text{otherwise} \end{array} \right. $$
Calculate the value of $p$ such that $\mathbb{E}(X) = 2$

**Giải đáp / SOLUTIONs.**

BT đã cho hàm MDXS và yêu cầu tìm tham số $p$ để $\mathbb{E}X = 2$, điều này tương đương với giải phương trình tích phân

$$ \displaystyle \int x f(x) dx = 2 \Leftrightarrow \int_1^{\infty} x \left( \dfrac{p-1}{x^p} \right)dx = 2 \Leftrightarrow \dfrac{p-1}{p-2} = 2 \Leftrightarrow p = 3 $$

#### Question 159.

#### Question 160.
- An actuary analyzes a company’s annual personal auto claims, $M$, and annual commercial auto claims, $N.$ 
- The analysis reveals that $\text{Var}(M) = 1600$, $\text{Var}(N) = 900$, and the correlation between $M$ and $N$ is $0.64.$

Calculate $\text{Var}(M + N).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Giả thiết cho ta biết $\text{Var}(M) = 1600$, $\text{Var}(N) = 900$, và hệ số tương quan 

$$ \text{Corr}(M, N) = 0.64 $$
- Nhắc lại rằng $ \text{Corr}(M, N) = \dfrac{\text{Cov}(M, N)}{\sqrt{\text{Var}(M) \text{Var}(N)}} $, do đó ta được 

$$ \text{Cov}(M, N) = (0.64) \sqrt{\text{Var}(M) \text{Var}(N)} = 768 $$

suy ra

$$ \text{Var}(M+N) = \text{Var}(M) + \text{Var}(N) + 2\text{Cov}(M, N) = 4036. $$
