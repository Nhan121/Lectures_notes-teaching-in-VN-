#### Question 111, 
Once a fire is reported to a fire insurance company, the company makes an initial estimate, $X$, of the amount it will pay to the claimant for the fire loss. When the claim is finally settled, the company pays an amount, $Y$, to the claimant. The company has determined that $X$ and $Y$ have the joint density function 
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{2}{x^2(x-1)} y^{-(2x-1)/(x-1)}, & x > 1, y >1 \\ 0, & \text{otherwise} \end{array} \right.$$

Given that the initial claim estimated by the company is 2, calculate the probability that
the final settlement amount is between 1 and 3.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta cần tính $P[1 < Y < 3 | X = 2]$, bằng định nghĩa của nó 
$$ P[1 < Y < 3 | X = 2] = \dfrac{P[X=2, Y \in (1,3)]}{P[X=2]} = \displaystyle \dfrac{\int_1^3 f(2, y) dy}{f_X(2)} $$
- Ta có

$$ \begin{array}{ccll} f(2, y) &=& \dfrac{2}{2^2(2 - 1)} y^{-(2\cdot 2 - 1)/(2 - 1)} = \dfrac{1}{2y^3} & \text{ khi } y > 1 \\ f_X(2) &=& \displaystyle \int_1^{\infty} f(2, y)dy = \int_1^{\infty} \dfrac{1}{2y^3}dy = \dfrac{1}{4} \\ \displaystyle \int_1^3 f(2,y)dy &=& \displaystyle \int_1^3 \dfrac{1}{2y^3}dy = \dfrac{2}{9} \end{array} $$
do đó
$$ P[1 < Y < 3 | X = 2] = \dfrac{8}{9} $$

#### Question 112.
- A company offers a basic life insurance policy to its employees, as well as a
supplemental life insurance policy. To purchase the supplemental policy, an employee
must first purchase the basic policy.

- Let $X$ denote the proportion of employees who purchase the basic policy, and $Y$ the
proportion of employees who purchase the supplemental policy. 
- Let $X$ and $Y$ have the
joint density function $f(x,y) = 2(x + y)$ on the region where the density is positive.

Given that 10% of the employees buy the basic policy, calculate the probability that
fewer than 5% buy the supplemental policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Trước tiên, lưu ý rằng $f(x, y)$ là hàm MDXS thể hiện tỷ lệ của nhóm "chính sách cơ bản ($X$)" và "chính sách bổ sung ($Y$)" nên miền giá trị là $0 < x, y < 1$. Hơn nữa, chỉ những ai đã mua gói cơ bản mới có thể mua thêm gói bổ sung nên tỷ lệ của gói bổ sung $y$ phải thấp hơn $x$, tức là

$$ f(x,y) = \left\lbrace \begin{array}{cl} 2(x + y), & 0 < y < x < 1, \\ 0 & \text{otherwise} \end{array} \right.$$
- Ta cần tính $P[Y < 0.05 | X = 0.10]$ và ta có
$$ P[Y < 0.05 | X = 0.10] = \dfrac{\displaystyle \int_0^{0.05} f(0.1, y)dy }{f_X(0.1)} $$
- Tương tự, ta phải thực hiện các thao tác sau

$$ \begin{array}{ccll} f(0.1, y) &=& 2(0.1 + y) = 2y + 0.2 & \text{khi } y \in (0, 0.1) \\ f_X(0.1) &=& \displaystyle \int_0^{0.1} f(0.1, y) dy = \int_0^{0.1} (2y + 0.2) dy = 0.03 \\ \displaystyle \int_0^{0.05} f(0.1, y) dy &=& \displaystyle \int_0^{0.05} (2y + 0.2)dy = 0.0125 \end{array} $$
suy ra
$$ P[Y < 0.05 | X = 0.10] = \dfrac{0.0125}{0.03} \approx 0.417. $$

#### Question 113.
Two life insurance policies, each with a death benefit of 10,000 and a one-time
premium of 500, are sold to a married couple, one for each person. 

The policies will expire at the end of the tenth year. The probability that only the wife will survive at least
ten years is 0.025, the probability that only the husband will survive at least ten years is
0.01, and the probability that both of them will survive at least ten years is 0.96.

Calculate the expected excess of premiums over claims, given that the husband survives
at least ten years. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện người sống sót sau 10 năm là vợ chồng hay cả hai; xác định bởi

$$ X = \left\lbrace \begin{array}{cll} W, & \text{neu chỉ nguoi vo con song} \\ H, & \text{ neu chỉ nguoi chong con song} \\ B, & \text{neu cả 2 còn sống} \\ N, & \text{neu ko co ai song sot} \end{array} \right. $$
- Theo giả thiết, ta có
$$ P[X = W] = 0.025, P[X = H] = 0.01, P[X = B] = 0.96, \text{ và } P[X = N] = 0.005 $$
- BT yêu cầu tính kỳ vọng khi số tiền BH phải thanh toán vượt quá yêu cầu; biết rằng người chồng còn sống sau 10 năm. Ta thấy rằng, từ giả thiết đầu tiên của BT, thì:

$\qquad \diamond$ Tổng số tiền BH sẽ chi rả cho mỗi cái chết là 10000

$\qquad \diamond$ Mỗi cặp vợ chồng sẽ mua BH với giá `2 x 500 = 1000`

$\qquad \diamond$ Khi người vợ (hoặc chồng) chết thì tiền BH sẽ trao cho người còn lại.

- Do đó, nếu BT cho biết người chồng còn sống sau 10 năm thì tiền thụ hưởng đó là khi người vợ chết đi; do đó trước tiên ta cần tính được xác suất có điều kiện `"tử số": chỉ người chồng còn sống` và `"mẫu số": XS ít nhất người chồng còn sống`, theo đó

        - XS để ít nhất người chồng còn sống là P[X = H] + P[X = B] = 0.01 + 0.96 = 0.97
        - XS chỉ có người chồng còn sống là P[X = H] = 0.01
nên kỳ vọng số tiền BH vượt quá yêu cầu; biết rằng người chồng còn sống sau 10 năm sẽ được tính bởi hiệu số của "số tiền mỗi cặp vợ chồng bỏ ra khi mua BH" và "số tiền mà BH phải thanh toán khi biết người chồng còn sống sau 10 năm."

$$ 1000 - 10000 P[X = H| (X = H) \cup (X = B)] = 1000 - \dfrac{10000}{97} \approx 1000 - 103 = 897. $$

#### Question 114.
A diagnostic test for the presence of a disease has two possible outcomes: 1 for
disease present and 0 for disease not present. Let $X$ denote the disease state (0 or 1) of a
patient, and let $Y$ denote the outcome of the diagnostic test. The joint probability function
of $X$ and $Y$ is given by: 
$$ \begin{array}{ccl} P[X = 0, Y = 0] &=& 0.800 \\ P[X = 1, Y = 0] &=& 0.050 \\ P[X = 0, Y = 1] &=& 0.025 \\ P[X = 1, Y = 1] &=& 0.125  \end{array} $$

Calculate $\text{Var}(Y |X = 1)$. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính $\text{Var}(Y |X = 1)$ thì ta cần biết chính xác phân phối của $Y$. Bằng công thức Bayes, ta có

$$ \begin{array}{ccl} \diamond P[Y = 0 | X = 1] &=& \dfrac{P[Y=0, X=1]}{P[X=1]} = \dfrac{P[X=1, Y=0]}{P[X=1,Y=0]+P[X=1,Y=1]} \approx 0.286 \\ \diamond P[Y = 1 | X = 1] &=& 1 - P[Y=0 | X=1] = 0.714 \end{array}$$
- Do đó $(Y | X=1) \sim \mathcal{B}(1, \theta = 0.286)$ nên có phương sai

$$ \text{Var}(Y | X=1) = \theta (1 - \theta) = 0.204. $$

#### Question 115.
The stock prices of two companies at the end of any given year are modeled with random
variables $X$ and $Y$ that follow a distribution with joint density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} 2x, & x \in (0, 1), 0 < x< y < x+1 \\ 0, & \text{otherwise} \end{array} \right.$$
Determine the conditional variance of $Y$ given that $X = x$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính $\text{Var}(Y | X = x)$ thì ta trước tiên cần xác định được hàm MDXS có điều kiện $f(y|x)$. 
- Ta biết rằng $$ f(y|x) = \dfrac{f(x, y)}{f_X(x)}; $$ ta có

$$ \begin{array}{rcl} f_X(x) &=& \displaystyle \int_{x}^{x+1} f(x,y)dy = \int_{x}^{x+1} 2xdy = 2x \\ \Rightarrow f(y|x) &=& 1_{\lbrace x < y < x+1 \rbrace} = 1_{\lbrace y \in (x, x+1) \rbrace} \end{array} $$
- Nhắc lại rằng một BNN có phân phối đều $\mathcal{U}([a, b])$ sẽ có phương sai là $\dfrac{(b - a)^2}{12}$
- Từ $f(y|x)$ suy ra BNN $(Y | X=x)$ có phân phối đều $\mathcal{U}([x, x+1])$ nên phương sai sẽ là

$$ \text{Var}(Y | X = x) = \dfrac{[(x+1) - x]^2}{12} = \dfrac{1}{12}. $$

#### Question 116.
An actuary determines that the annual number of tornadoes in counties P and Q are
jointly distributed as follows:

      .....................................................................
      |               .   |    Annual number of tornadoes in "county Q"   |
      |               .   |...............................................|
      |               .   |     0     |     1     |     2     |     3     |
      .................---|-----------------------------------------------|  
      | Annual number . 0 |    0.12   |    0.06   |    0.05   |    0.02   |            
      | of tornadoes  . 1 |    0.13   |    0.15   |    0.12   |    0.03   |
      | in "county P" . 2 |    0.05   |    0.15   |    0.10   |    0.02   |
      .....................................................................
Calculate the conditional variance of the annual number of tornadoes in county Q, given
that there are no tornadoes in county P.       

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ là BNN thể hiện số cơn bão hàng năm tại các nước "P", "Q" (việc ký hiệu lại như vậy để tránh nhầm lẫn "P" với xác suất P).
- Yêu cầu BT là tính phương sai có điều kiện về số cơn bão ở "Q" biết rằng chẳng có cơn nào ở "P"; tức là

$$ \text{Var}(Y | X=0) $$
- Từ bảng phân phối xác suất đồng thời về số cơn bão ở các quốc gia "P", "Q"; ta tìm được bảng phân phối xác suất lề cho $(Q|P=0)$ như sau

$$ \begin{array}{rcl} P[X = 0] &=& P[Y=0|X=0] + P[Y=1|X=1] + P[Y=2|X=1] + P[Y=3|X=1] = 0.25 \\ P[Y=0|X=0] &=& \dfrac{P[X=0, Y=0]}{P[X=0]} = \dfrac{0.12}{0.25} = \dfrac{12}{25} \\ P[Y=1|X=0] &=& \dfrac{P[X=0, Y=1]}{P[X=0]} = \dfrac{6}{25} \\ P[Y=2|X=0] &=& \dfrac{P[X=0, Y=2]}{P[X=0]} = \dfrac{1}{5} \\  P[Y=3|X=0] &=& \dfrac{P[X=0, Y=3]}{P[X=0]} = \dfrac{2}{25} \end{array} $$
từ đó suy ra

$$ \begin{array}{rcl} \mathbb{E}(Y|X=0) &=& \displaystyle \sum_{y=0}^3 y P[Y=y | X=0] = \dfrac{22}{25} \\ \mathbb{E}\left( Y^2 | X = 0 \right) &=& \displaystyle \sum_{y=0}^3 y^2 P[Y=y | X=0] = \dfrac{44}{25} \\ \text{Var}(Y|X=0) &=& \mathbb{E}\left( Y^2 | X = 0 \right) - \mathbb{E}\left( Y| X = 0 \right) = 0.9856 \end{array} $$

#### Question 117.
A company is reviewing tornado damage claims under a farm insurance policy.
Let $X$ be the portion of a claim representing damage to the house and let $Y$ be the portion
of the same claim representing damage to the rest of the property. The joint density
function of $X$ and $Y$ is

$$ f(x,y) = \left\lbrace \begin{array}{cl} 6[1 - (x+y)], & x,y \in (0, 1), 0 < x+y <1 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the probability that the portion of a claim representing damage to the house is
less than 0.2. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Bài toán yêu cầu tính $P[X < 0.2]$, do đó trước tiên ta cần tìm hàm MDXS lề của $X$, xác định bởi

$$ f_X(x) = \displaystyle \int f(x,y)dy = \int_0^{1-x} 6[1 - (x+y)]dy = 6\left. \left( y - xy + 0.5y^2 \right) \right\vert_{0}^{y = 1-x} = 6[0.5x^2 - x + 0.5], \quad \forall x \in (0,1) $$
- Suy ra

$$ P[X < 0.2] = \int_0^{0.2} f_X(x) dx = \int_0^{0.2} 6[0.5x^2 - x + 0.5]dx = 0.488. $$

#### Question 118.
Let X and Y be continuous random variables with joint density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} 15y, & x^2 \leq y \leq x \\ 0, & \text{otherwise} \end{array} \right. $$
Let $g$ be the marginal density function of $Y$.

Determine which of the following represents $g$. 

**Giải đáp / SOLUTIONs.**

BT đã cho hàm MDXS đồng thời của cặp $(X,Y)$ do đó hàm MDXS lề $g(y)$ sẽ được xác định bởi

$$ g(y) = \displaystyle \int f(x,y) dx = \int_A f(x,y)dx $$
trong đó
$$ \begin{array}{ccl} A &=& \lbrace x: x^2 \leq y \leq x \rbrace \\ &=& \lbrace x: y \leq x \leq \sqrt{y} \text{ với } y \in (0, 1) \rbrace \end{array} $$
Do vậy,
$$ g(y) = \displaystyle \int_y^{\sqrt{y}} 15y dx = 15y^{3/2} \left( 1 - y^{1/2} \right), \quad \forall y \in (0, 1) $$
hay
$$ g(y) = \left\lbrace \begin{array}{ccl} 15y^{3/2} \left( 1 - y^{1/2} \right), & \forall y \in (0, 1) \\ 0, & \text{otherwise} \end{array} \right. $$

#### Question 119.
An auto insurance policy will pay for damage to both the policyholder’s car and
the other driver’s car in the event that the policyholder is responsible for an accident.
The size of the payment for damage to the policyholder’s car, $X$, has a marginal density
function of 1 for $0 < x < 1$. Given $X = x$, the size of the payment for damage to the other
driver’s car, $Y$, has conditional density of 1 for $x < y < x + 1$.

 Given that the policyholder is responsible for an accident, calculate the probability that
the payment for damage to the other driver’s car will be greater than 0.5. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo giả thiết, $X$ có hàm MDXS $f_X(x) = 1$ với $x \in (0, 1)$; khi cố định $X = x$ thì hàm MDXS có điều kiện $f(y|x) = 1$ với $y \in (x, x+1)$. Do đó hàm MDXS đồng thời của cặp $(X, Y)$ sẽ xác định bởi

$$ f(x,y) = f_X(x) f(y|x) =\left\lbrace \begin{array}{cl} 1, & \text{ khi } x\in (0,1); x < y < x+1 \\ 0, & \text{otherwise} \end{array} \right. $$
- BT yêu cầu tính XS để khoản tiền bồi thường thiệt hại cho xe của người lái xe kia sẽ lớn hơn 0.5; tức là $P[Y > 0.5]$, do đó ta phải xác định hàm MDXS của $Y$, ta có

$$ \displaystyle f_Y(y) = \int f(x,y)dx = \int_{\max(0, y-1)}^{\min(1, y)} dx = \min(y,1) - \max(0, y-1), \quad \forall y \in (0, 2) $$
suy ra
$$ \begin{array}{ccl} P[Y>0.5] &=& 1 - P[Y \leq 0.5] \\ &=& \displaystyle 1 - \int_0^{0.5} \left[ \min(y,1) - \max(0, y-1) \right] dy \\ &=& \displaystyle 1 - \int_0^{0.5} ydy \\ &=& 1 - \dfrac{1}{8} \\ &=& \dfrac{7}{8} \end{array} $$

#### Question 120.
An insurance policy is written to cover a loss $X$ where $X$ has `density function`


$$ f(x) = \left\lbrace \begin{array}{cl} \dfrac{3x^2}{8}, &x \in (0,2) \\ 0, & \text{otherwise} \end{array} \right. $$

The time (in hours) to process a claim of size $x$, where $x \in [0, 2]$ , is `uniformly distributed` on the interval from $x$ to $2x$.

Calculate the probability that a randomly chosen claim on this policy is processed in three
hours or more. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện thời gian để xử lý một yêu cầu với cỡ (`size`) $x \in [0, 2]$ cố định; theo giả thiết thì $(Y|X=x)$ sẽ có phân phối đều trong khoảng $(x, 2x)$; tức là $$ f(y|x) = \left\lbrace \begin{array}{ccl} \dfrac{1}{x}, & x <y<2x \\ 0, & \text{otherwise} \end{array} \right. $$
- Do đó hàm MDXS đồng thời sẽ là

$$ f(x,y) = f(y|x)f(x) = \left\lbrace \begin{array}{ccl} \dfrac{3x}{8}, & x\in(0,2), x <y<2x \\ 0, & \text{otherwise} \end{array} \right. $$
suy ra hàm MDXS lề cho $Y$ 
$$ f_Y(y) = \displaystyle \int_{y/2}^{\min(2, y)} f(x,y)dx = \int_{y/2}^{\min(2, y)}\dfrac{3x}{8}dx=\dfrac{3}{16} \min(2,y)^2 - \dfrac{3y^2}{64}, \quad \forall y \in [0, 4] $$ 
- BT yêu cầu tính XS để một khiếu nại được chọn ngẫu nhiên về chính sách này được xử lý lâu hơn 3 giờ; tức là $P[Y \geq 3]$, ta có

$$ P[Y \geq 3] = \displaystyle \int_3^4 f_Y(y)dy = \int_3^4 \left( \dfrac{3}{16} \min(2,y)^2 - \dfrac{3y^2}{64} \right) dy = \int_3^4 \left( \dfrac{3}{4} - \dfrac{3y^2}{64} \right)dy = 0.172 $$
