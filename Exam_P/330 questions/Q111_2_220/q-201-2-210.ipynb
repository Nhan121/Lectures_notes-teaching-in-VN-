#### Question 201.
A theme park conducts a study of families that visit the park during a year. 

- The fraction of such families of size $m$ is $\dfrac{8-m}{28}$ for $m = 1,2,\ldots,7.$

- For a family of size $m$ that visits the park, the number of members of the family that ride
the roller coaster follows a discrete uniform distribution on the set $\lbrace 1, \ldots, m \rbrace$

Calculate the probability that a family visiting the park has exactly six members, given
that exactly five members of the family ride the roller coaster.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, N$ lần lượt là **số lần** gia đình đó đến công viên; và **số thành viên** trong đó chơi trò tàu lượn (`roller coaster`).
- Theo giả thiết, ta có $$ \begin{array}{rcllr} P[X = m] &=& \dfrac{8-m}{28}, & \forall m = 1,2,\ldots,7. \\ P[Y = k | X = m] &=& \dfrac{1}{m}, &\forall k = 1,2,\ldots,m; & \forall m = 1,2,\ldots,7 \\ P[Y = k | X = m] &=& 0, & \forall k \leq 0, \text{ or } \forall k > m; & \forall m = 1,2,\ldots,7\end{array} $$
- BT yêu cầu tính XS một gia đình có đúng 6 người ghé công viên biết rằng có đúng 5 người chơi trò tàu lượn; tức là $P[X=6|Y=5]$. Áp dụng công thức Bayes, ta có
$$ \begin{array} P[X=6|Y=5] &=& \dfrac{P[Y=5 | X=6] P[X=6]}{\displaystyle \sum_{m=1}^7 P[Y=5 | X=m] P[X=m] } \\ &=& \dfrac{ \dfrac{1}{6} \left( \dfrac{8 - 6}{28} \right) }{ 0 + 0 + 0 + 0 + \dfrac{1}{5} \left( \dfrac{8-5}{28} \right) + \dfrac{1}{6} \left( \dfrac{8-6}{28} \right) + \dfrac{1}{7} \left( \dfrac{8-7}{28} \right)}\\ &=& \dfrac{\dfrac{1}{3}}{\dfrac{3}{5}+\dfrac{1}{3}+\dfrac{3}{7}} \\ &=& \dfrac{35}{113} \\ & \approx & 0.3097 \end{array} $$

#### Question 202.
The following information is given about a group of high-risk borrowers.
- i) Of all these borrowers, 30% defaulted on at least one student loan.
- ii) Of the borrowers who defaulted on at least one car loan, 40% defaulted on at least
one student loan.
- iii) Of the borrowers who did not default on any student loans, 28% defaulted on at
least one car loan.

A statistician randomly selects a borrower from this group and observes that the selected borrower defaulted on at least one student loan.

Calculate the probability that the selected borrower defaulted on at least one car loan.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $S$ là sự kiện mà chọn được người vay (`borrower`) đã không thể trả được nợ ít nhất **khoản vay sinh viên (`student loan`)**.
- Đặt $C$ là sự kiện mà chọn được người vay đã không thể trả được nợ ít nhất **khoản vay mua oto (`car loan`)**.
- BT yêu cầu tính XS **"khi chọn ngẫu nhiên một người vay mà người này không trả được khoản vay mua oto biết rằng người ngày không trả được khoản vay sinh viên."** Tức là $P[C | S]$, ta có

$$ P[C|S] = \dfrac{P[S \cap C]}{P[S]} $$
nên ta cần xác định $x = P[S \cap C]$.

- Từ các giả thiết

$\qquad \diamond$ Theo `(i)`, ta có $P[S] = 0.3$ suy ra $$P[S^c] = 1 - P[S] = 0.7$$

$\qquad \diamond$ Theo `(ii)`, $P[S|C] = 0.4$ suy ra $$P[C] = \dfrac{P[C \cap S]}{P[S|C]} = \dfrac{x}{0.4}$$

$\qquad \diamond$ Theo `(iii)`, $P[C|S^c] = 0.28$ suy ra $$P[C \cap S^c] = P[S^c] P[C | S^c] = (0.7)(0.28) = 0.196$$

$\qquad \diamond$ Hơn nữa, ta có $$P[C] = P[C \cap S] + P[C \cap S^c] = x + 0.196$$

$\qquad \diamond$ Do đó, ta sẽ tìm $x = P[C \cap S]$ bằng cách giải phương trình (thông qua trung gian $P[C]$)

$$ \dfrac{x}{0.4} = x + 0.196 \Rightarrow x = 0.13076$$

- Do đó,
$$ P[C|S] = \dfrac{P[S \cap C]}{P[S]} = \dfrac{0.13067}{0.3} \approx 0.4356.$$

#### Question 203.
A machine has two components and fails when both components fail. The number of
years from now until the first component fails, $X$, and the number of years from now until
the machine fails, $Y$, are random variables with joint density function 

$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{18} e^{-(x+y)/6}, & 0 < x< y< \infty \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate $\text{Var}(Y | X = 2).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính phương sai có điều kiện 

$$\text{Var}(Y | X = 2) = \mathbb{E}\left[ Y^2 | X=2 \right] + \left( \mathbb{E}[Y|X=2] \right)^2,$$ 

ta cần xác định hàm MDXS có điều kiện của $Y$ khi $X=2$, tức là 

$$ f_{Y|X}(y|2) = \dfrac{f(2, y)}{f_X(2)}$$
trong đó
$$ f(2,y) = f(x=2, y) = = \left\lbrace \begin{array}{cl} \dfrac{1}{18} e^{-(2+y)/6}, & 2< y< \infty \\ 0, & \text{otherwise} \end{array} \right. $$
và
$$ f_X(2) = \int f(2,y) dy = \int_2^{\infty} \dfrac{1}{18} e^{-(2+y)/6}dy = \dfrac{1}{3} e^{-2/3}$$
suy ra
$$ f_{Y|X}(y|2) = \dfrac{f(2, y)}{f_X(2)} = \left\lbrace \begin{array}{cl} \dfrac{1}{6} e^{-(y-2)/6}, & 2 y< \infty \\ 0, & \text{otherwise} \end{array} \right. $$
- Ta có
$$ \mathbb{E}[Y|X=2] = \displaystyle \int y f_{Y|X} (y) dy = \dfrac{1}{6} \int_2^{\infty} y e^{-(y-2)/6} dy = 8, $$
và
$$ \mathbb{E}\left[ Y^2 | X=2 \right] = \displaystyle \int y^2 f_{Y|X} (y) dy = \dfrac{1}{6} \int_2^{\infty} y^2 e^{-(y-2)/6} dy = 100. $$
do đó
$$\text{Var}(Y | X = 2) = \mathbb{E}\left[ Y^2 | X=2 \right] + \left( \mathbb{E}[Y|X=2] \right)^2 = 100 - (8)^2 = 36.$$ 

#### Question 204.
- The elapsed time, $T$, between the occurrence and the reporting of an accident has probability density function

$$ f(t) = \left\lbrace \begin{array}{cl} \dfrac{8t - t^2}{72}, & t \in (0, 6), \\ 0, & \text{otherwise} \end{array} \right. $$
- Given that $T = t,$ the elapsed time between the reporting of the accident and payment by
the insurer is uniformly distributed on $[2 + t, 10].$

Calculate the probability that the elapsed time between the occurrence of the accident and payment by the insurer is less than $4.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là khoảng thời gian giữa 2 thời điểm **báo cáo tai nạn (`reporting an accident`)** và **chi trả BH (`payment`)**, khi đó 

$$(Y|T=t) \sim \mathcal{U}([2+t, 10]) $$
tức là hàm MDXS của $(Y|T=t)$ là 
$$ f(y|t) = \dfrac{1}{10 - (2+t)} \mathbf{1}_{\lbrace y \in [2+t, 10] \rbrace} = \dfrac{1}{8-t} \mathbf{1}_{\lbrace y \in [2+t, 10] \rbrace} $$
- Vì vậy hàm MDXS đồng thời $f(y,t)$ của cặp $(Y, T)$ sẽ xác định bởi

$$ f(y,t) = f(y|t) f(t) = \left\lbrace \begin{array}{cl} \dfrac{t}{72}, & t \in (0, 6); y \in [2+t, 10], \\ 0, & \text{otherwise} \end{array} \right. $$
- BT yêu cầu tính XS để thời gian từ lúc xảy ra tai nạn tính đến lúc nhận được chi trả (đã thông qua báo cáo) không vượt quá 4, tức là 

$$P[Y+T < 4] = P[(Y,T) \in A]$$
với 
$$ \begin{array}{ccl} A &=& \lbrace (y, t): t \in [0, 6]; y \in [2+t, 10] \text{ and } y+t < 4 \rbrace \\ &=& \lbrace (y, t): 2+t \leq y < 4-t, y \in [0,1] \rbrace \end{array}$$
- Như vậy,
$$P[Y+T < 4] = \displaystyle \iint_A f(y,t)dy dt = \int_0^1 \int_{2+t}^{4-t} \dfrac{t}{72} dy dt = \int_0^1 t \left[ \dfrac{(4-t) - (2+t)}{72} \right] dt = \dfrac{1}{216}.$$

#### Question 205.
- The time until failure, $T,$ of a product is modeled by a uniform distribution on [0, 10]. An extended warranty pays a benefit of $100$ if failure occurs between time $t = 1.5$ and $t = 8.$
- The present value, $W,$ of this benefit is 
$$ W = \left\lbrace \begin{array}{cl} 0, & 0 \leq T < 1.5 \\ 100 e^{-0.4T}, & 1.5 \leq T < 8 \\ 0, & 8 \leq T \leq 10 \end{array} \right. $$

Calculate $P(W < 79).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ biểu thức xác định của $W$, ta có $W \geq 0$ với mọi $T \in [0, 10]$
- Hơn nữa, $T = \mathcal{U}([0, 10])$ nên ta có lưu ý sau:
$$ P[a < T < b] = \dfrac{b - a}{10 - 0}, \qquad \forall 0 \leq a < b \leq 10 $$
- Do đó,
$$ P[W < 79] = P[W = 0] + P[0 < W < 79 ] $$
- Xét các biểu thức trong tổng,

$\qquad \diamond P[W=0].$ Lưu ý rằng, $W=0$ tương ứng với 2 trường hợp $T\in [0, 1.5)$ hoặc $T \in [8, 10]$, do đó

$$ \begin{array}{ccl} P[W = 0] &=& P[0 \leq T < 1.5] + P[8 \leq T \leq 10] \\ &=& \left( \dfrac{1.5 - 0}{10} \right) + \left( \dfrac{10 - 8}{10} \right) \\ &=& 0.35 \end{array} $$ 

$\qquad \diamond P[0 < W < 79].$ Ta có

$$ \begin{array}{ccl} P[0 < W < 79] &=& P\left[ 100 e^{-0.4T} < 79 \text{ and } T \in [1.5, 8) \right] \\ &=& P\left[ -0.4T < \log (0.79) \text{ and } T \in [1.5, 8) \right] \\  &=& P\left[ \dfrac{-\log 0.79}{0.04} < T \leq 8 \right] \\ &=& \dfrac{8 - \dfrac{-\log 0.79}{0.04}}{10} \\ &=& 0.211 \end{array} $$ 

Suy ra
$$ P[W < 79] = P[W = 0] + P[0 < W < 79 ] = 0.561. $$

#### Question 206.
- An insurance company issues policies covering damage to automobiles. The amount of damage is modeled by a uniform distribution on $[0, b].$
- The policy payout is subject to a deductible of $b/10.$
- A policyholder experiences automobile damage.

Calculate the **ratio** of the **standard deviation** of the `policy payout` to the standard deviation of the `amount of the damage`. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là khoản thiệt hại mà các oto gây ra, theo đó $X \sim \mathcal{U}([0, b])$; do đó
$$ \text{Var}X = \dfrac{(b - 0)^2}{12} $$
suy ra
$$ \text{sd}X = \sqrt{\text{Var}X} = \dfrac{b}{\sqrt{12}} \approx 0.28868 b.$$
- và $Y$ là khoản chi trả BH khấu trừ với mức $b/ 10$; tức là $$Y = (X - 0.1b)\mathbf{1}_{\lbrace X \geq 0.1b \rbrace} = \max(0, X - 0.1b)$$ nên

$$ \begin{array}{ccl} \mathbb{E}Y &=& \mathbb{E} [\max(0, X - 0.1b)] \\ &=& \displaystyle \int_{0}^b \max(0, x - 0.1b) \dfrac{1}{b} dx \\ &=& \displaystyle \int_{0.1b}^b \dfrac{(x-0.1b)}{b} dx \\ &=& \left. \left( \dfrac{x^2}{2b} - 0.1x \right)\right\vert_{x= (0.1b)}^{x=b} \\ &=& 0.405b \end{array} $$
và
$$ \begin{array}{ccl} \mathbb{E}Y^2 &=& \mathbb{E} [\max(0, X - 0.1b)^2] \\ &=& \displaystyle \int_{0.1b}^b \dfrac{(x-0.1b)^2}{b} dx \\ &=& \left. \left( \dfrac{x^3}{3b} - 0.1x^2 + (0.01b)y \right)\right\vert_{x= (0.1b)}^{x=b} \\ &=& 0.243b^2 \end{array} $$
suy ra
$$ \text{sd}Y = \sqrt{\text{Var}Y} = \sqrt{\mathbb{E}Y^2 - [\mathbb{E}Y]^2} \approx 0.28102b $$
- BT yêu cầu ta tính tỷ số độ lệch chuẩn của $X$ và $Y$, tức là

$$ \dfrac{\text{sd}Y}{\text{sd}X} \approx 0.97347 $$

#### Question 207.
A policyholder purchases automobile insurance for two years. Define the following events:
- F = the policyholder has exactly one accident in year one.
-  G = the policyholder has one or more accidents in year two.

Define the following events:
- i) The policyholder has exactly one accident in year one and has more than one
accident in year two.
- ii) The policyholder has at least two accidents during the two-year period.
- iii) The policyholder has exactly one accident in year one and has at least one
accident in year two.
- iv) The policyholder has exactly one accident in year one and has a total of two or
more accidents in the two-year period.
- v) The policyholder has exactly one accident in year one and has more accidents in
year two than in year one.

Determine the number of events from the above list of five that are the same as $F \cap G$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N_i$ là BNN thể hiện số vụ tai nạn xe cộ trong năm thứ $i$; theo đó $F, G$ lần lượt tương ứng với

$$ F = \lbrace \omega: N_1(\omega) = 1 \rbrace, \qquad G = \lbrace \omega: N_2(\omega) \geq 1 \rbrace $$
- Các sự kiện `(i)-(v)` của BT tương ứng với

$$ \begin{array}{rccl} \text{(i)} & A &=& \lbrace \omega: N_1(\omega) = 1, N_2(\omega) > 1 \rbrace \\ \text{(ii)} & B &=& \lbrace \omega: N_1(\omega) + N_2(\omega) \geq 2 \rbrace \\ \text{(iii)} & C &=& \lbrace \omega: N_1(\omega) = 1, N_2(\omega) \geq 1 \rbrace \\ \text{(iv)} & D &=& \lbrace \omega: N_1(\omega) = 1, N_1 (\omega) + N_2(\omega) \geq 2 \rbrace \\ \text{(v)} & E &=& \lbrace \omega: N_1(\omega) = 1, N_2(\omega) > N_1(\omega) \rbrace \end{array} $$
- BT hỏi sự kiện nào trong 5 sự kiện trên mô tả $F \cap G$; ở đây ta thấy rằng `(iii), (iv)` là 2 đáp án đúng nhất bởi

$$ C =  \lbrace \omega: N_1(\omega) = 1, N_2(\omega) \geq 1 \rbrace =  \lbrace \omega: N_1(\omega) = 1 \rbrace \cap  \lbrace \omega: N_2(\omega) \geq 1 \rbrace = F \cap G $$
do $N_1, N_2$ độc lập và

$$ D = \lbrace \omega: N(\omega) = 1, N_1 (\omega) + N_2(\omega) \geq 2 \rbrace = \lbrace \omega: N(\omega) = 1 \text{ and } N_2(\omega) \geq 2 - N_1(\omega) = 1 \rbrace = F \cap G$$

--------------

**Comment.**
- (i), (v) sai giống nhau; dễ thấy rằng 

$$A = E = \lbrace N_1 = 1 \rbrace \cap \lbrace N_2 > 1 \rbrace \neq F \cap G$$
- (ii) sai vì $B$ cũng có thể được viết thành

$$ B = \lbrace N_1 = 0 \rbrace \cap \lbrace N_2 \geq 2 \rbrace \neq F \cap G. $$

---------------

Do đó, BT này chỉ có 2 khẳng định đúng là `(iii)` và `(iv)`.

#### Question 208.
An insurance company categorizes its policyholders into three mutually exclusive groups: `high-risk`, `medium-risk`, and `low-risk`. 
- `(i)` An internal study of the company showed that `45%` of the policyholders are `low-risk` and `35%` are `medium-risk`. 
- `(ii)` The probability of **death** over the next year, **given that** a policyholder is `high-risk` is two times the probability of death of a `medium-risk` policyholder. 
- `(iii)` The probability of **death** over the next year, **given that** a policyholder is `medium-risk` is three times the probability of death of a `low-risk` policyholder.
- `(iv)`The probability of **death** of a randomly selected policyholder, over the next year, is 0.009.

Calculate the probability of death of a policyholder over the next year, given that the policyholder is high-risk.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $D$ là sự kiện nếu KH mua BH chết trong năm tiếp theo.
- Đặt $H, M, L$ lần lượt là các nhóm **`"rủi ro cao (High risk)"`**, **`"rủi ro trung bình (Medium risk)"`** và **`"rủi ro thấp (Low risk)"`**
- Theo các giả thiết, ta có các dữ kiện sau.

$$ \begin{array}{rlll} \text{Theo (i)} & P[L] = 0.45, & P[M] = 0.35; & \Rightarrow P[H] = 1 - P[L] - P[M] = 0.2. \\ \text{ (ii)} & P[D|H] = 2 P[D|M] & & \Rightarrow P[D|M] = \dfrac{1}{2} P[D|H] \\ \text{iii)} & P[D|M] = 3P[D|L]  & & \Rightarrow P[D|L] = \dfrac{1}{3} P[D|M] = \dfrac{1}{6} P[D|H] \\ \text{(iv)} & P[D] = 0.009 \end{array} $$
- BT yêu cầu tính $P[D|H],$ đặt $x = P[D|H]$.
- Áp dụng công thức xác suất đẩy đủ cho $H,M,L$ theo $D$; ta có

$$ \begin{array}{ccl} P[D] &=& P[D|H]\cdot P[H] + P[D|M]\cdot P[M] + P[D|L]\cdot P[L] \\ &=& x(0.2) + \left( \dfrac{x}{2} \right) (0.35) + \left( \dfrac{x}{6} \right) (0.45) \\ &=& 0.45x  \end{array} $$
suy ra
$$ P[D|H] = x = \dfrac{P[D]}{0.45} = 0.02. $$

#### Question 209.
A policy covers a gas furnace for one year. During that year, only one of three problems can occur:
- i) The igniter switch may need to be replaced at a cost of 60. There is a 0.10
probability of this.
- ii) The pilot light may need to be replaced at a cost of 200. There is a 0.05 probability of this.
- iii) The furnace may need to be replaced at a cost of 3000. There is a 0.01 probability of this.

Calculate the deductible that would produce an expected claim payment of 30.

**Giải đáp / SOLUTIONs.**

- Đặt $X,Y,Z$ lần lượt là BNN thể hiện chi phí thay thế cho bộ phận đánh lửa (`igniter`); đèn hoa tiêu (`pilot light`) và bộ đốt `furnace`. Theo giả thiết

$$ \begin{array}{rrcl} \text{Theo (i)} & P[X < 60] &=& 0.10 \\ \text{(ii)} & P[Y < 200] &=& 0.05 \\ \text{(iii)} & P[Z < 3000] &=& 0.01. \end{array} $$

- Một lò gas (`gas furnace`) sẽ có cả 3 bộ phận trên, do đó với mức khấu hao $d$ cho cả 3 bộ phận trên, thì số tiền phải chi trả bằng 30; tức là là

$$ (60 - d)P[X<60] + (200 - d)P[Y<200] + (3000 - d)P[Z < 3000] = 30 $$
Đặt
$$ \phi(d) = (60 - d)P[X<60] + (200 - d)P[Y<200] + (3000 - d)P[Z < 3000] $$
- Ta có các trường hợp có thể có như sau:

$\qquad \diamond$ Nếu $d \in [0, 60)$ thì 

$$ \phi(d) = 30 \Leftrightarrow (0.1)(60 - d) + (0.05)(200 - d) + (0.01)(3000 - d) = 30 \Leftrightarrow d = 100 \not \in (0, 60)$$ 

$\qquad \quad$ do đó ta loại trường hợp này.

$\qquad \diamond$ Nếu $d \in [60, 200)$ thì $P[X<60] = 0$, do đó

$$ \phi(d) = 30 \Leftrightarrow (0.05)(200 - d) + (0.01)(3000 - d) = 30 \Leftrightarrow d = \dfrac{500}{3} \in [60, 200)$$ 

$\qquad \quad$ do đó ta nhận $d \approx 166.67$

$\qquad \diamond$ Nếu $d \in [200, 3000)$ thì 

$$ \phi(d) = 30 \Leftrightarrow (0.01)(3000 - d) = 30 \Leftrightarrow d = 0 \not \in [200, 3000)$$ 

$\qquad \quad$ do đó ta loại trường hợp này.

- Trong 5 đáp án của câu hỏi này thì $C. 150 \leq d < 200$ là ước lượng chính xác nhất cho $d = 166.67$ ta đã giải ở đây.

#### Question 210.
- On a block of ten houses, $k$ are not insured. A tornado randomly damages three houses on the block.
- The probability that none of the damaged houses are insured is $1/120$.

Calculate the probability that at most one of the damaged houses is insured.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Tổng số ngôi nhà là $N=10$ và có $k \in 0,1,\ldots,N$ không được đảm bảo (đã mua BH) và $n=3$ ngôi nhà đã bị cơn bão tàn phá.
- Khi đó XS để không có ngôi nhà nào được đảm bảo bị thiệt hại
$$ \dfrac{\text{{select 0 in (10 - k) house which be insured}.{and 3 not-insured houses in k house} }}{\text{select 3 in total 10 houses}} = \dfrac{C_{10-k}^0 C_k^3}{C_{10}^3} $$
và theo BT thì xác suất này bằng 1/120, tức là 

$$ \dfrac{C_{10-k}^0 C_k^3}{C_{10}^3} = \dfrac{1}{120} \Rightarrow \dfrac{k(k-1)(k-2)}{720} = \dfrac{1}{120} \Rightarrow k = 3. $$
- Cuối cùng; BT yêu cầu tìm XS mà có nhiều nhất một ngôi nhà đã mua BH bị phá hủy, tức là

$$ \begin{array}{ccl} \displaystyle  \sum_{j=0}^1 P \left[ \text{total damaged house is insured} = j \right]  &=& \displaystyle \sum_{j=0}^1 \dfrac{C_{10-k}^j C_k^{3-j}}{C_{10}^3} \\ &=& \dfrac{C_{10-k}^0 C_k^3}{C_{10}^3} + \dfrac{C_{10-k}^1 C_k^2}{C_{10}^3} \\&=& \dfrac{C_{7}^0 C_3^3}{C_{10}^3} + \dfrac{C_{7}^1 C_3^2}{C_{10}^3} \\ &=& \dfrac{11}{60} \end{array} $$
