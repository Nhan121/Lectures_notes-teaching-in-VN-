#### Question 171.
The return on two investments, $X$ and $Y$, follows the joint probability density function 

$$ f(x,y) = \left\lbrace \begin{array}{ccl} 0.5, & 0 < |x| + |y| < 1 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate $\text{Var}(X).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đầu tiên, ta sẽ xác định rõ miền giới hạn trong hàm MDXS: 

$$ \lbrace (x,y): 0 < |x| + |y| < 1 \rbrace = A \cup B \cup C \cup D $$
trong đó
$$ \begin{array}{ll} A = \lbrace (x,y): x>0, y>0, x+y < 1 \rbrace, & \text{ do } |x| + |y| = x+y < 1, \quad \forall x,y > 0 \\ B = \lbrace (x,y): x>0, y< 0, x-y < 1 \rbrace, & \text{ do } |x| + |y| = x-y < 1, \quad \forall x > 0, y < 0 \\ C = \lbrace (x,y): x< 0, y>0, x-y>-1 \rbrace, & \text{ do } |x| + |y| = -x+y < 1, \quad \forall x < 0, y > 0 \\ D = \lbrace (x,y): x < 0, y < 0, x+y > -1 \rbrace, & \text{ do } |x| + |y| = -x-y < 1, \quad \forall x < 0, y < 0 \end{array} $$
- Do đó,

$$ \begin{array}{cll} \mathbb{E}X &=& \displaystyle \int_A xf(x)dx + \int_B xf(x)dx + \int_C xf(x)dx + \int_D xf(x)dx \\ \mathbb{E}X^2 &=& \displaystyle \int_A x^2f(x)dx + \int_B x^2f(x)dx + \int_C x^2f(x)dx + \int_D x^2f(x)dx\end{array}  $$
- Ta thấy rằng $xf(x)$ là hàm lẻ (vì $x$ là hàm lẻ và $f(x)$ là hằng số) và $A = -D, B = -C$ nên
$$ \displaystyle \int_A xf(x) dx = - \int_D xf(x)dx, \quad \text{ và } \int_B xf(x) dx = - \int_C xf(x)dx $$
do đó $\mathbb{E}X = 0$

- Hơn nữa, $x^2f(x)$ là hàm chẵn nên

$$ \begin{array}{l} \displaystyle \int_D x^2f(x)dx = \int_A x^2f(x)dx = \int_0^1 \int_0^{1-x} x^2(0.5) dy dx = \dfrac{1}{2} \int_0^1 x^2(1-x)dx = \dfrac{1}{24} \\ \displaystyle  \int_C x^2f(x)dx = \int_B x^2f(x)dx = \int_{-1}^0 \int_0^{1+x} x^2 (0.5)dy dx = \dfrac{1}{2} \int_{-1}^0 x^2(1 + x) dx = \dfrac{1}{24} \end{array}  $$
Do đó $ \mathbb{E}X^2 = \dfrac{1}{6} $ suy ra

$$ \text{Var}X = \mathbb{E}X^2 - [ \mathbb{E}X ]^2 = \dfrac{1}{6}. $$

#### Question 172.
- `(i)` A policyholder has probability 0.7 of having **no claims**, 0.2 of having exactly **one claim**,
and 0.1 of having **exactly two claims**. 
- `(ii)` Claim amounts are **uniformly distributed** on the interval [0, 60] and are independent. 
- `(iii)` The insurer covers 100% of each claim.

Calculate the probability that the total benefit paid to the policyholder is 48 or less.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số yêu cầu từ KH mua BH; theo `(i)`, ta có

$$ P[X = 0] = 0.7, \quad P[X = 1] = 0.2 \quad \text{ và } \quad P[X = 2] = 0.1 $$
- Đặt $Y$ là số tiền mà cty BH phải chi trả cho các yêu cầu từ KH, theo `(ii)` thì:

$\qquad \diamond (Y| X > 0) \sim \mathcal{U}([0, 60])$ hay 

$$P[Y < y | X = 1] = \displaystyle \int_0^y \dfrac{1}{60} dy = \dfrac{y}{60}, \quad \forall y \in [0, 60]$$
và
$$P[Y < y | X = 2] = \displaystyle \int_0^y\int_0^{y-s} \left( \dfrac{1}{60} \right)^2 dy ds = \dfrac{y^2}{7200}, \quad \forall y \in [0, 60]$$
$\qquad \diamond (Y | X = 0) = 0 $ thì XS cho số tiền chi trả của BH bằng 1; tức là

$$ P[Y < y | X = 0] = P[(Y|X=0) < y ] = P[0 < y] = 1, \quad \forall y > 0 $$

- Vì yêu cầu BT là tính XS để tổng tiền phải trả cho KH không vượt quá 48 đồng, tức là $P[Y \leq 48]$, ta có

$$ \begin{array}{ccl} P[Y \leq 48] &=& P[Y \leq 48 | X = 0] P[X = 0] + P[Y \leq 48 | X = 1] P[X = 1] + [Y \leq 48| X =2] P[X = 2] \\ &=& \quad \qquad \qquad (1).(0.7) \quad \quad + \quad \left( \dfrac{48}{60} \right) . (0.2) \quad \quad + \quad \left( \dfrac{48^2}{7200} \right) . (0.1) \\ & = & \qquad \qquad \qquad \qquad 0.892 \end{array} $$

#### Question 173.
In a given region, the number of tornadoes in a one-week period is modeled by a `Poisson
distribution` with mean 2. The numbers of tornadoes in different weeks are mutually
independent.

Calculate the probability that fewer than four tornadoes occur in a three-week period. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là số cơn lốc xoáy trong tuần thứ $i$, theo giả thiết ta có $X_i \overset{\text{i.i.d}}{\sim} \text{Pois}(2)$
- BT yêu cầu tính XS để tổng cơn lốc trong 3 tuần ít hơn 4, tức là $\displaystyle P\left[ \sum_{i=1}^3 X_i < 4 \right]$
- Đặt $X = \displaystyle \sum_{i=1}^3 X_i$; khi đó $X \sim \text{Pois}(6)$ ([xem lại lưu ý đầu tiên trong bài giải ở câu 164, link](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/Exam_P/330%20questions/Q111_2_220/q-161-2-170.ipynb) )
- Khi đó BT sẽ trở thành

$$ P[X < 4] = \displaystyle \sum_{k=0}^3 P[X = k] = \sum_{k=0}^3 e^{-6} \dfrac{6^k}{k!} =  e^{-6} \left( \dfrac{6^0}{0!} + \dfrac{6^1}{1!} + \dfrac{6^2}{2!} + \dfrac{6^3}{3!} \right) \approx 0.1512. $$

#### Question 174.
- An electronic system contains three cooling components that operate independently. 
- The probability of each component’s failure is 0.05. 
- The system will overheat if and only if at least two components fail.

Calculate the probability that the system will overheat.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_1, X_2, X_3$ tương ứng với 3 thành phần làm mát của hệ thống điện.
- Giả thiết thứ hai cho biết $X_1, X_2, X_3 \sim \mathcal{B}(1, 0.05)$, nghĩa là
$$ X_i = \left\lbrace \begin{array}{cl} 1, & \text{ neu thanh phan thu i bi loi voi xac suat } 0. 05 \\ 0, & \text{khong bi loi~} \end{array} \right. $$
- Giả thiết cuối cho biết hệ thống tản nhiệt này sẽ quá tải nếu có ít nhất 2 thành phần bị hỏng; tức là $\displaystyle \sum_{i=1}^3 X_i \geq 2$
- Đặt $X = \displaystyle \sum_{i=1}^3 X_i$ và kết hợp với $X_1, X_2, X_3 \sim \mathcal{B}(1, 0.05)$ ta được $X \sim \mathcal{B}(3, 0.05)$.
- BT yêu cầu tính XS để hệ thống tản nhiệt này bị hỏng, tức là

$$ P[X \geq 2] = P[X=2] + P[X =3] = C_3^2 (0.05)^2(1 - 0.05)^1 + C_3^3 (0.05)^3 = 0.00725. $$

#### Question 175.
- An insurance company’s annual profit is normally distributed with mean 100 and variance 400.
- Let $Z$ be normally distributed with mean 0 and variance 1 and let $F$ be the cumulative distribution function of $Z.$

Determine the probability that the company’s profit in a year is at most 60, given that the profit in the year is positive.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là lợi nhuận hàng năm của cty BH, và theo giả thiết đầu tiên thì $X \sim \mathcal{N}(\mu, \sigma^2)$ với $\mu=100$ và $\sigma^2 = 400.$
- BT yêu cầu tính XS để lợi nhuận cty không vượt quá $60$ biết rằng công ty có lãi (hay lợi nhuận là một số dương) tức là $P[X \leq 60 | X > 0]$. Ta có

$$ P[X < x] = P\left[ \mathcal{N}(0, 1) < \dfrac{x - 100}{20} \right] = \Phi\left( \dfrac{x - 100}{20} \right) $$

trong đó $\Phi(z) = P[Z < z]$ là hàm phân phối tích lũy của $Z=\mathcal{N}(0,1)$. Do đó,

$$ P[X \leq 60 | X > 0] = \dfrac{P[0 < X \leq 60]}{P[X > 0]} = \dfrac{P[X \leq 60] - P[X<0]}{1 -P[X < 0]} = \dfrac{\Phi(-2) - \Phi(-5)}{1 - \Phi(-5)} = \dfrac{\Phi(-2) - \Phi(-5)}{\Phi(5)} $$

#### Question 176.
In a group of health insurance policyholders, `20%` have high blood pressure and `30%` have high cholesterol. Of the policyholders with high blood pressure, `25%` have high cholesterol.

A policyholder is randomly selected from the group.

Calculate the probability that a policyholder has high blood pressure, given that the
policyholder has high cholesterol. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $B, C$ lần lượt là sự kiện chọn được nhóm người mua BH mà có vấn đề về `huyết áp cao (blood pressure)` và `high cholesterol`.
- Theo giả thiết (i) ta có

$$ P[B] = 0.2, \quad P[C] = 0.3, \quad \text{ và } \quad P[C | B] = 0.25 $$
- BT yêu cầu tính XS để chọn được người có vấn đề về huyết áp biết rằng có lượng cholesterol cao; tức là

$$ P[B | C] = \dfrac{(P[C | B]) (P[B])}{P[C]} = \dfrac{(0.25)(0.2)}{(0.3)} = \dfrac{1}{6}. $$

#### Question 177.
- In a group of 25 factory workers, 20 are low-risk and five are high-risk.
- Two of the 25 factory workers are randomly selected without replacement.

Calculate the probability that exactly one of the two selected factory workers is low-risk.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Giả thiết cho biết; mẫu được khảo sát $\Omega$ là một nhóm có 25 công nhân; tức $|\Omega| = 25$
- Trong đó có 2 nhóm, $H, L$ lần lượt ứng với nhóm rủi ro cao và rủi ro thấp. Theo đó
$$ |H| = 5, \quad \text{ va } \quad |L| = 5. $$
- BT yêu cầu tính XS để khi chọn ngẫu nhiên không hoàn lại 2 nhân công thì có đúng một người là nhóm rủi ro thấp, tức là 

$$ \dfrac{| \lbrace (x, y) : x\in H, y \in L \in \Omega \rbrace |}{| \lbrace (x, y) : x \neq y \forall x,y \in \Omega \rbrace |} = \dfrac{C_5^1 C_{20}^1}{C_{25}^2} = \dfrac{1}{3} $$

#### Question 178.
The proportion $X$ of yearly dental claims that exceed 200 is a random variable with probability density function 

$$ f(x) = \left\lbrace \begin{array}{cl} 60x^3(1-x)^2, & 0 < x < 1 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate $\text{Var}[X/(1 – X)]$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính $\text{Var}\varphi(X)$ với $\varphi$ là một ánh xạ đo được, ta chỉ việc xem $Z = \varphi(X)$ để triển khai

$$ \text{Var}\varphi(X) = \mathbb{E}(Z^2) - [\mathbb{E}Z]^2 $$
với lưu ý
$$ \mathbb{E}\varphi(X) = \displaystyle \int \varphi(x) f(x)dx $$
với $f$ là hàm MDXS của $X$.
- Lúc này, thay $\varphi(x) = \dfrac{x}{1-x} $, ta được
$$\mathbb{E}Z = \mathbb{E} \varphi(X) = \int \varphi(x) f(x)dx = \int_0^1 \dfrac{x}{1-x} \left( 60x^3(1-x)^2 \right)dx = 60\int_0^1 x^4(1-x)dx = 2$$
và
$$\mathbb{E}Z^2 = \mathbb{E} \varphi^2(X) = \int \varphi^2(x) f(x)dx = \int_0^1 \left( \dfrac{x}{1-x} \right)^2 \left( 60x^3(1-x)^2 \right)dx = 60\int_0^1 x^5dx = 10.$$
suy ra

$$ \text{Var}\left( \dfrac{X}{1-X} \right) = \mathbb{E}(Z^2) - [\mathbb{E}Z]^2 = 6.$$

#### Question 179.
This year, a medical insurance policyholder has probability 0.70 of having no emergency room visits, 0.85 of having no hospital stays, and 0.61 of having neither emergency room visits nor hospital stays

Calculate the probability that the policyholder has at least one emergency room visit and at least one hospital stay this year.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $E, H$ lần lượt là các BNN thể hiện số lần mà người bệnh `nằm phòng cấp cứu [emergency]` và `nằm viện [hospital stay]` trong năm nay. Theo đó

$\qquad \diamond$ XS người đó không vào phòng cấp cứu là $0.7$ tức là $P[E = 0] = 0.7$ suy ra $P[E \geq 1] = 1 - P[E=0] = 0.3.$

$\qquad \diamond$ XS không nằm viện bằng 0.85 tức là $P[H = 0] = 0.85$; tương tự $P[H \geq 1] = 0.15$

$\qquad \diamond$ XS không nằm viện cũng chẳng nằm phòng cấp cứu là 0.61; nghĩa là $P[H=0, E = 0] = 0.61$ cho nên

$$ 1 - P[H=0, E = 0] = P\left[ (H=0, E=0)^c \right] = P\left[ (H \geq 1) \cup (E \geq 1) \right] $$

- BT yêu cầu tính XS mà người đó ít nhất một lần nằm phòng cấp cứu và ít nhất một lần nhập viện năm nay; tức là $P[H \geq 1, E \geq 1]$. Ta có

$$ \begin{array}{ccl} P[H \geq 1, E \geq 1] &=& P[(H \geq 1)\cap(E \geq 1)] \\ &=& P[H \geq 1] + P[E \geq 1] - P[(H \geq 1) \cup (E \geq 1)] \\ &=& P[H \geq 1] + P[E \geq 1] - \left( 1 - P[H=0, E = 0] \right) \\ &=& 0.06. \end{array} $$

#### Question 180.
`(i)` An insurer offers a travelers insurance policy. Losses under the policy are **`uniformly distributed`** on the interval [0, 5].

`(ii)` The insurer reimburses a policyholder for a loss up to a maximum of 4.

Determine the **cumulative distribution function**, $F$, of the benefit that the insurer pays a policyholder who experiences exactly one loss under the policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là khoản bồi thường và mức thiệt hại của loại BH du lịch trong bài. Theo `(i)`

$$ Y \sim \mathcal{U}([0, 5]) $$

Với `(ii)` khoản tiền bồi thường $X$ sẽ được tính cho các tổn thất $Y$ cho đến khi tối đa bằng $4$, tức là

$$ P[X < x] = \left\lbrace \begin{array}{cl} P[Y < x], & \text{for } x < 4, \\ 1, & \text{otherwise}  \end{array} \right. $$

- Nhắc lại rằng, do $Y \sim \mathcal{U}([0, 5])$ nên $Y$ có hàm phân phối tích lũy xác định bởi
$$ F_Y(y) = P[Y < y] = \left\lbrace \begin{array}{cl} 0, & y < 0 \\ \dfrac{y}{5}, & y \in [0, 5] \\ 1, & y > 5  \end{array} \right. $$
- BT yêu cầu xác định hàm phân phối tích lũy của $X$
$$ F_X(x) = P[X < x] = \left\lbrace \begin{array}{cl} 0, & y < 0 \\ \dfrac{x}{5}, & y \in [0, 4) \\ 1, & y \geq 4  \end{array} \right. $$
