|-|-|-|-|-|

#### Question 142.
- An auto insurance company is implementing a new bonus system. In each month, if a policyholder does not have an accident, he or she will receive a cash-back bonus of $5$ from the insurer.
- Among the $1,000$ policyholders of the `auto insurance` company, $400$ are classified as **lowrisk** drivers and $600$ are classified as **high-risk** drivers.
- In each month, the probability of zero accidents for high-risk drivers is `0.80` and the probability of zero accidents for low-risk drivers is `0.90`.

Calculate the expected bonus payment from the insurer to the 1000 policyholders in one
year

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $H, L$ lần lượt là **khả năng không xảy ra tai nạn hằng tháng (`zero accidents`)** ứng với nhóm các mức độ về tai nạn (tương ứng với `high-risk` và `low-risk`) do tài xế gây ra.
- Theo giả thiết thì $P[H] = 0.8$ và $P[L] = 0,9$.
- Hơn nữa, nếu mỗi tháng không xảy ra tau nạn thì chủ xe phải trả cho BH một khoản phí 5 đồng. Đặt $X$ là BNN thể hiện chi phí mà BH sẽ nhận được trong một năm từ khách hàng khi không xảy ra tai nạn.
- $\Rightarrow$ Như vậy, nếu không xảy ra tai nạn trong một năm thì số tiền mà chủ hợp đồng phải trả cho BH là

$\qquad \diamond$ Với nhóm `Low-risk`; $\mathbb{E}[X | L] = 12 P[L] 5 = 12(0.8)5 = 48$ là kỳ vọng số tiền mà BH nhận được trong năm nếu biết KH là loại `L`.

$\qquad \diamond$ Với nhóm `High-risk`; $\mathbb{E}[X | H] = 12 P[H] 5 = 12(0.9)5 = 54$ là kỳ vọng số tiền nhận được nếu biết KH là loại `H`.

- Giả thiết còn lại cho biết trong tổng số 1000 hợp đồng thì có 400 là `Low-risk` và $600$ là `high-risk`; do đó kỳ vọng số tiền BH sẽ thu được từ chủ xe trong 1 năm chính là

$$ \mathbb{E}[400X_L + 600 X_H] = 400 \mathbb{E}[X | L] + 600 \mathbb{E}[X | H] = 50400. $$

#### Question 143.
The probability that a member of a certain class of homeowners with liability and
property coverage will file a liability claim is 0.04, and the probability that a member of
this class will file a property claim is 0.10. The probability that a member of this class
will file a liability claim but not a property claim is 0.01.

Calculate the probability that a randomly selected member of this class of homeowners
will not file a claim of either type.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $L, R$ lần lượt tương ứng với 2 nhóm sai phạm về `pháp lý (liability)` và `tài sản (property)`.
- BT yêu cầu ta tính XS để đơn kiện không thuộc loại nào trong 2 loại trên, tức là $P[L^c \cap R^c]$
- Giả thiết cho biết:
$$ P[L] = 0.04, \quad P[R] = 0.10, \quad \text{và} \quad P[L \cap R^c] = 0.01 $$ 
suy ra
$$ P[L \cap R] = P[L] - P[L \cap R^c] =0.03 $$
và
$$ P[R \cap L^c] = P[R] - P[L \cap R] =0.07 $$
- Do đó
$$ \begin{array}{ccl} P[L^c \cap R^c] &=& 1 - P[L \cup R] \\ &=& 1 - \left( P[L] + P[R] - P[L \cap R]  \right) \\ &=& 1 - \left( 0.04 + 0.10 - 0.03 \right) \\ &=& 0.89 \end{array} $$

#### Question 144.
A client spends $X$ minutes in an insurance agent’s waiting room and $Y$ minutes meeting with the agent. The joint density function of $X$ and $Y$ can be modeled by

$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{800} e^{-\frac{x}{40}} e^{-\frac{y}{20}}, & x >0, y > 0 \\ 0, & \text{otherwise} \end{array} \right. $$
Determine the expression of the probability that a client spends less than 60 minutes at the agent’s office. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo giả thiết, $X$ là thời gian mà KH chờ ở văn phòng và $Y$ là thời gian tiếp khách của phía BH.
- BT yêu cầu tính XS mà KH bỏ ra ít hơn 60 phút tại văn phòng, tức là tổng thời gian về chờ và tiếp khách là 60 phút; vì vậy miền giá trị của cặp BNN $(X,Y)$ chính là

$$ \mathcal{A} = \lbrace (x,y) \in \mathbb{R}^2: x > 0, y > 0, x+y < 60 \rbrace $$
nói cách khác, 
$$ \mathcal{A} = \lbrace (x,y) \in \mathbb{R}^2: 0 < y < 60 - x; 0 < x < 60 \rbrace $$
Do đó công thức BT yêu cầu chính là
$$ \displaystyle \iint_{\mathcal{A}} f(x,y)dx dy = \dfrac{1}{800} \int_0^{60} \int_0^{60-x} e^{-\frac{x}{40}} e^{-\frac{y}{20}} dy dx $$

#### Question 145.
New dental and medical plan options will be offered to state employees next year. An
actuary uses the following density function to model the joint distribution of the
proportion $X$ of state employees who will choose Dental Option 1 and the proportion $Y$
who will choose Medical Option 1 under the new plan options: 

$$ f(x,y) = \left\lbrace \begin{array}{cl} 0.5, & x\in [0, 0.5]; y \in [0, 0.5]\\ 1.25, & x\in [0, 0.5]; y \in [0.5, 1] \\ 1.5, & x\in [0.5, 1]; y \in [0, 0.5] \\ 0.75, & x\in [0.5, 1]; y \in [0.5, 1]  \end{array} \right. $$

Calculate $\text{Var} (Y | X = 0.75).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu tính phương sai có điều kiện $\text{Var} (Y | X = 0.75)$ do đó ta cần xác định được hàm MDXS có điều kiện $f(y|x=0.75)$.
- Ta có
$$ f(0.75, y) = f(x = 0.75, y) = \left\lbrace \begin{array}{ccl} 1.5, & y \in (0, 0.5), \\ 0.75, & y \in (0.5, 1) \end{array} \right. $$
và
$$ \displaystyle f_X(0.75) = \int_0^1 f(0.75, y) dy = \int_0^{0.5} 1.5 dy + \int_{0.5}^1 0.75dy = 1.125 $$
do đó
$$ f(y|x=0.75) = \left\lbrace \begin{array}{ccl} \dfrac{1.5}{1.125} = \dfrac{4}{3}, & y \in (0, 0.5), \\ \dfrac{0.75}{1.125} = \dfrac{2}{3}, & y \in (0.5, 1) \end{array} \right. $$
- Như vậy

$$ \begin{array}{ccl} \mathbb{E}[Y|X=0.75] &=& \displaystyle \int_0^1 y f(y| x=0.75)dy = \dfrac{4}{3}\int_0^{0.5}ydy + \dfrac{2}{3}\int_{0.5}^1 ydy = \dfrac{5}{12} \\ \mathbb{E}[Y^2|X=0.75] &=& \displaystyle \int_0^1 y^2 f(y| x=0.75)dy = \dfrac{4}{3}\int_0^{0.5}y^2dy + \dfrac{2}{3}\int_{0.5}^1 y^2dy = \dfrac{1}{4} \end{array} $$
do đó

$$ \text{Var} (Y | X = 0.75) = \mathbb{E}[Y^2|X=0.75] - \left( \mathbb{E}[Y|X=0.75] \right)^2 = \dfrac{11}{144} \approx 0.076 $$

#### Question 146.
A survey of 100 TV viewers revealed that over the last year:
- `i)` 34 watched CBS.
- `ii)` 15 watched NBC.
- `iii)` 10 watched ABC.
- `iv)` 7 watched CBS and NBC.
- `v)` 6 watched CBS and ABC.
- `vi)` 5 watched NBC and ABC.
- `vii)` 4 watched CBS, NBC, and ABC.
- `viii)` 18 watched HGTV, and of these, none watched CBS, NBC, or ABC.

Calculate how many of the 100 TV viewers did not watch any of the four channels `(CBS, NBC, ABC or HGTV)`. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C,N,A,H$ lần lượt tương ứng với các kênh `"CBS", "NBC", "ABC"` và `"HGTV"`
- Theo `(i), (ii), (iii)`, ta có
$$ |C| = 34, \quad |N| = 15, \quad \text{và} \quad |A| = 10 $$
- Theo `(iv), (v), (vi)`, 
$$ |C \cap N| =7, \quad |C \cap A| =6 \text{và} |N \cap A| = 5 $$
- Theo `(vii)` và `(viii)`,
$$ |C \cap N \cap A| = 4, \quad \text{và} \quad |H \cap (C \cup N \cup A)^c | = 18 $$
- BT yêu cầu ta tính
$$ |C^c \cap A^c \cap N^c \cap H^c|, \quad \text{ hay } \quad |(C \cup A \cup N \cup H)^c| $$
- Liên hệ. Ta có
$$ \begin{array}{ccl} |C \cup N \cup A| &=& |C| + |N| + |A| - \left[ |C \cap N| + |C \cap A| + |N \cap A| \right] + |C \cap N \cap A| \\ &=& 34 + 15 + 10 - \left[ 7 + 6 + 5 \right] - 4 = 45 \\ |C \cup N \cup A \cup H| &=& |H \cap (C \cup N \cup A)^c | + |C \cup N \cup A| \\ &=& 18 + 45 = 63 \end{array} $$
do đó
$$ |C^c \cap A^c \cap N^c \cap H^c| = |\Omega| -  |C \cup N \cup A \cup H| = 100 - 63 = 37. $$

#### Question 147.
The amount of a claim that a car insurance company pays out follows an exponential
distribution. By imposing a deductible of $d$, the insurance company reduces the expected
claim payment by `10%`.

Calculate the percentage reduction on the variance of the claim payment.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là số các yêu cầu trước và sau khi thực hiện khấu trừ.
- Theo giả thiết, $X$ có phân phối mũ với trung bình $\lambda > 0$ nào đó, khi đó 

$$\mathbb{E}X = \lambda, \quad \mathbb{E}X^2 = \text{Var}X + [\mathbb{E}X ]^2 = 2\lambda^2.$$
- Theo giả thiết còn lại sau khi khấu trừ một khoản phí $d > 0$ dẫn đến $Y=X-d$ thì kỳ vọng chi phí được giảm bớt `10%` so với ban đầu, do đó
$$ \mathbb{E}Y = 0.9 [\mathbb{E}X] = 0.9 \lambda, \quad \text{ và } \quad \mathbb{E}Y^2 = 0.9 [ \mathbb{E}X^2 ] = 1.8 \lambda^2 $$
suy ra 
$$ \text{Var}Y = \mathbb{E}Y^2 - [ \mathbb{E}Y ]^2 = 0.99 \lambda^2 $$
- Dẫn đến
$$ \text{Var}X - \text{Var}Y = 0.01 \lambda $$

vì vậy, phần trăm chi phí của phương sai sau khi khấu trừ là $0.01$ hay $1\%$

#### Question 148.
- The number of hurricanes that will hit a certain house in the next ten years is Poisson
distributed with mean 4.
- Each hurricane results in a loss that is exponentially distributed with mean 1000. Losses
are mutually independent and independent of the number of hurricanes.

Calculate the variance of the total loss due to hurricanes hitting this house in the next ten years.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là BNN thể hiện số cơn bão hàng năm. Theo đó, nó tuân theo phân phối Poisson với trung bình $4$; tức là

$$ \mathbb{E}N = 4 \text{ và } \text{Var}N = 4. $$
- Đặt $X_i$ là mức thiệt hại tương ứng với cơn bão thứ $i$. Theo đó, $X_i$ có cùng phân phối mũ với trung bình $1000$; các $X_i, X_j$ độc lập đôi một với nhau và độc lập với $N$; ta có
$$ \mathbb{E}X_i = 1000 \quad \text{ và } \quad \text{Var}X_i = 1000^2 = 1000000, \qquad \forall i $$
- BT yêu cầu tính phương sai của $X = \displaystyle \sum_{k=1}^N X_k$: tổng số thiệt hại gây ra bởi $N$ cơn bão trong năm. Ta có

$$ \begin{array}{ccll} \text{Var}X &=& \text{Var}[\mathbb{E}(X|N)] + \mathbb{E}[\text{Var}(X|N)] \\ &=& \text{Var}\left[ \mathbb{E}\left( \left. X_1 + \ldots + X_N \right\vert N \right) \right] + \mathbb{E}\left[ \text{Var} \left( \left. X_1 + \ldots + X_N \right\vert N \right) \right] \\ &=& \text{Var}\left[ \mathbb{E}\left( X_1 + \ldots + X_N \right) \right] + \mathbb{E}\left[ \text{Var} \left( X_1 + \ldots + X_N \right) \right]  & \text{do } X_1,\ldots X_N \text{ ĐL voi } N \\ &=& \text{Var}[N \mathbb{E}X_1] + \mathbb{E} [ N \text{Var}X_1 ] & \text{ do } X_1, \ldots, X_N \overset{\text{i.i.d}}{\sim} X_1 = \mathcal{E}(1000) \\ &=& \text{Var}[1000N] + \mathbb{E}[1000000N] & \text{do } \mathbb{E}X_1 = 1000 \text{ va Var}X_1 = 1000000 \\ &=& 1000^2 \text{Var}N + 1000000\mathbb{E}N & \text{do } \mathbb{E}(cZ) = c\mathbb{E}Z \text{ va Var}(cZ) = c^2\text{Var}Z \\ &=& (1000000)(4) + (1000000)(4) \\ &=& 8000000 \end{array} $$

#### Question 149.
- A motorist makes three driving errors, each independently resulting in an accident with
probability 0.25.
- Each accident results in a loss that is exponentially distributed with mean 0.80. Losses
are mutually independent and independent of the number of accidents.
The motorist’s insurer reimburses 70% of each loss due to an accident.

Calculate the variance of the total unreimbursed loss the motorist experiences due to
accidents resulting from these driving errors.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N$ là BNN thể hiện số vụ tai nạn. Theo đó, nó tuân theo phân phối nhị thức $\mathcal{B}(n=3, \theta = 0.25)$; tức là

$$ \mathbb{E}N = n\theta = 0.75 \text{ và } \text{Var}N = np(1-p) = 0.5625. $$
- Đặt $X_i$ là khoản bồi thường tương ứng với vụ tai nạn thứ $i$ được báo cáo. Theo đó, $X_i$ có cùng phân phối mũ với $30\%$ của trung bình $0.8$; các $X_i, X_j$ độc lập đôi một với nhau và độc lập với $N$; ta có

$$ \mathbb{E}X_i = \mathbb{E}[(0.3) \mathcal{E}(0.8)] = (0.3)(0.8) = 2.4$$ và 

$$ \text{Var}X_i = \text{Var}[(0.3) \mathcal{E}(0.8)] = (0.3)^2 \text{Var}[\mathcal{E}(0.8)] = (0.09)(0.8)^2 = 0.0576, \qquad \forall i $$
- BT yêu cầu tính phương sai của $X = \displaystyle \sum_{k=1}^N X_k$: tổng số thiệt hại gây ra bởi $N$ cơn bão trong năm. Ta có

$$ \begin{array}{ccll} \text{Var}X &=& \text{Var}[\mathbb{E}(X|N)] + \mathbb{E}[\text{Var}(X|N)] \\ &=& \text{Var}\left[ \mathbb{E}\left( \left. X_1 + \ldots + X_N \right\vert N \right) \right] + \mathbb{E}\left[ \text{Var} \left( \left. X_1 + \ldots + X_N \right\vert N \right) \right] \\ &=& \text{Var}\left[ \mathbb{E}\left( X_1 + \ldots + X_N \right) \right] + \mathbb{E}\left[ \text{Var} \left( X_1 + \ldots + X_N \right) \right]  & \text{do } X_1,\ldots X_N \text{ ĐL voi } N \\ &=& \text{Var}[N \mathbb{E}X_1] + \mathbb{E} [ N \text{Var}X_1 ] & \text{ do } X_1, \ldots, X_N \overset{\text{i.i.d}}{\sim} X_1 = (0.3)\mathcal{E}(0.8) \\ &=& \text{Var}[0.24N] + \mathbb{E}[0.576N] & \text{do } \mathbb{E}X_1 = 0.24 \text{ va Var}X_1 = 0.0576 \\ &=& (0.24)^2 \text{Var}N + 0.0576\mathbb{E}N & \text{do } \mathbb{E}(cZ) = c\mathbb{E}Z \text{ va Var}(cZ) = c^2\text{Var}Z \\ &=& (0.0576)(0.5625) + (0.0576)(0.75) \\ &=& 0.0756 \end{array} $$

#### Question 150.
- An automobile insurance company issues a one-year policy with a deductible of 500.
- The probability is 0.8 that the insured automobile has no accident and 0.0 that the
automobile has more than one accident. 
- If there is an accident, the loss before application of the deductible is exponentially distributed with mean 3000.

Calculate the 95th percentile of the insurance company payout on this policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện mức thiệt hại trước khi thực hiện khấu hao $500$ (với $X \geq 1$ vì ta chỉ bồi thường khi có tai nạn xảy ra) và theo giả thiết `(iii)` thì nó tuân theo phân phối mũ với trung bình $3000$; do đó $X$ có hàm phân phối tích lũy

$$ F_X(x) = 1 - e^{-\frac{x}{3000}} $$
- Đặt $Y$ là BNN thể hiện số vụ tai nạn và giả thiết `(ii)` cho biết 

$$ \left\lbrace \begin{array}{ccll} P[Y = 0] &=& 0.8, & \text{ tuc la khong co tai nan nao xay ra doi voi cac oto da mua BH} \\ P[Y > 1] &=& 0.0, & \text{ tuc la co nhieu hon 1 tai nan.} \\ \end{array} \right. $$
- Như vâyj; phân vị mức $95\% > 0.8$ là thuộc vào vùng xảy ra ít nhất một tai nạn; do đó trước tiên ta phải quy đổi nó về mức phân vị của $X$: mức bồi thường do thiệt hại trong vụ tai nạn đó gây ra, tức là

$$ \dfrac{0.95 - 0.80}{1 - 0.80} = 0.75 $$

và do đó mức phân vị $95\%$ cho khoản thanh toán sẽ tương ứng với mức $75\%$ trong mức bồi thường (trước khi khấu trừ với $500$). Tức là tìm $x_{\alpha}$ sao cho
$$ 0.75 = P[X < x_{\alpha}] \Leftrightarrow 1 - \exp \left( -\frac{x_{\alpha}}{3000} \right) = 0.75 \Leftrightarrow x_{\alpha} = 4159$$

Sau khi tìm được $x_{\alpha} = 4159$ ta phải khấu trừ thêm khoản $500$ để được chi phí sau cùng mà BH sẽ thanh toán. Do đó số tiền phải tính là 

$$x_{\alpha} - 500 = 3659.$$
