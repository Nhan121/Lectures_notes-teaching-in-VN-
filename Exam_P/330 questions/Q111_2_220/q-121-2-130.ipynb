#### Question 121.
Let $X$ represent the age of an insured automobile involved in an accident. Let $Y$ represent the length of time the owner has insured the automobile at the time of the accident. $X$ and $Y$ have joint probability density function 

$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{10 - xy^2}{64}, & x\in [2, 10], y\in [0,1] \\ 0, & \text{otherwise} \end{array} \right. $$
Calculate the expected age of an insured automobile involved in an accident. 

**Giải đáp / SOLUTIONs.**
- Từ hàm MDXS đồng thời $f(x,y)$, ta tìm được hàm MDXS lề $f_X$ như sau

$$ \displaystyle f_X(x) = \int f(x,y)dy = \int_0^1 \dfrac{10 - xy^2}{64} dy = \dfrac{1}{64} \left. \left[ 10y - \dfrac{x y^3}{3} \right] \right\vert_{y=0}^{y=1} = \dfrac{1}{64} \left[ 10 - \dfrac{x}{3} \right], \quad \forall x \in [2, 10] $$
- Suy ra kỳ vọng

$$ \mathbb{E}X = \displaystyle \int xf(x)dx = \int_2^{10} \dfrac{1}{64}\left[ 10x - \dfrac{x^2}{3} \right] dx \approx 5.78 $$

#### Question 122.
- A device contains two circuits. The second circuit is a backup for the first, so the second is used only when the first has failed. The device fails when and only when the second circuit fails.
- Let $X$ and $Y$ be the times at which the first and second circuits fail, respectively. $X$ and $Y$ have joint probability density function 

$$ f(x,y) = \left\lbrace \begin{array}{cl} 6e^{-x}e^{-2y}, & 0 < x< y < \infty \\ 0, & \text{otherwise} \end{array} \right. $$
Calculate the expected time at which the device fails. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo giả thiết, thiết bị chỉ hoạt động được nếu các chu trình thành phần $X, Y$ còn hoạt động và sẽ hỏng nếu $Y$ hỏng; hơn nữa $Y$ chỉ được dùng để thay thế cho $X$ khi $X$ hỏng.
- BT yêu cầu tính kỳ vọng về thời gian khi thiết bị hỏng, tức là $\mathbb{E}Y$.
- Do đó trước hết ta cần tìm được hàm MDXS cho $Y$, ta có

$$ \displaystyle f_Y(y) = \int f(x,y)dx = \int_0^{y} 6e^{-x} e^{-2y}dx = 6\left( e^{-2y} - e^{-3y} \right), \quad \forall y \in (0, +\infty) $$
suy ra
$$ \displaystyle \mathbb{E}Y = \int yf_Y(y)dy = \int_0^{\infty} 6y\left( e^{-2y} - e^{-3y} \right) dy = \dfrac{5}{6}. $$

#### Question 123.
You are given the following information about N, the annual number of claims for a randomly selected insured: 
$$ P[N = 0] = \dfrac{1}{2}, \quad P[N = 1] = \dfrac{1}{3}, \quad P[N > 1] = \dfrac{1}{6} $$
Let $S$ denote the total annual claim amount for an insured. 
- When $N = 1$, $S$ is `exponentially distributed` with mean 5. 
- When $N > 1$, $S$ is `exponentially distributed` with mean 8.

Calculate $P(4 < S < 8)$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đầu tiên, $N$: số các yêu cầu bồi thường hàng năm gồm $(N=0)$; $(N=1)$ và $(N > 1)$ đã tạo thành một hệ đầy đủ các biến cố; vì
$$ P[N=0] + P[N=1] + P[N>1] = 1. $$
- Hơn nữa, do $S$ là tổng số tiền từ các yêu cầu bồi thường hàng năm nên $S$ chỉ được tính khi $N > 0$ và 

$$ (S|N=1) \sim \mathcal{E}, \text{ và } (S| N > 1) \sim \mathcal{E}(8) $$
suy ra

$$ \begin{array}{ccl} P[4 < S < 8] &=& P[S \in (4,8) | N=1] \cdot P[N=1] + P[S \in (4, 8)|N > 1] \cdot P[N > 1] \\ &=& \displaystyle \left( P[\mathcal{E}(5) < 8] - P[\mathcal{E}(5) < 4] \right) \cdot P[N=1] + \left( P[\mathcal{E}(8) < 8] - P[\mathcal{E}(8) < 4] \right) \cdot P[N>1] \\ &=& \dfrac{1}{3} \cdot \left[ \left( 1 - e^{-8/5} \right) - \left( 1 - e^{-4/5} \right) \right] + \dfrac{1}{6} \cdot \left[ \left( 1 - e^{-8/8} \right) - \left( 1 - e^{-4/8} \right) \right] \\ &=& 0.122 \end{array} $$

#### Question 124.
The joint probability density for $X$ and $Y$ is
$$ f(x,y) = \left\lbrace \begin{array}{cl} 2e^{-(x+2y)}, & x > 0, y > 0 \\ 0, & \text{otherwise} \end{array} \right. $$
Calculate the variance of $Y$ given that $X > 3$ and $Y > 3$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài yêu cầu tính $\text{Var}\left[ Y|(Y>3) \right]$; trước tiên, ta cần xác định hàm MDXS có điều kiện
$$ f(y |y > 3) = \dfrac{f_Y(y)}{P[Y > 3]} $$
- Ta có
$$ \begin{array}{rcl} f_Y(y) &=& \displaystyle \int_0^{\infty} 2e^{-(x+2y)}dx = 2e^{-2y},\\ P[Y>3] &=& \displaystyle \int_3^{\infty} f_Y(y) dy = e^{-6} \end{array} $$
do đó $$ f(y |y > 3) = \left\lbrace \begin{array}{cl} 2e^{6-2y}, & y > 3 \\ 0, & \text{otherwise} \end{array} \right. $$
- Như vậy

$$ \begin{array}{rcl} \mathbb{E}[Y|Y>3] &=& \displaystyle \int y f_Y(y)dy = \int_3^{\infty} 2ye^{6-2y} dy = 3.5 \\ \mathbb{E}[Y^2| Y>3] &=& \displaystyle \int y^2 f_Y(y)dy = \int_3^{\infty} 2y^2e^{6-2y} dy = 12.5 \end{array} $$
do đó
$$ \text{Var}[Y| Y>3] = \mathbb{E}[Y^2| Y>3] - \left[ \mathbb{E}[Y| Y>3] \right]^2 = 0.25 $$

#### Question 125.
The distribution of $Y$, given $X$, is `uniform` on the interval $[0, X]$. The marginal density of $X$ is

$$ f_X(x) = \left\lbrace \begin{array}{cl} 2x, & 0 < x< 1 \\ 0, & \text{otherwise} \end{array} \right. $$
Determine the conditional density of $X$, given $Y = y$ where positive.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính hàm MDXS có điều kiện $f(x|y)$ khi $y > 0$, ta phải xác định hàm MDXS đồng thời của cặp $(X, Y)$.
- Giả thiết đầu tiên cho biết $f(y|x)$ là phân phối đều trên $[0, x]$ do đó

$$ f(x,y) = f(y|x) f_X(x) = \left\lbrace \begin{array}{cl} x, & 0 <y< x< 1 \\ 0, & \text{otherwise} \end{array} \right. $$
- Ta có
$$ f_Y(y) = \int f(x,y)dx = \int_y^1 2dx = 2(1 - y), \quad \forall y \in (0, 1). $$
suy ra
$$ f(x|y) = \dfrac{f(x,y)}{f_Y(y)} = \dfrac{1}{1-y}, \quad \forall y \in (0, 1). $$

#### Question 126.
Under an insurance policy, a maximum of five claims may be filed per year by a policyholder. Let $p(n)$ be the probability that a policyholder files $n$ claims during a given year, where $n = 0,1,2,3,4,5$. An actuary makes the following observations:
- `(i)` $p(n) \geq p(n+1)$ for $n = 0, 1, 2, 3, 4,$.
- `(ii)` The difference between $p(n)$ and $p(n+ 1)$ is the same for $n = 0,1,2,3,4$.
- `(iii)` Exactly $40\%$ of policyholders file fewer than two claims during a given year.

Calculate the probability that a random policyholder will file more than three claims during a given year. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $p_n = p(n)$ với $n = 0,1,\ldots,5$. Điều kiện đầu tiên để $p_0,p_1,\ldots,p_5$ phải tạo thành một hệ các trọng số xác suất đầy đủ, tức 
$$ \sum_{k=0}^5 p_k = 1 $$
- 2 giả thiết (i) và (ii) cho biết tồn tại một hằng số dương $c > 0$ sao cho
$$ p_{k} - p_{k+1} = c, \quad \forall k = 0,1,2,3,4 $$
suy ra
$$ \begin{array}{ccl} p_1 &=& p_0 - c \\ p_2 &=& p_1 - (p_1 - p_2) = (p_0 - c) - c = p_0 - 2c \\ \ldots \\ p_5 &=& p_0 - 5c \end{array} $$
liên hệ với ràng buộc $\displaystyle \sum_{k=0}^5 p_k = 1 $, ta được
$$ \displaystyle \sum_{k=0}^5 \left[ p_0 - kc \right] = 1 \Leftrightarrow 6p_0 - 15c = 1 $$
- Giả thiết (iii) cho biết 2 yêu cầu đầu tiên chiếm 40% tổng số; tức là
$$ p_0 + p_1 = 0.4  \Leftrightarrow 2p_0 - c = 0.4 $$
- Giải hệ phương trình $ \left\lbrace \begin{array}{ccl} 6p_0 - 15c &=& 1 \\ 2p_0 - c &=& 0.4 \end{array} \right. $ ta được $ \left\lbrace \begin{array}{ccl} p_0 &=& \dfrac{5}{24} \\ c &=& \dfrac{1}{60} \end{array} \right. $
- BT yêu cầu tính XS có nhiều hơn 3 yêu cầu trong năm; tức là

$$ p_4 + p_5 = (p_0 - 4c) + (p_0 - 5c) = 2p_0 - 9c = 0.267 $$

#### Question 127.
- The amounts of automobile losses reported to an insurance company are mutually
independent, and each loss is uniformly distributed between 0 and 20,000. 
- The company covers each such loss subject to a deductible of 5,000.

Calculate the probability that the total payout on 200 reported losses is between
1,000,000 and 1,200,000. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là số phương tiện trong báo cáo thứ $i$ bị hỏng được báo về cty BH. Theo giả thiết đầu tiên thì $X_i \overset{\text{i.i.d}}{\sim} X = \mathcal{U}([0, 20000])$ 
- Đặt $Y_i$ là số tiền mà BH phải chi trả cho báo cáo thứ $i$, với khấu hao $5000$, tức là $Y_i = \max(X_i - 5000, 0)$ do đó
$$ \begin{array}{ccl} \mathbb{E}Y_i &=& \displaystyle \int_0^{20000} \dfrac{\max(x-5000, 0)}{20000}dx = \int_{5000}^{20000} \dfrac{x-5000}{20000} dx = 5625 \\ \mathbb{E}Y_i^2 &=& \displaystyle \int_0^{20000} \dfrac{(\max(x-5000, 0))^2}{20000}dx = \int_{5000}^{20000} \dfrac{(x-5000)^2}{20000}dx = 5625000 \\ \text{Var}Y_i &=& \mathbb{E}Y_i^2 - [\mathbb{E}Y_i]^2 = 24609375. \end{array} $$
- BT yêu cầu tính XS để trong 200 bản báo cáo thì chi phí phải trả dao động từ 1000000 đến 1200000.
- Đặt $S = \sum_{k=1}^{200} Y_i$; áp dụng Định lý giới hạn trung tâm, ta có
$$ S \sim \mathcal{N}(200 \mathbb{E}Y_1, 200 \text{Var}Y_1) = \mathcal{N}( 1125000, , 4921875000) $$
và ta phải tính $P(1000000 < S < 1200000)$. Ta có

$$ \begin{array}{ccl} P(1000000 < S < 1200000) &=& P \left[ \dfrac{1000000 - 1125000}{\sqrt{4921875000}} < \mathcal{N}(0, 1) < \dfrac{1200000 - 1125000}{\sqrt{4921875000}} \right] \\ &=& \Phi(1.069)- \Phi(-1.7817) \\ &=& 0.8575 – (1 – 0.9626)\\ &=& 0.8201 \end{array} $$

#### Question 128.
An insurance agent offers his clients auto insurance, homeowners insurance and renters
insurance. The purchase of homeowners insurance and the purchase of renters insurance
are mutually exclusive. The profile of the agent’s clients is as follows:
- i) 17% of the clients have none of these three products.
- ii) 64% of the clients have auto insurance.
- iii) Twice as many of the clients have homeowners insurance as have renters insurance.
- iv) 35% of the clients have two of these three products.
- v) 11% of the clients have homeowners insurance, but not auto insurance.

Calculate the percentage of the agent’s clients that have both auto and renters insurance.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt nhóm các KH mua BH xe cộ (`auto`), chủ nhà (`homeowners`) và người thuê nhà (`renters`) lần lượt là $A, H, R$
- Theo (i) ta có $P[A^c \cap H^c \cap R^c] = 0.17$ hay $P[A \cup H \cup R] = 0.83$
- Theo (ii) $P[A] = 0.64$ và (iii) cho biết $P[H] = 2P[R]$
Đến đây, ta được hệ giả định sau

$$ \left\lbrace \begin{array}{lcl} P[A^c] &=& 0.36 \\ P[H] - 2P[R] &=& 0 \\ P[A \cup H \cup R] &=& 0.83 \\ P[H \cap R] &=& 0 \end{array} \right. $$
- Theo (iv) thì
$$ P[A\cap H] + P[A\cap R] + P[ H \cap R] = 0.35 $$
và (v) cho biết
$$ P[H \cap A^c] = 0.11 $$
- Từ các giả thiết trên, suy ra
$$ \begin{array}{ccl} P[R \cap A^c] &=& P[A^c] - P[H \cap A^c] - P[A^c \cap R^c \cap H^c] = 0.08 \\ P[H \cap A] &=& P[H] - P[H \cap A^c] = P[H] - 0.11\\ P[R \cap A] &=& P[R] - P[R \cap A^c] = P[R] - 0.08 \end{array} $$

kết hợp $P[H \cap R] = 0$ thì phương trình $$ P[A\cap H] + P[A\cap R] + P[ H \cap R] = 0.35 $$ tương đương với
$$ P[H]-0.11 + P[R]-0.08 = 0.35 \Leftrightarrow P[H] + P[R] = 0.54  $$

Lại kết hợp với $P[H] = 2P[R]$ suy ra $P[H] = 0.36$ và $P[R] = 0.18$ suy ra

$$ P[R \cap A] = P[R] - 0.08 = 0.1 $$

và đây chính là yêu cầu của BT.

#### Question 129.
- `(i)` The `cumulative distribution function` for `health care costs experienced by a policyholder` is modeled by the function 

$$ F(x) = \left\lbrace \begin{array}{cl} 1 - e^{-(x/100)}, & x > 0 \\ 0, & \text{otherwise} \end{array} \right. $$

- `(ii)` The policy has a `deductible of 20`. An `insurer reimburses` the policyholder for `100%` of health care costs between 20 and 120. Health care costs above 120 are reimbursed at `50%`.

- `(iii)` Let $G$ be the `cumulative distribution function` of `reimbursements` **given that the reimbursement is positive**.

Calculate $G(115)$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là số tiền hoàn trả (`reimbursement`) thì theo định nghĩa của $G$ và các giả thiết `(iii)` và `(ii)` về chính sách, ta có 

$$ G(y) = P[Y < y| X > 20].$$
và do $y = 115 < 120$ nên theo `(ii)` $$(Y < 115) = \left( X < 120 + \dfrac{115 - 100}{0.5} \right) = (X < 150)$$
do đó
$$ G(115) = P[X < 150 | X > 20] = \dfrac{P[(X < 150) \cap (X > 20)]}{P[X > 20]} $$
- Bấy giờ, với `(i)` và nhớ rằng $F(x) = P(X < x)$, ta được

$$ \begin{array}{rcl} P[(X < 150) \cap (X > 20)] &=& P[20<X<150] = F(150) - F(20) = e^{-0.2} - e^{-1.5} \\ P[X > 20] &=& 1 - P[X \leq 20] = 1 - F(20) = e^{-0.2} \end{array} $$
do đó
$$ G(115) = P[X < 150 | X > 20] = \dfrac{e^{-0.2} - e^{-1.5}}{e^{-0.2}} \approx 0.727. $$

#### Question 130.
The value of a piece of factory equipment after three years of use is $100(0.5)^X$ where $X$ is a random variable having moment generating function 

$$ M_X(t) = \dfrac{1}{1 - 2t}, \quad t < \dfrac{1}{2}. $$
Calculate the expected value of this piece of equipment after three years of use.

**Giải đáp / SOLUTIONs.**

Bằng định nghĩa hàm sinh moment, ta có
$$ M_X(t) = \mathbb{E} e^{Xt} $$
nên kỳ vọng trị giá của thiết bị sau 3 năm sử dụng sẽ được tính bởi

$$ \begin{array}{ccl} \mathbb{E} \left[ 100(0.5)^X \right] &=& 100 \mathbb{E} \left[ \exp[ \log((0.5)^X) \right] \\ &=& 100 \mathbb{E} \left[ e^{ X\log(0.5)} \right] \\ &=& 100 M_X( \log(0.5) ) \\ &=& \dfrac{100}{1 - 2\log(0.5)} \\ &=& 41.9 \end{array} $$
