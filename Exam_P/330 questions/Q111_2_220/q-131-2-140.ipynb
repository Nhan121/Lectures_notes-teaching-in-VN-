#### Question 131.
Let $N_1, N_2$ represent the numbers of claims submitted to a life insurance company in `April` and `May`, respectively. The joint probability function of $N_1, N_2$ is

$$ p(n_1, n_2) = \left\lbrace \begin{array}{ll} \dfrac{3}{4} \left( \dfrac{1}{4} \right)^{n_1 - 1} e^{-n_1} \left( 1 - e^{-n_1} \right)^{n_2 - 1}, & n_1 = 1,2,\ldots; \text{ and } n_2 = 1,2,\ldots \\0, & \text{otherwise} \end{array} \right. $$

Calculate the **expected number of claims** that will be submitted to the company in `May`, **given that exactly 2 claims** were submitted in `April`.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu tính kỳ vọng có điều kiện $\mathbb{E}[N_2| N_1 = 2]$ nên trước hết ta phải tính được hàm phân phối có điều kiện
$$ \begin{array}{ccl} P[N_2 = n_2 | N_1 = 2] &=& \dfrac{P[N_1 = 2, N_2 = n_2]}{P[N_1 = 2]} \\ &=& \dfrac{p(2, n_2)}{\displaystyle \sum_{n_2 = 1}^{\infty} p(2, n_2) } \end{array} $$
- Ta có
$$ \begin{array}{ccl} p(2, n_2) &=& \dfrac{3}{4} \left( \dfrac{1}{4} \right)^{2 - 1} e^{-2} (1 - e^{-2})^{n_2 - 1}  \\ &=&  \dfrac{3}{16 e^{2}} \left( 1- e^{-2} \right)^{n_2 - 1} \\ \displaystyle \sum_{n_2 = 1}^{\infty} p(2, n_2) &=& \displaystyle \dfrac{3}{16 e^{2}} \sum_{n_2 = 1}^{\infty} \left( 1- e^{-2} \right)^{n_2 - 1} \\ &=& \dfrac{3}{16e^2} \cdot \dfrac{1}{1 - \left( 1- e^{-2} \right)} = \dfrac{3}{16} \end{array} $$
suy ra
$$ P[N_2 = n_2 | N_1 = 2] = e^{-2} \left( 1- e^{-2} \right)^{n_2 - 1}, \quad \forall n_2 = 1,2,\ldots $$
- Như vậy BNN $(N_2| N_1 = 2)$ có `phân phối hình học` $\text{Geom}(\theta)$ bởi

$$ P[\text{Geom}(\theta) = k] = \theta (1 - \theta)^{k-1}, \quad \forall k = 1,2, \ldots $$
với $\theta = e^{-2}$.
- Vì vậy BNN có phân phối hình học sẽ có kỳ vọng xác định bởi công thức
$$ \mathbb{E}[N_2| N_1 = 2] = \mathbb{E} \left[ \text{Geom}(\theta) \right] = \dfrac{1}{\theta} = e^2. $$

#### Question 132.
A store has 80 modems in its inventory, 30 coming from Source A and the remainder
from Source B. Of the modems from Source A, 20% are defective. Of the modems from
Source B, 8% are defective.

Calculate the probability that exactly two out of a sample of five modems selected
without replacement from the store’s inventory are defective.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Cửa hàng có 80 sản phẩm; trong đó 30 đến từ $A$ và 50 đến từ $B$.
- Vì tỷ lệ phế phẩm (sản phẩm hỏng / defective) là 20% từ A và 8% từ B nên tổng số phế phẩm trong 80 sản phẩm tại cửa hàng là

$$ (0.2)\cdot 30 + (0.08)\cdot 50 = 10 $$
- Do đó, xác suất để đúng 2 trong số 5 mẫu modem được chọn mà không bị lỗi của cửa hàng là $\dfrac{C_{10}^2 C_{70}^3}{C^5_{80}}$

#### Question 133.
A man purchases a life insurance policy on his 40th birthday. The policy will pay 5000 if
he dies before his 50th birthday and will pay 0 otherwise. The length of lifetime, in years
from birth, of a male born the same year as the insured has the `cumulative distribution
function`

$$ F(t) = \left\lbrace \begin{array}{ll} 0, & t \leq 0 \\ 1 - \exp \left( \dfrac{1 - (1.1)^t}{1000} \right), & t > 0 \end{array} \right. $$
Calculate the expected payment under this policy.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y$ là BNN thể hiện khoản phí phải trả từ BH cho khách hàng và $X$ là tuổi thọ của tính theo năm sinh của những khách hàng được khảo sát đã sinh vào năm đó. 
- Theo hợp đồng, thì BH sẽ trả 5000 nếu khách hàng chết trước sinh nhật lần thứ 50 và 0 nếu không chết; do đó ta có liên hệ
$$ Y = \left\lbrace \begin{array}{cl} 5000, & \text{ if } X \in (40, 50), \\ 0, & \text{otherwise} \end{array} \right. $$
- BT cần tính $\mathbb{E}[Y | X > 40]$, ta có

$$ \mathbb{E}[Y | X > 40] = 5000 P[X < 50 | X>40] + 0 P[X > 50 | X > 40] = \dfrac{5000 P[40<X<50]}{P[X > 40]} $$
- Theo giả thiết về hàm phân phối tích lũy của $X$, thì
$$ \begin{array}{ccl}  P[ 40 <  X < 50 ] &=& F(50) - F(40) \\ &=& \left[ 1 - \exp \left( \dfrac{1 - (1.1)^{50}}{1000} \right) \right] - \left[ 1 - \exp \left( \dfrac{1 - (1.1)^{40}}{1000} \right) \right] \\ &=& \exp \left( \dfrac{1 - (1.1)^{40}}{1000} \right) - \exp \left( \dfrac{1 - (1.1)^{50}}{1000} \right) \\ &=& 0.9567 - 0.8901 \\ &=& 0.0666 \\ \\ P[X > 40] &=& 1 - P[X \leq 40] \\ &=& 1 - F(40) \\ &=& 1 - \left[ 1 - \exp \left( \dfrac{1 - (1.1)^{40}}{1000} \right) \right] \\ &=& 0.9567 \end{array} $$
do đó
$$ \mathbb{E}[Y | X > 40] = \dfrac{5000 P[40<X<50]}{P[X > 40]} = 348 $$

#### Question 134.
A mattress store sells only king, queen and twin-size mattresses. Sales records at the
store indicate that the number of queen-size mattresses sold is one-fourth the number of
king and twin-size mattresses combined. Records also indicate that three times as many
king-size mattresses are sold as twin-size mattresses.

Calculate the probability that the next mattress sold is either king or queen-size. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $K, Q, T$ lần lượt là các BNN thể hiện nhóm các loại nệm được bán là loại nào trong `King, Queen` và `Twin`.
- Giả thiết cho biết:
$$ \begin{array}{ccl} P[Q] &=& \dfrac{1}{4} \left( P[K] + P[T] \right) \\ P[K] &=& 3 P[T] \end{array} $$
- BT yêu cầu tính XS để nệm tiếp theo được bán có thể là loại `King` hoặc `Queen`; tức là $P[K \cup Q]$, ta có các lưu ý sau: 

$\qquad \qquad \diamond$ 3 nhóm nệm này là rời nhau vì một người chỉ có thể mua `"Queen" (nếu single women)`; `King (single man)` và `Twin (for couple)` nên chúng đôi một rời nhau, tức là 
$$ \displaystyle Q \cap K = K \cap T = T \cap Q = \emptyset $$
$\qquad \qquad \diamond$ 3 nhóm nệm này được khảo sát từ một cửa hàng nên tổng các trọng số xác suất phải bằng 1.
- Vì vậy, ta được hệ sau
$$ \left\lbrace \begin{array}{lcl} P[K] + P[T] + P[Q] &=& 1 \\ P[K] - 3P[T]&=& 0 \\ P[K] + P[T] - 4P[Q] &=& 0 \end{array} \right. $$
suy ra 
$$ P[T] = 0.2; P[Q] = 0.2; P[K] = 0.6 $$
do đó

$$ P[K \cup Q] = P[Q] + P[K] = 0.8 $$

#### Question 135.
The number of workplace injuries, $N$, occurring in a factory on any given day is `Poisson`
distributed with mean $\lambda$. The parameter  $\lambda$ is a random variable that is determined by the
level of activity in the factory, and is uniformly distributed on the interval [0, 3].

Calculate $\text{Var}(N)$.

**Giải đáp / SOLUTIONs.**

Theo giả thiết, ta có:
- $(N | \lambda) \sim \text{Pois}(\lambda)$ là số các ca chấn thương xảy ra trong ngày tại một nhà máy.
- $\lambda \sim \mathcal{U}([0, 3])$ là số các mức độ hoạt động tại nhà máy (`level of activity`)

Áp dụng công thức phương sai có điều kiện, ta có

$$ \begin{array}{ccl} \text{Var}N &=& \mathbb{E}[ \text{Var}(N | \lambda)] +  \text{Var}[\mathbb{E}(N | \lambda)] \\ &=& \mathbb{E}[ \text{Var}(\text{Pois}(\lambda)) ] + \text{Var}[ \mathbb{E}(\text{Pois}(\lambda)) ] \\ &=& \mathbb{E} (\lambda) + \text{Var}(\lambda) \\ &=& \dfrac{3 + 0}{2} + \dfrac{(3 - 0)^2}{12} \\ &=& 2.25 \end{array} $$

#### Question 136.
A fair die is rolled repeatedly. Let $X$ be the number of rolls needed to obtain a 5 and Y
the number of rolls needed to obtain a 6.

Calculate $\mathbb{E}(X | Y = 2).$

**Giải đáp / SOLUTIONs.**

Theo giả thiết; 
- $X$ là số lần cần thiết để gieo được mặt 5 và 
- $Y$ là số lần cần thiết để gieo được mặt 6.

Để tính $\mathbb{E}[X | Y= 2]$ ta cần xác định được phân phối xác suất của BNN $(X | Y=2)$.

Ta thấy rằng khi $Y = 2$ tức là ở lần gieo thứ 2 mới được mặt 6 và 
- lần gieo đầu tiên nếu được mặt 5 sẽ có xác suất $0.2$
- lần gieo thứ ba (sau Y=2) mới được mặt 5 với xác suất $0.8$

Như vậy

$$ \begin{array}{ccl} \mathbb{E}[X | Y= 2] &=& \mathbb{E}[X | X < 3].P[X < 3] + \mathbb{E}[X|X \geq 3].P[X \geq 3] \\ &=& \mathbb{E}[X | X < 3].P[X = 1] + \mathbb{E}[X|X \geq 3].P[X =3] \\ &=&\left( \dfrac{1}{\theta} + 2\right) (0.8) + 1(0.2) \\ &=& \left( \dfrac{1}{1/6} + 2 \right) (0.8) +  0.2 \\ &=& 6.6 \end{array} $$
trong đó $\theta = \dfrac{1}{6}$ là trung bình của BNN hình học $X$ về số lần gieo cho đến khi đạt được mặt 5.

#### Question 137.
Let $X$ and $Y$ be identically distributed independent random variables such that the
moment generating function of $X + Y$ is 

$$ M_{X+Y}(t) = 0.09e^{-2t} + 0.24e^{-t} + 0.34 + 0.24e^t + 0.09e^{2t}, \quad t \in \mathbb{R} $$
Calculate $P[X \leq 0]$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Vì $X, Y$ độc lập có cùng phân phối nên sẽ có cùng hàm sinh moment và 
$$ M_{X+Y}(t) = M_X(t) M_Y(t) = M_X^2(t) $$
- Như vậy, 
$$ M_X(t) = \sqrt{M_{X+Y}(t)}, $$
ta có
$$ \begin{array}{ccl} M_{X+Y}(t) &=& 0.09e^{-2t} + 0.24e^{-t} + 0.34 + 0.24e^t + 0.09e^{2t}\\ &=& \left( 0.3e^t + 0.3e^{-t} + 0.4 \right)^2 \end{array} $$
suy ra
$$ M_X(t) = 0.3e^t + 0.3e^{-t} + 0.4 $$
- Ta thấy rằng bậc mũ trong hàm sinh moment của $X$ có bậc âm và có 3 giá trị $0, 1, -1$ vì vậy đây là BNN rời rạc chỉ nhận 3 giá trị $0, 1, -1$; và hàm sinh moment này tương ứng với công thức
$$ M_X(t) = \sum_{k=-1}^1 e^{kt} P[X = k] $$
do đó 
$$ P[X=0] = 0.4; \quad P[X=1] = 0.3, \quad P[X = -1] = 0.3 $$
- Vậy
$$ P[X \leq 0] = P(X = 0) + P(X = -1) = 0.7. $$

#### Question 138.
A machine consists of two components, whose lifetimes have the joint density function 
$$ f(x, y) = \left\lbrace \begin{array}{cl} \dfrac{1}{50}, & x > 0, y > 0, x+y < 10 \\ 0, & \text{otherwise} \end{array} \right. $$
The machine operates until both components fail.

Calculate the expected operational time of the machine.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt 2 thành phần thứ nhất và thứ hai của cỗ máy lần lượt là $X,Y$, khi đó chúng có hàm MDXS đồng thời như đề.
- Giả thiết sau cùng cho biết máy chỉ hoạt động nếu cả 2 chi tiết máy còn hoạt động.
- Để tính kỳ vọng về thời gian hoạt động của cỗ máy này, tức là hoặc $X$, hoặc $Y$ bị hỏng sau cùng 
- Với $X$ bị hỏng sau cùng ($Y$ có thể hỏng trước); ta có

$$ \begin{array}{ccl} (X, Y) &\in & \lbrace x \in [0, 10], y \in [0, 5] : x+ y < 10 \rbrace \\ f_X(x) &=& \displaystyle \int_0^{5} \dfrac{1}{50} dy = \dfrac{1}{10} \\ \mathbb{E}X &=& \displaystyle \int xf_X(x) dx = \int_0^{10} \dfrac{x}{10}dx = 5\end{array} $$
- Tương tự cho $Y$ là chi tiết bị hỏng sau cùng ($X$ có thể hỏng trước);

$$ \begin{array}{ccl} (X, Y) &\in & \lbrace y \in [0, 10], x \in [0, 5] : x+ y < 10 \rbrace \\ f_Y(y) &=& \displaystyle \int_0^{5} \dfrac{1}{50} dx = \dfrac{1}{10} \\ \mathbb{E}Y &=& \displaystyle \int yf_Y(y) dx = \int_0^{10} \dfrac{y}{10}dy = 5\end{array} $$

- Do đó, kỳ vọng về thời gian hoạt động của cổ máy này là $5$.

#### Question 139.
A `driver` and a `passenger` are in a `car accident`. 
- Each of them independently has probability 0.3 of being hospitalized. 
- When a `hospitalization occurs`, the `loss` is `uniformly distributed` on $[0, 1]$. When two hospitalizations occur, the losses are independent.

Calculate the expected number of people in the car who are hospitalized, given that the total loss due to hospitalizations from the accident is less than 1. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Trên xe chỉ có 2 người là hành khách và tài xế, mỗi người nhập viện độc lập với nhau với xác suất tai nạn $0.3$.
- Đặt $X_1, X_2$ lần lượt tương ứng với khả năng người nhập viện là tài xế hay hành khách; ta có
$$ X_i = \left\lbrace \begin{array}{cl} 0, & \text{không nhap vien} \\ 1, & \text{nhap vien} \end{array} \right. $$
và $P(X_1 = 1) = P(X_2 = 1) = 0.3$
- Đặt $X = X_1 + X_2$ là số người sẽ phải nhập viện trên chiếc xe đó, và $Z$ là viện phí phải trả. Ta có
$$ X = \left\lbrace \begin{array}{cl} 0, & \text{không có ai nhap vien} \\ 1, & \text{hoac tài xe, hoac hành khách} \\ 2, & \text{ca 2 cùng nhap vien} \end{array} \right.$$
Từ 3 giả thiết này, ta được
$$ \begin{array}{ccl} P[X = 2] &=& P(X_1 = 1).P(X_2 = 1)\\ &=& (0.3)^2 = 0.09 \\ P[X = 1] &=& P[X_1 = 1, X_2 = 0] + P[X_1 = 0, X_2 = 1] \\ &=& P[X_1 = 1]P[X_2 = 0] + P[X_1 = 0]P[X_2 = 1] \\ &=& (0.3)(0.7) + (0.7)(0.3) \\ &=& 0.21 \\ P[X = 0] &=& P[X_1=0]P[X_2=0] \\ &=& 0.49 \end{array} $$

Hơn nữa, khi đã nhập viện thì; 
- Viện phí $(Z | X \leq 1)$ tuân theo phân phối đều $\mathcal{U}([0, 1])$ (nếu số người nhập viện là $0, 1$)
- Nếu cả hai cùng nằm viện $(Z | X=2)$ thì viện phí này tính ở mức $0.5$ (do độc lập).

Do đó, kỳ vọng về số người nhập viện khi biết viện phí thấp hơn 1 sẽ được tính bởi

$$ \begin{array}{ccl} \mathbb{E}(X | Z < 1) &=& \displaystyle \sum_{x = 0}^2 x P[X = x | Z < 1] \\ &=& 0 \cdot \dfrac{P[X=0, Z<1]}{P[Z<1]} + 1 \cdot \dfrac{P[X=1, Z<1]}{P[Z<1]} + 2 \cdot \dfrac{P[X=2, Z<1]}{P[Z<1]} \\ &=& 0 \cdot \dfrac{0.49}{0.49 + 0.42 + (0.5)0.09}+ 1\cdot \dfrac{0.42}{0.49 + 0.42 + (0.5)0.09} + 2 \cdot \dfrac{(0.5)0.09}{0.49 + 0.42 + (0.5)0.09} \\ &=& 0.534 \end{array} $$

#### Question 140.
Each time a hurricane arrives, a new home has a 0.4 probability of experiencing damage.
The occurrences of damage in different hurricanes are mutually independent.

Calculate the mode of the number of hurricanes it takes for the home to experience
damage from two hurricanes. 

**Giải đáp / SOLUTIONs.**

Đặt $X$ là số cơn bão mà trong đó có 2 cơn bão gây thiệt hại trực tiếp đến nhà cửa; khi đó nó có phân phối xác suất

$$ P[X = n] = (n-1)(0.4)^2(0.6)^{n-2}, \quad \forall n \geq 2 $$
vì 
- $0.4$ là XS mà cơn bão trực tiếp phá hoại nhà cửa
- $(n-1)$ là số trường hợp có thể xảy ra với 2 cơn bão gây thiệt hại về nhà cửa trong số $n$ cơn bão.

BT yêu cầu tính `mode` của $X$, tức là tìm $n_0 \geq 2$ sao cho $P[X = n_0]$ đạt giá trị lớn nhất trong $n_0 \geq 2$.

Lúc này, chú ý tỷ số đơn điệu $ r_n = \dfrac{P[X = n+1]}{P[X = n]} $, ta có

$$ r_n = \dfrac{ n(0.4)^2(0.6)^{n-1}}{ (n-1)(0.4)^2(0.6)^{n-2}} = \dfrac{0.6n}{n-1} $$
ta thấy rằng
$$ \left\lbrace \begin{array}{clll} r_n & \geq & 1, & \quad n = 2, \\ r_n & < & 1, & \quad \forall n \geq 3 \end{array} \right. $$
- Do đó tỷ số $r_n$ chỉ lớn hơn $1$ khi $n=2$, tức là 

$$ P[X = 3] > P[X = 2] $$ và sau đó nó sẽ giảm khi $n \geq 3$, tức là

$$ P[X = n+1] < P[X = n] < P[X = 3], \quad \forall n \geq 3 $$
- Như vậy, `mode` của $X$ là khi $X = 3$
