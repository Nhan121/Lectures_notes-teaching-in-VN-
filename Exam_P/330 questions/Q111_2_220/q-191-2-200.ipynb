#### Question 191.
- A fire in an apartment building results in a loss, $X$, to the owner and a loss, $Y$, to the
tenants. 
- The variables X and Y have a bivariate normal distribution with 

$$ \mathbb{E}(X) = 40, \quad\text{Var}(X) = 76,\quad \mathbb{E}(Y) = 30, \quad\text{Var}(Y) = 32, \quad \text{and} \quad \text{Var}(X | Y = 28.5) = 57. $$

Calculate $\text{Var}(Y | X = 25)$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Vì $X, Y$ là một BNN có phân phối chuẩn 2 chiều $\mathcal{N}_2$ nên ta có 

$$ \begin{array}{ccl} \text{Var}(X|Y=y) & = & \left(1 - \rho^2 \right) \text{Var}(X), \qquad \forall y \\ \text{Var}(Y|X=x) &=& \left(1 - \rho^2 \right) \text{Var}(Y), \qquad \forall x \end{array} $$
trong đó
$$ \rho = \text{Corr}(X, Y) = \dfrac{\text{Cov}(X,Y)}{\sqrt{\text{Var}X \cdot \text{Var}Y} } $$
Thay $y=28.5$ suy ra
$$ 1 - \rho^2 = \dfrac{\text{Var}(X|Y=y)}{\text{Var}(X)} = \dfrac{57}{76} = 0.75. $$
- Để tính $\text{Var}(Y|X=25)$, ta có
$$ \text{Var}(Y|X=25) = \left(1 - \rho^2 \right) \text{Var}(Y) = (0.75)32 = 24. $$

#### Question 192.
- `(i)` Losses covered by a flood insurance policy are uniformly distributed on the interval $[0, 2].$ 
- `(ii)` The insurer pays the amount of the loss in excess of a deductible $d$.
- `(iii)` The probability that the insurer pays at least 1.20 on a random loss is $0.30$.

Calculate the probability that the insurer pays at least 1.44 on a random loss

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là các BNN về khoản tiền mà BH phải trả do lũ lụt và mức thiệt hại do lũ gây ra.
- Theo `(i)` thì $Y \sim \mathcal{U}([0, 2])$ và theo `(ii)` ta có $X = Y-d$ hay $$P[X > t] = P[Y > t+d]$$
- `(iii)` cho biết $$ P[X \geq 1.2] = 0.3 $$
suy ra
$$ P[Y = \mathcal{U}([0, 2]) \geq 1.2+d] = 0.3 \Leftrightarrow \dfrac{2-(1.2+d)}{2 - 0} = 0.3 \Leftrightarrow d = 0.2 $$
- BT yêu cầu tính XS để BH chi trả ít nhất 1.44 cho các thiệt hại, tức là $P[X \geq 1.44]$, ta có

$$ P[X \geq 1.44] = P[Y \geq d + 1.44] = \dfrac{2 - (d+1.44)}{2} = 0.18.$$

#### Question 193.
The lifespan, in years, of a certain computer is **exponentially distributed**. The probability that its lifespan exceeds four years is 0.30.

Let $f(x)$ represent the density function of the computer’s lifespan, in years, for $x > 0.$

Determine which of the following is an expression for $f(x)$ w.r.t $c^{q(x)}$ where $c$ is a constant.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện tuổi thọ của máy tính, theo giả thiết thì $X \sim \mathcal{E}(\lambda)$ với $\lambda > 0$, tức $X$ có hàm phân phối tích lũy $$ F_X(x) = P[X < x] = 1 - e^{-\lambda x} $$
- Vì XS mà máy tính xài được trên 4 năm bằng 0.3 nên $$ P[X > 4] = 0.3 $$ suy ra $$ F_X(x) = 0.3 \Rightarrow 1 - e^{-4\lambda} = 0.3 \Rightarrow \lambda = \dfrac{-\ln(0.3)}{4} $$
- Do đó $X$ có hàm MDXS là
$$ f(x) = \lambda e^{-\lambda x}, \quad \forall x > 0 $$
- Vì BT yêu cầu biểu diễn $f$ dưới dạng hàm của $c^{q(x)}$ với $c$ là một hằng số nào đó nên

$$\begin{array}{cllll} f(x) &=& \lambda e^{-\lambda x} &=& \left( \dfrac{-\ln(0.3)}{4} \right) e^{-\left( \frac{-\ln(0.3)}{4} \right) x} \\ &&&=& \left( \dfrac{-\ln(0.3)}{4} \right) \left( e^{\ln(0.3)} \right)^{x/4} \\ &&&=& \left( \dfrac{-\ln(0.3)}{4} \right) (0.3)^{x/4} \end{array} $$

#### Question 194.
The lifetime of a light bulb has density function, $f$, where $f(x)$ is proportional to $\dfrac{x^2}{1+x^3}, \quad 0 < x < 5$, and $0$, otherwise.

Calculate the mode of this distribution. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN về tuổi thọ của bóng đèn $c$ là hằng số mà hàm MDXS $f$ của $X$, theo đó

$$ f(x) = c \left( \dfrac{x^2}{1+x^3} \right) $$
- Do đó, để $f$ thỏa một hàm MDXS thì $$ f \geq 0 , \quad \text{ and } \quad \displaystyle \int f(x)dx = 1$$
tương đương với $$ c > 0 , \quad \text{ and } \quad \displaystyle c \int_0^5 \dfrac{x^2}{1+x^3}dx = 1 $$
do đó $$ c = \dfrac{3}{\ln(126)} $$
- Để xác định **mode** của $X$, tức là tìm $x_m$ thỏa $$ \dfrac{df}{dx}(x_m) = 0 $$
do đó $x_m$ là nghiệm của phương trình

$$\dfrac{d}{dx} \left( \dfrac{3}{\ln 126} \dfrac{x^2}{1 + x^3} \right) = 0 \Rightarrow \dfrac{d}{dx} \left( \dfrac{x^2}{1 + x^3} \right) = 0 \Rightarrow \dfrac{x(2 - x^3)}{(1+x^3)^2} = 0 $$
vậy $x_m = 0$ hoặc $x_m = 2^{1/3}.$
- Lúc này, do $x \in (0, 5)$ nên ta sẽ loại $x = 0$ và mode của $X$ chính là $$ x_m = 2^{1/3}. $$

#### Question 195.
An insurer’s medical reimbursements have density function $f$, where $f(x)$ is proportional to $xe^{-x^2},$ for $0 < x < 1$ and $0$, otherwise.

Calculate the mode of the medical reimbursements.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Rút kinh nghiệm từ BT 194, ta cũng không cần tìm hằng số $C$ sao cho $f$ là hàm MDXS mà thay vào đó ta sẽ tìm trực tiếp **mode** bằng cách xét phương trình

$$ \dfrac{d}{dx} \left( xe^{-x^2} \right) = 0,\quad \forall x \in (0, 1) $$
- Giải phương trình trên, ta có

$$ \dfrac{d}{dx} \left( xe^{-x^2} \right) = 0 \Rightarrow e^{-x^2} (1 - 2x^2) = 0 \Rightarrow 1 - 2x^2 = 0 \Rightarrow x_m = \pm \dfrac{1}{\sqrt{2}}$$
- Lúc này chú ý miền xác định $x \in (0, 1)$ nên mode của $X$ chính là $\dfrac{1}{\sqrt{2}}$

#### Question 196.
A company has five employees on its health insurance plan. 
- `(i)` Each year, each employee independently has an 80% probability of no hospital admissions. 
- `(ii)` If an employee requires one or more hospital admissions, the number of admissions is modeled by a geometric
distribution with a mean of 1.50. 
- `(iii)` The numbers of hospital admissions of different employees are mutually independent.
- `(iv)` Each hospital admission costs 20,000.

Calculate the probability that the company’s total hospital costs in a year are less than 50,000. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện số lần mà một nhân viên sẽ nhập viện ở công ty hàng năm. 
- Theo `(i)`, ta có $$P[X = 0] = (0.8)$$ vì mỗi nhân viên không nhập viện với XS 0.8.
- Theo `(ii)`, khi $X > 0$ (nhân viên nhập viện nhiều lần) thì $X$ sẽ tuân theo phân phối hình học $\text{Geom}(\lambda)$ với $\lambda = 1.5$, ta có

$$ P[X>0|X = k] = \lambda^{-1} \left( 1 - \lambda^{-1} \right)^{k-1}, \quad \forall k = 1,2,\ldots $$
- BT yêu cầu tính XS để số tiền mà cty chi trả cho phí nhập viện không vượt quá 50000. Mà theo `(iv)` thì mỗi nhân viên nằm viện sẽ tốn 20000 mỗi lần nên tối đa 5 nhân viên này chỉ có 2 lần vào viện. Do đó, ta có 4 trường hợp sau:

$\qquad \diamond$ TH1. Không có nhân viên nào nằm viện. tức là $$P\left[ S = \sum_{k=1}^5 X_i = 0 \right] = (0.8)^5$$
trong đó $\displaystyle S = \sum_{k=1}^5 X_i$ là số nhân viên nhập viện trong tổng số $5$ nhân viên, xác suất nhập viện mỗi nhân viên độc lập nhau và bằng 0.2. Vì vậy $$ S \sim \mathcal{B}(5, 0.2) $$

$\qquad \diamond$ TH2. Một nhân viên vào viện 1 lần, 4 đứa còn lại vẫn khỏe. tức là 

$$ \begin{array}{ccl} \displaystyle P\left[ S = 1, X = 1 \right] &=& P[\mathcal{B}(5, 0.2) = 1] \cdot \left( P[X = 1] \right)^{S=1} \\ &=& P[\mathcal{B}(5, 0.2) = 1] \cdot P[X = 1] \\ &=& \left( C_5^1 (0.2) (0.8)^4 \right) \left( (1.5)^{-1} \left( 1 - (1.5)^{-1} \right)^{1-1} \right) \\ &=& 5 (0.2)(0.8)^4 (2/3) \\ &=& 0.27307. \end{array} $$

$\qquad \diamond$ TH3. Một nhân viên vào viện 2 lần (chiếm suất 4 đứa kia nên 4 đứa kia không còn cơ hội nằm viện chùa)

$$ \begin{array}{ccl} \displaystyle P\left[ S = 1, X = 2 \right] &=& P[\mathcal{B}(5, 0.2) = 1] \cdot P[X = 2] \\ &=& \left( C_5^1 (0.2) (0.8)^4 \right) \left( (1.5)^{-1} \left( 1 - (1.5)^{-1} \right)^{2-1} \right) \\ &=& 5 (0.2)(0.8)^4 (2/9) \\ &=& 0.09102. \end{array} $$

$\qquad \diamond$ TH4. Có 2 nhân viên chia nhau một lần vào viên (3 nhân viên còn lại vẫn khỏe),

$$ \begin{array}{ccl} \displaystyle P\left[ S = 2, X = 1 \right] &=& P[\mathcal{B}(5, 0.2) = 1] \cdot (P[X = 2])^2 \\ &=& \left( C_5^2 (0.2)^2 (0.8)^3 \right) \left( (1.5)^{-1} \left( 1 - (1.5)^{-1} \right)^{1-1} \right)^2 \\ &=& 5 (0.2)(0.8)^4 (4/9) \\ &=& 0.09102. \end{array} $$

Cuối cùng, tổng 4 trường hợp trên, ta được XS cần tìm bằng $0.78279.$

#### Question 197.

- `(i)` On any given day, a certain machine has either no malfunctions or exactly one malfunction. 
- `(ii)` The probability of malfunction on any given day is 0.40. 
- `(iii)` Machine malfunctions on different days are mutually independent.

Calculate the probability that the machine has its third malfunction on the fifth day, given
that the machine has not had three malfunctions in the first three days. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $$X_i = \left\lbrace \begin{array}{cl} 1, & \text{neu ngay thu i co xay ra su co} \\ 0, & \text{otherwise} \end{array} \right. $$
- Theo `(i)`-`(iii)`, ta được $$ X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.4) $$
- BT yêu cầu tính XS mà máy lần thứ 3 bị trục trặc vào ngày thứ 5 biết rằng 3 ngày đầu tiên có ít nhất một sự cố, tức là

$$ \displaystyle P[S_5 = 3, X_5 = 1 | S_3 \neq 0], \quad \text{where } \quad S_n = \sum_{k=1}^n X_i $$
- Ta có
$$ S_n \sim \mathcal{B}(n, 0.4), \quad \forall  X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.4) $$
nên
$$ \begin{array}{rclcl} P[S_3 \neq 0] &=& 1 - P[\mathcal{B}(3, 0.4) = 0] = 1 - (0.4)^3 &=& 0.936 \\ P[S_5 = 3, X_5 = 1, S_3 \neq 0] &=& P[S_5 - X_5 = 2] P [X_5 = 1] \\ &=& P[S_4 = 2] P[X_5 = 1] \\ &=& \left( C_4^2 (0.4)^2 (1 - 0.4)^2 \right) \left( 0.4 \right) &=& 0.13824 \end{array} $$
suy ra
$$ P[S_5 = 3, X_5 = 1 | S_3 \neq 0] = \dfrac{P[S_5 = 3, X_5 = 1, S_3 \neq 0]}{P[S_3 \neq 0]} = 0.1477$$

#### Question 198.

In a certain group of cancer patients, each patient's cancer is classified in exactly one of
the following five stages: stage 0, stage 1, stage 2, stage 3, or stage 4.
- `i)` 75% of the patients in the group have stage 2 or lower.
- `ii)` 80% of the patients in the group have stage 1 or higher.
- `iii)` 80% of the patients in the group have stage 0, 1, 3, or 4.

One patient from the group is randomly selected.

Calculate the probability that the selected patient's cancer is stage 1.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $p_i$ là XS mà bệnh nhân bị ung thư bước vào giai đoạn thứ $i$ với $i = 0,1,2,3,4$; theo đó $$\displaystyle \sum_{i=0}^4 p_i = 1$$
- Từ `i`, ta có $$p_0 + p_1 + p_2 = 0.75$$
- Từ `ii`, ta có $$p_1 + p_2 + p_3 + p_4 = 0.8$$
- Từ `iii`, ta có $$p_0 + p_1 + p_3 + p_4 = 0.8$$
- BT yêu cầu ta tìm $p_1$, ta có
$$ p_0 = \displaystyle \left( \sum_{i=0}^4 p_i\right) - \left( p_1 + p_2 + p_3 + p_4\right) = 0.2 $$
và
$$ p_2 = \displaystyle \left( \sum_{i=0}^4 p_i\right) - \left( p_0 + p_1 + p_3 + p_4\right) = 0.2 $$
suy ra
$$ p_1 = \left( p_0 + p_1 + p_2 \right) - p_0 - p_2 = 0.75 - 0.2 - 0.2 = 0.35. $$

#### Question 199.

A car is new at the beginning of a calendar year. 

The time, in years, before the car experiences its first failure is **exponentially distributed** with mean 2.

Calculate the probability that the car experiences its first failure in the last quarter of
some calendar year.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là thời gian tính đến khi chiếc xe gặp hư hỏng đầu tiên trong năm, theo đó $X \sim \mathcal{E}(2)$ và $X$ có hàm phân phối tích lũy
$$ F(x) = P[X < x] = 1 - e^{-x/\lambda}, \qquad \lambda = 2 $$
- BT yêu cầu tính XS để chiếc xe bị hỏng vào quý cuối cùng của năm ở tất cả các năm. Theo đó,

$\qquad \diamond$ Với mỗi năm, ta cho $k$ chạy từ $0$ đến $\infty$

$\qquad \diamond$ Vì mỗi năm, đều hỏng vào quý cuối, tức là $X \in (k + 0.75, k+1]$ vào năm thứ $k$, do đó XS cần tìm chính là 

$$ \begin{array}{ccl} P \left( X \in (0.75, 1] \right) &=& \displaystyle \sum_{k=0}^{\infty} P\left( X \in (k + 0.75, k+1], X \in [k, k+1] \right) \\ &=& \displaystyle \sum_{k=0}^{\infty} P\left( X \in (k + 0.75, k+1] | X \in (k, k+1] \right) P(k<X\leq k+1)\end{array} $$
Mà
$$ \begin{array}{ccl} P\left( X \in (k + 0.75, k+1] | X \in (k, k+1] \right) &=& \dfrac{P[k+0.75 < X < k+1]}{P[k< X \leq k+1]} \\ &=& \dfrac{F(k+1) - F(k+0.75)}{F(k+1) - F(k)} \\ &=& \dfrac{\left( 1 - e^{-(k+1)/2} \right) - \left(1 - e^{-(k+0.75)/2} \right)}{\left( 1 - e^{-(k+1)/2} \right) - \left(1 - e^{-k/2} \right)} \\ &=& \dfrac{e^{-k/2}\left[ -e^{-1/2} + e^{-(0.75)/2} \right]}{e^{-k/2} \left[ -e^{-1/2} + 1 \right]} \\ &=& 0.205. \end{array} $$

suy ra

$$ P \left( X \in (0.75, 1] \right) = \displaystyle 0.205 \sum_{k=0}^{\infty} P(k<X\leq k+1) = 0.205 $$

#### Question 200.
- In a shipment of 20 packages, 7 packages are damaged. 
- The packages are randomly
inspected, one at a time, without replacement, until the fourth damaged package is
discovered.

Calculate the probability that exactly 12 packages are inspected.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $$X_i = \left\lbrace \begin{array}{cl} 1, & \text{goi hang thu i bi hong} \\ 0, & \text{otherwise} \end{array} \right. $$ là kiện hàng thứ $i$ có bị hỏng hay không và $$S_n = \displaystyle \sum_{i=1}^n X_i$$ là tổng số kiện hàng bị hỏng trong $n$ kiện.
- Đề cho biết có 20 món hàng và có 7 kiện bị hỏng và công việc kiểm tra chỉ dừng lại khi thấy có 4 kiện hàng bị hỏng
- yêu cầu BT là tính XS để chỉ cần đúng 12 lần kiểm tra, tức là lần thứ 12 tìm thấy kiện hàng thứ tư bị hỏng và 3 kiện hỏng còn lại nằm trong 11 lần trước đó, tức là
$$ P[X_{12} = 1, S_{11} = 3] = P\left[ \left. (X_{12} = 1) \right\vert (S_{11} = 3) \right] P[S_{11} = 3] $$
- Lưu ý rằng ta không có giả thiết về Bernoulli ở đây với $X_i$!! Do đó,

$\qquad \diamond P[S_{11} = 3]$ là XS trong 11 lần lấy đầu tiên, có 3 kiện bị hỏng (tổng là 7 kiện hỏng trong tổng số kiện 20) nên

$$ P[S_{11} = 3] = \dfrac{C_{7}^3 C_{20 - 7}^{11- 3}}{C_{20}^{11}} = \dfrac{C_7^3 C_{13}^8}{C_{20}^{11}} \approx 0.2682 $$

$\qquad \diamond  P\left[ \left. (X_{12} = 1) \right\vert (S_{11} = 3) \right]$ là XS mà kiểm được kiện hàng bị hỏng ở lần thứ 12; biết rằng 11 lần trước có 3 lần hỏng, tức là lúc này: **trong mẫu 9 phần tử còn lại chỉ còn 4 kiện hỏng**

$$  P\left[ \left. (X_{12} = 1) \right\vert (S_{11} = 3) \right] = \dfrac{7 - 3}{20 - 11} = \dfrac{4}{9} $$
- Như vậy,
$$ P[X_{12} = 1, S_{11} = 3] \approx 0.119. $$
