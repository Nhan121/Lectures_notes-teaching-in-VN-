#### Question 281, 
- For a certain health insurance policy, losses are uniformly distributed on the interval
[0, 450]. 
- The policy has a deductible of $d$ and the expected value of the unreimbursed
portion of a loss is $56$.

Calculate $d.$


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là tổn thất từ chính sách BH, và $Y$ là khấu hao từ $X$ với mức khấu hao là $d$, khi đó 

$$ X \sim \mathcal{U}([0, 450]), \qquad Y = \max(X, d) $$
- BT yêu cầu tìm $d$ với $d \in [0, 450]$ khi biết $\mathbb{E}Y = 56.$ Ta có

$$ \begin{array}{cclrc} \mathbb{E}Y &=&  \displaystyle \int \max(d, x) f(x)dx \\ &=& \displaystyle \int_0^d \dfrac{x}{450}dx + \int_d^{450} \dfrac{d}{450}dx &=& \dfrac{d(900 - d)}{900} \end{array} $$
do đó $d$ là nghiệm của phương trình

$$ \dfrac{d(900 - d)}{900} = 56 \Leftrightarrow d^2 -900d + 50400 = 0 \Leftrightarrow \left[ \begin{array}{ccl} d &=& 60 \in (0, 450) \\ d &=& 840 \not \in (0, 450) \end{array}  \right. $$

do đó nhận $d = 60.$

#### Question 282.
A motorist just had an accident. The accident is **minor** with probability $0.75$ and is otherwise **major**.

Let $b$ be a positive constant. 
- If the accident is minor, then the loss amount follows a uniform distribution on the interval $[0, b]$. 
- If the accident is major, then the loss amount follows a uniform distribution on the interval $[b, 3b]$.

The median loss amount due to this accident is 672.

Calculate the mean loss amount due to this accident

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt 
$$ X = \left\lbrace \begin{array}{cl} X_a , & \text{if the accident is major} \\ X_c, & \text{if the accident is minor} \end{array} \right. $$
khi đó
$$ P[X = X_c] = 0.75, \qquad P[X = X_a] = 0.25. $$
- Đặt $Y$ là mức tổn thất do tai nạn motor gây ra, theo đó

$$ (Y|X = X_c) \sim \mathcal{U}([0, b]) \qquad (Y|X = X_a) \sim \mathcal{U}([b, 3b]) . $$
- Giả định cuối cùng cho biết trung vị của $Y$ là 672; tức là
$$ P[Y < 672] = 0.5 $$
- Ta có các trường hợp sau về $b$,

$\qquad \diamond$ Nếu $672 \in [0, b]$ thì 
$$ \begin{array}{ccrcl} P[Y < 672] = 0.5 & \Leftrightarrow & P[Y < 672|X=X_c] \cdot P[X = X_c] &=& 0.5 \\ & \Leftrightarrow & \dfrac{672 - 0}{b - 0} \cdot (0.75) &=& 0.5 \\ & \Leftrightarrow & b &=& 1008. \end{array} $$
$\qquad \quad$ Như vậy $b > 672$ nên thỏa.

$\qquad \diamond$ Nếu $672 \in [b, 3b]$ thì 
$$ \begin{array}{ccrcl} P[Y < 672] = 0.5 & \Leftrightarrow & P[Y < 672|X=X_a] \cdot P[X = X_a] &=& 0.5 \\ & \Leftrightarrow & \dfrac{672 - b}{b - 0} \cdot (0.25) &=& 0.5 \\ & \Leftrightarrow & b &=& 168. \end{array} $$
$\qquad \quad$ Như vậy $[b, 3b] = [168, 504]$ nên $672 \not \in [b, 3b]$

$\qquad \diamond$ Ở ngoài khoảng $[0, 3b]$ thì ta không có giá trị của $Y$ nên ta không cần bàn đến trường hợp này; do đó $$b = 1008$$ là giá trị thích hợp của bài toán.
- Cuối cùng, BT yêu cầu tính kỳ vọng của $Y$, ta có

$$ \begin{array}{ccl} \mathbb{E}Y &=& \mathbb{E}(Y|X=X_c) P(X=X_c) + \mathbb{E}(Y|X=X_a) P(X=X_a) \\ &=& \mathbb{E}\left[ \mathcal{U}([0, b]) \right] \cdot (0.75) + \mathbb{E}\left[ \mathcal{U}([b, 3b]) \right] \cdot (0.25) \\ &=& (0.75) \left( \dfrac{0 + b}{2} \right) + (0.25) \left( \dfrac{b + 3b}{2} \right) \\ &=& 882 \end{array} $$

#### Question 283.
An insurance policy will reimburse only one claim per year.
- For a random policyholder, there is a 20% probability of no loss in the next year, in
which case the claim amount is 0. 
- If a loss occurs in the next year, the claim amount is
normally distributed with mean 1000 and standard deviation 400.

Calculate the median claim amount in the next year for a random policyholder. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số tiền bồi thường các tổn thất và $L$ là mức tổn thất, theo đó 

$$ (X|L) \sim \left\lbrace \begin{array}{cl} \mathcal{N}\left( 1000, 400^2 \right), & \text{if } L > 0 \\ 0, & \text{if } L = 0 \end{array} \right. $$
và $$P[L = 0] = 0.2, \qquad P[L > 0] = 1 - P[L=0] = 0.8. $$
- BT yêu cầu xác định trung vị của $X$, tức là tìm $m_X > 0$ sao cho $$ P[X < m_X] = 0.5 $$
Ta có

$$ \begin{array}{ccl} P[X < m_X] &=& P\left[X < m_X | L = 0 \right] \cdot P[L = 0] + P[X < m_X | L > 0] \cdot P[L > 0] \\ &=& P\left[(X|L=0) < m_X \right] \cdot (0.2) + P[(X | L > 0) < m_X]. (0.8) \\ &=& \quad P [0 < m_X] \cdot (0.2) \quad + \quad P\left[ \mathcal{N}\left( 1000, 400^2 \right) < m_X \right]  \cdot (0.8) \\ &=& \qquad 0.2 \quad + \quad (0.8) \cdot \Phi \left( \dfrac{m_X - 1000}{400} \right) \end{array}$$
do đó $m_X$ phải thỏa
$$ \begin{array}{ccrcl} 0.2 + (0.8) \cdot \Phi \left( \dfrac{m_X - 1000}{400} \right) = 0.5  & \Leftrightarrow & \dfrac{m_X - 1000}{400} &=& \Phi^{-1} \left( \dfrac{0.5 - 0.2}{0.8} \right) \\ & \Leftrightarrow & \dfrac{m_X - 1000}{400} &=& -0.3187 \\ & \Leftrightarrow & m_X &=& 873 \end{array} $$

#### Question 284.
- Losses incurred by a policyholder follow a normal distribution with mean 20,000 and
standard deviation 4,500. 
- The policy covers losses, subject to a deductible of 15,000.

Calculate the 95th percentile of losses that exceed the deductible.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là **mức tổn thất từ bên mua BH** và **khoản chi trả của bên BH khi vượt quá khấu hao $15000$**. Khi đó

$$ X \sim \mathcal{N}\left(20000, 4500^2 \right), \qquad Y = \max(0, X - 15000) $$
- BT yêu cầu tìm phân vị mức $95\%$ của $Y$, tức là tìm $q > 15000$ sao cho $P[Y < q] = 0.95$, ta có

$\qquad \diamond$ Tại giá trị khấu hao $X = 15000$ thì mức phần trăm phân vị tại đây là

$$ P[X \geq 15000] = P\left( \mathcal{N}(0,1) \geq \dfrac{15000 - \mathbb{E}X}{\sqrt{\text{Var}X}} \right) = 1 - \Phi \left( \dfrac{15000 - 20000}{4500} \right) = 0.8665. $$

$\qquad \diamond$ Bởi vì, $P[Y < q] = 0.95$, nên phân vị mức $95\%$ của $Y$ sẽ bằng

$$ 1 - (1 - P[Y < q])\cdot P[X \geq 15000] = 1 - (0.05)(0.8665) = 0.9567 $$

là phân vị mức $95.67\%$ của tổng tổn thất $X$ (khi không có khấu hao); tức là

$$ 0.9567 = P[X < q] = P\left( \mathcal{N}(0,1) < \dfrac{q - \mathbb{E}X}{\sqrt{\text{Var}X}} \right) = \Phi \left( \dfrac{q - 20000}{4500} \right) $$
suy ra

$$ q = 4500 \cdot \Phi^{-1}(0.9567)  + 20000 \approx 27711. $$

#### Question 285.
- A gun shop sells gunpowder. Monthly demand for gunpowder is normally distributed, averages 20 pounds, and has a standard deviation of 2 pounds. 
- The shop manager wishes to stock gunpowder inventory at the beginning of each month so that there is only a 2%
chance that the shop will run out of gunpowder (i.e., that demand will exceed inventory) in any given month.

Calculate the amount of gunpowder to stock in inventory, in pounds. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là lượng tiêu thụ thuốc súng tại cửa hàng, khi đó $X \sim \mathcal{N}\left(20, 2^2 \right)$
- Vì quản lý cửa hàng muốn lượng dự trữ kho thuốc súng vào mỗi đầu tháng để chỉ có $2\%$ khả năng làm cửa hàng hết hàng (tức là nhu cầu sẽ vượt quá lượng hàng tồn kho) trong tháng đó. 
- Do đó, lượng hàng cần trữ trong kho $q$ phải đạt phân vị mức $(100 - 2\%) = 98\%$ của lượng tiêu thụ $X$, tức là

$$ P[X < q] = 0.98 \Leftrightarrow q = \sqrt{\text{Var}X} \Phi^{-1}(0.98) + \mathbb{E}X = 2(2.054) + 20 = 24.108 \text{ (pounds).}$$

#### Question 286.
- A large university will begin a 13-day period during which students may register for that semester’s courses. 
- Of those 13 days, the number of elapsed days before a randomly selected student registers has a continuous distribution with density function $f(t)$ that is symmetric about $t = 6.5$ and proportional to $1/(t + 1)$ between days $0$ and $6.5$.

A student registers at the $60^{th}$ percentile of this distribution. Calculate the number of elapsed days in the registration period for this student. 

**Giải đáp / SOLUTIONs.**

- Đặt $T$ là thời gian đăng ký (`registration time`) của SV.
- Hàm MDXS $f(t)$ của $T$, theo các giả thiết thì:

$\qquad \diamond$ $t$ có giá trị trong khoảng $[0, 13]$ với $13$ là số ngày để SV đăng ký các môn học của học kỳ mới.

$\qquad \diamond$ $f$ là hàm liên tục trên $[0, 13]$ và đối xứng qua trung điểm $t_0 = 6.5$, tức là 

$$f(6.5 - t) = f(6.5 + t), \qquad \forall t \in [0, 6.5]$$
$\qquad \quad$ và
$$P(T < 6.5 - t) = P(T > 6.5 + t), \qquad \forall t \in [0, 6.5]$$
$\qquad \diamond$ $f$ tỷ lệ với $\dfrac{1}{1+t}$ khi $t \in [0, 6.5]$, tức là tồn tại một hằng số $c > 0$ sao cho $$ f(t) = \dfrac{c}{1 + t}, \qquad \forall t \in [0, 6.5] $$
- Từ đây, suy ra

$\qquad \diamond$ $P[T < 6.5] = P[T > 6.5] = 0.5$ bởi tính đối xứng của $f$ tại $t_0 = 6.5$

$\qquad \diamond$ Tìm $c > 0$ để $f$ là hàm MDXS, tương đương với

$$ \displaystyle \int_0^{6.5} f(t) = 0.5 \Leftrightarrow \int_0^{6.5} \dfrac{c}{1+t}dt = 0.5 \Leftrightarrow c = \dfrac{0.5}{\ln(7.5)} $$
- Cuối cùng, BT yêu cầu tìm phân vị $q$ mức $60\%$ của $T$; ta có

$$ P[T < q] = 0.6 \Longrightarrow P[T > q] = 0.4 $$

mà , bởi tính đối xứng của $f$ (xem dấu bằng ở giữa trong khai triển dưới đây), 

$$ P[T > q] = P[T > 6.5 + (q - 6.5)] = P[T < 6.5 - (q - 6.5)] = P[T < 13 - q] $$
do đó ta cần tìm $q$ sao cho 

$$ \begin{array}{ccrcl} P[T < 13 - q] = 0.4 & \Leftrightarrow & \displaystyle \int_0^{13 - q} \dfrac{c}{1+t}dt &=& 0.4 \\ & \Leftrightarrow & \dfrac{(0.5) \ln(14 - q)}{\ln (7.5)} &=& 0.4 \\ & \Leftrightarrow & q & \approx & 8.99 \end{array} $$

#### Question 287.
- The loss L due to a boat accident is exponentially distributed.
- Boat insurance policy A covers up to 1 unit for each loss. Boat insurance policy B covers
up to 2 units for each loss.
- The probability that a loss is fully covered under policy B is 1.9 times the probability that
it is fully covered under policy A.

Calculate the variance of L. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo các giả thiết, $L \sim \mathcal{E}(\lambda)$, với $\lambda > 0$ chưa biết, tức là

$$ F(x) = P[L < x] = 1 - e^{-\lambda x}, \qquad \mathbb{E}L = \lambda^{-1}, \qquad \text{Var}L = \lambda^{-2} $$
- Đặt $T$ là phí chi trả các tổn thất $L$ của BH và $$ X = \left\lbrace \begin{array}{cl} X_A, & \text{neu BH cua thuyen la loai A} \\ X_B, & \text{neu BH cua thuyen la loai B} \end{array} \right. $$ khi đó $$ T(X) = \left\lbrace \begin{array}{cl} 1, & X = X_A \\ 2, & X = X_B \end{array} \right. $$
- Giả thiết cuối cùng cho biết

$$ \begin{array}{ccrcl} P[L  < T|X = X_B] = (1.9) \cdot P[L < T| X = X_A] & \Leftrightarrow & P[L < 2] &=& (1.9) \cdot P[L < 1] \\ & \Leftrightarrow & 1 - e^{-2\lambda} &=& (1.9) \cdot (1 - e^{-\lambda}) \\ & \Leftrightarrow & (1 - e^{-\lambda})(1 + e^{-\lambda}) &=& (1.9)(1 - e^{-\lambda}) \\ & \Leftrightarrow & (1 - e^{-\lambda})(e^{-\lambda} - 0.9) &=& 0 \\ & \Leftrightarrow & \lambda & = & \lbrace 0, -\ln(0.9) \rbrace \end{array} $$

vì ta chỉ xét $\lambda > 0$ nên ta nhận $\lambda = -\ln(0.9)$
- BT yêu cầu tính phương sai của $L$ và sau khi tìm được $\lambda$, suy ra

$$ \text{Var} L = \lambda^{-2} \approx 90.1 $$

#### Question 288.
- Losses, $X,$ under an insurance policy are exponentially distributed with mean 10. 
- For each loss, the claim payment $Y$ is equal to the amount of the loss in excess of a deductible $d > 0.$

Calculate $\text{Var}(Y).$

**Giải đáp / SOLUTIONs.**

- Giả thiết cho biết 

$$X \sim \mathcal{E}(10), \qquad Y = \max(0, X-d)$$
- BT yêu cầu ta tính $\text{Var}Y$ theo khấu hao $d > 0$, ta có

$$ \begin{array}{ccl} \mathbb{E}Y &=& \displaystyle \int_d^{\infty} (x-d) f(x) dx \\ &=&  \displaystyle \int_d^{\infty}(x-d) \left( \dfrac{1}{10} e^{-\frac{x}{10}} \right) = \displaystyle \dfrac{1}{10} \int_0^{\infty} t e^{-\frac{(t+d)}{10}} dt = \displaystyle e^{-0.1d} \left( (0.1) \int_0^{\infty} t e^{-(0.1t)}dt \right) = 10 e^{-(0.1 d)} \\ \mathbb{E}Y^2 &=& \displaystyle \int_d^{\infty} (x-d)^2 f(x) dx \\ &=&  \displaystyle \int_d^{\infty}(x-d)^2 \left( \dfrac{1}{10} e^{-\frac{x}{10}} \right) = \displaystyle \dfrac{1}{10} \int_0^{\infty} t^2 e^{-\frac{(t+d)}{10}} dt = \displaystyle e^{-0.1d} \left( (0.1) \int_0^{\infty} t^2 e^{-(0.1t)}dt \right) = 200 e^{-(0.1 d)} \end{array} $$
suy ra

$$ \text{Var}Y = \mathbb{E}Y^2 - [\mathbb{E}Y]^2 = 100 \left( 2e^{-(0.1d)} - e^{-(0.2d)} \right). $$

#### Question 289.
- For a certain insurance company, 10% of its policies are Type A, 50% are Type B, and 40% are Type C.
- The annual number of claims for an individual Type A, Type B, and Type C policy follow Poisson distributions with respective means 1, 2, and 10.
- Let $X$ represent the annual number of claims of a randomly selected policy.

Calculate the variance of $X. $

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số các yêu cầu (giống như đề) và $$T = \left\lbrace \begin{array}{cl} A, & \text{if policy is Type A} \\ B, & \text{if policy is Type B} \\ C, & \text{if policy is Type C} \end{array} \right.$$
- Ta có 
$$ (X|T=A) \sim \text{Pois}(1), \qquad (X|T=B) \sim \text{Pois}(2), \qquad (X|T=C) \sim \text{Pois}(10) $$
và
$$ P[T = A] = 0.10, \qquad P[T = B] = 0.50, \qquad P[T = C] = 0.40 $$
suy ra

$$ \begin{array}{ccl} \mathbb{E}X &=& \mathbb{E}(X|T=A) \cdot P[T=A] + \mathbb{E}(X|T=B) \cdot P[T=B] + \mathbb{E}(X|T=C)\cdot P[T=C] \\ &=& \mathbb{E}(\text{Pois}(1)) \cdot (0.1) + \mathbb{E}(\text{Pois}(2)) \cdot (0.5) + \mathbb{E}(\text{Pois}(10)) \cdot (0.4) \\ &=& 1 \cdot (0.1) + 2 \cdot (0.5) + 10 \cdot (0.4) \\ &=& 5.1 \\ \mathbb{E}X^2 &=& \mathbb{E}(\text{Pois}(1))^2 \cdot (0.1) + \mathbb{E}(\text{Pois}(2))^2 \cdot (0.5) + \mathbb{E}(\text{Pois}(10))^2 \cdot (0.4) \\ &=& \left( 1 + 1^2 \right) \cdot (0.1) + \left( 2 + 2^2 \right) \cdot (0.5) + \left( 10 + 10^2 \right) \cdot (0.4) \\ &=& 47.2 \end{array} $$
do đó

$$ \text{Var}X = \mathbb{E}X^2 - [\mathbb{E}X]^2 = 21.19. $$

#### Question 290.
The number of tornadoes in a given year follows a Poisson distribution with mean 3.

Calculate the variance of the number of tornadoes in a year given that at least one tornado
occurs.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số cơn bão trong năm, theo đó $X \sim \text{Pois}(3)$ với $$ P[X = k] = e^{-3} \dfrac{3^k}{k!}, \qquad \forall k = 0,1,2, \ldots $$ và $$Y = (X | X \geq 1)$$ là số cơn bão trong năm mà năm đó có ít nhất một cơn bão đổ bộ.
- Khi đó, $Y$ sẽ nhận giá trị dựa trên $X \geq 1$, tức là $\forall y = 1,2,\ldots$ ta có

$$ \begin{array}{ccl} P[Y = y] = P[X = y | X \geq 1] = \dfrac{P[(X = y)\cap(X \geq 1)]}{P[X \geq 1]} = \dfrac{P[Y = y]}{1 - P[Y=0]} &=& \dfrac{e^{-3} \frac{3^y}{y!}}{1 - e^{-3}} \end{array} $$
- BT yêu cầu tính $\text{Var}Y$, do đó ta lại lần lượt tính các moment cấp 1 và cấp 2 của nó, ta có
$$ \begin{array} \mathbb{E}Y = \displaystyle \sum_{y=1}^{\infty} y P[Y=y] &=& \displaystyle \dfrac{1}{1 - e^{-3}} \left[ \sum_{y=1}^{\infty} y \left( \dfrac{e^{-3} \cdot 3^y}{y!} \right) \right] \\ &=& \displaystyle \dfrac{1}{1 - e^{-3}} \left[ \sum_{y=0}^{\infty} y \left( \dfrac{e^{-3} \cdot 3^y}{y!} \right) \right] \\  &=& \dfrac{1}{1 - e^{-3}} (\mathbb{E}X) \\ &=& \dfrac{3}{1 - e^{-3}} \\ \mathbb{E}Y^2 = \displaystyle \sum_{y=1}^{\infty} y^2 P[Y=y] &=& \displaystyle \dfrac{1}{1 - e^{-3}} \left[ \sum_{y=0}^{\infty} y^2 \left( \dfrac{e^{-3} \cdot 3^y}{y!} \right) \right] \\ &=& \displaystyle \dfrac{1}{1 - e^{-3}} \cdot \left( \mathbb{E} X^2 \right) \\ &=& \dfrac{3 + 3^2}{1 - e^{-3}} \\ &=& \dfrac{12}{1 - e^{-3}} \end{array}$$

do đó

$$ \text{Var}Y = \mathbb{E}Y^2 - [\mathbb{E}Y]^2 \approx 2.6609 $$
