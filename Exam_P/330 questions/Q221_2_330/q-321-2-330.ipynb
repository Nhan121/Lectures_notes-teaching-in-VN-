#### Question 321, 
Losses follow an exponential distribution with mean 1. Two independent losses are observed.

Calculate the expected value of the smaller loss.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt các tổn thất trong đó lần lượt là $X, Y$, theo đó $X, Y \overset{\text{i.i.d.}}{\sim} \mathcal{E}(1)$ và

$$ P[X > t] = P[Y > t] = e^{-t}, \qquad \forall t \geq 0$$
- BT yêu cầu tính $\mathbb{E}\min(X,Y)$, do đó, ta cần xác định được phân phối của $\min(X,Y)$. Ta có

$$ P[\min(X,Y) > t] = P[(X > t) \cap (Y > t)] = P[X > t] \cdot P[Y > t] = e^{-2t} $$
suy ra
$$ P[\min(X,Y) \leq t] = 1 - P[\min(X,Y) > t] = 1 - e^{-2t} $$

và đây là dạng hàm phân phối tích lũy của $\mathcal{E}\left( \frac{1}{2} \right)$ nên
$$ \mathbb{E}\min(X,Y) = \dfrac{1}{2}. $$

#### Question 322.
A delivery service owns two cars that consume 15 and 30 miles per gallon. 
- Fuel costs 3 per gallon. 
- On any given business day, each car travels a number of miles that is
independent of the other and is normally distributed with mean 25 miles and standard
deviation 3 miles.

Calculate the probability that on any given business day, the total fuel cost to the delivery
service will be less than 7.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ là quãng đường đi được của 2 chiếc xe, tính bằng dặm (`miles`)
- Đặt $F$ là chi phí cho tiền xăng dầu và nó được tính bởi 3 USD trên mỗi gallon
- Theo đề, mỗi gallon xăng tương ứng với $X, Y$ lần lượt là $15$ và $30$ dặm, vì thế số tiền $F$ sẽ được tính bởi tích của $3$ trên tổng số gallons đã tiêu thụ: $\dfrac{X}{15}$ cho $X$ (`số gallons` = `quãng đường` / `mức tiêu thụ xăng mỗi gallon`); và, tương tự: $\dfrac{Y}{30}$ cho $Y$; do đó
$$ F = 3 \left( \dfrac{X}{15} + \dfrac{Y}{30} \right) = 0.2 X + 0.1Y $$
- Hơn nữa,
$$ X, Y \overset{\text{i.i.d}}{\sim} \mathcal{N}(25, 3^2) $$
suy ra
$$ \begin{array}{rcl} \mathbb{E}F &=& (0.2)\mathbb{E}X + (0.1)\mathbb{E}Y = 7.5 \\ \text{Var}F &=& (0.2)^2\text{Var}X + (0.1)^2\text{Var}Y = 0.45 \end{array}$$
và 
$$ F \sim \mathcal{N}(7.5, 0.45) $$
- BT yêu cầu tính $P[F < 7]$, ta có

$$ P[F < 7] = P\left[ \mathcal{N}(0,1) < \dfrac{7 - 7.5}{\sqrt{0.45}} \right] = \Phi(-0.7454) = 0.23. $$

#### Question 323.
- Two independent estimates are to be made on a building damaged by fire. 
- Each estimate is **normally distributed** with mean $10b$ and variance $b^2$.

Calculate the probability that the first estimate is at least 20 percent higher than the second.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt tương ứng với các ước lượng thứ nhất và thứ hai (về các thiệt hại do đám cháy gây ra). Theo đó,

$$ X,Y  \overset{\text{i.i.d}}{\sim} \mathcal{N}(10b, b^2) $$
- BT yêu cầu tính XS mà $X$ cao hơn $Y$ ít nhất $20\%$, tức là $P[X > (1.2)Y]$
- Từ các giả thiết trên, suy ra
$$ \begin{array}{ccl} \mathbb{E}[X - (1.2)Y] &=& \mathbb{E}X - (1.2) \mathbb{E}Y = -2b \\ \text{Var}[X - (1.2)Y] &=& \text{Var}X + (1.2)^2 \text{Var}Y = 2.44 b^2 \end{array} $$
và
$$ X - (1.2)Y \sim \mathcal{N}(-2b, 2.44b^2) $$
- Vậy
$$ P[X > (1.2)Y] = P[X-(1.2)Y>0] = 1 - P\left[ \mathcal{N}(0,1) < \dfrac{0 - (-2b)}{\sqrt{(2.44)b^2}} \right] = 1 - \Phi(1.28) \approx 0.1 $$

#### Question 324.
- The independent random variables $X$ and $Y$ have the same mean. 
- The coefficients of variation of $X$ and $Y$ are $3$ and $4$ respectively.

Calculate the coefficient of variation of $\dfrac{1}{2}(X+Y)$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Vì $X,Y$ độc lập được giả định có trung bình bằng nhau nên ta đặt $m$ là trung bình của $X, Y$, khi đó :

$$ \mathbb{E}X = \mathbb{E}Y = m $$
- Giả định còn lại (về tỷ số giữa các độ lệch chuẩn trên trung bình) cho biết
$$ \dfrac{\text{sd}X}{\mathbb{E}X} = 3, \qquad \dfrac{\text{sd}Y}{\mathbb{E}Y} = 4. $$
suy ra
$$ \text{Var}X = (3m)^2, \qquad \text{Var}Y = (4m)^2. $$
- Do đó
$$ \text{Var} \left( \dfrac{X+Y}{2} \right) = \dfrac{1}{4} \text{Var}(X+Y) = \dfrac{1}{4}\left( \text{Var}X + \text{Var}Y \right) = \dfrac{25m^2}{4} $$
- BT yêu cầu tính hệ số biến phân
$$ \dfrac{\text{sd} \left( \dfrac{X+Y}{2} \right)}{\mathbb{E}\left( \dfrac{X+Y}{2} \right)} = \dfrac{\left( \sqrt{\dfrac{25m^2}{4}} \right)}{\dfrac{m + m}{2}} = \dfrac{5}{2}. $$

#### Question 325.
Points scored by a game participant can be modeled by $Z = 3X + 2Y – 5$. 

$X$ and $Y$ are independent random variables with $\text{Var } (X) = 3$ and $\text{Var } (Y) = 4.$

Calculate $\text{Var }(Z)$

**Giải đáp / SOLUTIONs.**

Nhắc lại rằng:
$$ \begin{array}{rcclr} 1) & \text{Var}(X + c) &=& \text{Var}X & \forall c \in \mathbb{R} \\ 2) & \text{Var}(cY) &=& c^2 \text{Var}X & \forall c \in \mathbb{R} \\ 3) & \text{Var}(X + Y) &=& \text{Var}X + \text{Var}Y + 2\text{Cov}(X,Y) \\  4) & \text{Var}(X + Y) &=& \text{Var}X + \text{Var}Y & \text{ if }X,Y \text{ are independent} \end{array} $$
Do đó với $X, Y$ độc lập, ta có

$$ \begin{array}{crll} \text{Var}(Z) &=& \text{Var}(3X+2Y-5) \\ &=& \text{Var}(3X + 2Y) & \text{[ Theo (1)]} \\ &=& \text{Var}(3X) + \text{Var}(2Y) & \text{[ Do (4)]} \\ &=& 9 \text{Var}X + 4 \text{Var}Y & \text{[ Theo (2)]} \\ &=& 43. \end{array} $$

#### Question 326.
An actuary is studying hurricane models. 
- A year is classified as a high, medium, or low hurricane year with probabilities 0.1, 0.3, and 0.6, respectively. 
- The numbers of hurricanes in high, medium, and low years follow Poisson distributions with means 20, 15, and 10, respectively.

Calculate the variance of the number of hurricanes in a randomly selected year

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $H,M,L$ lần lượt tương ứng với quy mô cơn bão là `High, Medium` và `Low`, theo đó

$$ P[H] = 0.1, \qquad P[M] = 0.3;\qquad P[L] = 0.6. $$
- Đặt $N$ là số cơn bão xảy ra trong năm, theo đó

$$ (N|H) \sim \text{Pois}(20); \qquad (N|M) \sim \text{Pois}(15); (N|L) \sim \text{Pois}(10) $$
- BT yêu cầu tính $\text{Var}N$; ta có

$$ \begin{array}{cccrl} \mathbb{E}N &=& \mathbb{E}(N|H) \cdot P(H) + \mathbb{E}(N|M) \cdot P[M] + \mathbb{E}(N|L) \cdot P[L] \\ &=& (20) \cdot (0.1) + (15) \cdot (0.3) + (10) \cdot (0.6) &=& 12.5 \end{array} $$
tương tự
$$ \begin{array}{cccrl} \mathbb{E}N^2 &=& \mathbb{E}(N|H)^2 \cdot P(H) + \mathbb{E}(N|M)^2 \cdot P[M] + \mathbb{E}(N|L)^2 \cdot P[L] \\ &=& \left( 20^2 + 20 \right) \cdot (0.1) + \left(15^2 + 15\right) \cdot (0.3) + \left(10^2 + 10 \right) \cdot (0.6) &=& 180. \end{array} $$
do đó
$$ \text{Var}N = \mathbb{E}N^2 - [\mathbb{E}N]^2 = 23.75. $$

#### Question 327.
- A dental insurance company pays 100% of the cost of fillings and 70% of the cost of root canals. 
- Fillings and root canals cost 50 and 500 each, respectively.
- The tables below show the probability distributions of the annual number of fillings and annual number of root canals for each of the company’s policyholders.

| # of Fillings | $0$ | $1$ | $2$ | $3$ |
|-|-|-|-|-|
|Probability| $0.60$ | $0.20$ | $0.15$ | $0.05$ |

and

| # Root of Cannals | $0$ | $1$ |
|-|:-:|:-:|
|Probability| $0.80$ | $0.20$ |

Calculate the expected annual payment per policyholder for fillings and root canals.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $F, R$ lần lượt là số lần trám răng (`# of fillings`) và số lần lấy tủy răng (`# root of cannals`).
- Vì chi phí sẽ được tính là $50$ cho mỗi lần trám răng và $100$ cho mỗi lần lấy tủy; nên tổng chi phí sẽ được xác định bởi

$$ T = 50F + 100R $$
- BT yêu cầu tính kỳ vọng về chi phí hàng năm tại phòng nha khoa này, tức là $\mathbb{E}(T)$
- Dựa vào bảng phân phối xác suất của chúng, ta được

$$ \begin{array}{ccl} \mathbb{E}F &=& 0\cdot P[F=0] + 1\cdot P[F=1] + 2\cdot P[F=2] + 3\cdot P[F=3] = 0.65 \\ \mathbb{E}R &=& 0\cdot P[R=0] + 1\cdot P[R=1] = 0.2 \end{array} $$
do đó

$$ \mathbb{E}T = 50 \mathbb{E}F + 100 \mathbb{E}R = 102.5. $$

#### Question 328.
- This year, the number of accidents, $X,$ at an amusement park has probability generating function $P_X(t) = e^{-0.2(1-t)},$ where defined.
- The amusement park owner’s insurance company reimburses up to a maximum of one accident this year. 
- Let $Y$ be the number of unreimbursed accidents.

Determine the probability generating function, $P_Y(t)$, of the number of unreimbursed
accidents at this amusement park this year, where defined.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Nhắc lại rằng, hàm sinh xác suất (probability generating function) của BNN không âm $X$ sẽ được xác định bởi
$$ P_X(t) = \displaystyle \sum_{x=0}^{\infty} t^x P[X = x] $$
đặc biệt, khi $t = 0$ và theo hàm $_PX(t)$ như đề thì
$$ P[X = 0]= P_X(0) = e^{-0.2}. $$
- Vì $Y$ là số vụ tai nạn không được bồi thường, do đó

$$ \left\lbrace \begin{array}{crll} P[Y=0] &=& P[X \leq 1] \\ P[Y = y] &=& P[X = y + 1], & \forall y \geq 1 \end{array} \right. $$
do đó,
$$ \begin{array}{crll} P_Y(t) = \displaystyle \sum_{y=0}^{\infty} t^y P[Y = y]  &=& \displaystyle P[Y=0] + \sum_{y=1}^{\infty} t^y P[Y = y] \\ &=& \displaystyle \left( P[X=0] + P[X=1] \right) + \sum_{y=1}^{\infty} t^y P[Y = y] \\ &=& P[X= 0] + \left[ t^0 P[X=1] + \displaystyle \sum_{y=1}^{\infty} t^y P[Y = y] \right] \\ &=& \displaystyle P[X=0] + \sum_{y=0}^{\infty} t^y P[X = y+1] \\ &=& \displaystyle P[X=0] + \sum_{x=1}^{\infty} t^{x-1} P[X = x] & \text{ by letting } x = y + 1 \\ &=& \displaystyle P[X=0] + \dfrac{1}{t} \sum_{x=1}^{\infty} t^x P[X = x] \\ &=& \displaystyle P[X=0] + \dfrac{1}{t} \left( P_X(t) - P[X = 0] \right) \\ &=& e^{-0.2} - \dfrac{1}{t} \left( e^{-0.2(1-t)} - e^{-0.2} \right) \\ &=& \dfrac{e^{-0.2(t-1)} + (t-1)e^{-0.2}}{t} \end{array}  $$

#### Question 329.
- Let $H$ be the number of hurricanes in a year in Florida. 
- $H$ follows a Poisson distribution with mean 0.30.

Calculate the coefficient of $t^2$ in the probability generating function, $P_H(t)$

**Giải đáp / SOLUTIONs.**

- Theo đề, $H \sim \text{Pois}(0.3)$ và BT yêu cầu ta xác định hệ số của $t^2$ trong hàm sinh xác suất $P_H(t)$
- Xem lại định nghĩa trên bài 328, hệ số của $t^2$ ứng với bậc $x = 2$ nên nó chính là xác suất của $H$ tại đó: $P[H = 2]$,

$$ P[H = 2] = \dfrac{e^{-0.3}\cdot (0.3)^2}{2!} \approx 0.333 $$

#### Question 330.
A loss under a liability policy is modeled by an **exponential distribution**. 
- The insurance company will cover the amount of that loss in excess of a deductible of 2000. 
- The probability that the reimbursement is less than 6000, given that the loss exceeds the deductible, is 0.50.

Calculate the probability that the reimbursement is greater than 3000 but less than 9000, given that the loss exceeds the deductible. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là khoản lỗ của cty và khoản tiền mà cty sẽ bồi thường khi vượt mức khấu hao $2000$, theo đó 

$$X \sim \mathcal{E}(\lambda), \qquad Y = \max(0, X-d). $$
- Giả thiết cho biết $P[Y < 6000 | Y > 2000] = 0.5$ suy ra

$$ 1 - e^{-6000/\lambda} = 0.5 \Longrightarrow \lambda = 8656.17 $$
- BT yêu cầu tính $P[3000<Y<9000|Y>2000]$, do đó

$$ P[3000<Y<9000 | Y > 2000] = P[X < 9000] - P[X < 3000] = (1 - e^{-9000\lambda}) - (1 - e^{-3000\lambda}) \approx 0.35. $$
