#### Question 221, 
Insurance companies A and B each earn an annual profit that is normally distributed with
the same positive mean. The standard deviation of company A’s annual profit is one half
of its mean.

In a given year, the probability that company B has a loss (negative profit) is 0.9 times
the probability that company A has a loss.

Calculate the ratio of the standard deviation of company B’s annual profit to the standard
deviation of company A’s annual profit. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X,Y$ lần lượt là các BNN thể hiện mức lợi nhuận ở 2 cty BH $A, B$. Theo đó, chúng có phân phối chuẩn với cùng trung bình nhưng độ lệch chuẩn $\sigma_A$ ở cty $A$ bằng một nửa trung bình của nó $\mu_A$; như vậy 

$$\mu_B = \mu_A = 2\sigma_A.$$ Cụ thể
$$ X \sim \mathcal{N}\left( 2\sigma_A, \sigma_A^2 \right), \qquad Y \sim \mathcal{N}\left( 2\sigma_A, \sigma_B^2 \right) $$
- Giả thiết cuối cho biết XS mà cty B có tổn thất (lợi nhuận âm) bằng 0.9 khi cty A bị âm lợi nhuận; tức là

$$ P[Y < 0] = 0.9 P[X < 0] \Leftrightarrow P \left[ \mathcal{N}(0,1) < \dfrac{-2\sigma_A}{\sigma_B} \right] = 0.9 P \left[ \mathcal{N}(0,1) < \dfrac{-2\sigma_A}{\sigma_A} \right] $$
tức là
$$ \Phi \left( \dfrac{-2\sigma_A}{\sigma_B} \right) = 0.9 \Phi \left( -2 \right) = 0.9 (0.0228) = 0.02052 $$
- BT yêu cầu ta tính tỷ số của 2 độ lệch chuẩn từ 2 cty; tức là $\dfrac{\sigma_B}{\sigma_A}$. Ta có
$$ \begin{array}{ccl} \Phi \left( \dfrac{-2\sigma_A}{\sigma_B} \right) = 0.02052 & \Leftrightarrow & -\dfrac{-2\sigma_A}{\sigma_B} = \Phi^{-1}(0.0252) = -2.04 \\ &\Leftrightarrow & \dfrac{\sigma_B}{\sigma_A} = \dfrac{1}{1.02} \approx 0.98 \end{array} $$
- Do đó, tỷ số cần tìm bằng $0.98$

#### Question 222.
Let $X$ represent the number of policies sold by an agent in a day. The moment generating function of $X$ is

$$ M_X(t) = 0.45e^t + 0.35e^{2t} + 0.15e^{3t} + 0.05 e^{4t} $$
Calculate the standard deviation of $X.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Nhắc lại rằng, nếu một BNN rời rạc $X$ nhận các giá trị trong $\Omega$ thì hàm sinh moment của nó sẽ có dạng một đa thức của các hàm e mũ; thật vậy

$$ M_X(t) = \mathbb{E} e^{Xt} = \displaystyle \sum_{x \in \Omega} e^{tx} P[X = x] $$
- Ta thấy rằng hàm sinh moment của $X$ có các bậc $e^t$ đến $e^{4t}$ nên $X$ là BNN rời rạc nhận các giá trị $x\in \Omega = \lbrace 1,2,3,4 \rbrace$ và các xác suất tương ứng, cụ thể

|$x$|1 | 2 | 3 | 4 |
|-|-|-|-|-|
|$P(X=x)$| 0.45 | 0.35 | 0.15| 0.05|

- BT yêu cầu tính độ lệch chuẩn của $X$, tức là $$\text{sd}X = \sqrt{\text{Var}X}$$
ta có
$$ \begin{array}{ccl} \text{Var}X &=& \mathbb{E}X^2 - [\mathbb{E}X]^2 \\ &=&  \displaystyle \left( \sum_{x=1}^4 x^2 P[X = x] \right) - \left( \sum_{x=1}^4 x P[X = x] \right)^2 \\ &=& \left( 0.45 + 2^2(0.35) + 3^2(0.15) + 4^2(0.05) \right) - \left( 0.45 + 2(0.35) + 3(0.15) + 4(0.05) \right)^2 \\ &=& 4 - (1.8)^2 \\ &=& 0.76 \end{array} $$
do vậy,
$$\text{sd}X = \sqrt{\text{Var}X} = 0.87.$$


#### Question 223.
Claim amounts at an insurance company are independent of one another. In year one,
claim amounts are modeled by a normal random variable X with mean 100 and standard
deviation 25. In year two, claim amounts are modeled by the random variable
$Y = 1.04X + 5.$

Calculate the probability that a random sample of 25 claim amounts in year two average
between 100 and 110. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ giả thiết, $X,Y$ lần lượt là các BNN độc lập thể hiện số các yêu cầu trong năm thứ nhất và năm thứ hai; theo đó 

$$X \sim \mathcal{N}(100, 25^2), \qquad Y = 1.04X+5 $$
- BT yêu cầu tính XS để trung bình một mẫu ngẫu nhiên $25$ yêu cầu trong năm 2 rơi vào khoảng $[100, 110]$.
- Đặt $$Z = \dfrac{1}{25} \displaystyle \sum_{i=1}^25 Y_i,$$ khi đó $Z$ chính là trung bình mẫu 25 yêu cầu trong năm thứ hai với $Y_i$ là yêu cầu thứ $i^{th}$ trong năm 2. Như vậy, ta cần tính 

$$ P[100 < Z < 110] $$
- Ta có

$$ Y = 1.04X + 5 \sim \mathcal{N}\left( (1.04)\mathbb{E}X + 5, (1.04)^2 \text{Var}X \right) = \mathcal{N} \left( 109, 26^2 \right) $$
và
$$ Y_i \overset{\text{i.i.d}}{\sim} Y \Longrightarrow Z \sim \mathcal{N} \left( 109, \dfrac{26^2}{25} \right)  $$
Do đó

$$ \begin{array}{ccl} P[100 < Z < 110] &=& P \left( \dfrac{100 - \mathbb{E}Z}{\sqrt{\text{Var}Z}} < \mathcal{N}(0,1) < \dfrac{110 - \mathbb{E}Z}{\sqrt{\text{Var}Z}} \right) \\ &=& \Phi \left( \dfrac{110 - 109}{\dfrac{26}{5}} \right) - \Phi \left( \dfrac{100 - 109}{\dfrac{26}{5}} \right) \\ &=& \Phi(0.19) - \Phi(-1.73) \\ &=& \Phi(0.19) - [ 1 - \Phi(1.73)] \\ &=& 0.5753 - (1 - 0.9582) \\ &=& 0.5335. \end{array} $$

#### Question 224.
Random variables $X$ and $Y$ are uniformly distributed on the region bounded by the x and y
axes, and the curve $y = 1 - x^2$. 

Calculate $\mathbb{E}XY$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $f$ là hàm MDXS của cặp BNN có phân phối đều $(X, Y)$ trên miền $A$ giới hạn bởi các trục tọa độ và đường cong $y = 1-x^2$; khi đó

$$ f(x,y) = \dfrac{1}{|A|} \textbf{1}_{\lbrace (x,y) \in A \rbrace} $$

trong đó $|A|$ là diện tích của miền $A$ xác định bởi

$$ \begin{array}{ccl} A &=& \left \lbrace (x,y): x \geq 0, y \geq 0, y = 1 - x^2 \right \rbrace \\ &=& \left\lbrace (x,y): 0\leq y \leq 1-x^2 \quad  \text{ and } \quad x \in [0, 1] \right\rbrace \end{array} $$
- BT yêu cầu tính $\mathbb{E}XY$ nên trước tiên ta cần xác định chính xác hàm MDXS của $f$, do đó cần tìm cụ thể giá trị diện tích của miền $A$; ta có
$$ |A| = \displaystyle \iint_A 1 dydx = \int_0^1 \int_0^{1-x^2} dy dx = \int_0^1 \left( 1-x^2 \right)dx = \dfrac{2}{3}. $$
do đó
$$ f(x,y) = \dfrac{1}{|A|} \textbf{1}_{\lbrace (x,y) \in A \rbrace} = \dfrac{3}{2} \textbf{1}_{\lbrace (x,y) \in A \rbrace} $$
và
$$ \begin{array}{ccl} \mathbb{E}(XY) &=& \displaystyle \iint_A xy f(x,y)dy dx \\ &=&  \displaystyle \dfrac{3}{2} \int_0^1 \int_0^{1-x^2} xy dy dx \\ &=& \dfrac{3}{2} \displaystyle \int_0^1 \left.\left( \dfrac{xy^2}{2} \right)\right\vert_{y=0}^{1-x^2} dx \\ &=&  \dfrac{3}{2} \displaystyle \int_0^1 \left( \dfrac{x(x^2 - 1)^2}{2} \right) dx \\ &=& \dfrac{1}{8} \end{array} $$

#### Question 225.
An insurance company will cover losses incurred from tornadoes in a single calendar
year. However, the insurer will only cover losses for a maximum of three separate
tornadoes during this timeframe. 

Let $X$ be the number of tornadoes that result in at least 50 million in losses, and let $Y$ be the total number of tornadoes. The joint probability
function for $X$ and $Y$ is

$$ f(x,y) = \left\lbrace \begin{array}{cl} c(x+2y), & x,y \in \lbrace 0,1,2,3 \rbrace \text{ such that } x \leq y \\ 0, & \text{otherwise} \end{array} \right. $$
where $c$ is a constant.

Calculate the expected number of tornadoes that result in fewer than 50 million in losses.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo đề, cty chỉ trả tiền BH cho tối đa 3 cơn bão trong suốt thời gian này, nên $(X, Y)$ phải thỏa

$$ \begin{array}{ccl} (x,y) \in \Omega &=& \lbrace x,y = 0,1,2,3 : 0 \leq x \leq y \leq 3 \rbrace \\ &=& \lbrace (0,0),(0,1),(0,2),(0,3); \;(1,1), (1,2),(1,3);\; (2,2), (2,3) \rbrace \end{array} $$
với hàm MDXS như trong đề, ta có
$$ \begin{array}{ccrcl} \displaystyle \sum_{(x,y) \in \Omega} P[(X,Y) = (x,y)] = 1 & \Leftrightarrow & \displaystyle \sum_{(x,y) \in \Omega} c(x+2y) =1 \\ & \Leftrightarrow & c \left[ 0 + 2 + 4 + 6 + (1 + 2) + (1 + 4) + (1 + 6) + (2 + 4) + (2 + 6) + (3 + 6) \right] &=& 1 \\ & \Leftrightarrow & c &=& \dfrac{1}{50} \end{array} $$
- BT yêu cầu tính kỳ vọng số cơn bão $Y$ ít hơn 50 (triệu) so với tổng thiệt hại $X$, tức là
$$ \begin{array}{ccl} \mathbb{E}[Y - X |  X \geq 50] &=& \displaystyle \sum_{(x, y) \in \Omega} (y - x) P[(X, Y) = (x,y)] \\ &=& \displaystyle \sum_{(x, y) \in \Omega} (y - x) [c(x + 2y)] \\ &=& \dfrac{1}{50} \left[ 0(0) ++ 1(2) + 2(4) + 3(6) + 0(3) + 1(5) + 2(7) + 0(6) + 1(8) + 0(9) \right] \\ &=& 1.1 \end{array} $$

#### Question 226.
At a polling booth, ballots are cast by ten voters, of whom three are Republicans, two are
Democrats, and five are Independents. A local journalist interviews two of these voters,
chosen randomly.

Calculate the expectation of the absolute value of the difference between the number of
Republicans interviewed and the number of Democrats interviewed. 

**Giải đáp / SOLUTIONs.**

Đặt $I, R, D$ lần lượt tương ứng với các sự kiện chọn được người bỏ phiếu cho đảng **độc lập `(Independent)`**, **cộng hòa `(Republican)`** và **dân chủ `(Democrats)`**.

BT cho biết nếu chọn ngẫu nhiên 2 người trong đó và tính kỳ vọng cho hiệu số tuyệt đối giữa Đảng Cộng hòa và phe Dân chủ, tức là $\mathbb{E}|R-D|$

Đặt $X_1, X_2$ tương ứng với sự kiện về kết quả chọn được người phỏng vấn trong lần thứ nhất và lần thứ hai; khi đó,

$\qquad \diamond$ TH1, $|R-D| = 0$, khi đó $(X_1, X_2) = \lbrace (R,D), (D, R), (I, I) \rbrace$ và TH này có XS bằng

$$ \begin{array}{ccl} P[|R-D| = 0] &=& P[X_1 = R, X_2 = D] + P[X_1 = D, X_2 = R] + P[X_1 = I, X_2 = I] \\ &=& \dfrac{3}{10} \cdot \dfrac{2}{9} +  \dfrac{2}{10} \cdot \dfrac{3}{9} + \dfrac{5}{10} \cdot \dfrac{4}{9} \\ &=& \dfrac{16}{45} \end{array} $$

$\qquad \diamond$ TH2, $|R-D| = 1$, khi đó $(X_1, X_2) = \lbrace (R,I), (D, I) \rbrace$ và TH này có 

$$ \begin{array}{ccl} P[|R-D| = 1] &=& P[X_1 = R, X_2 = I] + P[X_1 = I, X_2 = R] + P[X_1 = D, X_2 = I] + P[X_1 = I, X_2 = D] \\ &=& \dfrac{3}{10} \cdot \dfrac{5}{9} +  \dfrac{5}{10} \cdot \dfrac{3}{9} + \dfrac{2}{10} \cdot \dfrac{5}{9} + \dfrac{5}{10} \cdot \dfrac{2}{9} \\ &=& \dfrac{5}{9} \end{array} $$

$\qquad \diamond$ TH3, $|R-D| = 2$, khi đó $(X_1, X_2) = \lbrace (R,R), (D, D) \rbrace$ và TH này có 

$$ \begin{array}{ccl} P[|R-D| = 2] &=& P[X_1 = R, X_2 = R] + P[X_1 = D, X_2 = D] \\ &=& \dfrac{3}{10} \cdot \dfrac{2}{9} + \dfrac{2}{10} \cdot \dfrac{1}{9} \\ &=& \dfrac{8}{90} \end{array} $$

Suy ra

$$ \mathbb{E}|R-D| = \displaystyle \sum_{k = 0}^2 k P[|R-D| = k] = \dfrac{16}{45} \cdot 0 + \dfrac{5}{9} \cdot 1 + \dfrac{8}{90} \cdot 2 = \dfrac{11}{15}$$

---------------------

**Giải thích.** KGMẫu ta có 10 phần tử, phép chọn này là chọn không hoàn lại nên cứ mỗi lần rút ra thì phần tử còn lại trong mẫu sẽ giảm đi một và phần tử lấy ra cũng biến mất; cụ thể
- $P[X_1 = R, X_2 = D]$ là XS mà lần đầu chọn $R$ (có $3R$ trong $10$ khả năng: $3R, 2D,5I$), lần hai chọn $D$ (bỏ lần đầu và $D$ có $2$ trong $9$ khả năng còn lại: $2R, 2D, 5I$) nên KQ sẽ được tính bởi $$P[X_1 = R, X_2 = D] = P[X_1 = R] \cdot P[X_2 = D | X_1 = R] = \dfrac{3}{10} \cdot \dfrac{2}{9}.$$
- $P[X_1 = I, X_2 = I]$ là XS mà lần đầu chọn $I$ (có 5 trong 10 khả năng), lần hai chọn $I$ (bỏ lần đầu và $I$ còn 4 $I$ trong 9 khả năng còn lại: $2R, 3D, 4I$) nên KQ sẽ được tính bởi $\dfrac{5}{10} \cdot \dfrac{4}{9}$.

#### Question 227.
The random variables X and Y have joint probability function $p(x,y)$ for $x = 0,1$ and $y = 0,1,2.$

Suppose $3p(1,1) = p(1,2)$, and $p(1,1)$ maximizes the variance of $XY$.

Calculate the probability that $X$ or $Y$ is $0. $

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Z = XY$ thì $Z$ chỉ có thể nhận các giá trị $\lbrace 0,1,2 \rbrace$ do $X \in \lbrace 0,1 \rbrace$ và $Y \in \lbrace 0,1,2 \rbrace$.
- Đặt
$$ a = P[Z=0];\quad b = P[Z=1];\quad c = P[Z=2] $$
khi đó
$$ c = p(1, 2), \quad b = p(1,1), \quad a = p(1,0)+p(0,1)+p(0,2) $$
- Theo giả thiết,
$$ \left\lbrace \begin{array}{ccl} a+b+c &=& 1 \\ 3b &=& c \end{array} \right. $$

$\quad$ và $Z = XY$ đạt cực đại tại $b = p(1,1)$; vì vậy, trước tiên ta sẽ biểu diễn $\text{Var}Z$ theo $b$.
- Ta có
$$ \begin{array}{cclclrl} \mathbb{E}Z &=& \displaystyle \sum_{k=0}^2 kP[Z=k] &=& b + 2c & = & 7b \\ \mathbb{E}Z^2 &=& \displaystyle \sum_{k=0}^2 k^2 P[Z=k] &=& b + 4c &=& 13b \end{array} $$
suy ra
$$ \text{Var}Z = \mathbb{E}Z^2 - [ \mathbb{E}Z ]^2 = 13b - (7b)^2 = 13b - 49b^2. $$

- Để $\text{Var}Z$ đạt cực đại theo $b$ thì biểu thức $13b - 49b^2$ phải đạt cực đại, do đó $b$ là nghiệm của phương trình

$$ \left\lbrace \begin{array}{ccl} \dfrac{d}{db} \left( 13b - 49b^2 \right) &=& 0 \\ \dfrac{d^2}{db^2} \left( 13b - 49b^2 \right) &<& 0 \end{array} \right. \Longrightarrow b = \dfrac{13}{98} $$
- BT yêu cầu tính XS khi $X$ hoặc $Y$ bằng 0; tương đương với $P[Z=0]$, ta có

$$ P[Z=0] = a = 1 - b - c = 1 - 4b = \dfrac{23}{49}. $$

#### Question 228.
As a block of concrete is put under increasing pressure, engineers measure the pressure $X$
at which the first fracture appears and the pressure $Y$ at which the second fracture
appears. $X$ and $Y$ are measured in tons per square inch and have joint density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} 24x(1 - y), & 0 < x < y < 1 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the average pressure (in tons per square inch) at which the second fracture
appears, given that the first fracture appears at 1/3 ton per square inch. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Do $X,Y$ lần lượt tương ứng với vết nứt (`fracture`) đầu tiên và thứ hai, như vậy BT yêu cầu tính $\mathbb{E}[Y | X = 1/3]$
- Theo đó, trước tiên ta cần xác định hàm MDXS có điều kiện $f_{\left( Y|X>\frac{1}{3} \right)}(y)$ cho $Y$ khi biết $X = 1/3$, xác định bởi

$$ f_{\left( Y|X=\frac{1}{3} \right)}(y) = \dfrac{f\left( \frac{1}{3}, y \right)}{P\left[ X > \frac{1}{3} \right]} $$
- Ta có
$$ \begin{array}{ccl} f\left( \frac{1}{3}, y \right) & = & 24 \cdot \dfrac{1}{3} \cdot (1-y) \cdot \textbf{1}_{\left\lbrace \frac{1}{3} < y < 1 \right\rbrace} = 8(1-y) \textbf{1}_{\left\lbrace \frac{1}{3} < y < 1 \right\rbrace} \\ P\left[ X > \frac{1}{3} \right] &=& \displaystyle \int_{1/3}^1 f\left( \frac{1}{3}, y \right) dy = 8 \displaystyle \int_{1/3}^1 (1-y)dy = \dfrac{16}{9} \end{array} $$
nên

$$ f_{\left( Y|X=\frac{1}{3} \right)}(y) = 4.5 (1-y) \textbf{1}_{\left\lbrace \frac{1}{3} < y < 1 \right\rbrace}. $$
- Do đó,

$$ \mathbb{E}[Y | X = 1/3] = \displaystyle \int y f_{\left( Y|X=\frac{1}{3} \right)}(y) dy = (4.5) \int_{1/3}^1 y(1-y)dy = \dfrac{5}{9}. $$

#### Question 229.
The number of severe storms that strike city $J$ in a year follows a binomial distribution
with $n = 5$ and $p = 0.6$. 

Given that $m$ severe storms strike city $J$ in a year, the number
of severe storms that strike city $K$ in the same year is $m$ with probability 1/2, $m +1$
with probability $1/3$, and $m + 2$ with probability $1/6.$

Calculate the expected number of severe storms that strike city $J$ in a year during which 5
severe storms strike city $K.$


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Theo đề, $J$ là số cơn bão nghiêm trọng xuất hiện ở thành phố $J$ và $J \sim \mathcal{N}(n,p)$ với $n=5$ và $p=0.6$, như vậy

$$ P[J = x] = C^x_n p^x (1-p)^{n-x}, \qquad \forall x = 0,1,\ldots,n. $$
- Giả thiết kế tiếp cho biết:
$$ P[K = m| J = m] = \dfrac{1}{2}, \quad P[K = m+1| J = m] = \dfrac{1}{3}, \quad P[K = m+2| J = m] = \dfrac{1}{6}; \qquad \forall m \in \mathbb{N}. $$
tương đương với

$$ P[K = m| J = m] = \dfrac{1}{2}, \quad P[K = m| J = m-1] = \dfrac{1}{3}, \quad P[K = m| J = m-2] = \dfrac{1}{6}; \qquad \forall m \geq 2. $$
- BT yêu cầu ta tính $\mathbb{E}[J | K=5]$, ta có
$$ \mathbb{E}[J | K=5] = \displaystyle \sum_{x=3}^5 x P[J=x | K = 5] = \sum_{x=3}^5 x \left( \dfrac{P[K=5|J=x] \cdot P[J=x]}{P[K=5]} \right) $$
- Từ các giả thiết trên, ta có
$$ \begin{array}{rcl} P[K=5] &=& \displaystyle \sum_{x=3}^5 P[K=5|J=x] \cdot P[J=x] = \left( \dfrac{1}{6}\cdot C_5^3 0.6^3 0.4^2 + \dfrac{1}{3}\cdot C_5^4 0.6^4 0.4 + \dfrac{1}{2}\cdot C_5^5 0.6^5 \right) = 0.18288 \\ P[J=3|K=5] &=& \dfrac{P[K=5|J=3] P[J=3]}{P[K=5]} = \dfrac{\dfrac{1}{6}\cdot C_5^3 0.6^3 0.4^2}{0.18288} \approx 0.31496 \\  P[J=4|K=5] &=& \dfrac{P[K=5|J=4] P[J=4]}{P[K=5]} = \dfrac{\dfrac{1}{3}\cdot C_5^4 0.6^4 0.4}{0.18288} \approx 0.47244 \\  P[J=5|K=5] &=& \dfrac{P[K=5|J=5] P[J=5]}{P[K=5]} = \dfrac{\dfrac{1}{2}\cdot C_5^5 0.6^5}{0.18288} \approx 0.21260 \end{array} $$

suy ra

$$ \mathbb{E}[J | K=5] = 3(0.31496) + 4(0.47244) + 5(0.21260) = 3.89764 . $$

#### Question 230.

Let 
- $X$ denote the **proportion of employees at a large firm who will choose to be covered
under the firm’s medical plan**, and 
- $Y$ denote the **proportion who will choose to be
covered under both the firm’s medical and dental plans**.

Suppose that for $0 \leq y \leq x \leq 1,$ $X$ and $Y$ have the joint cumulative distribution function

$$ F(x,y) = y(x^2 + xy - y^2) $$

Calculate the expected proportion of employees who will choose to be **covered under
both plans.**

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu tính kỳ vọng cho tỷ lệ những nhân viên sẽ chọn thực thi cả 2 phương án, tức là $\mathbb{E}Y$ (nên nhớ $X$ trong bài chỉ nói về những nhân viên chọn phương án về y tế.)
- Để tính $\mathbb{E}Y$ ta cần xác định được hàm MDXS của $Y$ từ hàm phân phối tích lũy đồng thời $F(x,y)$.
- Vì miền xác định của $(x,y)$ là 

$$ \Omega = \lbrace (x,y): 0 \leq y \leq x \leq 1 \rbrace$$
nên hàm phân phối tích lũy lề $F_Y(y)$ của $Y$ sẽ xác định bởi

$$ F_Y(y) = F(1, y) = y + y^2 - y^3, \qquad y \in (0, 1) $$
suy ra hàm MDXS $f_Y(y)$ của $Y$
$$ f_Y(y) = \dfrac{dF_Y}{dy} = 1 + 2y - 3y^2. $$
- Do đó,

$$ \mathbb{E}Y = \displaystyle \int y f_Y(y)dy = \int_0^1 y \left( 1 + 2y - 3y^2 \right) dy = \dfrac{5}{12}. $$
