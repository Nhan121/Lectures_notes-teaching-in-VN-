#### Question 261, 
An insurance company studies back injury claims from a manufacturing company. The
insurance company finds that 40% of workers do no lifting on the job, 50% do moderate
lifting and 10% do heavy lifting.

During a given year, the probability of filing a claim is 0.05 for a worker who does no
lifting, 0.08 for a worker who does moderate lifting and 0.20 for a worker who does
heavy lifting.

A worker is chosen randomly from among those who have filed a back injury claim.
Calculate the probability that the worker’s job involves moderate or heavy lifting. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C, N, M, H$ lần lượt là các sự kiện **"đã gửi yêu cầu", "không nâng đồ vật (no lifting)", "có nâng vật vừa phải (moderate)" & "nâng vật nặng (heavy lifting)"**
- BT yêu cầu tính XS mà khi chọn ngẫu nhiên một người thì công nhân đó có nâng vật vừa hoặc vật nặng biết rằng người này đã gửi yêu cầu hợp đồng BH, tức là $ P[(M \cap H) | C] $
- Giả định trong đoạn thứ hai cho biết 
$$ P[C|N] = 0.05, \qquad P[C|M] = 0.08; \qquad P[C|H] = 0.2 $$
- Theo giả thiết trong đoạn đầu tiên, ta có
$$ P[N] = 0.4, \qquad P[M] = 0.5 , \qquad P[H] = 0.1 $$
và bộ ba sự kiện $N, H, M$ đã tạo thành một hệ đầy đủ (hệ các sự kiện rời nhau có tổng xác suất bằng một); do đó

$$ \begin{array}{ccl} P[(M \cap H) | C] &=& 1 - P[N | C] \\ &=& 1 - \dfrac{P[C|N] \cdot P[N]}{P[C|N] \cdot P[N] + P[C|M] \cdot P[M] + P[C|H] \cdot P[H]} \\ &=& 0.75 \end{array} $$

#### Question 262.
The number of traffic accidents occurring on any given day in Coralville is Poisson
distributed with mean 5. The probability that any such accident involves an uninsured
driver is 0.25, independent of all other such accidents.

Calculate the probability that on a given day in Coralville there are no traffic accidents
that involve an uninsured driver. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $U, N$ lần lượt là **"nếu tài xế không mua BH xảy ra tai nạn"** và **"số vụ tai nạn xảy ra"**, theo đó

$$ P[U | X = 1] = 0.25, \qquad N \sim \text{Pois}(\lambda = 5) $$
hay
$$ \begin{array}{ccl} P[U^c| X = k] &=& P[\text{"0 accidents that involve 1 uninsured driver" | "k accidents"}] = (0.75)^k \\ P[N = k] &=& e^{-5} \dfrac{5^k}{k!}; \qquad \forall k = 0,1,2,\ldots \end{array} $$
- BT yêu cầu **tinh xác suất để vào một ngày nhất định ở Coralville không có vụ tai nạn giao thông nào liên quan đến người lái xe không có bảo hiểm**; tức là

$$ \begin{array}{ccl} P[U^c] &=& \displaystyle \sum_{k=0}^{\infty} P[U^c | X = k] \cdot P[X = k] \\ &=& \displaystyle \sum_{k=0}^{\infty} (0.75)^k \cdot \left( e^{-5} \dfrac{5^k}{k!} \right) \\ &=& e^{-5} \left( \displaystyle \sum_{k=0}^{\infty} \dfrac{(3.75)^k}{k!} \right) \\ &=& e^{-5} \cdot e^{-3.75} \\ &=& 0.287. \end{array} $$

#### Question 263.
A group of 100 patients is tested, one patient at a time, for three risk factors for a certain
disease until either all patients have been tested or a patient tests positive for more than
one of these three risk factors. 
- For each risk factor, a patient tests positive with
probability $p$, where $0 < p < 1$. 
- The outcomes of the tests across all patients and all risk
factors are mutually independent.

Determine an expression for the probability that exactly n patients are tested, where n is a
positive integer less than 100.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $Y_k$ là kết quả kiểm định triệu chứng thứ $k$ có dương tính hay không trong 3 triệu chứng, ta có $Y_k \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, p)$ và $$ Y = \displaystyle \sum_{i=k}^3 Y_k \sim \mathcal{B}(3, p)$$
là số lần dương tính trong 3 kiểm định các triệu chứng.
- Một bệnh nhân được chẩn đoán là có bệnh (certain disease) nếu có nhiều hơn 1 trong 3 triệu chứng (risk factor); do đó XS để một bệnh nhân mắc bệnh là $$ \begin{array}{ccl} \theta &=& P[Y > 1] \\ &=& P[Y=2] + P[Y=3] \\ &=& C_3^2 p^2 (1-p) + C_3^3 p^3 \\&=& 3p^2(1-p) + p^3. \end{array}$$
- Đặt $X_i$ là kết quả khám nghiệm 3 triệu chứng của bệnh nhân thứ $i^{th}$ với

$$ X_i = \left\lbrace \begin{array}{cl} 1, & \text{ neu ket qua kiem dinh la co benh voi xac suat } \theta \\ 0, & \text{otherwise} \end{array} \right. $$

và $X$ là số lần thử để tìm được người có kết quả **có bệnh** trong $100$ bệnh nhân $X_1, \ldots, X_{100}$.
- Như vậy, ta có $X \sim \text{Geom}(\theta)$ và XS để tìm được đúng 1 bệnh nhân **có bệnh** trong $n$ bệnh nhân là

$$ P[X = n] = \theta (1 - \theta)^{n-1} = \left[ 3p^2(1-p) + p^3 \right] \cdot \left[ 1 - \left( 3p^2(1-p) + p^3 \right) \right]. $$

#### Question 264.
A representative of a market research firm contacts consumers by phone in order to
conduct surveys. The specific consumer contacted by each phone call is randomly
determined. The probability that a phone call produces a completed survey is 0.25.

Calculate the probability that more than three phone calls are required to produce one
completed survey. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $$X_i = \left\lbrace \begin{array}{cl} 1, & \text{ neu cuoc dien thoai thu i hoan thanh khao sat }\\ 0, & \text{otherwise} \end{array} \right. $$ 

là kết quả cuộc điện thoại khảo sát thứ $i$ có hoàn thành hay không, biết XS để hoàn thành một cuộc khảo sát bằng $0.25$, tức là

$$ P[X_i = 1] = 0.25, \qquad \forall i $$
- Vì BT yêu cầu tính XS để có nhiều hơn 3 cuộc điện thoại được thực thi để hoàn thành khảo sát, đồng nghĩa với 3 cuộc gọi đầu tiên $X_1, X_2, X_3$ đều không hoàn thành khảo sát, như vậy XS để 3 cuộc gọi đầu tiên không hoàn thành bằng với

$$ P[X_1 = 0, X_2 = 0, X_3 = 0] = \displaystyle \prod_{k=1}^3 P[X_k = 0] = \prod_{k=1}^3 \left( 1 - P[X_k = 1] \right) = (1 - 0.25)^3 = 0.42. $$

#### Question 265.
Four distinct integers are chosen randomly and without replacement from the first twelve
positive integers. $X$ is the random variable representing the second smallest of the four
selected integers, and $p$ is the probability function of $X$.

Determine $p(x)$ for $x = 2,3,\ldots,10.$ 

**Giải đáp / SOLUTIONs.**

Đặt $\Omega = \lbrace 1,2, \ldots, 12 \rbrace$ là KG mẫu chứa $12$ số nguyên đầu tiên.

$X$ là số nhỏ thứ hai trong 4 số được chọn từ $\Omega$ nên $X$ chỉ nhận các giá trị $2,3, \ldots, 10$ vì:
- Nếu $X = 1$ thì đây là số nhỏ nhất nên không còn số nhỏ thứ hai nhỏ hơn nó.
- Nếu $X = 11, 12$ thì đây là 2 số lớn nhất trong $\Omega$ trong khi $X$ là số nhỏ thứ hai nên sẽ còn 2 số lớn hơn $11$ hoặc $12$; do đó điều này là không thể xảy ra.

Như vậy, với tập giá trị $X = x \in \lbrace 2,3, \ldots, 10 \rbrace$ thì ta sẽ có $x - 1$ lựa chọn cho số nhỏ hơn $x$ và $12 - x$ lựa chọn cho 2 số lớn hơn $x$; do đó

$$ \begin{array}{ccl} p(x) &=& \dfrac{|\text{smallest from (x - 1) numbers}| \cdot |\text{2 larger from (12 - x) numbers}|}{|\text{"select 4 numbers from" } \Omega|} \\ &=& \dfrac{C_{x-1}^1 C_{12-x}^2}{C_{12}^4} = \dfrac{\left( \dfrac{(x-1)(12 - x)(11 - x)}{2} \right)}{495} = \dfrac{(x-1)(12 - x)(11 - x)}{990} \end{array} $$

#### Question 266.
Losses due to burglary are exponentially distributed with mean 100.

The probability that a loss is between 40 and 50 equals the probability that a loss is
between $60$ and $r$, with $r > 60.$

Calculate $r.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là BNN thể hiện mức tổn thất từ các vụ trộm, theo đó $X \sim \mathcal{E}(100)$ với 

$$ F(x) = P[X < x] = 1 - e^{-x/100}; \qquad x \geq 0 $$
- BT yêu cầu tìm $r > 60$ biết rằng XS để mức tổn thất trong khoảng $(40, 50)$ bằng với XS trên khoảng $(60, r)$; tức là tìm $r > 60$ thông qua phương trình

$$ P[40 < X < 50] = P[60 < X < r] $$
ta có

$$ \begin{array}{lrcl} & P[40 < X < 50] &=& P[60 < X < r] \\ \Rightarrow & F(50) - F(40) &=& F(r) - F(60) \\ \Rightarrow & (1 - e^{-50/100}) - ( e^{-40/100}) &=& (1 - e^{-r/100}) - (1 - e^{-60/100}) \\ \Rightarrow & e^{-r/100} &=& e^{-60/100} - e^{-40/100} + e^{-50/100} \\ \Rightarrow & r &=& 72.36. \end{array} $$

#### Question 267.
The time until the next car accident for a particular driver is exponentially distributed
with a mean of 200 days.

Calculate the probability that the driver has no accidents in the next 365 days, but then
has at least one accident in the 365-day period that follows this initial 365-day period.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là thời gian cho đến khi có một vụ tai nạn xe oto xảy ra, theo đó $X \sim \mathcal{E}(200)$, với

$$ F(x) = P[X < x] = 1 - e^{-x/200}, \qquad \forall x \geq 0. $$
- BT yêu cầu tính xác suất để người lái xe không bị tai nạn trong 365 ngày tới $(X > 365)$, nhưng sau đó bị ít nhất một tai nạn trong khoảng thời gian 365 ngày tiếp theo khoảng thời gian 365 ngày ban đầu này $(X < 365 + 365)$; tức là

$$ \begin{array}{ccl} P[(X > 365) \cap (X < 365 + 365) ] &=& P[365 < X < 730] \\ &=& F(730) - F(365) \\ &=& \left( 1 - e^{-730/200} \right) - \left( 1 - e^{-365/200} \right) \\ &=& 0.1352. \end{array} $$

#### Question 268.
- The annual profit of a life insurance company is normally distributed.
- The probability that the annual profit does not exceed 2000 is 0.7642. The probability
that the annual profit does not exceed 3000 is 0.9066.

Calculate the probability that the annual profit does not exceed 1000.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là lợi nhuận hàng năm của cty BH đó, theo giả thiết thì $X \sim \mathcal{N}(\mu, \sigma^2)$ với trung bình $\mu$, phương sai $\sigma^2$ chưa biết.
- Biết rằng **XS $X < 2000$ bằng 0.7642** và **XS để $X < 3000$ = 0.9066**, từ đây ta được hệ

$$ \left\lbrace \begin{array}{cclrl} 0.7642 &=& P[X < 2000] &=& P\left[ \mathcal{N}(0,1) < \dfrac{2000 - \mu}{\sigma} \right] = \Phi \left( \dfrac{2000 - \mu}{\sigma} \right) \\ 0.9066 &=& P[X < 3000] &=& P\left[ \mathcal{N}(0,1) < \dfrac{3000 - \mu}{\sigma} \right] = \Phi \left( \dfrac{3000 - \mu}{\sigma} \right) \end{array} \right. $$
suy ra
$$ \left\lbrace \begin{array}{cclrl} \dfrac{2000 - \mu}{\sigma} &=& \Phi^{-1}(0.7642) &=& 0.72 \\ \dfrac{3000 - \mu}{\sigma} &=& \Phi^{-1}(0.9066) &=& 1.32 \end{array} \right.  $$
do đó $ \left\lbrace \begin{array}{ccl} \mu &=& 800 \\ \sigma &=& 1666.67 \end{array} \right.$

- BT yêu cầu tính XS để $X < 1000$; tức là
$$ P[X < 1000] = \Phi \left( \dfrac{1000 - \mu}{\sigma} \right) = \Phi(0.12) \approx 0.5478. $$

#### Question 269.
Individuals purchase both collision and liability insurance on their automobiles. The
value of the insured’s automobile is $V$. Assume the loss $L$ on an automobile claim is a
random variable with cumulative distribution function 

$$ F(l) = \left\lbrace \begin{array}{cl} \dfrac{3}{4} \left( \dfrac{l}{V} \right)^3, & 0 \leq l \leq V \\ 1 - \dfrac{1}{10} e^{-\frac{(l-V)}{V}}, & l \geq V \end{array} \right. $$

Calculate the probability that the loss on a randomly selected claim is greater than the
value of the automobile. 

**Giải đáp / SOLUTIONs.**

- BT yêu cầu tính **xác suất để tổn thất trên một yêu cầu được chọn ngẫu nhiên lớn hơn giá trị của ô tô.** Tức là $P[L > V]$
- Giả thiết đã cho $F(l)$ là hàm phân phối tích lũy của mức tổn thất $L$, tức là 

$$ F(x) = P[L < x], \qquad \forall x \geq 0 $$ do đó

$$ P[L > V] = 1 - P[L \geq V] = 1 - \left( 1 - \dfrac{1}{10} e^{-\frac{(V - V)}{V}} \right) = \dfrac{1}{10} = 0.1 $$

#### Question 270.
The lifetime of a machine part is **exponentially distributed** with a **mean of five years**.

Calculate the mean lifetime of the part, given that it survives less than ten years.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $T$ là tuổi thọ của chi tiết máy (`machine part`), theo đó $T \sim \mathcal{E}(5)$ với hàm MDXS

$$ f(t) = \dfrac{1}{5} e^{-t/5} \textbf{1}_{\lbrace t \geq 0 \rbrace} $$
và hàm phân phối tích lũy
$$ F(t) = 1 - 0.2 e^{-0.2 t}, \qquad \forall t \geq 0 $$
- BT yêu cầu tính kỳ vọng có điều kiện $\mathbb{E}[T | T < 10]$, do đó trước tiên ta phải xác định được hàm mật độ XS $g$ của $(T | T < 10)$, ta có

$$ g(t) = \dfrac{f(t)}{P[T < 10]} = \dfrac{f(t)}{F(10)} = \dfrac{0.2 e^{-0.2t}}{1 - e^{-2}}, \qquad \forall t \in [0, 10) $$
do đó
$$ \mathbb{E}[T | T < 10] = \displaystyle \int t g(t) dt = \int_0^{10} \dfrac{0.2 t e^{-0.2t}}{1 - e^{-2}} dt \approx 3.435. $$
