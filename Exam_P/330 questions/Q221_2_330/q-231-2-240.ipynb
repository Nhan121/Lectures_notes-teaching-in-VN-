#### Question 231, 
Let $N$ denote the number of accidents occurring during one month on the northbound side of a highway and let S denote the number occurring on the southbound side.

Suppose that $N$ and S are jointly distributed as indicated in the table



|$ \begin{array}{ccl} N & & \\ .. & \ &  S =  \end{array} $| $0$ |$1$ |$2$ |$3$ or more |
|-|-|-|-|-|
|$0$| $0.04$| $0.06$| $0.10$| $0.04$|
|$1$| $0.10$| $0.18$| $0.08$| $0.03$|
|$2$| $0.12$| $0.06$| $0.05$| $0.02$|
|$3$ or more| $0.05$| $0.04$| $0.02$| $0.01$|


- Calculate $\text{Var} (N | N + S = 2).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề bài yêu cầu tính phương sai có điều kiện

$$\text{Var} (N | N + S = 2) = \mathbb{E}\left( N^2 | N + S = 2 \right) - \left[ \mathbb{E}(N | N + S = 2) \right]^2 .$$

trong đó $N, S$ là các bnn rời rạc với giá trị là các số nguyên lớn hơn hoặc bằng 0 và phân phối xác suất đồng thời như bảng đã cho, do đó
$$ \begin{array}{ccl} \mathbb{E}\left( N | N + S = 2 \right) &=& \displaystyle \sum_{n=0}^{2} nP[N = n| N+S = 2] \\ \mathbb{E}\left( N^2 | N + S = 2 \right) &=& \displaystyle \sum_{n=0}^{2} n^2P[N = n| N+S = 2] \end{array} $$

vì vậy; trước hết, ta cần tính các xác suất có điều kiện $P[N = n| N+S = 2]$ với $n = 0,1,2$. 
- Ta có
$$ \begin{array}{ccl} P[N = 0| N+S = 2] &=& \dfrac{P[(N= 0) \cap (N+S=2)]}{P[N+S=2]} \\ &=& \dfrac{P[N=0, S=0]}{P[N=0,S=2]+P[N=1,S=1]+P[N=2,S=2]} \\&=& \dfrac{0.1}{0.1 +0.18 + 0.12} \\ &=& 0.25\end{array} $$
Tương tự
$$ \begin{array}{cclcr} P[N = 1| N+S = 2] &=& \dfrac{0.18}{0.1 +0.18 + 0.12} &=& 0.45 \\ P[N = 1| N+S = 2] &=& \dfrac{0.12}{0.1 +0.18 + 0.12} &=& 0.30 \end{array}$$
suy ra
$$ \begin{array}{lclcr} \mathbb{E}\left( N | N + S = 2 \right) &=& 0(0.25) + 1(0.45) + 2(0.3) &=& 1.05 \\ \mathbb{E}\left( N^2 | N + S = 2 \right) &=& 0(0.25) + 1^2(0.45) + 2^2(0.3) &=& 1.65 \end{array} $$
do đó
$$\text{Var} (N | N + S = 2) = \left( 1.65 \right) - \left( 1.05 \right)^2 = 0.5475.$$

#### Question 232.
An insurance company sells automobile liability and collision insurance. 

Let $X$ denote the percentage of liability policies that will be renewed at the end of their terms and $Y$ the percentage of collision policies that will be renewed at the end of their terms. $X$ and $Y$ have the joint cumulative distribution function

$$ F(x,y) = \dfrac{xy(x+y)}{2000000}, \quad 0 \leq x, y \leq 100.$$

Calculate $\text{Var}(X).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu tính $\text{Var}X$ khi chỉ biết hàm phân phối tích lũy đồng thời $F(x,y)$, do đó ta cần xác định được hàm MDXS của $X$.
- Theo đó, ta sẽ tìm hàm MDXS $f_X$ thông qua hàm phân phối lề $F_X(x)$.
- Lưu ý rằng ở đây miền xác định của $(x,y)$ bị giới hạn bởi hình vuông

$$ \Omega = \lbrace (x,y): 0 \leq x,y \leq 100 \rbrace = \lbrace x: 0 \leq x \leq 100 \rbrace \times \lbrace y: 0 \leq y \leq 100 \rbrace $$
nên
$$ F_X(x) = F\left( x, y_{\max} \right) = F(x, 100) = \dfrac{100x(100+x)}{2000000}, \qquad x \in [0, 100] $$
suy ra
$$ f_X(x) = \dfrac{dF_X(x)}{dx} = \dfrac{x}{10000} + \dfrac{1}{200} $$
- Lúc này, mọi thứ sẽ trở nên dễ dàng
$$ \begin{array}{ccl} \mathbb{E}X &=& \displaystyle \int x f_X(x)dx = \int_0^{100} x\left( \dfrac{x}{10000} + \dfrac{1}{200} \right)dx = \dfrac{175}{3} \\ \mathbb{E}X^2 &=& \displaystyle \int x^2 f_X(x)dx = \int_0^{100} x^2\left( \dfrac{x}{10000} + \dfrac{1}{200} \right)dx = \dfrac{12500}{3} \end{array} $$
do đó

$$ \text{Var}X = \mathbb{E}X^2 - [ \mathbb{E}X ]^2 = 764. $$

#### Question 233.
A hurricane policy covers both water damage, $X,$ and wind damage, $Y,$ where $X$ and $Y$ have joint density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} 0.13e^{-0.5x-0.2y} - 0.06e^{-x-0.2y} - 0.06e^{-0.5x-0.4y} + 0.12e^{-x-0.4y} , & x >0, y >0 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the standard deviation of $X.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu xác định độ lệch chuẩn của $X$, theo đó ta sẽ tìm nó thông qua căn bậc 2 của phương sai

$$ \text{sd}X = \sqrt{\text{Var}X} = \sqrt{\mathbb{E}X^2 - [\mathbb{E}X]^2} $$
- Như vậy, trước tiên ta phải tìm được hàm MDXS của $X$ từ hàm MDXS đồng thời của nó; ta có

$$ \begin{array}{ccl} f_X(x) &=& \displaystyle \int f(x,y)dy \\ &=& \displaystyle \int_0^{\infty} \left( 0.13e^{-0.5x-0.2y} - 0.06e^{-x-0.2y} - 0.06e^{-0.5x-0.4y} + 0.12e^{-x-0.4y} \right) dy \\ &=& \displaystyle \left( 0.13e^{-0.5x} - 0.06e^{-x} \right) \left( \int_0^{\infty} e^{-0.2y} dy \right) + \left( - 0.06e^{-0.5x} + 0.12e^{-x} \right) \left( \int_0^{\infty} e^{-0.4y} dy \right) \\ &=& 5\left( 0.13e^{-0.5x} - 0.06e^{-x} \right) + 2.5\left( - 0.06e^{-0.5x} + 0.12e^{-x} \right) \\ &=& 0.5 e^{-0.5x} \end{array}$$
- Lưu ý rằng: **mọi phân phối mũ $\mathcal{E}(\lambda)$ đều có dạng hàm MDXS là $\lambda e^{-\lambda x}$** nên 

$$ X \sim \mathcal{E}(\lambda = 0.5) $$
do đó

$$ \text{Var}X = \dfrac{1}{\lambda} = 2. $$

#### Question 234.
An insurance company has an equal number of claims in each of three territories. In each territory, only three claim amounts are possible: 100, 500, and 1000. Based on the company’s data, the probabilities of each claim amount are: 

| Claim Amount := | 100 | 500 | 1000 |
|-|-|-|-|
|Territory 1. $T_1$| 0.90| 0.08| 0.02 |
|Territory 2. $T_2$| 0.80| 0.11| 0.09|
|Territory 3. $T_3$| 0.70| 0.20| 0.10|

Calculate the standard deviation of a randomly selected claim amount. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A$ là số yêu cầu mà cty BH nhận được và $T$ là khu vực, lãnh thổ (`territory`) đã yêu cầu với các khu vực $T_1,T_2,T_3$.
- Dựa vào bảng phân phối XS, ta có:

$$ \begin{array}{lll} P[A=100|T=T_1] = 0.90, &  P[A=500|T=T_1] = 0.08, &  P[A=1000|T=T_1] = 0.02 \\ P[A=100|T=T_2] = 0.80, &  P[A=500|T=T_2] = 0.11, &  P[A=1000|T=T_2] = 0.09 \\ P[A=100|T=T_3] = 0.70, &  P[A=500|T=T_1] = 0.20, &  P[A=1000|T=T_1] = 0.10 \end{array} $$

- Hơn nữa, các lãnh thổ bằng nhau về số phần tử **`(equal number of claims in each of three territories.)`** nên

$$ P[T = T_1] = P[T = T_2] = P[T = T_3] = \dfrac{1}{3}. $$
do đó
$$ \begin{array}{rcl} P[A=100]&=& \displaystyle \sum_{i=1}^3 P[T=T_i]\cdot P[A=100|T=T_i] = \dfrac{0.90 + 0.80 + 0.70}{3} = 0.80 \\ P[A=500]&=& \displaystyle \sum_{i=1}^3 P[T=T_i]\cdot P[A=500|T=T_i] = \dfrac{0.08 + 0.11+ 0.20}{3} = 0.13 \\ P[A=1000]&=& \displaystyle \sum_{i=1}^3 P[T=T_i]\cdot P[A=1000|T=T_i] = \dfrac{0.02+0.09+0.10}{3} = 0.07  \end{array} $$
- BT yêu cầu tính độ lệch chuẩn của $A$, tức là $\text{sd}A$, ta có

$$ \begin{array}{ccl} \mathbb{E}A &=& 100 P[A=100] + 500 P[A=500] + 1000P[A=1000] = 215 \\ \mathbb{E}A^2 &=& 100^2 P[A=100] + 500^2 P[A=500] + 1000^2 P[A=1000] = 110500 \\ \text{sd}A &=& \sqrt{\text{Var}A} \; = \; \sqrt{\mathbb{E}A^2 - [\mathbb{E}A]^2} = \sqrt{110500 - (215)^2} \approx 253.33 \end{array}$$

Do đó $\text{sd}A = 253.33$.

#### Question 235.
At the start of a week, a coal mine has a high-capacity storage bin that is half full.
- During the week, 20 loads of coal are added to the storage bin. Each load of coal has a
volume that is normally distributed with mean 1.50 cubic yards and standard deviation
0.25 cubic yards.

- During the same week, coal is removed from the storage bin and loaded into 4 railroad
cars. The amount of coal loaded into each railroad car is normally distributed with mean
7.25 cubic yards and standard deviation 0.50 cubic yards.

- The amounts added to the storage bin or removed from the storage bin are mutually
independent.

Calculate the probability that the storage bin contains more coal at the end of the week
than it had at the beginning of the week. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Tại thời điểm đầu tuần, có 20 bao được tải vào kho chứa, ta đặt bao tải thứ $i^{th}$ trong này là $X_i$. Theo giả thiết thì 

$$X_i \overset{\text{i.i.d}}{\sim} \mathcal{N}(1.50, 0.25^2), \qquad \forall i = 1,\ldots, 20.$$
$\qquad \diamond$ Như vậy, $X = \displaystyle \sum_{i=1}^{20}X_i$ là tổng kho lúc đầu tuần và do đó
$$ X \sim \mathcal{N}\left( 20 \mathbb{E}X_1, 20 \text{Var}X_1 \right) = \mathcal{N}(30, 1.25) $$
- Cũng tuần đó nhưng đến khi có, có một tàu hỏa 4 toa được cho vào kho chứa lấy hàng, ta đặt toa tàu thứ $i^{th}$ trong này là $Y_i$. Theo giả thiết thì 

$$ Y_i \overset{\text{i.i.d}}{\sim} \mathcal{N}(7.25, 0.5^2), \qquad \forall i = 1,\ldots, 4.$$
$\qquad \diamond$ Như vậy, $Y = \displaystyle \sum_{i=1}^{4}X_i$ là tổng trọng của tàu hỏa sau khi đã tải than lên, do đó
$$ Y \sim \mathcal{N}\left( 4 \mathbb{E}Y_1, 20 \text{Var}Y_1 \right) = \mathcal{N}(29, 1) $$
- BT yêu cầu **xác suất để kho chứa cuối tuần nhiều than hơn đầu tuần.** Điều này có nghĩa là tổng kho chứa đầu tuần $X$ trừ đi tổng lượng than tải lên tàu $Y$, phải là số dương (vẫn còn dư than), tức là $P[X-Y > 0]$. Vì $X, Y$ là độc lập, ta lại có

$$ X-Y \sim \mathcal{N}\left( \mathbb{E}X - \mathbb{E}Y, \text{Var}X + \text{Var}Y \right) = \mathcal{N}(1, 1.5) $$
do đó

$$ P[X-Y > 0] = P\left[ \mathcal{N}(0,1) > \; \dfrac{0 - 1}{\sqrt{1.5}} \right] = 1 - P\left[ \mathcal{N}(0,1) \leq \; -0.67 \right] = 1 - \Phi(-0.67) = \Phi(0.67) \approx 0.7486. $$

#### Question 236.
An insurance company insures a good driver and a bad driver on the same policy. 
- The table below gives the probability of a given number of claims occurring for each of these drivers in the next ten years. 

| Number of claims | Probability for the good driver | Probability for the bad driver |
|-|-|-|
|0|  0.5| 0.2|
|1| 0.3 |0.3|
|2| 0.2 |0.4|
|3| 0.0 |0.1|

- The number of claims occurring for the two drivers are independent.

Calculate the mode of the distribution of the total number of claims occurring on this policy in the next ten years. 


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_G, X_B$ là số các yêu cầu trong 10 năm kế tiếp (`the next ten years`) tương ứng với sự kiên tài xế đó là loại tệ(bad) hay tốt (good).
- Theo bảng phân phối xác suất; ta có

$$ \begin{array}{llll} P[X_G=0] = 0.5 & P[X_G = 1] = 0.3 & P[X_G=2] = 0.2 & P[X_G=3] = 0.0 \\ P[X_B=0] = 0.2 & P[X_B=1] = 0.3 & P[X_B=2] = 0.4 & P[X_B=3] = 0.1 \end{array} $$
- Đặt $S = X_G + X_B$ là tổng số các yêu cầu từ 2 nhóm tài xế $B, G$ (mỗi nhóm có tối đa 3 yêu cầu) nên có tất cả 7 giá trị $0,1,2,3,4,5,6$.
- BT yêu cầu tìm mode của tổng số người từ 2 nhóm, tức là phần tử có xác suất cao nhất trong 7 giá trị của $S$. Ta có

$$ \begin{array}{cclrl} P[S=0] &=& P[X_G=0, X_B=0] \\ &=& P[X_G = 0] \cdot P[X_B = 0] = (0.5)(0.2) &=& 0.1 \\ P[S=1] &=& P[X_G=1]\cdot P[X_B=0] + P[X_G = 0] \cdot P[X_B=1] &=& 0.21 \\ P[S=2] &=& P[X_G=1]\cdot P[X_B=1] + P[X_G=2]\cdot P[X_B=0] + P[X_G=2]\cdot P[X_B=0] &=& 0.33 \end{array} $$

tương tự,
$$ \begin{array}{cclrl} P[S=3] &=& (0.5)(0.1) + (0.3)(0.4) + (0.2)(0.3) + (0.0)(0.2) &=& 0.23 \\ P[S=4] &=& (0.0)(0.3) + (0.1)(0.3) + (0.2)(0.4) &=& 0.11 \\ P[S=5] &=& (0.0)(0.4) + (0.1)(0.2) &=& 0.02 \\ P[S=6] &=& (0.0)(0.1) &=& 0 \end{array} $$

do đó, $P[X=2] = 0.33$ là lớn nhất trong tất cả các xác suất của $S$ nên $2$ là mode của $S$.

#### Question 237.
- `(i)` In a group of 15 health insurance policyholders diagnosed with cancer, each policyholder
has probability 0.90 of receiving radiation and probability 0.40 of receiving
chemotherapy. 
- `(ii)` Radiation and chemotherapy treatments are independent events for each
policyholder, and the treatments of different policyholders are mutually independent.
- `(iii)` The policyholders in this group all have the same health insurance that pays 2 for
radiation treatment and 3 for chemotherapy treatment.

Calculate the variance of the total amount the insurance company pays for the radiation
and chemotherapy treatments for these 15 policyholders. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lươt là số các KH đã thực hiện xạ trị (`radiation`) và hóa trị (`chemotherapy`).
- Theo `(i)`, tổng số KH (dựa trên hợp đồng) là 15 và do đó

$$ X \sim \mathcal{B}(15, 0.9), \qquad Y \sim \mathcal{B}(15, 0.4) $$
bởi XS để có được một hợp đồng điều trị bằng xạ trị bằng 0.9 trong tổng số 15, do đó $X$ tuân theo luật nhị thức $\mathcal{B}(15, 0.9)$ và lập luận tương tự cho $Y$.
- Theo `(ii)`, việc trị liệu theo 2 liệu pháp $X, Y$ là độc lập
- Theo `(iii)`, số tiền chi trả của nhóm này cho việc điều trị là $2$ cho xạ trị và $3$ cho hóa trị; do đó
$$ Z = 2X + 3Y $$
sẽ là số tiền tương ứng mà nhóm này phải chi trả.
- BT  yêu cầu ta tính phương sai của số tiền chi trả này, là $\text{Var}Z$; ta có

$$ \begin{array}{crll} \text{Var}Z &=& \text{Var}[2X+3Y] \\ &=& 2^2 \text{Var}X + 3^2 \text{Var}Y & \text{, do } X, Y \text{ doc.lap.} \\ &=& 4\left( 15 \cdot 0.9 \cdot (1 - 0.9) \right) + 9\left( 15 \cdot 0.4 \cdot (1 - 0.4) \right) & \text{, cong thuc phuong sai cua BNN nhi thuc } \\ &=& 37.8 \end{array} $$

-------------

**Comment.** 
- Trong BT này, hoặc các BT liên quan đến các phân phối phổ biến như nhị thức, chuẩn, poisson, mũ, ... thì độc giả cần lưu ý các công thức về kỳ vọng, phương sai của dạng tổng quát để rút ngắn tính toán khi cần thiết.
- Chẳng hạn ở dấu bằng thứ 3 trong biểu thức sau cùng, $\text{Var}Z$; ta đã dùng công thức phương sai của BNN nhị thức: Nếu $T \sim \mathcal{B}(n, \theta)$ thì 

$$ \text{Var}T = n \theta (1 - \theta). $$

#### Question 238.
In a large population of patients, 20% have early stage cancer, 10% have advanced stage
cancer, and the other 70% do not have cancer. 

Six patients from this population are
randomly selected.

Calculate the expected number of selected patients with advanced stage cancer, given that at least one of the selected patients has early stage cancer.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là số các bệnh nhân ung thư (`patient-cancer`) đang ở thời kỳ đầu (`early stage`) và giai đoạn sau (`advanced stage`).
- BT yêu cầu ta tính $\mathbb{E}[Y | X \geq 1]$ **khi họ đã chọn ngẫu nhiên 6 bệnh nhân**, ta có

$$ \mathbb{E}[Y | X \geq 1] = \dfrac{\mathbb{E}Y - \mathbb{E}[Y | X=0]\cdot P[X=0]}{P[X \geq 1]} $$
- Hơn nữa, theo giả thiết, $X \sim \mathcal{B}(6, 0.2)$ và $Y\sim \mathcal{B}(6, 0.1)$ nên

$$ \begin{array}{cclrl} P[X = 0] &=& (1 - 0.2)^6 &=& (0.8)^6 \\ P[X \geq 0] &=& 1 - P[X = 0] &=& 1 - (0.8)^6 \end{array} $$ 
- Tiếp theo, ta sẽ tìm $\mathbb{E}Y$ và $\mathbb{E}[Y | X=0]$; ta có

$\qquad \diamond \; \mathbb{E}Y = 6(0.1) = 0.6$ (Vì $Y\sim \mathcal{B}(6, 0.1)$ hay chỉ có $10\%$ bệnh nhân bị ung thư ở giai đoạn sau và có 6 bệnh nhân được chọn).

$\qquad \diamond \; $ Hơn nữa, $\mathbb{E}[Y | X=0]$ là kỳ vọng có điều kiện cho $Y$ (các bệnh nhân ở giai đoạn sau) khi biết $X=0$ (không có bệnh nhân nào đang ung thư giai đoạn đầu trong 6 bệnh nhân) do đó

$$ (Y|X=0) \sim \mathcal{B} \left( 6, \dfrac{P[\text{ late stage cancer}]}{P[\text{no patient have early stage cancer}]} \right) = \mathcal{B} \left( 6, \dfrac{0.1}{1 - 0.2} \right) $$
$\qquad \quad$ suy ra
$$ \mathbb{E}[Y|X=0] = 6 \left( \dfrac{0.1}{1 - 0.2} \right) = 0.75 $$

$\qquad \quad$ và

$$ \mathbb{E}[Y | X \geq 1] = \dfrac{0.6 - (0.75)(0.8)^6}{1 - (0.8)^6} \approx 0.547. $$

#### Question 239.
- Four distinct integers are chosen randomly and without replacement from the first twelve positive integers.
- Let $X$ be the random variable representing the second largest of the four selected integers, and let $p$ be the probability function for $X$.

Determine $p(x)$, for integer values of $x$, where $p(x) > 0.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $\Omega = \lbrace 1,2,\ldots,12 \rbrace$ là KG mẫu.
- Theo đề, ta sẽ chọn ngẫu nhiên không hoàn lại $4$ số phân biệt từ $\Omega$ nên **tổng số các trường hợp** sẽ có $C_{12}^4 = 495$.
- Vì $X$ là số lớn thứ hai trong tập $4$ số đã chọn từ $\Omega$ nên nó không thể rơi vào các trường hợp sau:

$\qquad \diamond X = 1,2$ **(loại)** vì đây là 2 số nhỏ nhất trong $\Omega$.

$\qquad \diamond X = 12$ **(loại)** vì đây là số lớn nhất. 

do đó, với $X = x \in \lbrace 3,4,\ldots,11 \rbrace$; ta sẽ biện luận như sau:

$\qquad \diamond$ Khi $x=3$; ta còn $3$ lựa chọn (trong tổng $4$ lựa chọn) và phải có $2$ số nhỏ hơn $x$ $($là $1,$ $2)$ và $1$ lựa chọn cho số lớn hơn $x$ trong các số $4,5,6,\ldots,11$. Như vậy, ta sẽ có $C_2^2 C_9^1 = 9$ trường hợp khi $x=3$.

#### Question 240.
- An insurance policy covers losses incurred by a policyholder, subject to a deductible of $10,000.$ 
- Incurred losses follow a normal distribution with mean $12,000$ and standard deviation $c.$ 
- The probability that a loss is less than $k$ is $0.9582$, where $k$ is a constant.
- Given that the loss exceeds the deductible, there is a probability of $0.9500$ that it is less than $k.$

Calculate $c.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là mức tổn thất mà BH sẽ chi trả, theo 2 giả định đầu tiên $$X \sim \mathcal{N}\left( 12000, c^2 \right)$$ 
- Giả thiết (iii) cho biết XS mà tổn thất này bé hơn $k$ bằng $0.9582$; tức là $ P[X < k] = 0.9582 $
- Giả thiết cuối cho biết: khi có khấu hao ($10000$, nghĩa là khấu hao chỉ xuất hiện khi $X > 10000$) thì XS để $X < k$ bằng

$$ P[X < k | X > 10000] = 0.95 $$
$\qquad$ Ta có 

$$ P[X < k | X > 10000] = \dfrac{P[(X<k) \cap (X > 10000) ]}{P[X > 10000]} = \dfrac{P[X < k] - P[X \leq 10000]}{1 - P[X \leq 10000]} $$
$\qquad$ suy ra

$$ \begin{array}{crcl} & 0.95\left( 1 - P[X \leq 10000] \right) &=& P[X < k] - P[X \leq 10000] \\ \Rightarrow & 0.95 - 0.95P[X \leq 10000] &=& 0.9582 - P[X \leq 10000] \\ \Rightarrow & P[X \leq 10000] &=& \dfrac{0.9582 - 0.95}{1 - 0.95} = 0.164 \end{array}  $$
- Để tìm $c$, ta sẽ dùng bảng tra phân phối chuẩn; ta có

$$ \begin{array}{crcll} P[X \leq 10000] = 0.164 &\Leftrightarrow & P \left( \mathcal{N}(0,1) < \dfrac{10000 - \mathbb{E}X}{\sqrt{\text{Var}X}} \right) &=& 0.164 \\ &\Leftrightarrow & \Phi \left( \dfrac{10000 - 12000}{c} \right) &=& 0.164 \\ &\Leftrightarrow & \dfrac{-2000}{c} &=& -0.978 \end{array} $$
do đó 
$$c = \dfrac{2000}{0.978} \approx 2045.$$
