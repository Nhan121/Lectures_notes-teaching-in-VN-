#### Question 241,
Losses covered by an insurance policy are modeled by a **uniform distribution** on the interval $[0, 1000]$. An insurance company reimburses losses in excess of a deductible of $250.$

Calculate the difference between the median and the $20^{th}$ percentile of the insurance company reimbursement, over all losses. 

**Giải đáp / SOLUTIONs.**

Đặt $X$ là mức tổn thất, theo đó $X \sim \mathcal{U}([0, 1000])$ và $Y = \max(0, X-250)$ là mức bồi thường mà BH sẽ chi trả. Tổng quát, ta có

$$ P[X \leq x] = \left\lbrace \begin{array}{rl} 0,& x < 0 \\ \dfrac{x}{1000} & x \in [0, 1000] \\ 1, & x > 1000. \end{array} \right. $$
do đó
- phân vi mức $20\%$ của $X$ sẽ thỏa $\dfrac{x_p}{1000} = 0.20 \Longrightarrow x_p = 200$.
- tương tự; trung vị của $X$ sẽ thỏa $\dfrac{x_m}{1000} = 0.50 \Longrightarrow x_m = 500$.

Vì $Y = \max(0, X-250)$ nên các phân vị tương ứng của $Y$ sẽ xác định theo $X$ thông qua hệ thức 

$$ y_p = \max(0, x_p - 250) = 0 $$
và
$$ y_m = \max(0, x_m - 250) = 250 $$
Do đó; hiệu số cần tìm là 

$$ y_m - y_p = 250. $$

#### Question 242.
The annual profits that company A and company B earn follow a **bivariate normal distribution**.
- Company A's annual profit has mean 2000 and standard deviation 1000.
- Company B's annual profit has mean 3000 and standard deviation 500.
- The correlation coefficient between these annual profits is 0.80.

Calculate the probability that company B's annual profit is less than 3900, given that company A's annual profit is 2300. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là lợi nhuận hàng năm ở các cty $A$ và $B$; khi đó $(X,Y) \sim \mathcal{N}_2 \left( \mu, \Sigma \right)$ với

$$ \mu = \left( \mu_X, \mu_Y \right)^T, \qquad \Sigma = \left( \begin{array}{cc} \sigma_X^2 & \rho_{XY} \\ \rho_{XY} & \sigma_Y^2 \end{array} \right) $$
- 3 giả thiết còn lại cho biết
$$ \begin{array}{rll} \text{Theo (i),} & \mu_X = 2000, & \sigma_X = 1000 \\ \text{(ii),} & \mu_Y = 3000, & \sigma_Y = 500 \\ \text{(iii),} & \dfrac{\rho_{XY}}{\sigma_X \sigma_Y} = 0.8 \end{array} $$
- BT yêu cầu tính XS để lợi nhuận cty B thấp hơn 3900 biết rằng lợi nhuận cty A bằng 2300; tức là $P[Y < 3900 | X = 2300]$
- **Kiến thức ở thống kê nhiều biến [Multivariate Statistics]. Nếu $Z = (X,Y)$ có phân phối chuẩn 2 chiều $\mathcal{N}_2\left( \mu, \Sigma \right)$ thì BNN có điều kiện $(Y|X=x_0)$ sẽ có phân phối chuẩn và xác định bởi**

$$ (Y|X=x_0) \sim \mathcal{N} \left( \mu_Y + \dfrac{r_{XY} \cdot \sigma_Y}{\sigma_X}\left(x_0 - \mu_X \right), \;\; \sigma_Y^2 \left( 1 - r_{XY}^2 \right) \right) $$
trong đó $$ r_{XY} = \dfrac{\rho_{XY}}{\sigma_X \sigma_Y} $$
- Quay lại bài toán; thay các giá trị trong giả thiết vào; ta có $r_{XY} = 0.8$ và

$$ \begin{array}{rllcl} \mu_Y + \dfrac{r_{XY} \cdot \sigma_Y}{\sigma_X}\left(x_0 - \mu_X \right) &=& 3000 + \dfrac{(0.8)\cdot 500}{1000}(2300 - 2000) &=& 3120 \\ \sigma_Y^2 \left( 1 - r_{XY}^2 \right) &=& (500)^2 (1 - (0.8)^2) &=& 90000. \end{array} $$
do đó

$$ \begin{array}{ccl} P[Y < 3900 | X = 2300] &=& P\left( \mathcal{N}(3120, 90000) < 3900 \right) \\ &=& P\left( \mathcal{N}(0, 1) < \dfrac{3900 - 3120}{\sqrt{90000}} \right) \\ &=& \Phi(2.6) \\ &=& 0.9953. \end{array}$$

#### Question 243.
An insurance agent’s files reveal the following facts about his policyholders:
- i) 243 own auto insurance.
- ii) 207 own homeowner insurance.
- iii) 55 own life insurance and homeowner insurance.
- iv) 96 own auto insurance and homeowner insurance.
- v) 32 own life insurance, auto insurance and homeowner insurance.
- vi) 76 more clients own only auto insurance than only life insurance.
- vii) 270 own only one of these three insurance products.

Calculate the total number of the agent’s policyholders who own at least one of these three insurance products.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $L,A,H$ là 3 sự kiện tương ứng với việc chọn được hợp đồng BH là loại nào trong `"nhân thọ (life ins)"`, `"xe cộ (auto ins)"` và `"nhà cửa (home ins)"`. Theo các giả thiết, ta có (về số người)

$$ \begin{array}{rll} \text{Theo (i) & (ii).} & |A| = 243, & |H| = 207 \\ \text{(iii) & (iv).} & |L \cap H| = 55, & |A\cap H| = 96 \\ \text{(v) & (vi).} & |A \cap L \cap R| = 32, & |A \cap L^c \cap H^c| - |L \cap A^c \cap H^c| = 76 \\ \text{(vii).} & |\text{co dung 1 goi BH}| = & |A \cap L^c \cap H^c| + |A^c \cap L \cap H^c| + |A^c \cap L^c \cap H| = 270 \end{array} $$
- BT yêu cầu tính tổng số hợp đồng của những KH đã mua ít nhất một trong 3 loại BH trên, tức là $|A \cup L \cup H|$
- Ta có
$$ \begin{array}{crlrl} |L \cap H \cap A^c| &=& |L \cap H| - |L \cap H \cap A| &=& 55 - 32 = 23 \\ |A \cap H \cap L^c| &=& |A \cap H| - |A \cap H \cap L| &=& 96 -32 = 64 \end{array} $$
kết hợp với $\text{(ii)}$; suy ra

$$ \begin{array}{crlrl} |H \cap A^c \cap L^c| &=& |H \cap (A \cup L)^c| \\ &=& |H| - |H \cap (A \cup L)| \\  &=& |H| - |(H \cap L) \cup (H \cap A)| \\ &=& |H| - \left( |H \cap L| + |H \cap A| - |(H\cap A) \cap (H \cap L)| \right) \\ &=& |H| - \left( |H \cap L| + |(H \cap A) \cap L^c | \right) \\ &=& 207 - (55 + 64) \\ &=& 88 \end{array}$$

- Đặt $x = |L \cap A^c \cap H^c|$ và $y = |A \cap L^c \cap H^c|$; ta có hệ phương trình sau:

$$ \left\lbrace \begin{array}{ccl} y - x &=&  |A \cap L^c \cap H^c| - |L \cap A^c \cap H^c| = 76 \\ x+y &=& 270 - |H \cap A^c \cap L^c| = 182 \end{array} \right. $$
do đó $ \left\lbrace \begin{array}{ccl} x &=& 53 \\ y &=& 129 \end{array} \right. $
- Vì vậy,
$$ \begin{array}{ccl} |L \cap A \cap H^c| &=& |A| - |A \cap (L^c \cup H)| \\ &=& |A| - (|A \cap H| + |A \cap L^c| - |A \cap L^c \cap H| ) \\ &=& |A| - (|A \cap H | + |(A \cap L^c) \cap H^c|) \\ &=& 243 - (96 + 129) \\ &=& 18. \end{array} $$
- Cuối cùng,
$$ \begin{array}{crlrl} |H \cup A \cup L| &=& |\text{co dung 1 goi BH}| + |\text{co dung 2 goi BH}| + |\text{co ca 3 goi BH}| \\ &=& (53 + 129 + 88) + (18 + 64 + 23) + 32 \\ &=& 407. \end{array} $$

#### Question 245.
An **insurance policy pays** 100 per day for up to three days of hospitalization and 50 per day for each day of hospitalization thereafter.

The **number of days of hospitalization**, $X$, is a discrete random variable with probability function

$$ P[X=k]= \left\lbrace \begin{array}{cl} \dfrac{6-k}{15}, & k = 1,2,3,4,5. \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the expected payment for hospitalization under this policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đầu tiên, dễ dàng kiểm chứng $\displaystyle \sum_{k=1}^{\infty} P[X=k] = \sum_{k=1}^5 \dfrac{6 - k}{15} = 1.$
- Đặt $Y=g(X)$ là chi phí trả tiền viện phí theo ngày; theo đó

$$ g(x) = \left\lbrace \begin{array}{cll} 100x, & x = 1,2,3, & \text{vien.phi trong 3 ngay dau tiên: 100 per day} \\ 50x , & x = 4,5,\ldots & \text{vien phi trong nhung ngay ke tiep: 50 per day} \end{array} \right. $$
- BT yêu cầu tính kỳ vọng số tiền chi trả, tức là

$$ \mathbb{E}Y = \displaystyle \sum_{k=0}^5 g(k) \cdot P[X = k] = 100 \left( \dfrac{5}{15} \right) + 100 \left( \dfrac{4}{15} \right) + 100 \left( \dfrac{3}{15} \right) + 50 \left( \dfrac{2}{15} \right) + 50 \left( \dfrac{1}{15} \right) = 220. $$

#### Question 248.
- **Bowl I** contains eight red balls and six blue balls. **Bowl II** is empty. 
- Four balls are selected at random, without replacement, and transferred from bowl I to bowl II. 
- One ball is then selected at random from bowl II.

Calculate the conditional probability that two red balls and two blue balls were transferred from bowl I to bowl II, given that the ball selected from bowl II is blue.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là sự kiện lấy được bát I (`Bowl I`) hay bát II (`Bowl II`); và $R, B$ lần lượt là màu của bi trong bát
- Theo `(i)`

$$ X = \lbrace 8 R, 6B \rbrace, \qquad Y = \lbrace 0 R, 0B \rbrace $$
- Theo `(ii)-(iii)`, rút 4 bi không hoàn lại từ $X$ bỏ sang $Y$; rồi lại lấy ngẫu nhiên một viên từ $Y$
- BT yêu cầu tính XS có điều kiện: **chọn được 2 bi đỏ (2R) và2 xanh được chuyển từ X sang Y biết rằng bi chọn được từ Y là xanh (B)**, tức là 

$$ P\left[ 2R, 2B \text{ transf from X to Y} | \text{B drawn from Y} \right] = \dfrac{P[2R, 2B \text{ (transf from X to Y)} \cap \text{(B drawn from Y)}]}{P[\text{B drawn from Y}]} $$
ta có
$$ \begin{array}{ccl} P[2R, 2B \text{ (transf from X to Y)} \cap \text{(B drawn from Y)}] &=& \dfrac{|\text{select 2R, 2B from X = (8R, 6B)}|\cdot |\text{select B from Y = (2R, 2B)}| }{|\text{select 4 balls from 14 balls of X}|} \\ &=& \dfrac{|\text{select (2R from 8R) and (2B from 6B)}|\cdot |\text{select B from Y = (2R, 2B)}| }{|\text{select 4 balls from 14 balls of X}|} \\ &=& \dfrac{\left( C_8^2 C_6^2 \right) \cdot \left( \dfrac{2}{2+2} \right)}{C_{14}^4} \\ &=& \dfrac{840}{4004}. \end{array} $$

và
$$ \begin{array}{ccl} P[\text{B drawn from Y}] &=& \dfrac{|\text{select B from Y which at least 1 blue ball}|}{|\text{select 4 balls from 14 balls of X}|} \\ &=& \dfrac{|\text{select B from 4 cases of Y (0R, 4B), (1R, 3B), (2R, 2B) and (3R, 1B)}|}{|\text{select 4 balls from 14 balls of X}|} \\ &=& \displaystyle \sum_{k=1}^4 \dfrac{|\text{select [k.B and (4 - k)R] from X}| \cdot |\text{select B from [k.B and (4 - k)R]}|}{|\text{select 4 balls from 14 balls of X}|} \\ &=& \displaystyle \sum_{k=1}^4 \dfrac{\left( C_8^{4-k} C_6^k \right) \cdot \left( \dfrac{k}{4} \right) }{C_{14}^4} \\ &=& \dfrac{\left( C_8^{3} C_6^1 \right) \cdot \left( \dfrac{1}{4} \right) + \left( C_8^{2} C_6^2 \right) \cdot \left( \dfrac{2}{4} \right) + \left( C_8^{1} C_6^3 \right) \cdot \left( \dfrac{3}{4} \right) + \left( C_8^{0} C_6^4 \right) \cdot \left( \dfrac{4}{4} \right)}{C_{14}^4} \\ &=& \dfrac{1716}{4004} \end{array} $$

suy ra

$$ P\left[ 2R, 2B \text{ transf from X to Y} | \text{B drawn from Y} \right] = \dfrac{\frac{840}{4004}}{\frac{1716}{4004}} = \dfrac{840}{1716} = 0.49 $$
