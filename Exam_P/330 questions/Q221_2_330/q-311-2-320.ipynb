#### Question 311, 
- An individual experiences a loss due to property damage and a loss due to bodily injury.
- Losses are independent and **uniformly distributed** on the interval $[0, 3]$.

Calculate the expected loss due to bodily injury, given that at least one of the losses is less than 1. 

**Giải đáp / SOLUTIONs.**

Đặt $X, Y$ lần lượt là các tổn thất về tài sản và thể chất (`bodily injury`) biết rằng:
- $X, Y$ độc lập và có cùng phân phối đều $\mathcal{U}([0, 3])$, do đó có hàm MDXS đồng thời

$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{9} , & 0\leq x,y \leq 3 \\ 0, & \text{otherwise} \end{array}  \right. $$

- BT yêu cầu tính $\mathbb{E}[Y| (X < 1) \text{ or } (Y < 1)]$, đặt $g$ là hàm mật độ của $Y$ với điều kiện $(X < 1) \cup (Y < 1)$, khi đó 

$$ g(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{|A|} , & 0\leq x,y \leq 3 \\ 0, & \text{otherwise} \end{array}  \right. $$
trong đó
$$ A = \lbrace 0\leq x < 1, y \in [0, 3] \rbrace \cup \lbrace 0 \leq y < 1, x \in [1, 3] \rbrace $$

Như vậy, do $(X,Y)$ vẫn là phân phối đều trên $$A \subset \lbrace x\in [0, 3], y \in [0, 3] \rbrace,$$ và dễ thấy $|A|$ là diện tích của miền $A$; giới hạn bởi 2 hình chữ nhật rời nhau với tổng diện tích 

$$|A| = (3-0)\cdot (1-0) \; + \; (3-1)\cdot (1-0) = 5$$
- Do đó
$$ \begin{array}{ccl}\mathbb{E}Y[Y| (X<1) \cup (Y<1)] &=& \displaystyle \iint_{\lbrace 0\leq y < 1, x \in [1, 3] \rbrace} yg(x,y)dxdy + \iint_{\lbrace 0 \leq x < 1, y \in [0, 3] \rbrace} yg(x,y)dxdy \\ &=& \displaystyle \int_0^1 \int_1^3  y \cdot \dfrac{1}{5} dxdy + \int_0^3 \int_0^1 y \cdot \dfrac{1}{5} dxdy \\ &=& 0.2 + 0.9 \\ &=& 1.1. \end{array} $$

#### Question 312.
The table below shows the joint probability function of a sailor’s number of boating
accidents and number of hospitalizations from these accidents this year. 

      |===============================================================|
      |              |     Number of Hospitalizations from Accidents  |
      |---------------------------------------------------------------|
      |              |        |    0    |    1    |    2    |    3    | 
      |              |........|.........|.........|.........|.........|
      |  Number of   |    0   |  0.700  |         |         |         |
      |  Accidents   |    1   |  0.150  |  0.050  |         |         |
      |              |    2   |  0.060  |  0.020  |  0.010  |         |
      |              |    3   |  0.005  |  0.002  |  0.002  |  0.001  |
      |==============|========|=========|=========|=========|=========|


Calculate the sailor’s expected number of hospitalizations from boating accidents this year.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là số các vụ tai nạn (`X: number of accidents`) và số các ca nhập viện từ các vụ tai nạn (`Y: # of hospitalizations from accidents`). 
- Dựa vào bảng phân phối xác suất, ta có thể thấy rằng $X, Y$ nhận các giá trị trên $0 \leq Y \leq X \leq 3$, chẳng hạn $P[X=0, Y=0] = 0.7$
- BT yêu cầu tính $\mathbb{E}Y$, ta có
$$ \mathbb{E}Y = \displaystyle \sum_{y=0}^3 y \cdot P[Y=y] $$
vì vậy, ta cần tính $P[Y = y]$ bằng các tổng cột (`phân phối xác suất lề`), ta có
$$ \begin{array}{ccl} P[Y=0] &=& \displaystyle \sum_{k=0}^4 P[Y= 0, X = k] \\ &=& P[X=0, Y=0] + P[X=1,Y=0] + P[X=2,Y=0]+P[X=3,Y=0] \\ &=& 0.70 + 0.15 + 0.06 + 0.005 \\ &=& 0.915 \\ P[Y=1] &=& 0.050 + 0.020 + 0.002 \\ &=& 0.072 \\ P[Y=2] &=& 0.012 \\ P[Y=3] &=& 0.001 \end{array} $$
do đó

$$ \mathbb{E}Y = 0 \cdot (0.915) + 1 \cdot (0.072) + 2 \cdot (0.012) + 3 \cdot (0.001) = 0.099. $$

#### Question 313.
Every day, the 30 employees at an auto plant each have probability 0.03 of having one
accident and zero probability of having more than one accident. 
- Given there was an accident, the probability of it being major is 0.01. 
- All other accidents are minor. The
numbers and severities of employee accidents are mutually independent.
- Let $X$ and $Y$ represent the numbers of major accidents and minor accidents, respectively, occurring in the plant today.

Determine the joint moment generating function $M_{X,Y}(s , t)$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A_i$ là BNN thể hiện nhân viên thứ $i^{th}$ có xảy ra tai nạn hay không?
- Đặt $X_i, Y_i$ là BNN quan sát một nhân viên $i^{th}$  có xảy ra tai nạn nghiêm trọng (`major`) hay tai nạn cấp nhẹ (`minor`) hay không? 
- Đặt
$$ \displaystyle S = \sum_{i=1}^{30} A_i  $$
- Theo các giả thiết, ta được $A_i = X_i + Y_i$ (khi có tai nạn xảy ra, hoặc nó là nặng; hoặc nó là nhẹ, không thể đồng thời xảy ra)

$\qquad \diamond P[A_i = 1] = 0.03 $ là XS xảy ra tai nạn đối với nhân viên thứ $i^{th}$

$\qquad \diamond P[X_i = 1| A_i = 1] = 0.01$ là XS tai nạn xảy ra với nhân viên $i^{th}$ là loại nghiêm trọng khi có tai nạn xảy ra.

$\qquad \diamond P[Y_i = 1| A_i = 1] = P[X_i = 0| A_i = 1] = 0.99$ là XS tai nạn xảy ra với nhân viên $i^{th}$ là loại nhẹ khi có tai nạn xảy ra.

- BT yêu cầu tính hàm sinh moment đồng thời $M_{X,Y}(s,t)$, ta thấy rằng bộ ba BNN $X, Y, 30 - S$ tạo thành BNN dạng `tri-nomial (tam thức)` với

$$ (X, Y, 30-S) \sim \mathcal{T}(n, p_X, p_Y, 1 - p_X - p_Y) $$
trong đó

$\qquad \diamond$ $n = 30$ và $S = X+Y$ là tổng số nhân viên bị tai nạn (bao gồm dạng nghiêm trọng [`major` $X$] và dạng nhẹ [`minor` $Y$])

$\qquad \diamond$ $p_X = P[X_i = 1|A_i=1]\cdot P[A_i=1] = 0.003$ là XS xảy ra một tai nạn nghiêm trọng (`major`).

$\qquad \diamond$ $p_Y = P[X_i=0|A_i=1] \cdot P[A_i =1] = 0.0297$ là XS xảy ra một tai nạn nhẹ (`minor`).

$\qquad \diamond$ Suy ra $1 - p_X - p_Y = 0.97$ là XS không có tai nạn xảy ra.

- Nhắc lại rằng nếu $ (X, Y, 30-X-Y) \sim \mathcal{T}(n, p_X, p_Y, 1 - p_X - p_Y) $ thì

$$ M_{X,Y}(s,t) = \left( p_X e^s + p_Y e^t + 1-p_X - p_Y \right)^n $$
do đó, ta được
$$ M_{X,Y}(s,t) = \left( 0.003 e^s + 0.0297 e^t + 0.97 \right)^{30}. $$

#### Question 314.
The returns on two investments, $X$ and $Y$, follow the joint probability density function

$$ f(x,y) = \left\lbrace \begin{array}{cl} k, & 0 < |x| + |y| < 1, \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate the maximum value of $\text{Var}[Y | X = r]$ where $-1 <r < 1.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Để tính phương sai có điều kiện $\text{Var}[Y|X=r]$, ta cần xác định hàm mật độ của $(Y|X=r)$ với $r \in (-1. 1)$
- Từ giả thiết $f(x,y)$ suy ra với mỗi $r \in (-1, 1)$ thì
$$ f_{(Y|X=r)}(y) = \left\lbrace \begin{array}{cl} c, & 0 < |r| + |y| < 1, \\ 0 & \text{otherwise} \end{array} \right. $$

nên $(Y|X=r)$ có $ f_{(Y|X=r)}(y)$ có dạng hàm MDXS của phân phối đều trên đoạn $|y| \in (-|r|, 1 - |r|)$. 

Như vậy, với mọi $r \in (-1, 1)$ thì $y \in (-1, 1)$, do đó phương sai của một phân phối đều $\mathcal{U}([a, b])$ sẽ được tính bởi

$$ \dfrac{(b - a)^2}{12} $$
- Thay $b = 1, a = -1$ ta được 
$$ \text{Var}[Y|X=r] = \dfrac{(1 - (-1))^2}{12} = \dfrac{1}{3}. $$

#### Question 315.
- On Main Street, a driver’s speed just before an accident is uniformly distributed on [5, 20]. 
- Given the speed, the resulting loss from the accident is exponentially distributed with mean equal to three times the speed.

Calculate the variance of a loss due to an accident on Main Street.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $S, L$ lần lượt là tốc độ và thiệt hại (tổn thất) do tài xế gây ra. Theo đó,

$$ S \sim \mathcal{U}([5, 20]), \qquad (L | S = v) \sim \mathcal{E}(3v), \quad \forall v \in [5, 20]. $$
vì vậy
$$ \mathbb{E}(L|S) = 3S, \qquad \text{Var}(L|S) = (3S)^2 = 9S^2 $$
và
$$ \mathbb{E}(S) = \dfrac{20-5}{2} = \dfrac{15}{2}, \qquad \text{Var}(S) = \dfrac{(20 - 5)^2}{12} = \dfrac{225}{12} $$
- BT yêu cầu tính phương sai $\text{Var}L$, theo đó ta sẽ dùng công thức phương sai có điều kiện:

$$ \begin{array}{ccl} \text{Var}L &=& \mathbb{E}[\text{Var}(L | S)] + \text{Var}[ \mathbb{E}(L | S) ] \\ &=& \mathbb{E}[9S^2] + \text{Var}(3S) \\ &=& 9 \left( [ \mathbb{E}S ]^2 + \text{Var}S \right) + 9 \text{Var} S \\ &=& 9 \left( \left( \dfrac{15}{2} \right)^2 + \dfrac{225}{12} \right) + 9 \cdot \dfrac{225}{12} \\ &=& \dfrac{6975}{4} = 1743.75. \end{array} $$

#### Question 316.
- Let $X$ be the annual number of hurricanes hitting Florida, and let $Y$ be the annual number of hurricanes hitting Texas. 
- $X$ and $Y$ are independent Poisson variables with respective means 1.70 and 2.30.

Calculate $\text{Var}[X - Y | X+Y=3]$

**Giải đáp / SOLUTIONs.**

- Theo giả thiết, $X,Y$ độc lập với
$$ X \sim \text{Pois}(1.7), \qquad Y \sim \text{Pois}(2.3) $$
- BT yêu cầu tính phương sai có điều kiện của $X - Y$ trên $X+Y=3$; nhưng vì $X,Y$ là các BNN Poisson nên chỉ nhận các giá trị $0,1,2,\ldots;$ do đó

$$ \lbrace (x, y): x+y = 3; \quad \forall x,y = 0,1,\ldots \rbrace = \left\lbrace (0,3), (1,2), (2,1), (3,0) \right\rbrace $$
- Đặt 
$$p(x, y) = P[X = x, Y = y] = P[X = x] \cdot P [Y = y]$$ 

là phân phối của cặp $(X, Y)$, khi đó

$$ \begin{array}{cclclcl} p(0, 3) &=& \left( e^{-1.7} \right) \cdot \left( \dfrac{e^{-2.3} (-2.3)^3}{3!} \right) &=& 2.02728 e^{-4} \\ p(1,2) &=& \left( \dfrac{e^{-1.7} (-1.7)^1}{1!} \right) \cdot \left( \dfrac{e^{-2.3} (-2.3)^2}{2!} \right) &=& 4.4965 e^{-4} \\ p(2,1) &=& \left( \dfrac{e^{-1.7} (-1.7)^2}{2!} \right) \cdot \left( \dfrac{e^{-2.3} (-2.3)^1}{1!} \right) &=& 3.3235 e^{-4} \\ p(3,0) &=& \left( \dfrac{e^{-1.7} (-1.7)^3}{3!} \right) \cdot \left( e^{-2.3} \right) &=& 0.8188 e^{-4} \end{array} $$

- Để giới hạn các xác suất trên xuống tập  $\left\lbrace (0,3), (1,2), (2,1), (3,0) \right\rbrace$, ta phải chuẩn hóa nó như sau

$$ \begin{array}{cclclcl} P(0, 3) &=& \dfrac{p(0,3)}{p(0,3) + p(1,2) + p(2,1) + p(3,0)} &=& 0.1901 \\ P(1, 2) &=& \dfrac{p(1,2)}{p(0,3) + p(1,2) + p(2,1) + p(3,0)} &=& 0.4215 \\ P(2, 1) &=& \dfrac{p(2,1)}{p(0,3) + p(1,2) + p(2,1) + p(3,0)} &=& 0.3116 \\ P(3, 0) &=& \dfrac{p(3,0)}{p(0,3) + p(1,2) + p(2,1) + p(3,0)} &=& 0.0768\end{array} $$
- Lúc này,
$$ \begin{array}{rclclcl} \mathbb{E}[X-Y|X+Y] &=& (0-3)\cdot P(0, 3) + (1-2)\cdot P(1, 2) + (2-1)\cdot P(2, 1) + (3 - 0)\cdot P(3, 0) &=& -0.4498 \\ \mathbb{E}\left[ (X-Y)^2|X+Y \right] &=& (0-3)^2\cdot P(0, 3) + (1-2)^2\cdot P(1, 2) + (2-1)^2\cdot P(2, 1) + (3 - 0)^2\cdot P(3, 0) &=& 3.1352 \end{array} $$
nên

$$ \text{Var}[X-Y|X+Y] = \mathbb{E}\left[ (X-Y)^2|X+Y \right] - \left( \mathbb{E}\left[X-Y|X+Y \right] \right)^2 = 2.9329. $$

#### Question 317.
- The intensity of a hurricane is a random variable that is uniformly distributed on the interval $[0, 3].$ 
- The damage from a hurricane with a given intensity $y$ is exponentially distributed with a mean equal to $y.$

Calculate the variance of the damage from a random hurricane. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là số cơn bão và sức phá hủy của cơn bão (`intensity of damage`). Theo đó,

$$ Y \sim \mathcal{U}([0,3]), \qquad (X| Y=y) \sim \mathcal{E}(y), \qquad \forall y \in [0, 3]. $$
- BT yêu cầu tính $\text{Var}X$, lại có
$$ \begin{array}{ccl} \text{Var}X &=& \mathbb{E}[\text{Var}(X | Y)] + \text{Var}[ \mathbb{E}(X | Y) ] \\ &=& \mathbb{E}[Y^2] + \text{Var}(Y) \\ &=& \left( [ \mathbb{E}Y ]^2 + \text{Var}Y \right) + \text{Var} Y \\ &=& \left( \left( \dfrac{3 - 0}{2} \right)^2 + \dfrac{(3 - 0)^2}{12} \right) + \cdot \dfrac{(3 - 0)^2}{12} \\ &=& 3.75. \end{array} $$

#### Question 318.
- Random variables $X$ and $Y$ have joint distribution 

|  | $X = 0$ | $X = 1$ | $X = 2$ |
|-|-|-|-|
|$Y=0$| $1/15$ | $a$ | $2 / 15$ |
|$Y=1$| $a$ | $b$ | $a$ |
|$Y=2$| $2/15$ | $a$ | $1/15$ |

- Let $a$ be the value that minimizes the variance of $X.$

Calculate the variance of $Y.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ bảng phân phối XS đồng thời ở trên, ta được: tổng các XS phải bằng 1, tức
$$ 4a + b + \dfrac{6}{15} = 1 \Leftrightarrow b = \dfrac{3}{5} - 4a $$
và bảng phân phối XS lề cho $X$ như dưới đây

| range of $X$ | $X = 0$ | $X = 1$ | $X = 2$ |
|-|-|-|-|
|prob $X$| $\dfrac{3}{15} + a$ | $2a + b$ | $\dfrac{3}{15} + a$ |

và tương tự cho $Y$.

| range of $Y$ | $Y = 0$ | $Y = 1$ | $Y = 2$ |
|-|-|-|-|
|prob $Y$| $\dfrac{3}{15} + a$ | $2a + b$ | $\dfrac{3}{15} + a$ |

- Vì $a$ là giá trị làm $\text{Var}X$ đạt cực tiểu, nên ta sẽ tìm $a$ bằng cực tiểu hóa phương sai của $X$, ta có
$$ \mathbb{E}X = \displaystyle \sum_{x=0}^2 x P[X = x] = 0 + (2a+b) + 2 \left( a + \dfrac{1}{5} \right) = 4a + b + \dfrac{2}{5} = 1. \qquad \text{[Note that: ] } \dfrac{6}{15} = \dfrac{2}{5} .$$
và
$$ \mathbb{E}X^2 = \displaystyle \sum_{x=0}^2 x^2 P[X = x] = 0 + (2a+b) + (2^2) \left( a + \dfrac{1}{5} \right) = 6a + b + \dfrac{4}{5} = \left( 4a + b + \dfrac{2}{5} \right) + 2a = 2a + \dfrac{2}{5}. $$
suy ra
$$ \text{Var}X = \mathbb{E}X^2 - [ \mathbb{E}X ]^2 = 2a + \dfrac{2}{5} $$
và phương sai này nhỏ nhất khi $a = 0$.
- Hơn nữa, ta thấy rằng $X, Y$ có cùng phân phối xác suất; do 

$$P[X=k]=P[Y=k], \qquad \forall k = 0,1,2$$
vì vậy
$$ \text{Var}Y = \text{Var}X = \dfrac{2}{5} $$

#### Question 319.
Annual windstorm losses, $X$ and $Y,$ in two different regions are independent, and each is **uniformly distributed** on the interval `[0, 10]`.

Calculate the covariance of $X$ and $Y$, given that $X + Y < 10.$

**Giải đáp / SOLUTIONs.**

Theo các giả thiết, $X, Y \overset{\text{i.i.d}}{\sim} \mathcal{U}([0, 10])$ và có hàm MDXS đồng thời
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{100}, & 0 \leq x, y \leq 10 \\ 0, & \text{otherwise} \end{array} \right. $$
BT yêu cầu tính $\text{Cov}(X,Y)$ trên miền 

$$ A = \lbrace (X,Y): X+Y < 10, \qquad 0 \leq X,Y \leq 10 \rbrace $$

Do đó ta sẽ tính lại hàm MDXS của cặp $(X, Y)$ trên miền này, dễ thấy rằng

$$ f_A(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{50}, & 0 \leq x, y \leq 10, x+y < 10 \\ 0, & \text{otherwise} \end{array} \right. $$

Ta có
$$ \mathbb{E}X = \displaystyle \iint_A xf_A(x,y)dydx = \int_0^{10} \int_0^{10 - x} x(0.02) dydx = \dfrac{10}{3}$$
tương tự
$$ \mathbb{E}Y = \displaystyle \iint_A yf_A(x,y)dydx = \int_0^{10} \int_0^{10 - y} y(0.02) dxdy = \dfrac{10}{3}$$
và
$$ \mathbb{E}(XY) = \displaystyle \iint_A xyf_A(x,y)dydx = \int_0^{10} \int_0^{10 - y} xy(0.02) dxdy = \dfrac{25}{3}$$
Vậy

$$ \text{Cov}(X,Y) = \mathbb{E}(XY) - \mathbb{E}X \cdot \mathbb{E}Y = -\dfrac{25}{9}  $$

#### Question 320.
Let $X$ be a random variable that takes on the values $–1, 0,$ and $1$ with **equal probabilities.** Let $Y = X^2$

Which of the following is true? 
- A. $\text{Cov}(X,Y)>0$; the random variables $X$ and $Y$ are **dependent**
- B. $\text{Cov}(X,Y)>0$; the random variables $X$ and $Y$ are **independent**
- C. $\text{Cov}(X,Y)=0$; the random variables $X$ and $Y$ are **dependent**
- D. $\text{Cov}(X,Y)=0$; the random variables $X$ and $Y$ are **independent**
- E. $\text{Cov}(X,Y)<0$; the random variables $X$ and $Y$ are **dependent.**

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Ta thấy rằng $X$ chỉ nhận các giá trị $-1, 0, 1$ nên $Y = X^2$ chỉ nhận 2 giá trị $0, 1$ và
$$ P[X=0] = P[X=1] = P[X=-1] = \dfrac{1}{3} $$
tương tự
$$ P[Y = 0] = P[X = 0] = \dfrac{1}{3} $$
và
$$ P[Y=1] = P[X^2 = 1] = P[X=1] + P[X=-1] = \dfrac{2}{3} $$
Suy ra
$$ P[X=0, Y= 0] = P[X=0, X^2 = 0] = P[X=0] = \dfrac{1}{3} $$
nhưng
$$ P[X=0] \cdot P[Y=0] = \dfrac{1}{9} \neq \dfrac{1}{3} = P[X=0, Y= 0] $$
nên **$X, Y$ không thể độc lập, mà nó là phụ thuộc (dependent)** Lúc này ta **loại các đáp án `independent`: `B, D`**
- Hơn nữa, ta có
$$ \begin{array}{rcl} \mathbb{E}X &=& (-1) \cdot P[X=-1] + 0 + (1) \cdot P[X=1] = 0 \\ \mathbb{E}X^2 = \mathbb{E}Y &=& 0\cdot P[Y=0] + 1 \cdot P[Y=1] = \dfrac{2}{3} \\ \mathbb{E}X^3 &=& (-1)^3 \cdot P[X=-1] + 0 + (1)^3 \cdot P[X=1] =0 \end{array} $$
suy ra
$$ \begin{array}{rcl}  \text{Cov}(X, Y) &=& \mathbb{E}(XY) - \mathbb{E}X \mathbb{E}Y \\ &=& \mathbb{E}(X \cdot X^2) - (\mathbb{E}X) \cdot (\mathbb{E}X^2) \\ &=& \mathbb{E}(X^3) - (\mathbb{E}X) \cdot (\mathbb{E}X^2) \\ &=& 0 \end{array} $$
do đó chỉ còn `C` là đáp án đúng.
