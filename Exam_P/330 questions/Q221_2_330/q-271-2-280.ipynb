#### Question 271, 
Let $X$ be a random variable with density function 

$$ f(x) = \left\lbrace \begin{array}{cl} 2e^{-2x}, & x > 0 \\ 0, & \text{otherwise} \end{array} \right. $$

Calculate $P[X \leq 0.5 | X \leq 1.0]$

**Giải đáp / SOLUTIONs.**

- Với hàm MDXS $f$ và $x_0 \geq 0$, ta tính được

$$ P\left[ X \leq x_0 \right] = \int_0^{x_0} f(t) dt = 1 - e^{-2x_0}$$

- Áp dụng công thức xác suất có điều kiện, ta có

$$ P[X \leq 0.5 | X \leq 10] = \dfrac{P[(X \leq 0.5) \cap (X \leq 1.0)]}{P[X \leq 1.0]} = \dfrac{P[X \leq 0.5]}{P[X \leq 1.0]} = \dfrac{1 - e^{-2(0.5)}}{1 - e^{-2}} \approx 0.731. $$

#### Question 272.
Events $E$ and $F$ are independent. $P[E] = 0.84$ and $P[F] = 0.65.$

Calculate the probability that exactly one of the two events occurs.

**Giải đáp / SOLUTIONs.**
- Vì $E, F$ độc lập nên
$$ P[E \cap F] = P[E] \cdot P[F] $$
và hơn nữa $E^c, F$ và $F^c, E$ cũng độc lập
- BT cần tính XS để có đúng một trong 2 sự kiện trên xảy ra, tức là 
$$ \begin{array}{ccl} P[\text{"exactly 1 event"}] &=& P[E \cap F^c] + P[E^c \cap F] \\ &=& P[E]\cdot (1 - P[F]) + P[F] \cdot (1 - P[E]) \\ &=& (0.84)(1 - 0.65) + (0.65)(1 - 0.84) \\ &=& 0.398 \end{array} $$

#### Question 273.
- A flood insurance company determines that $N,$ the number of claims received in a month, is a random variable with 
$$P[N=n] = \dfrac{2}{3^{n+1}}, \qquad \text{ for }n = 0,1,2,\ldots .$$ 
- The numbers of claims
received in different months are mutually independent.

Calculate the probability that more than three claims will be received during a
consecutive two-month period, given that fewer than two claims were received in the first
of the two months. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $N_1, N_2$ là số các yêu cầu được nhận trong tháng thứ nhất và tháng thứ hai. Theo đó $N_1, N_2 \overset{\text{i.i.d}}{\sim} N$ với $N$ là số yêu cầu nhận được trong một tháng mà đề cho.
- Giả định còn lại cho biết $N_1, N_2$ độc lập, tức là

$$ P[N_1 = n_1, N_2 = n_2] = P[N_1 = n_1] \cdot P[N_2 = n_2], \qquad \forall n_1, n_2 = 0,1,2,\ldots $$
- BT yêu cầu tính XS để có nhiều hơn 3 yêu cầu trong 2 tháng liên tiếp, biết rằng tháng đầu tiên nhận được ít hơn 2 yêu cầu, tức là

$$ \begin{array}{ccl} P[N_1 + N_2 > 3 | N_1 < 2] &=& 1 - P[N_1 + N_2 \leq 3 | N_1 < 2] \\ &=& 1 - \displaystyle \sum_{0\leq n + m \leq 3}^{m < 2} \dfrac{P[N_1 = n, N_2 = m]}{P[N_2 < 2]} \end{array} $$
Ta có
$$ P[N_2 < 2] = \displaystyle \sum_{m=0}^1 P[N_2 = m] = \sum_{m=1}^2 \dfrac{2}{3^{m+1}} = \dfrac{8}{9}.$$
v
$$ \begin{array}{ccl} \displaystyle \sum_{0\leq n + m \leq 3}^{m < 2} P[N_1 = n, N_2 = m] &=& P(N_2=0)\cdot \left[ P(N_1=0) + P(N_1=1) + P(N_1=2) +P(N_1=3)\right] + \\ & & + P(N_2=1)\cdot \left[ P(N_1=0) + P(N_1=1) + P(N_1=2)\right] \\ &=& \dfrac{2}{3} \cdot \left[ \dfrac{2}{3} + \dfrac{2}{9} + \dfrac{2}{27} + \dfrac{2}{81} \right] \quad + \quad \dfrac{2}{9} \cdot  \left[ \dfrac{2}{3} + \dfrac{2}{9} + \dfrac{2}{27} \right] \\ &=& \dfrac{220}{243}  \end{array} $$
do đó
$$  P[N_1 + N_2 > 3 | N_1 < 2] = 1 - \displaystyle \sum_{0\leq n + m \leq 3}^{m < 2} \dfrac{P[N_1 = n, N_2 = m]}{P[N_2 < 2]} = 1 -  \dfrac{\left( \dfrac{220}{243} \right)}{\left(\dfrac{8}{9} \right)} = 0.0185. $$

#### Question 274.
- Patients in a study are tested for sleep apnea, one at a time, until a patient is found to have
this disease. 
- Each patient independently has the same probability of having sleep apnea.
- Let r represent the probability that at least four patients are tested.

Determine the probability that at least twelve patients are tested given that at least four
patients are tested.


**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số bệnh nhân đã được kiểm định về việc có triệu chứng `ngưng thở (apnea)` khi ngủ hay không; và theo giả thiết đầu tiên việc tìm các triệu chứng **sẽ dừng cho đến khi phát hiện** được chứng ngưng thở nên 

$$X \sim \text{Geom}(\lambda), \qquad \lambda \in (0, 1).$$ 
Do đó, XS để có ít nhất $n$ bệnh nhân được kiểm định đồng nghĩa với $n-1$ người trước đó không phát hiện được triệu chứng này, do đó

$$ P[X \geq n] = (1- \lambda)^{n-1}, \qquad \forall n \geq 1 $$

- BT đã đặt $r = P[X \geq 4] = (1 - \lambda)^3$ và yêu cầu ta xác định $P[X \geq 12 | X \geq 4]$ theo tỷ lệ $r$ này. Ta có

$$ P[X \geq 12 | X \geq 4] = \dfrac{P[(X \geq 12) \cap (X \geq 4)]}{P[X \geq 4]} = \dfrac{(1-\lambda)^{11}}{(1-\lambda)^3} = \left[ \left( 1 - \lambda \right)^3 \right]^{8/3} = r^{8/3} $$

#### Question 275.
A factory tests 100 light bulbs for defects. The probability that a bulb is defective is 0.02.
The occurrences of defects among the light bulbs are mutually independent events.

Calculate the probability that exactly two are defective given that the number of defective
bulbs is two or fewer. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_i$ là kết quả kiểm định bóng đèn thứ $i^{th}$ có phải phế phẩm (có sai sót) hay không, tức

$$ X_i = \left\lbrace \begin{array}{cl} 1, & \text{la phe pham (co sai sot) voi xac suat } 0.02 \\ 0, & \text{otherwise} \end{array} \right. $$
- Do đó với giả thiết 100 bóng đèn thì $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.02)$ và
$$ X = \displaystyle \sum_{i=1}^{100} X_i \sim \mathcal{B}(100, 0.02) $$
là số phế phẩm trong 100 bóng đèn.
- BT yêu cầu tính XS có đúng 2 phế phẩm biết rằng số phế phẩm ít hơn hoặc bằng 2; tức là

$$ \begin{array}{ccl} P[X = 2|X \leq 2] &=& \dfrac{P[(X = 2) \cap (X \leq 2)]}{P[X \leq 2]} \\ &=& \dfrac{P[X=2]}{P[X=0]+P[X=1] +P[X=2]} \\ &=& \dfrac{C_{100}^2 (0.02)^2 (1 - 0.02)^{100 - 2} }{C_{100}^0 (0.98)^{100} + C_{100}^1 (0.02) (0.98)^{99} + C_{100}^2 (0.02)^2 (0.98)^{98}} \\ & \approx & 0.404. \end{array} $$

#### Question 276.
- A certain town experiences an average of 5 tornadoes in any four year period. 
- The number of years from now until the town experiences its next tornado as well as the
number of years between tornados have identical exponential distributions and all such
times are mutually independent

Calculate the median number of years from now until the town experiences its next
tornado. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là thời gian (tính bằng số năm) cho đến khi thị trấn xuất hiện cơn lốc kế tiếp và giả định thứ hai cho biết $X \sim \mathcal{E}(\lambda)$
- Theo giả thiết, cứ 4 năm sẽ có 5 cơn bão nên trung bình $\lambda = \dfrac{4}{5} = 0.8$ (số cơn bão / năm); do đó

$$ P[X < x] = 1 - e^{-x/(0.8)}, \qquad \forall x \geq 0. $$
- BT yêu cầu tìm trung vị về thời gian đến khi xuất hiện tiếp tục một cơn bão khác, tức là tìm $x_m$ thỏa

$$ \begin{array}{ccrclrc} 0.5 = P[X < x_m] &\Leftrightarrow & 1 - e^{-x_m/(0.8)} &=& 0.5 \\ &\Leftrightarrow & x_m &=& -(0.8) \ln (0.5) &=& 0.55 \end{array} $$

#### Question 277.
Losses under an insurance policy are exponentially distributed with mean 4. The
deductible is 1 for each loss.

Calculate the median amount that the insurer pays a policyholder for a loss under the
policy. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là khoản tổn thất từ BH **khi chưa có khấu hao**; theo đó $X \sim \mathcal{E}(4).$
- và $Y = \max(0, X - 1)$ là khoản tổn thất **khi có khấu hao** $1$
- BT yêu cầu xác định trung vị $m_y$ của $Y$, theo đó ta sẽ xác định trung vị $m_x$ bởi sau đó 

$$m_y = \max(0, m_x - 1) $$
- Lại có $m_x$ là trung vị của $X$ nên
$$ \begin{array}{ccrclrc} 0.5 = P[X < x_m] &\Leftrightarrow & 1 - e^{-\frac{x_m}{4}} &=& 0.5 \\ &\Leftrightarrow & x_m &=& -4 \ln (0.5) &=& 2.77 \end{array} $$
do đó

$$m_y = \max(0, 2.77 - 1) = 1.77 .$$

#### Question 278.
- A company has purchased a policy that will compensate for the loss of revenue due to
severe weather events.
- The policy pays 1000 for each severe weather event in a year
after the first two such events in that year. 
- The number of severe weather events per year
has a Poisson distribution with mean 1.

Calculate the expected amount paid to this company in one year.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số lần mà thời tiết trở nên khắc nghiệt và $Y = g(X)$ là số tiền bồi thường do các hiện tượng đó gây ra trong năm.
- Theo các giả thiết đề bài, ta có
$$ g(X) = \left\lbrace \begin{array}{cl} 1000(X - 2), & X = 3,4,\ldots \\ 0, & X = \lbrace 0,1,2 \rbrace \end{array} \right. $$
và 
$$ X \sim \text{Pois}(1), \text{ i.e } P[X = k] = e^{-1} \dfrac{1}{k!} $$
- BT yêu cầu tính kỳ vọng số tiền bồi thường trong năm, tức là

$$ \begin{array}{ccl} \mathbb{E}Y &=& \displaystyle \sum_{k=0}^{\infty} g(x) P[X = x] \\ &=& \displaystyle 1000 \left[ \left( \sum_{k=0}^{\infty} (x-2)P[X = x] \right) - \left( \sum_{k=0}^{2} (x-2) P[X=x] \right) \right] \\ &=& 1000 \left[ \mathbb{E}(X-2) - \left( (0-2)\cdot \dfrac{e^{-1}}{0!} +  (1-2)\cdot \dfrac{e^{-1}}{1!} +  (2-2)\cdot \dfrac{e^{-1}}{2!}  \right) \right] \\ &=& 1000 \left[ \mathbb{E}(X) - 2 - \left( -2e^{-1} - e^{-1} \right) \right] \\ &=& 104 \end{array} $$

vì $\mathbb{E}X = 1$ với $X \sim \text{Pois}(1)$.
- Do đó kỳ vọng số tiền mà BH sẽ bồi thường là $ \mathbb{E}Y = 104$.

#### Question 279.
- A company provides each of its employees with a death benefit of 100. The company
purchases insurance that pays the cost of total death benefits in excess of 400 per year.
- The number of employees who will die during the year is a Poisson random variable with
mean 2.

Calculate the expected annual cost to the company of providing the death benefits,
excluding the cost of the insurance. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số nhân viên chết đi hàng năm ở cty; theo giả thiết thì $X \sim \text{Pois}(2)$
- $Y = g(X)$ là số tiền cty sẽ chi trả hằng năm để bù đắp cho những nhân viên này, khi đó

$$ g(X) =  \left\lbrace \begin{array}{cl} 100 \cdot X, & X = 1,2,3 \\ 400, & X = 4,5,\ldots \end{array} \right.  $$
- BT yêu cầu tính kỳ vọng số tiền chi trả, tức là


$$ \begin{array}{ccl} \mathbb{E}Y &=& \displaystyle \sum_{k=0}^{\infty} g(x) P[X = x] \\ &=& 100 P[X=1] + 200 P[X=2] + 300 P[X=3] + 400 P[X \geq 4] \\ &=& 100 \left( \dfrac{e^{-2} 2}{1!} \right) + 200 \left( \dfrac{e^{-2} 2^2}{2!} \right) + 300 \left( \dfrac{e^{-2} 2^3}{3!} \right) + 400 \left( 1 - \dfrac{e^{-2} 2^0}{0!} - \dfrac{e^{-2} 2}{1!} - \dfrac{e^{-2} 2^2}{2!} - \dfrac{e^{-2} 2^3}{3!} \right) \\ &=& 192. \end{array} $$

#### Question 280.
The number of burglaries occurring on Burlington Street during a one-year period is
Poisson distributed with mean 1.

Calculate the expected number of burglaries on Burlington Street in a one-year period,
given that there are at least two burglaries.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số vụ trộm (`burglaries`) tại đường Burlington trong một năm; theo đó $X \sim \text{Pois}(1)$
- BT yêu cầu tính kỳ vọng có điều kiện $\mathbb{E}[X | X \geq 2]$, ta có

$$ \begin{array}{ccl} \mathbb{E}[X | X \geq 2] &=& \displaystyle \sum_{x=0}^{\infty} x \left( \dfrac{P[(X = x) \cap (X \geq 2)]}{P[X \geq 2]} \right) \\ &=& \dfrac{1}{P[X \geq 2]} \displaystyle \left( \sum_{x=2}^{\infty} x P[X = x] \right) \\ &=&  \dfrac{1}{1 - P[X = 0] - P[X = 1]} \displaystyle \left[ \left( \sum_{x=0}^{\infty} x P[X = x] \right) - 0 \cdot P[X = 0] - 1 \cdot P[X = 1] \right] \\ &=& \dfrac{1}{1 - e^{-1} - e^{-1} } \left[ \mathbb{E}X - 0 - e^{-1} \right] \\ &=& \dfrac{1}{1 - 2e^{-1}} \left[ 1 - e^{-1} \right] \\ &=& 2.39 \end{array} $$
