#### Question 291, 
- A government employee’s yearly dental expense follows a **uniform distribution** on the interval from $200$ to $1200.$ 
- The government’s primary dental plan reimburses an employee for up to 400 of dental expense incurred in a year, while a supplemental plan pays up to 500 of any remaining dental expense.
- Let $Y$ represent the yearly benefit paid by the supplemental plan to a government employee.

Calculate $\text{Var}(Y)$.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là khoản chi phí hàng năm tại phòng khám của nhân viên, theo đó $X \sim \mathcal{U}([200, 1200])$ với hàm MDXS
$$ f(x) = \dfrac{1}{1200 - 200} \textbf{1}_{\lbrace x \in [200, 1200] \rbrace} = \dfrac{1}{1000} \cdot \textbf{1}_{\lbrace x \in [200, 1200] \rbrace}  $$
- Các giả thiết còn lại cho biết
$$ Y = h(X) = \left\lbrace \begin{array}{cl} 0, & X \in [200, 400] \\ X - 400, & X \in (400, 900] \\ 500, & X \in (900, 1200] \end{array} \right. $$
- BT lại yêu cầu tính $\text{Var}Y$, ta có
$$ \begin{array}{ccl} \mathbb{E}Y &=& \displaystyle \int h(x) f(x) dx \\ &=& \displaystyle \int_{200}^{400} 0 \cdot f(x) dx + \int_{400}^{900} (x-400) f(x)dx + \int_{900}^{1200} 500 f(x)dx \\ &=& \displaystyle 0 + \int_{400}^{900} \dfrac{x-400}{1000} dx + \int_{900}^{1200} \dfrac{500}{1000}dx \\ &=& 275 \end{array}$$
và tương tự,
$$ \begin{array}{ccl} \mathbb{E}Y^2 &=& \displaystyle \int_{200}^{400} 0^2 \cdot f(x) dx + \int_{400}^{900} (x-400)^2 f(x)dx + \int_{900}^{1200} 500^2 f(x)dx \\ &=& \displaystyle 0 + \int_{400}^{900} \dfrac{(x-400)^2}{1000} dx + \int_{900}^{1200} \dfrac{500^2}{1000}dx \\ &\approx& 116667.667 \end{array}$$
do đó
$$ \text{Var}Y = \mathbb{E}Y^2 - [\mathbb{E}Y]^2 \approx 41041.667. $$

#### Question 292.
- Under a liability insurance policy, losses are uniformly distributed on $[0, b],$ where $b$ is a positive constant. 
- There is a deductible of $b/2.$

Calculate the ratio of the variance of the claim payment (greater than or equal to zero)
from a given loss to the variance of the loss. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là khoản tổn thất mà BH phải chi trả, theo đó $X \sim \mathcal{U}([0, b])$ với $b > 0$ là một hằng số,
- Đặt $L$ là phí chi trả sau khi khấu trừ mức $b/2$ vào nguồn tiền tổn thất $X$ (mã vẫn lớn hơn hoặc bằng $0$), theo đó $C = \max\left( 0, X - \dfrac{b}{2} \right).$ Như vậy, ta đã có
$$ \mathbb{E}X = \dfrac{b - 0}{2} = \dfrac{b}{2}, \quad \text{ and } \quad \text{Var}X = \dfrac{(b - 0)^2}{12} = \dfrac{b^2}{12}. $$
- BT yêu cầu tính tỷ số các phương sai: $\dfrac{\text{Var}Y}{\text{Var}X}$ và ta chỉ cần tìm $\text{Var}Y.$ Ta có

$$ \begin{array}{ccl} \mathbb{E}Y &=& \displaystyle \int_0^{b/2} 0 \cdot \frac{1}{b} dx + \int_{b/2}^b \left( x - \frac{b}{2} \right) \cdot \frac{1}{b} dx &=& 0 + \left. \left[ \dfrac{\left( x - \frac{b}{2} \right)^2}{2b} \right] \right\vert_{x = b/2}^b = \dfrac{b}{8} \end{array} $$
và tương tự
$$ \begin{array}{ccl} \mathbb{E}Y^2 &=& \displaystyle \int_0^{b/2} 0 \cdot \frac{1}{b} dx + \int_{b/2}^b \left( x - \frac{b}{2} \right)^2 \cdot \frac{1}{b} dx &=& 0 + \left. \left[ \dfrac{\left( x - \frac{b}{2} \right)^3}{3b} \right] \right\vert_{x = b/2}^b = \dfrac{b^2}{24} \end{array} $$
do đó
$$ \text{Var}Y = \mathbb{E}Y^2 - [\mathbb{E}Y]^2 = \dfrac{5b^2}{192} $$
suy ra

$$ \dfrac{\text{Var}Y}{\text{Var}X} = \dfrac{\left( \frac{5b^2}{192} \right)}{\left( \frac{b^2}{12} \right)} = \dfrac{5}{16}. $$

#### Question 293.
- A company’s annual profit, in billions, has a **normal distribution** with variance equal to the cube of its mean. 
- The probability of an annual loss is $5\%.$

Calculate the company’s expected annual profit. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là mức lợi nhuận hàng năm của cty, theo đó $X \sim \mathcal{N}\left( \mu, \sigma^2 \right)$ và theo giả định đầu tiên

$$ \sigma^2 = \mu^3. $$
- Giả thiết còn lại cho biết 
$$ \begin{array}{ccl} P[X < 0] = 0.05 \Leftrightarrow P \left[ \mathcal{N}(0,1) < \dfrac{0 - \mu}{\sigma} \right] = 0.05 & \Leftrightarrow & \dfrac{-\mu}{\sigma} = \Phi^{-1}(0.05) = -1.645 \\ & \Leftrightarrow & \mu = (1.645) \sigma \end{array} $$
- BT yêu cầu tìm $\mu$ với lưu ý rằng đơn vị đang tính là hàng tỷ (`billions`). Thay $\mu = (1.645) \sigma$ vào $\mu^3 = \sigma^2$, ta tìm được $$\left\lbrace \begin{array}{ccl} \mu &=& 0.3695 \\ \sigma &=& 0.2249 \end{array} \right.$$
Do đó, kỳ vọng số tiền lãi của cty là vào khoảng $369.5$ triệu (hay $0.3695$ tỷ.)

#### Question 294.
The number of claims $X$ on a health insurance policy is a random variable with
$\mathbb{E}\left[ X^2 \right] = 61$ and $\mathbb{E}\left[(X - 1)^2 \right] = 47$.

Calculate the standard deviation of the number of claims. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu tính độ lệch chuẩn 

$$ \text{sd}X = \sqrt{\text{Var}X} = \sqrt{\mathbb{E}X^2 - [\mathbb{E}X]^2} $$
- Do đó, ta chỉ cần tính được $\mathbb{E}X$. Ta có

$$ \mathbb{E}\left[(X - 1)^2 \right] - \mathbb{E}\left[X^2 \right] = \left( \mathbb{E}\left[X^2 \right] - 2 \mathbb{E}\left[X \right] + 1 \right) - \mathbb{E}\left[X^2 \right] = 1 - 2 \mathbb{E}\left[X \right] $$

Thay $\mathbb{E}[X^2] = 61$ và $\mathbb{E}\left[(X - 1)^2 \right] = 47$, ta được

$$ 1 - 2 \mathbb{E}\left[X \right] = 47 - 61 \Longrightarrow \mathbb{E}[X] = 7.5 $$
do đó
$$ \text{sd}X = \sqrt{\text{Var}X} = \sqrt{\mathbb{E}X^2 - [\mathbb{E}X]^2} \approx 2.18. $$


#### Question 295.
- Ten cards from a deck of playing cards are in a box: two diamonds, three spades, and five hearts. 
- Two cards are randomly selected without replacement.

Calculate the variance of the number of diamonds selected, given that no spade is
selected.

**Giải đáp / SOLUTIONs.**

- **Tóm tắt BT & suy diễn**

$\qquad \diamond$ Đặt $\Omega = \lbrace 2D, 3S, 5H \rbrace$ là KG mẫu chứa các lá bài trong hộp với $D, S, H$ lần lượt là ký hiệu các loại bài (`diamond` [rô], `spades`[chuồng] và `hearts` [cơ]).

$\qquad \diamond$ Người chơi sẽ rút ngẫu nhiên 2 lá bài không hoàn lại.

$\qquad \diamond$ BT yêu cầu tính phương sai có điều kiện về số quân $D$ rút được biết rằng không có lá $S$ nào được chọn
- **Sơ lược mô hình & giải đáp**.

$\qquad \diamond$ Đặt $N_S, N_D, N_H$ lần lượt là số lá $S, D, H$ mà người chơi rút được, khi đó BT sẽ trở thành phép tính $\text{Var}\left( N_D | N_S = 0 \right)$.

$\qquad \diamond$ Vì đề cho ta chỉ có 2 lần rút bài nên giá trị của $N_D = n$ chỉ gồm $n \in \lbrace 0,1,2 \rbrace$

$\qquad \diamond$ Đầu tiên, ta cần xác định $P[N_S = 0]$ rồi $P[N_D = n]$ với $n \in \lbrace 0,1,2 \rbrace$ bởi ta sẽ tính $\text{Var}\left( N_D | N_S = 0 \right)$ dựa trên các moment

$$ \mathbb{E}\left( N_D^k | N_S = 0 \right) = \displaystyle \sum_{n = 0}^2 n^k P[N_D = n | N_S = 0], \qquad k = 1,2,\ldots.$$

$\qquad \diamond$ Xét $P[N_S = 0]$, đây là XS để **sau 2 lần rút, không chọn được lá $S$ nào từ $\Omega$**, theo đó

$$ \begin{array}{ccrcl} P[N_S = 0] &=& \dfrac{|\text{"select 0 S from 3S of "} \Omega| \cdot |\text{"select 2 from 7 cards of (2D, 5H)"} |}{|\text{"select a set of 2 in 10 cards from "} \Omega|} &=& \dfrac{C_3^0 \cdot C_{7}^2}{C_{10}^2} = \dfrac{7}{15} \\ \end{array} $$

$\qquad \diamond$ Với $P[N_D = n | N_S = 0]$, ta lại dùng xác suất có điều kiện (với lưu ý tổng số lá được rút là 2 và tổng số phần tử là 10)

$$ \dfrac{P[N_D = n, N_S = 0]}{P[N_S = 0]}$$
ta có
$$ \begin{array}{ccl} P[N_D = n, N_S = 0] &=& \dfrac{|\text{"select 0 S from 3S of"} \Omega| \cdot |\text{"sel n D from 2D of "} \Omega| \cdot |\text{"sel (2-n) H from 2H of "} \Omega|}{|\text{"select a set of 2 in 10 cards from "} \Omega|} \\ &=& \dfrac{C^0_2 \cdot C_2^n \cdot C_7^{2-n}}{C_{10}^2}. \qquad \forall n = 0,1,2. \end{array} $$
suy ra
$$ \begin{array}{ccl} \mathbb{E}\left( N_D | N_S = 0 \right) &=& \displaystyle \sum_{n = 0}^2 n \dfrac{P[N_D = n, N_S = 0]}{P[N_S = 0]} \\ &=& \displaystyle \sum_{n = 0}^2 n \left( \dfrac{ \left( \dfrac{C^0_2 \cdot C_2^n \cdot C_7^{2-n}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) \\ &=& 0 \cdot \left( \dfrac{ \left( \dfrac{C^0_2 \cdot C_2^0 \cdot C_7^{2}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) + 1 \cdot \left( \dfrac{ \left( \dfrac{C^1_2 \cdot C_2^1 \cdot C_7^{1}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) + 2 \cdot \left( \dfrac{ \left( \dfrac{C^2_2 \cdot C_2^2 \cdot C_7^{0}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) = \dfrac{4}{7} \end{array}$$

và tương tự
$$ \begin{array}{ccl} \mathbb{E}\left( N_D^2 | N_S = 0 \right) &=& (0^2) \cdot \left( \dfrac{ \left( \dfrac{C^0_2 \cdot C_2^0 \cdot C_7^{2}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) + (1^2) \cdot \left( \dfrac{ \left( \dfrac{C^1_2 \cdot C_2^1 \cdot C_7^{1}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) + (2^2) \cdot \left( \dfrac{ \left( \dfrac{C^2_2 \cdot C_2^2 \cdot C_7^{0}}{C_{10}^2} \right)}{\dfrac{7}{15}} \right) = \dfrac{2}{3} \end{array}$$
do đó
$$ \text{Var}\left( N_D | N_S = 0 \right) = \mathbb{E}\left( N_D^2 | N_S = 0 \right) - \left[ \mathbb{E}\left( N_D | N_S = 0 \right) \right]^2 = \dfrac{50}{147} \approx 0.34. $$

#### Question 296.
A homeowners insurance policy covers losses due to theft, with a deductible of 3. Theft
losses are uniformly distributed on $[0, 10].$

Determine the moment generating function, $M(t),$ for $t \neq 0,$ of the claim payment on a
theft.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là khoản thiệt hại của chủ hộ do lũ trộm gây ra và $Y$ là khoản chi trả từ BH với khấu hao bằng $3$. Theo đó 

$$ X \sim \mathcal{U}([0, 10]), \qquad \text{and} \qquad Y = \max(0, X - 3)$$
- BT yêu cầu tính hàm sinh moment cho $Y$, ta có
$$ M(t) = \mathbb{E}[e^{Yt}] = \displaystyle \int e^{t \cdot \max(0, x-3)} f(x)dx $$
trong đó $f(x) = \dfrac{1}{10} \textbf{1}_{\left\lbrace x \in [0, 10] \right\rbrace}$ là hàm MDXS của $X$, do đó

$$ \begin{array}{ccrcl} M(t) &=& \displaystyle \int_0^3 e^{0} \dfrac{1}{10} dx + \int_3^{10} e^{t(x-3)} \dfrac{1}{10}dx \\ &=& \displaystyle \dfrac{1}{10} \left[ \int_0^3 1 \cdot dx + \int_0^{7} e^{t \cdot u} du \right] \\ &=& \displaystyle \dfrac{1}{10} \left[ 3 + \left.\left( \dfrac{e^{ut}}{t} \right)\right\vert_{u=0}^7 \right] &=& \dfrac{1}{10} \left[ 3 + \dfrac{e^{7t} - 1}{t} \right] \end{array} $$
Vậy
$$ M(t) = \dfrac{3}{10} + \dfrac{e^{7t} - 1}{10 t} $$

#### Question 297.
The random variable $X$ has moment generating function $M(t)$.

Determine which of the following is the moment generating function of some random variable.

- A. $M(t)M(5t)$
- B. $2M(t)$
- C. $e^tM(t)$

**Giải đáp / SOLUTIONs.**

- Cho $X, Y$ là các BNN độc lập với các hàm sinh moment $M_X(t)$ và $M_Y(t)$, khi đó hàm sinh moment của $Z = X+ Y$ sẽ xác định bởi

$$ M_Z(t) = M_X(t) M_Y(t) $$

dựa vào đây, cả 3 đáp án trong bài đều có dạng tích các hàm sinh với $M(t)$ nên **ta phải chọn hàm sinh moment $M_Y(t)$ hợp lệ trong số các $M(5t), 2$ và $e^t$.**
- Nếu $Y$ là BNN có các giá trị rời rạc $y \in \Omega$ thì
$$ M_Y(y) = \displaystyle \sum_{y \in \Omega} e^{yt} P[Y = y] $$
từ đây suy ra 

$\qquad \diamond$ Nếu $M_Y(t) = e^t,$ thì tương ứng với $Y$ nhận giá trị $y = 1$ và là giá trị duy nhất của $Y$ (hợp lệ)

$\qquad \diamond$ Nếu $M_Y(t) = 2,$ thì lưu ý rằng nếu $y = 0$ thì $M_Y(t) = 1$ (lưu ý rằng $P[Y = y] \in [0, 1]$) do đó không tồn tại giá trị $y$ nào hợp lệ để $M_Y(t) = 2$

- Nếu $Y$ là hàm tuyến tính theo $X$, tức là tồn tại $a \neq 0$ sao cho $Y = aX$ thì 

$$ M_Y(t) = \mathbb{E}e^{Yt} = \mathbb{E}e^{(aX)t} = \mathbb{E}e^{X(at)} = M_X(at) $$

$\qquad \diamond$ Do đó $M_Y(t) = M(5t)$ tương ứng với $a = 5$ hay $Y = 5X$ và đây cũng là BNN hợp lệ.

- Như vậy trong số 3 hàm sinh moment đã cho,, chỉ có 2 đáp án hợp lệ :

$$ M(t)M(5t), \qquad \text{and} \qquad e^tM(t). $$

#### Question 298.
- The number of boating accidents a policyholder experiences this year is modeled by a **Poisson** random variable with variance $0.10.$
- An insurer reimburses only the first accident.
- Let $Y$ be the number of unreimbursed accidents the policyholder experiences this year and let $p$ be the probability function of $Y$.

Determine $p(y).$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số vụ tai nạn xảy ra và $Y$ là số vụ tai nạn **không được nhận bồi thường**.
- Theo giả thiết, $X \sim \text{Pois}(\lambda)$ với phương sai $\text{Var}X = 0.10$ do đó $\lambda = 0.1$ hay

$$ P[X = k] = \dfrac{e^{-0.1} (0.1)^k}{k!}, \qquad \forall k = 0,1,2, \ldots. $$
- Vì BH chỉ bồi thường cho tai nạn đầu tiên; do đó 
$$ Y = \max(0, X - 1). $$ 
- BT yêu cầu tính xác suất $p(y) = P[Y = y]$ của $Y$ và với biểu thức trên, $Y$ có cùng tập giá trị với $X$, như vậy:

$\qquad \diamond$ Khi $y = 0$, ta có
$$ \begin{array}{crlcl} p(0) = P[\max(0, X - 1) = 0] &=& P[X - 1 \leq 0] \\ &=& P[X = 0] + P[X = 1] \\ &=& e^{-0.1} + \dfrac{e^{-(0.1)} (0.1)}{1!} &=& (1.1) e^{-(0.1)}\end{array} $$
$\qquad \diamond$ Khi $y > 0$, ta có
$$ \begin{array}{crlcl} p(y) = P[\max(0, X - 1) = y] &=& P[X - 1 = y] \\ &=& P[X = 1+y] &=& \dfrac{e^{-(0.1)} \cdot (0.1)^{1+y}}{(1+y)!} \end{array} $$
Như vậy
$$ p(y) = \left\lbrace \begin{array}{cl} (1.1)e^{-(0.1)}, & y = 0 \\ \dfrac{e^{-(0.1)} \cdot (0.1)^{1+y}}{(1+y)!}, & y = 1,2,\ldots \end{array} \right. $$

#### Question 299.
Let $X$ be a continuous random variable with probability density function 

$$ f(x) = \left\lbrace \begin{array}{cl} \lambda e^{-\lambda x}, & x > 0, \\ 0, & \text{otherwise} \end{array} \right. $$
where $\lambda > 0.$

Let $Y$ be the **smallest integer greater than or equal** to $X.$

Determine the probability function of $Y.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Vì $Y$ là **số nguyên dương nhỏ nhất lớn hơn hoặc bằng $X$** nên $Y$ nhận giá trị là các số nguyên dương.
- Theo đó, với mỗi $n \in \mathbb{N}$ thì $X \leq n$ và ràng buộc $X > n-1$, do đó

$$ \displaystyle P[Y = n] = P[n-1 < X \leq n] = \int_{n-1}^n f(x)dx = \int_{n-1}^n \lambda e^{-\lambda x} dx = -e^{-n\lambda}+e^{-(n-1)\lambda} = e^{-n\lambda} \left( e^{\lambda} - 1 \right) $$

#### Question 300.
- Batteries $A$ and $B$ have lifetimes that are independent and exponentially distributed with a common mean of $m$ years. 
- The probability that battery $B$ outlasts battery $A$ by more than one year is $0.33.$

Calculate $m.$

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X, Y$ lần lượt là tuổi thọ của các loại pin $A, B$.
- Theo các giả thiết, thì $X, Y$ độc lập với hàm MDXS đồng thời 
$$ f(x,y) = \left\lbrace \begin{array}{cl} \dfrac{1}{m^2} e^{-\frac{(x+y)}{m}}, & x \geq0, y \geq 0 \\ 0, & \text{otherwise} \end{array} \right. $$
- BT yêu cầu tìm $m$ và ta dựa vào giả thiết cuối cùng, nó đã cho biết pin loại $B$ có hạn dùng lâu hơn loại $A$ một năm với XS bằng $0.33$, tức là

$$ \begin{array}{ccrcl} P[Y > X + 1] = 0.33 & \Leftrightarrow & \displaystyle \iint_{\left\lbrace (x,y): x,y \geq 0, y>x-1 \right\rbrace} f(x,y)dydx &=& 0.33 \\ & \Leftrightarrow & \displaystyle \iint_{\left\lbrace 0\leq x < \infty, x+1 < y < \infty \right\rbrace} f(x,y)dydx &=& 0.33 \\ & \Leftrightarrow & \displaystyle \int_{0}^{\infty} \int_{x+1}^{\infty} \left( \dfrac{1}{m^2} e^{-\frac{(x+y)}{m}} \right) dydx &=& 0.33 \\  & \Leftrightarrow & \displaystyle \int_{0}^{\infty} \left. \left( \dfrac{e^{-\frac{(x+y)}{m}}}{m} \right) \right\vert_{y = x+1}^{y \to \infty} dx &=& 0.33 \\ & \Leftrightarrow & \displaystyle \int_{0}^{\infty} \dfrac{e^{-\frac{(2x+1)}{m}}}{m} dx &=& 0.33 \\ & \Leftrightarrow & \displaystyle \left. \left( \dfrac{e^{-\frac{(2x+1)}{m}}}{2} \right) \right\vert_{x = 0}^{x \to \infty} &=& 0.33 \\  & \Leftrightarrow & (0.5) e^{-(1/m)} &=& 0.33 \\  & \Leftrightarrow & m &=& \dfrac{-1}{\ln (0.66)} = 2.41 \end{array} $$
