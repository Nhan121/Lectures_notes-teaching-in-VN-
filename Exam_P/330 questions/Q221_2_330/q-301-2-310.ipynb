#### Question 301, 
- A car and a bus arrive at a railroad crossing at times independently and uniformly
distributed between `7:15` and `7:30`. 
- A train arrives at the crossing at `7:20` and halts traffic
at the crossing for `five minutes`.

Calculate the probability that the waiting time of the car or the bus at the crossing
exceeds three minutes

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là thời gian (tính bằng phút) đến đường tàu và $Y, Z$ thời gian chờ của oto, bus.
- BT yêu cầu tính XS để $Y$ hoặc $Z$ phải đợi hơb $3$ phút, tức là có ít nhất một phương tiện phải đợi hơn 3 phút,

$$ P[Y \geq 3] + P[Z \geq 3] + P[Y \geq 3 \cap Z \geq 3] $$
- Vì $Y$ hoặc $Z$ phải đợi lâu hơn $3$ phút cùng với giả định tàu sẽ chạy qua ngã tư vào lúc `7:20`, cùng với thời gian dừng hoạt động $5$ phút là sẽ giao thông trở lại vào lúc `7:25` nên để $Y, Z$ đợi hơn $3$ phút thì các phương tiện này phải đến trước `7:22` và không được sớm hơn `7:20`, nghĩa là
$$ P[Y \geq 3] = P[Z \geq 3] = P[X \in [20: 22]] = \dfrac{22 - 20}{30 -15} = \dfrac{2}{15}, \qquad \text{for } X \sim \mathcal{U}([15, 30]) $$
- Hơn nữa, thời gian chờ $Y, Z$ là độc lập nên

$$ P[Y \geq 3 \cap Z \geq 3] = P[Y \geq 3] \cdot P[Z \geq 3] = \dfrac{4}{225} $$
do đó

$$ P[Y \geq 3] + P[Z \geq 3] + P[Y \geq 3 \cap Z \geq 3] = \dfrac{2}{15} + \dfrac{2}{15} + \dfrac{4}{225} = \dfrac{56}{225} $$

#### Question 302.
An insurance company sells automobile liability and collision insurance. 
- Let $X$ denote the percentage of liability policies that will be renewed at the end of their terms and $Y$ the
percentage of collision policies that will be renewed at the end of their terms. 
- $X$ and $Y$ have a joint cumulative distribution
$$ F(x,y) = \dfrac{xy(x+y)}{2000000}, \qquad 0 \leq x, y \leq 100. $$
Calculate the probability that at least 80% of liability policies and 80% of collision
policies will be renewed at the end of their terms. 

**Giải đáp / SOLUTIONs.**

- Từ hàm phân phối tích lũy đồng thời $F$ trên miền compact $[0, 100] \times [0, 100]$, ta có các quan sát sau:

$$ \left\lbrace \begin{array}{rcl} P(X < x) &=& F(x, 100), \qquad x \in [0, 100) \\ P(Y < y) &=& F(100, y), \qquad y \in [0, 100) \\ P(X < x, Y < y) &=& F(x, y), \qquad x,y \in [0, 100] \end{array} \right. $$

- BT yêu cầu tính 

$$ \begin{array}{ccl} P[(X \geq 80) \cap (Y \geq 80)] &=& 1 - P[(X < 80) \cup (Y < 80)] \\ &=& 1 - \left( P[X < 80] + P[Y < 80] - P[(X < 80) \cap (Y < 80)]  \right) \\ &=& 1 - \left( F(80, 100) + F(100, 80) - F(80, 80) \right) \\ &=& 0.072. \end{array} $$

#### Question 303.
Skateboarders $A$ and $B$ practice one difficult stunt until becoming injured while attempting the stunt. 
- On each attempt, the probability of becoming injured is $p,$ independent of the outcomes of all previous attempts.
- Let $F(x, y)$ represent the probability that skateboarders $A$ and $B$ make no more than $x$ and $y$ attempts, respectively, where $x$ and $y$ are positive integers.
- It is given that $F(2, 2) = 0.0441$.

Calculate $F(1, 5).$ 

**Giải đáp / SOLUTIONs.**

- Đặt $T$ là số lần thử động tác của một VDV cho đến khi có chấn thương (với XS xảy ra bằng $p$), theo đó $T \sim \text{Geom}(p)$, với

$$ P[T = k] = p(1-p)^{k-1}, \qquad \forall k = 1,2,\ldots $$
và
$$ P[T > k] = P\left[ \text{"No injures on the first k attemps"} \right] = (1-p)^{k}, \qquad \forall k = 1,2,\ldots $$
- Đặt $X, Y$ lần lượt ứng với số lần thử của 2 VDV $A, B$, với hàm phân phối xác suất 

$$F(x, y) = P[X \leq x, Y \leq y], \qquad X, Y \overset{\text{i.i.d}}{\sim} T $$
nghĩa là **số lần thử của $A, B$ lần lượt không quá $x, y$**
- Hơn nữa, sự kiện xảy ra chấn thương của các lần thực hiện động tác là độc lập nên

$$ \begin{array}{ccl} P[X \leq x, Y \leq y] &=& P[X \leq x] \cdot P[Y \leq y]\\ &=& (1 - P[X > x]) \cdot (1 - P[Y > y]) \\&=& \left(1 - (1-p)^x \right) \cdot \left(1 - (1-p)^y\right) \end{array}$$
- Giả thiết cuối cùng cho biết $F(2,2) = 0.0441$, tức là

$$ \begin{array}{ccrcl} \left(1 - (1-p)^2 \right) \cdot \left(1 - (1-p)^2 \right) = 0.0441 & \Leftrightarrow & \left(1 - (1-p)^2 \right) &=& 0.21 \\ & \Leftrightarrow & (1-p)^2 &=& 0.79 \\ & \Leftrightarrow & p &=& \lbrace 0.11118, 1.888832 \rbrace \\ & \Rightarrow & p &=& 0.11118 \in (0, 1)  \end{array} $$
- BT yêu cầu tính $F(1, 5)$, lúc này ta chỉ cần thay giá trị $p = 0.11118$ tìm được,

$$ F(1,5) = (1 - (1-p)) \cdot \left( 1-(1-p)^5 \right) = p \left( 1-(1-p)^5 \right) = 0.0495 $$

#### Question 304.
The number of minor surgeries, $X,$ and the number of major surgeries, $Y,$ for a policyholder, this decade, has joint cumulative distribution function 
$$ F(x,y) = \left[ 1 - (0.5)^{x+1} \right] \left[ 1 - (0.2)^{y+1} \right] $$
for nonnegative integers $x$ and $y.$

Calculate the probability that the policyholder experiences exactly three minor surgeries and exactly three major surgeries this decade.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đề đã đặt $X, Y$ là số ca **tiểu phẫu (`minor surgeries`)** và **đại phẫu (`major surgeries`)** trong một thập niên (decade) với hàm phân phối tích lũy đồng thời 
$$F(x ,y) = P[X \leq x, Y \leq y] . $$
- BT yêu cầu tính XS có đúng 3 ca tiểu phẫu và 3 ca đại phẩu trong thập niên này, tức là
$$ P[X = 3, Y = 3] $$
- Vì $X,Y$ nhận giá trị là các số nguyên $\left\lbrace 0, 1, \ldots \right\rbrace$ do đó

$$ \begin{array}{ccl} P[X = 3, Y = 3] &=& P[(2 < X \leq 3), (2 < Y \leq 3)] \\ &=& P[2<X\leq 3, Y \leq 3] \quad - \quad P[2<X\leq 3, Y \leq 2] \\ &=& \left( P[X \leq 3, Y \leq 3] - P[X \leq 2, Y \leq 3] \right) - \left( P[X \leq 3, Y \leq 2] - P[X \leq 2, Y \leq 2] \right) \\ &=& F(3, 3) - F(2, 3) - F(3, 2) + F(2, 2) \\ &=& 0.0004 \end{array} $$

#### Question 305.
A company provides a death benefit of 50,000 for each of its 1000 employees.
- There is a 1.4% chance that any one employee will die next year, independent of all other
employees. 
- The company establishes a fund such that the probability is at least 0.99 that the fund will cover next year’s death benefits.

Calculate, using the Central Limit Theorem, the smallest amount of money, rounded to
the nearest 50 thousand, that the company must put into the fund. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt
$$ X_i = \left\lbrace \begin{array}{cl} 1, & \text{if the employee will die next year with prob } 1.4\% \\ 0, & \text{otherwise} \end{array} \right. $$
khi đó $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.014)$
- Theo các số liệu đầu đề, cty có 1000 nhân viên, ta đặt $$X = \displaystyle \sum_{i=1}^{1000} X_i$$ là số nhân viên tử vong trong năm tới trên tổng số 5000 nhân viên; và

$$ S = 50000 \cdot X $$

$\quad$ là tổng số tiền cty trả cho những nhân viên sẽ chết, khi đó $X \sim \mathcal{B}(1000, 0.014).$
- Giả thiết sau cùng cho biết, cty muốn thành lập một quỹ tiền để chi trả cho các nhân viên tử vong vào năm tới với XS tối thiểu $$0.99$$
- BT yêu cầu sử dụng Định lý giới hạn trung tâm; để tìm khoản tiền nhỏ nhất mà cty phải bỏ ra. Kết hợp với giả thiết sau cùng, ta cần tìm $q$ sao cho
$$ P[S < q] = 0.99 $$
bằng định lý giới hạn trung tâm. Ta có

$$ \mathbb{E}S = 50000 \mathbb{E}X = 50000 \cdot \left( 1000 \cdot (0.014)\right) = 700000$$
$\qquad$ và
$$ \text{Var}S = \left( 50000^2 \right) \text{Var}X = \left( 50000^2 \right) \cdot \left( 1000 \cdot (0.014) \cdot (1-0.014) \right) = 34510000000.$$

$\qquad$ Áp dụng Định lý giới hạn trung tâm khi $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.014)$ và $S = 50000X$, ta được 

$$ S \sim \mathcal{N} \left( 700000, 34510000000 \right) $$

$\qquad$ do đó,

$$ \begin{array}{ccrcl} P[S < q] = 0.99 & \Leftrightarrow & P \left[ \mathcal{N}(0,1) < \dfrac{q - \mathbb{E}S}{\sqrt{\text{Var}S}} \right] &=& 0.99 \\ & \Leftrightarrow & \dfrac{q - 700000}{\sqrt{34510000000}} &=& \Phi^{-1}(0.99) \\ & \Leftrightarrow & q &=& 700000 + \Phi^{-1} (+185 769) \\ & \Leftrightarrow & q &=& 1132099 \end{array} $$

- Vì BT có quy ước làm tròn lên $50000$ nên $q \approx 1 150 000.$

#### Question 306.
An investor invests 100 dollars in a stock. 
- Each month, the investment has probability
0.5 of increasing by 1.10 dollars and probability 0.5 of decreasing by 0.90 dollars.
- The changes in price in different months are mutually independent.

Calculate the probability that the investment has a value greater than 91 dollars at the end
of month 100. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X_k$ là số tiền trong khoản đầu tư ở tháng thứ $k$ với $$ X_k = \left\lbrace \begin{array}{cl} 1.1, & \text{co lai~ voi XS } 0.5 \\ -0.9, & \text{lo~ voi XS } 0.5  \end{array} \right. $$
khi đó $$ S = \displaystyle \sum_{k=1}^{100} X_k $$
là số tiền tích lũy sau 100 tháng,
- Ta có $$ \mathbb{E}X_1 = (1.1)P[X_1 = 1.1] + (-0.9)P[X = -0.9] = 0.1 $$ và $$ \mathbb{E}X_1^2 = (1.1)^2P[X_1 = 1.1] + (-0.9)^2P[X_1 = -0.9] = 1.01 $$

do đó $X_k \overset{\text{i.i.d}}{\sim} X_1$ với $\mathbb{E}X_1 = 0.1$ và $\text{Var}X_1 = \mathbb{E}X_1^2 - [\mathbb{E}X_1]^2 = 1$ nên

$$ S \sim \mathcal{N} \left( 100 \mathbb{E}X_1, (100^2) \text{Var}X_1 \right) = \mathcal{N}(10, 100) $$
- BT yêu cầu tính XS để khoản đầu tư nhiều hơn 91 USD sau khi kết thúc tháng thứ 100, tức là
$$ P[S + 100 \geq 91] $$
trong đó $S + 100$ gồm $100$ USD có sẵn ban đầu và $S$ là số tiền sau $100$ tháng. Ta có

$$ P[S + 100 \geq 91] = P[S \geq -9] = 1 - P \left[ \mathcal{N}(0,1) < \dfrac{-9 - 10}{\sqrt{100}} \right] = 1 - \Phi(-1.9) = \Phi(1.9) \approx 0.9713.$$

#### Question 307.
- The number of workplace accidents occurring in a factory on any given day is **Poisson** distributed with mean $\lambda$. 
- The parameter $\lambda$ is a random variable that is determined by the level of activity in the factory and is uniformly distributed on the interval $[0,3].$

Calculate the probability of one accident on a given day. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $X$ là số các vụ tai nạn, theo đó

$$ (X | \lambda) \sim \text{Pois}(\lambda), \qquad \lambda \sim \mathcal{U}([0,3]) $$
- BT yêu cầu tính XS chỉ có 1 tai nạn trong ngày, tức là
$$ \begin{array}{ccl} P[X = 1] &=& \displaystyle \int_0^3 P[X=1| \lambda] \cdot f(\lambda) d\lambda \\ &=& \displaystyle \int_0^3 \left( \dfrac{e^{-\lambda} \lambda}{1!} \cdot \dfrac{1}{3} d\lambda \right) \\ &=& \dfrac{1}{3} \displaystyle \int_0^3 \lambda e^{-\lambda} d\lambda \\ &=& \dfrac{1 - 4e^{-3}}{3}\\ &\approx & 0.267 \end{array} $$

#### Question 308.
- Let $X$ denote the loss amount sustained by an insurance company’s policyholder in an auto collision. 
- Let $Z$ denote the portion of X that the insurance company will have to pay.
- An actuary determines that X and Z are independent with respective density and probability functions $ f(x) = \left\lbrace \begin{array}{cl} (1/8) e^{-x/8}, & x > 0 \\ 0, & \text{otherwise} \end{array} \right. $  and $ P[Z = z] = \left\lbrace \begin{array}{cl} 0.45, & z = 1 \\ 0.55, & z = 0 \end{array} \right. $

Calculate the variance of the insurance company’s claim payment $ZX.$ 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Từ định nghĩa hàm MDXS của $X$, ta được $X \sim \mathcal{E}(8)$, ta có 
$$ \mathbb{E}X = 8, \qquad \mathbb{E}X^2 = 128 $$
- Từ phân phối xác suất của $Z$, ta có
$$ \begin{array}{ccl} \mathbb{E}Z &=& \displaystyle \sum_{z=0}^1 z P[Z = z] = 0.45 \\ \mathbb{E}Z^2 &=& \displaystyle \sum_{z=0}^1 z^2 P[Z = z] = 0.45 \end{array} $$
- Vì $X, Z$ độc lập nên

$$ \mathbb{E}[XZ] = \mathbb{E}X \cdot \mathbb{E}Z = 3.6 $$
và
$$ \mathbb{E}\left[(XZ)^2 \right] = \mathbb{E}X^2 \cdot \mathbb{E}Z^2 = 57.6 $$
suy ra
$$ \text{Var}(XZ) = \mathbb{E}\left[(XZ)^2 \right] - \left[ \mathbb{E}(XZ) \right]^2 = 44.64. $$

#### Question 309.
- A city with borders forming a square with sides of length 1 has its city hall located at the
origin when a rectangular coordinate system is imposed on the city so that two sides of
the square are on the positive axes. The density function of the population is 

$$ f(x,y) = \left\lbrace \begin{array}{cl} 1.5(x^2 + y^2), & 0 < x,y < 1 \\ 0, & \text{otherwise} \end{array} \right. $$
- A resident of the city can travel to the city hall only along a route whose segments are
parallel to the city borders.

Calculate the expected value of the travel distance to the city hall of a randomly chosen resident of the city

**Giải đáp / SOLUTIONs.**
- BT yêu cầu tính $\mathbb{E}(X + Y).$ Lưu ý rằng, nếu hàm MDXS đồng thời của BNN nhiều chiều đối xứng trên miền xác định của nó thì 

$$ \mathbb{E}X^k = \mathbb{E}Y^k, \qquad \forall k = 1,2,\ldots. $$
- Ở đây, $f(x,y) = f(y,x)$ với mọi $x,y \in [0,1]$ nên
$$ \mathbb{E}X = \mathbb{E}Y $$
do đó
$$ \mathbb{E}(X+Y) = 2 \mathbb{E}X = 2 \displaystyle \int_0^1 \int_0^1 x \left[ (1.5)\left( x^2 + y^2 \right) \right] dx dy = \dfrac{5}{4} $$

#### Question 310.
A couple takes out a medical insurance policy that reimburses them for days of work **missed due to illness.** 
- Let $X$ and $Y$ denote the number of days missed during a given
month by the wife and husband, respectively.
- The policy pays a monthly benefit of 50 times the maximum of $X$ and $Y$, subject to a benefit limit of 100. 
- $X$ and $Y$ are independent, each with a discrete uniform distribution on the set **{0,1,2,3,4}.**

Calculate the expected monthly benefit for missed days of work that is paid to the couple. 

**Giải đáp / SOLUTIONs.**

Theo các giả thiết: 
- $X, Y$ lần lượt là số các ngày vắng (nghỉ làm vì bệnh) trong tháng của vợ và chồng và có phân phối đều trên tập $\lbrace 0,1,2,3,4 \rbrace$, tức là $(X,Y)$ có phân phối đều đồng thời 
$$ f(x, y) = \left\lbrace \begin{array}{cl} \dfrac{1}{|\Omega|}, & \forall (x,y) \in \Omega \\ 0, & \text{otherwise} \end{array} \right. $$
trong đó $|\Omega|$ là số phần tử trong $\Omega$ với
$$ \Omega = \lbrace (x, y) : x,y = 0,1,2,3,4. \rbrace $$
do đó $|\Omega| = 5^2 = 25$ và $$ f(x, y) = \left\lbrace \begin{array}{cl} \dfrac{1}{25}, & \forall (x,y) \in \Omega \\ 0, & \text{otherwise} \end{array} \right. $$
- Đặt $Z=\varphi(X, Y)$ là chi phí hàng tháng (tối đa là $100$ và tăng thêm $+50$ cho theo $\max(X, Y)$) mà BH sẽ trả cho họ, theo đó

$$ \varphi(X, Y) = \left\lbrace \begin{array}{clr} 0, & (X,Y) = (0, 0) \\ 50, & \max(X, Y) = 1, & \text{or, } (X, Y) = \lbrace (1, 0); (0, 1); (1,1) \rbrace \\ 100, & \text{otherwise} \end{array} \right. $$
- Do đó, kỳ vọng số tiền mà BH sẽ trả cho cặp vợ chồng này là

$$ \begin{array}{ccl} \mathbb{E}Z &=& 0.P[(X,Y) = (0, 0)] + 50 \cdot P[(X,Y) \in \lbrace (1, 0); (0, 1); (1,1) \rbrace ] + 100 \cdot P[\max(X, Y) \geq 2] \\ &=& 0 \cdot \dfrac{1}{25} + 50 \cdot \dfrac{3}{25} + 100 \cdot \left(1 - \dfrac{1}{25} - \dfrac{3}{25} \right) \\ &=& 90 \end{array} $$
