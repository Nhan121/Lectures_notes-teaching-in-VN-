#### Question 251, 
A state is starting a lottery game. To enter this lottery, a player uses a machine that randomly selects six distinct numbers from among the first 30 positive integers. 

- The lottery randomly selects six distinct numbers from the same 30 positive integers. A winning entry must match the same set of six numbers that the lottery selected.
- The entry fee is 1, each winning entry receives a prize amount of 500,000, and all other entries receive no prize.

Calculate the probability that the state will lose money, given that 800,000 entries are purchased. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- **Tóm tắt luật chơi & biện giải.**

$\qquad \diamond$ Có 30 số trong $\Omega = \lbrace 1, \ldots, 30 \rbrace$ và người chơi phải chọn 6 số khác nhau từ $\Omega$; ta có tổng cộng $C_{30}^6$ trường hợp.

$\qquad \diamond$ **Trung tâm xổ số (TTXS)** cũng sẽ chọn 6 số riêng biệt từ $\Omega$ và người chơi sẽ chiến thắng nếu 6 số họ chọn trùng khớp với 6 số từ **TTXS**

$\qquad \diamond$ Mỗi lượt chơi, người chơi phải đóng 1 đồng; nếu thắng sẽ có 500000 và thua sẽ không có tiền thưởng.

- **Yêu cầu BT. Tính XS để TTXS bị lỗ, biết rằng những người chơi đã đặt tất thảy 800000 vào giải.**
- **Mô hình hóa BT.** Đặt:

$\qquad \diamond$ $n$ là số người chơi tham gia trò xổ số. Theo giả thiết thì $n = 800000$

$\qquad \diamond$ $X_i$ là sự kiện người chơi thứ $i^{th}$ giành chiến thắng trò xổ số; theo đó xác suất chiến thắng trò chơi giành giải là như nhau và được xác định bởi

$$ P[\text{Winner}] = \dfrac{|\text{chon.dung 6 so tu TTXS da chon.}| \cdot |\text{ko chon.so nao tu 24 so con lai}|}{|\text{select 6 distinct numbers from } \Omega|} = \dfrac{C_6^6 C_{24}^6}{C_{30}^6} $$

như vậy mỗi người chơi $X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, \theta_W)$ với $\theta_W = P[\text{Winner}] = \dfrac{1}{593,775}$

$\qquad \diamond$ $X = \displaystyle \sum_{i=1}^n X_i$ là tổng số người thắng giải trong $n = 800000$ người; do đó $X \sim \mathcal{B}(n, \theta_W)$

$\qquad \diamond$ Vì để **TTXS** bị lỗ vốn nên chỉ cần có **ít nhất 2 người thắng giải trong $n$ người chơi**, như vậy; ta cần tính

$$ \begin{array}{ccl} P[X \geq 2] &=& 1 - P[X=0] - P[X=1] \\ &=& 1 - C_n^0 \theta_W (1 - \theta_W)^n - C_n^1 \theta_W (1 - \theta_W)^{n-1} \\ &=& 1 - 0.2599 - 0.3502 \\ &=& 0.39. \end{array} $$

#### Question 252.
A life insurance company has found there is a 3% probability that a randomly selected
application contains an error. Assume applications are mutually independent in this
respect.

An auditor randomly selects 100 applications.

Calculate the probability that 95% or less of the selected applications are error-free.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $$X_i = \left\lbrace \begin{array}{cl} 1, & \text{neu ung dung. thu i co loi~ (voi xac suat 0.03)} \\ 0, & \text{otherwise} \end{array} \right.$$ 
và $X = \displaystyle \sum_{i=1}^n X_i$ là số các ứng dụng bị lỗi trong $n$ sản phẩm đã kiểm định. Khi đó
$$ X_i \overset{\text{i.i.d}}{\sim} \mathcal{B}(1, 0.03), \qquad X \sim \mathcal{B}(n, 0.03) $$
- BT cho biết $n = 100$ và yêu cầu tính XS mà số sản phẩm `không lỗi (error-free)` chọn được ít hơn $95\%$ trong tổng số $n = 100$ sản phẩm, tức là có nhiều nhất 5 sản phẩm bị lỗi trong 100 sản phẩm. Ta có

$$ P[X \geq 5] = 1 - P[X < 5] = 1 - \displaystyle \sum_{k=0}^4 C_{100}^k (0.03)^k (1 - 0.03)^{100 - k} \approx 0.1821. $$

#### Question 253.
Let $A,$ $B,$ and $C$ be events such that $P[A] = 0.2,$ $P[B] = 0.1,$ and $P[C] = 0.3$. 

The events $A$ and $B$ are independent, the events $B$ and $C$ are independent, and the events $A$ and $C$ are mutually exclusive.

Calculate $P[A \cup B \cup C]$

**Giải đáp / SOLUTIONs.**

Theo giả thiết, ta có
- $P[A] = 0.2, \qquad P[B] = 0.1,\quad$ và $\quad P[C] = 0.3$
- Vì $A, B$ độc lập nên $P[A \cap B] = P[A] \cdot P[B] = 0.02$.
- Do $B, C$ độc lập nên $P[B \cap C] = P[B] \cdot P[C] = 0.03$
- Mà $A, C$ xung khắc nên $P[A \cap C] = 0$ và $P[(A \cap C) \cap B] = 0$

BT yêu cầu tính $P[A \cup B \cup C]$. Ta có

$$ \begin{array}{ccl} P[A \cup B \cup C] &=& (P[A] + P[B] + P[C]) - (P[A \cap B] + P[A \cap C] + P[C \cap B]) + P[A \cap B \cap C] \\ &=& 0.2 + 0.1 + 0.3 - (0.02 + 0.03 + 0) - 0 \\ &=& 0.55. \end{array} $$

#### Question 254.
The annual numbers of thefts a homeowners insurance policyholder experiences are analyzed over three years.

Define the following events:
- i) A = the event that the policyholder experiences no thefts in the three years.
- ii) B = the event that the policyholder experiences at least one theft in the second year.
- iii) C = the event that the policyholder experiences exactly one theft in the first year.
- iv) D = the event that the policyholder experiences no thefts in the third year.
- v) E = the event that the policyholder experiences no thefts in the second year, and at least one theft in the third year.

Determine which three events satisfy the condition that the probability of their union equals the sum of their probabilities. 

**Giải đáp / SOLUTIONs.**
- **Yêu cầu BT. Tìm 3 sự kiện nào thỏa điều kiện XS của hội 3 sự kiện đó bằng với tổng XS từng sự kiện.** Điều này đồng nghĩa với 3 sự kiện này phải xung khắc hay rời nhau (nghĩa là khi sự kiện này xảy ra thì sự kiện còn lại không thể xảy ra)
- **Liên hệ BT.** Đặt $N_1, N_2, N_3$ lần lượt tương ứng số vụ trộm xảy ra trong năm 1, năm 2 và năm 3. Khi đó $$N = \displaystyle \sum_{i=1}^3 X_i$$ là tổng số vụ trộm trong 3 năm.
- **Các sự kiện mà BT đã cho.**

$\qquad \diamond$ $A$. Không có vụ trộm nào xảy ra trong vòng 3 năm; viết lại là $A = \lbrace \omega: N(\omega) = 0 \rbrace$

$\qquad \diamond$ $B$. Có ít nhất 1 vụ trộm xảy ra trong năm thứ hai, hay $B = \lbrace \omega: N_2(\omega) \geq 1 \rbrace$

$\qquad \diamond$ $C$. Có đúng 1 vụ trộm xảy ra trong năm đầu tiên, $C = \lbrace \omega: N_1(\omega) = 1 \rbrace$

$\qquad \diamond$ $D$. Không có vụ trộm nào xảy ra trong năm thứ ba; $D = \lbrace \omega: N_3(\omega) = 0 \rbrace$

$\qquad \diamond$ $E$. Không có vụ trộm nào xảy ra trong năm thứ 2, đồng thời thì có ít nhất một vụ trong năm thứ ba;

$$ E = \lbrace \omega: N_2(\omega) = 0 \text{ and } N_3(\omega) \geq 1 \rbrace $$

- Trong số 5 sự kiện trên; chỉ có $A, B, E$ thỏa yêu cầu vì

$$ A \cap B = B \cap E = A \cap E = \emptyset $$

-------------

**Comments** Các bộ 3 sự kiện còn lại không thể thỏa mãn vì chúng không đôi một xung khắc; chẳng hạn (ở đây tôi chỉ lập luận trong 4 đáp án còn lai của [sách-trang 107-câu 254](https://github.com/Nhan121/Lectures_notes-teaching-in-VN-/blob/master/Exam_P/330%20questions/330-questions.pdf)) 
- $A, C, E$ thì 
$$ A \cap C = \lbrace \omega: N(\omega) = 0; \; N_1 (\omega) = 1 \rbrace = \lbrace \omega: (N_2 + N_3)(\omega) = (N - N_1) (\omega) = -1 < 0 \rbrace = \emptyset $$
nhưng
$$ C \cap E = \lbrace \omega: N_1 (\omega) = 1, \text{ and } N_2(\omega) = 0; \text{ and } N_3(\omega) \geq 1 \rbrace \neq \emptyset $$
- $A, D, E$ thì 
$$ D \cap A = \lbrace \omega: N = 0; \text{ and } N_3 = 0 \rbrace = \lbrace \omega: N_1 + N_2 = N - N_3 = 0 \rbrace \neq \emptyset$$
- $B, C, D$ thì
$$ B \cap C = \lbrace \omega: (N_2 \geq 1); \text{ and } (N_1 = 1) \rbrace \neq \emptyset $$
- $B, C, E$ thì
$$ B \cap C = \lbrace \omega: (N_2 \geq 1); \text{ and } (N_1 = 1) \rbrace \neq \emptyset$$

#### Question 255.
Four letters to different insureds are prepared along with accompanying envelopes. The
letters are put into the envelopes randomly.

Calculate the probability that at least one letter ends up in its accompanying envelope.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT yêu cầu **Tính xác suất để có ít nhất một lá thư được đựng trong phong bì kèm theo.**
- Ta có 4 lá thư tất cả và 4 phong bì sẽ dùng để bao nó; do đó sẽ có 2 trường hợp sau có thể xảy ra:

$\qquad \diamond$ **TH1,** Lá thư đầu tiên có phong bì kèm theo. XS xảy ra sự kiện này là $$P[\text{"first envelope correct"}] = P[\text{"first letters is in 1 envelope among 4 envelopes"}] = \dfrac{1}{4}.$$

$\qquad \diamond$ **TH2,** Nếu lá thư đầu tiên bị lỗi (thiếu phong bì) thì trong 3 lá thư còn lại (lúc này có đến 4 phong bì) thì phải có ít nhất một lá thư được bỏ vào phong bì do số bì thư nhiều hơn số thư. Do đó, :

$\qquad \qquad - $ Có 3 trong 4 cách khác nhau để làm lá thư đầu tiên thiếu phong bì!

$\qquad \qquad - $ Sẽ có 3 cách chọn 3 lá thư còn lại vào $3! = 6$ cách thức khác nhau để hoán vị các phong bì nên XS để bỏ đúng ít nhất một thư (trong 3) vào 4 phong bì là $\dfrac{3}{3!} = \dfrac{1}{2}$

$\qquad \qquad + $ Như vậy, xác suất xảy ra của TH2 là $P[\text{"next letter correct | given first envelope incorrect"}]$ và xác định bởi 

$$ P[\text{"the first envelope is incorrect"}] \cdot P [\text{"Among 6 permutations, 3 will place 1 letter correctly"}] = \dfrac{3}{4} \cdot \dfrac{1}{2} = \dfrac{3}{8}$$

- Vì 2 trường hợp này là xung khắc (TH1 xảy ra thì TH2 ko thể xảy ra và ngược lại), do đó XS cần tính bằng

$$ P[\text{"first envelope correct"}] +  P[\text{"next letter correct | given first envelope incorrect"}] = \dfrac{1}{4} + \dfrac{3}{8} = \dfrac{5}{8}. $$

#### Question 256.
A health insurance policy covers visits to a doctor’s office. Each visit costs 100. The
annual deductible on the policy is 350. 
- For a policy, the number of visits per year has the
following probability distribution: 

| Number of Visits | $0$ | $1$ | $2$ | $3$ | $4$ | $5$ | $6$ |
|-|-|-|-|-|-|-|-|
|Probability| $0.60$| $0.15$| $0.10$| $0.08$| $0.04$| $0.02$| $0.01$|

- A policy is selected at random from those where costs exceed the deductible.

Calculate the probability that this policyholder had exactly five office visits.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đăt $N$ là số lượt ghé phòng khám, theo bảng phân phối thì $N$ có 7 giá trị $0,1,\ldots,6$ và các XS tương ứng.
- Mỗi lượt đến phòng khám; bệnh nhân sẽ chi trả $100$; nên phí chi trả là $100N$ cho $N$ lượt đến.
- Khấu hao hàng năm là $350$ nên để vượt quá hạn mức này thì $N$ phải thỏa 

$$ 100 N \geq 350 \Longrightarrow N \geq 4 $$
- BT yêu cầu tính XS có đúng $5$ lượt ghé phòng khám sau khi chi phí chi trả đã vượt mức khấu hao, tức là

$$ \dfrac{P[N=5]}{P[N \geq 4]} = \dfrac{0.02}{0.04 + 0.02 + 0.01} \approx 0.286. $$

#### Question 257.
A machine has two parts labelled A and B. 

- The probability that part A works for one year is 0.8 and the probability that part B works for one year is 0.6. 
- The probability that at least one part works for one year is 0.9.

Calculate the probability that part B works for one year, given that part A works for one
year.

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- BT cho biết 1 cỗ máy có 2 chi tiết máy $A, B$.
- Đặt $A, B$ lần lượt là các sự kiện về các chi tiết máy $A, B$ hoạt động trong 1 năm. Theo giả thiết, ta có 

$$ P[A] = 0.8; \qquad P[B] = 0.6; \qquad P[A \cup B] = 0.9 $$
- BT yêu cầu tính XS để chi tiết $B$ làm việc được 1 năm biết rằng chi tiết $A$ cũng làm việc được 1 năm, tức là

$$ P[B | A] = \dfrac{P[A \cap B]}{P[A]} = \dfrac{P[A] + P[B] - P[A \cup B]}{P[A]} = \dfrac{5}{8}. $$

#### Question 258.
Six claims are to be randomly selected from a **group of thirteen** different claims, which
includes two **workers compensation** claims, four **homeowners** claims and seven **auto**
claims.

Calculate the probability that the six claims selected will include one workers
compensation claim, two homeowners claims and three auto claims. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $C, H, A$ lần lượt là các yêu cầu về hợp đồng BH về `"bồi thường (Compensation)"`, `"nhà cửa (Homeowners)"` và `"xe cộ (Auto)"`
- Theo đó, $13$ yêu cầu từ KH gồm: $2C, 4H, 7A$ và BH sẽ chỉ chọn 6 trong số đó. Do đó ta có $C_{13}^6$ cách để chọn được nhóm $6$ người từ $13$ người này.
- BT yêu cầu tính XS để trong 6 người được chọn; thì có $1C, 2H, 3A$, như vậy, ta cần tính

$$ \dfrac{P[\text{"select a set of (1C, 2H, 3A) from population of (2C, 4H, 7A)"}]}{P[\text{"select a group of 6 people from population of 13."}]}= \dfrac{C_2^1 \cdot C_4^2 \cdot C_7^3}{C_{13}^6} = 0.245 $$

#### Question 259.
A drawer contains four pairs of socks, with each pair a different color. One sock at a
time is randomly drawn from the drawer until a matching pair is obtained.

Calculate the probability that the maximum number of draws is required.

**Giải đáp / SOLUTIONs.**

- **Tóm tắt & mô hình BT**

$\qquad \diamond$ Một ngăn kéo có 4 đôi vớ, mỗi đôi có các màu khác nhau, quy ước là $A, B, C, D$.

$\qquad \diamond$ BT yêu cầu tính XS tương ứng với số lần tối đa sẽ mở các ngăn kéo.

$\qquad \diamond$ Mỗi lần (đi chơi), ta sẽ một chiếc tất (vớ) được rút ngẫu nhiên từ ngăn kéo cho đến khi nhận được một đôi phù hợp. Do đó, tối đa sẽ tốn 5 lần mở ngăn kéo (gồm **4 lần đầu tiên bị xui nhận 4 màu khác nhau** và **lần cuối chọn màu phù hợp** nhất trùng màu với cái đầu tiên)

- **Thiết lập liên hệ và kết luận.** Ta đã suy ra số lần tối đa sẽ mở ngăn kéo là $5$, trong đó:

$\qquad \diamond$ Lần đầu, chọn ngẫu nhiên một màu bất kỳ trong 8 chiếc tất (vì là 4 đôi) nên ta có XS cho lần đầu là $\dfrac{8}{8} = 1$.

$\qquad \diamond$ Lần 2, đã rút ra 1 chiếc, trong ngăn còn 7 chiếc và lựa chọn thứ hai (để kiểm tra từng chiếc trong 7 chiếc đó có cùng màu với cái ban đầu hay không nhưng lại xui và khác màu) có XS bằng $\dfrac{6}{7}$ 

$\qquad \diamond$ Lần 3, còn 6 chiếc; ta phải lựa 4 trong 6 chiếc này (4 bởi vì ta tránh 2 đôi của 2 trường hợp đầu), nên XS bằng $\dfrac{4}{6}$

$\qquad \diamond$ Lần 4, còn 5 chiếc;tương tự XS bằng $\dfrac{2}{5}$

$\qquad \diamond$ Lần 5. Lúc này, ta đã chính thức chọn được một đôi vớ khớp nhau về màu (nên XS cũng bằng $1$).

$\qquad \diamond$ Vì các lần này là độc lập với nhau, nên XS để thực hiện được điều này bằng

$$ 1 \cdot \dfrac{6}{7} \cdot \dfrac{4}{6} \cdot \dfrac{2}{5} = 0.2286. $$

#### Question 260.
At a mortgage company, 60% of calls are answered by an attendant. 

The remaining 40% of callers leave their phone numbers. Of these 40%, 75% receive a return phone call the
same day. The remaining 25% receive a return call the next day.

- Of those who initially spoke to an attendant, 80% will apply for a mortgage. 
- Of those who received a return call the same day, 60% will apply. 
- Of those who received a return call the next day, 40% will apply.

Calculate the probability that a person initially spoke to an attendant, given that he or she
applied for a mortgage. 

**Giải đáp / SOLUTIONs.**

Trình tự lý luận:
- Đặt $A,S,R,N$ lần lượt tương ứng với các sự kiện **"đăng ký thế chấp", "lúc đầu sẽ nói với tiếp viên (tư vấn lúc đang rãnh)", "nhận được cuộc gọi trong ngày (đang bận)", "được hẹn gọi lại vào ngày kế tiếp (quá bận)"**
- Từ các giả thiết, ta có

$$ \begin{array}{lll} P[S] = 0.6; & P[R] = (0.4)(0.75) = 0.3; & P[N] = (0.4)(0.25) = 0.1 \\ P[A|S] = 0.8; & P[A|R] = 0.6; & P[A | N] = 0.4 \end{array} $$
- BT yêu cầu **tính xác suất để một người ban đầu nói chuyện với một tiếp viên, khi người đó nộp đơn vay thế chấp.** Tức là 

$$ P[S|A] = \dfrac{P[A|S]\cdot P[S]}{P[A|S]\cdot P[S] + P[A|R]\cdot P[R] + P[A|N]\cdot P[N]} = 0.69 $$
