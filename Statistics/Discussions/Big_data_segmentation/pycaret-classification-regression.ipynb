```python
import seaborn as sns
data = sns.load_dataset("iris")
data.head()
```

```python
data_unseen = data.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.classification import *

exp_clf101 = setup(data = data, target = 'species', session_id=123)
```

https://pycaret.readthedocs.io/en/latest/api/classification.html

#### return top 3 models based on 'Accuracy

```python
%time top3 = compare_models(n_select = 3)
```

#### Return best model based on AUC

```python
%time best_model = compare_models(sort = 'AUC')
print(best_model)
```
